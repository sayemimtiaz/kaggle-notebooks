```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pydot
import tensorflow as tf
import keras
import seaborn as sb
from sklearn.metrics import confusion_matrix
```

<div align='left'><font size="5" color="#000080">Data Exploration</font></div>
<hr>
<div align='left'><font size="4" color="#000080">Loading the dataset</font></div>


```python
train = pd.read_csv('../input/fashionmnist/fashion-mnist_train.csv')
test = pd.read_csv('../input/fashionmnist/fashion-mnist_test.csv')
train.tail()
```

<div align='left'><font size="4" color="#000080">Missing value detection</font></div>
<hr>
<div align='left'><font size="3" color="#000000">There is no missing value in this dataset</font></div>


```python
df = pd.concat([train,test], ignore_index=True)
df.isnull().sum().sort_values(ascending=True)
```

```python
X_train = train.drop('label',axis = 1)
y_train = train['label']
X_test = test.drop('label',axis = 1)
y_test = test['label']
X_train.shape
```

```python
x_train_reshape = X_train.values.reshape(-1,28,28)
x_test_reshape = X_test.values.reshape(-1,28,28)
x_train_reshape.shape
```

<div align='left'><font size="4" color="#000080">Data Balancing</font></div>
<hr>
<div align='left'><font size="3" color="#000000">A balanced dataset is the one that contains equal or almost equal number of samples from the each classes. Data balancing play an importent role in accuracy of the model. In this dataset All training classes are equally distributed, So no data imbalance here.</font></div>



```python
y_train.value_counts()
```

```python
col_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

```python
def visualize_input(img, ax):
    ax.imshow(img, cmap='summer')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],2)), xy=(y,x),
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')

fig = plt.figure(figsize = (12,12)) 
ax = fig.add_subplot(111)
visualize_input(x_train_reshape[0], ax)
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i in range(15):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train_reshape[i], cmap='gray')
    plt.xlabel(col_names[y_train[i]])
plt.show()
```


<div align='left'><font size="5" color="#000080">Data Preprocessing</font></div>
<hr>


```python
print(col_names[y_train[0]])
plt.imshow(x_train_reshape[0], cmap=plt.cm.binary)
plt.show()
```

<div align='left'><font size="4" color="#000080">Normalization</font></div>
<hr>
<div align='left'><font size="3" color="#000000">In this dataset pixel values for each image in the dataset are unsigned integers in the range between 0 and 255. In order to scale the each value in the dataset we devide the dataset by 255.After scaling, dataset has valuesin the range of 0 to 1..</font></div>


```python
x_train=X_train/255
x_test=X_test/255
```

<div align='left'><font size="6" color="#000080">Artificial Neural Network</font></div>
<hr>
<div align='left'><font size="3" color="#000000"> An artificial neural network is an interconnected group of nodes, inspired by a simplification of neurons in a brain. It consists of 3 layers ‚Äì input, hidden and output. The input layer accepts the inputs, the hidden layer processes the inputs, and the output layer produces the result. </font></div>
<hr>

<hr>
<div align='left'><font size="3" color="#000080">Steps involving in creating ANN model</font></div>
<hr>

* [<div align='left'><font size="3" color="##00000">1. Defining sequencial the model</font></div>](#Defining)
- [<div align='left'><font size="3" color="##00000">2. Adding hidden layer</font></div>](#Adding)
    * <div align='left'><font size="3" color="##00000">* Fully-connected layer</font></div>
    * <div align='left'><font size="3" color="##00000">* Second hidden layer</font></div>
    * <div align='left'><font size="3" color="##00000">* Dropout layer</font></div>
    * <div align='left'><font size="3" color="##00000">* Output layer</font></div>
- [<div align='left'><font size="3" color="##00000">3. Compiling the model</font></div>](#Adding)
- [<div align='left'><font size="3" color="##00000">4. Training the model</font></div>](#Adding)
- [<div align='left'><font size="3" color="##00000">5. Model  evaluation</font></div>](#Adding)                

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
model.add(keras.layers.Dense(units=24, activation='relu'))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(units=10, activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
model_history = model.fit(x_train,y_train,epochs = 30)
model_history.params
```

```python
model_history.history
```

```python
pd.DataFrame(model_history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("\n Accuracy of ANN model:",test_accuracy)
```

```python
y_predict = model.predict(X_test)
y_predict=np.argmax(y_predict, axis=1)
```

```python
con_mat=confusion_matrix(y_test,y_predict)
plt.style.use('seaborn-deep')
plt.figure(figsize=(10,10))
sb.heatmap(con_mat,annot=True,annot_kws={'size': 15},linewidths=0.5,fmt="d",cmap="gray")
plt.title('True or False predicted clothing\n',fontweight='bold',fontsize=15)
plt.show()
```
