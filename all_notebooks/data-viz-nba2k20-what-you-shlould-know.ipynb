We will transform the string values into numerical or categorical values. Calculate the age and the numbers of years played. Indicate if the player has attended college.
Also we will calcalculate the Body Mass Index (BMI) given by the formula \\(\frac{Weight}{Height^2}\\) and use a standard rule to estime the shape of players however muscular individuals and highly trained athletes may have higher BMIs due to large muscle mass.

```python
def from_date_to_age(date):
    born=datetime.datetime.strptime(date, '%m/%d/%y')
    today = datetime.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
df_nba["current_age"]=df_nba["b_day"].apply(lambda x: from_date_to_age(x))


def int_weight(weight):
    return weight.split("/")[1].split(" ")[1]

def int_height(height):
    return height.split("/")[1].split(" ")[1]

df_nba["height"]=df_nba["height"].apply(lambda x: int_height(x)).astype("float")
df_nba["weight"]=df_nba["weight"].apply(lambda x: int_weight(x)).astype("float")

df_nba.rename({"height":"height_in_m","weight":"weight_in_kg"},axis='columns',inplace=True)
df_nba["salary"] = df_nba["salary"].str[1:].astype("int64")
df_nba["year_played"] = df_nba["current_age"] - (df_nba["draft_year"] - pd.to_datetime(df_nba["b_day"]).dt.year)

df_nba["draft_round"] = df_nba["draft_round"].replace({"Undrafted": 0}).astype("int8")
df_nba["draft_peak"] = df_nba["draft_peak"].replace({"Undrafted": 0}).astype("int8")

df_nba.drop(columns=['b_day'])

df_nba["body_mass_index"] = np.round(df_nba["weight_in_kg"] / ((df_nba["height_in_m"])**2),1)
df_nba.loc[(df_nba["body_mass_index"]>=18.5) & (df_nba["body_mass_index"]<=24.9),"bmi_class"] = "Normal"
df_nba.loc[(df_nba["body_mass_index"]>=25) & (df_nba["body_mass_index"]<=29.9),"bmi_class"] = "Overweight"
df_nba.loc[df_nba["body_mass_index"]>=30,"bmi_class"] = "Obese"

df_nba['college'].isna().astype('int').value_counts() 
df_nba["attended_college"] = df_nba['college'].isna().astype('int')

```

# What about the players positions ?
For the position the letters are used for : G - Guard,  F - Forward, C - Center. We commonly use just five position so lets try to use the stardand will use assume that position C-F = F-C and F-G = G-F.
![Position in baskball](https://upload.wikimedia.org/wikipedia/commons/a/ac/Basketball_Positions.png)

- The leader also known as **Point guard** is responsible for distributing the ball and organizing the attacking play. He mounts the ball from his own side and announces the tactics to be used3 : he is thus frequently called as "the coach on the field". In addition to good dribbling skills, he must have excellent vision of the game to be able to distribute the ball to his teammates.

- The **Shooting guard** is a tough, fast, agile player with good athletic ability. His main role is to score shots and three-point baskets, so he must have good shooting ability. He also makes a few penetrations into the racket for lay-ups or dunks.

- The **Small forward**, also known as a shooting winger or small winger, is a versatile player capable of both shooting at three points and playing in the racket, for example helping the insides on the rebound. His game is usually a compromise between the agility and speed of a leader and the size and power of a pivot. The position thus mixes an offensive and a defensive dimension. 

- The pivot also called **Center** is usually the tallest and thinnest player (often around two meters ten) and the strongest physically. He is therefore frequently nicknamed the big man. His role is to protect the racket by using his size and stature to prevent opponents from shooting or dunking the racket by countering them. Their size also allows them to capture many bounces and dunker easily. 

- The **Power forward** plays a role similar to that of the pivot and forms the so-called "inner" sector with it. His role is mainly defensive: he must prevent the opposing attackers from approaching the basket, capture defensive rebounds and counter the opponents' shots. 

Source : Wikipedia
