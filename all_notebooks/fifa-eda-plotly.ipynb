```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

#### Reading the Data

```python
fifa15=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_15.csv')
fifa16=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_16.csv')
fifa17=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_17.csv')
fifa18=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_18.csv')
fifa19=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_19.csv')
fifa20=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/players_20.csv')
# teams=pd.read_csv('/kaggle/input/fifa-20-complete-player-dataset/teams_and_leagues.csv')
```

```python
all_df=[fifa15,fifa16,fifa17,fifa18,fifa19,fifa20]
names=["fifa15","fifa16","fifa17","fifa18","fifa19","fifa20"]
for df,name in zip(all_df,names):
    df["Fifa"]=name
```

```python
df=pd.concat(all_df)
df.head()
```

## Lets start with the most important question ever!!! 
### Messi Vs Ronaldo Debate

How has their rattings Changed over time?

```python
data=df[(df.short_name=='L. Messi') | (df.short_name=='Cristiano Ronaldo')]


fig=px.bar(data, x="short_name", y="overall", animation_frame="Fifa",
           color="club", hover_name="value_eur",range_y=[90,96])




fig.update_layout(title={
        'text': "How their overall ratting has changed overtime?",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
    },
    showlegend=True,
    xaxis_title='Name',
    yaxis_title='Overall Rating'
)
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 800
fig.show()
```

How do their stats compare?

```python
messi=df[df.short_name=='L. Messi']
ronaldo=df[df.short_name=='Cristiano Ronaldo']


fig = go.Figure()  
for i,name in enumerate(names):
    # Fifa 15
    if(i==0):
        vis=True
    else:
        vis=False
        
    messi_data=messi[messi.Fifa==name]
    ronaldo_data=ronaldo[ronaldo.Fifa==name]
    fig.add_trace(go.Scatterpolar(
            r = [messi_data["pace"].values[0],messi_data["shooting"].values[0],
                 messi_data["dribbling"].values[0],messi_data["passing"].values[0],
                 messi_data["defending"].values[0],messi_data["physic"].values[0]],
            theta = ["Pace","Shooting","Dribbling","Passing","Defending","Physique"],
            fill='toself',name="Messi",visible=vis
    ))
    fig.add_trace(go.Scatterpolar(
            r = [ronaldo_data["pace"].values[0],ronaldo_data["shooting"].values[0],
                 ronaldo_data["dribbling"].values[0],ronaldo_data["passing"].values[0],
                 ronaldo_data["defending"].values[0],ronaldo_data["physic"].values[0]],
            theta = ["Pace","Shooting","Dribbling","Passing","Defending","Physique"],
            fill='toself',name="Ronaldo",visible=vis
    ))




fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="Fifa15",
                     method="update",
                     args=[{"visible": [True, True, False, False,False,False,False, False, False, False,False,False]},
                           {"title": "Messi Vs Ronaldo: Fifa 15"}]),
                
                dict(label="Fifa16",
                     method="update",
                     args=[{"visible": [False, False, True, True,False,False,False, False, False, False,False,False]},
                           {"title": "Messi Vs Ronaldo: Fifa 16"}]),
                
                dict(label="Fifa17",
                     method="update",
                     args=[{"visible": [False, False, False, False,True,True,False, False, False, False,False,False]},
                           {"title": "Messi Vs Ronaldo: Fifa 17"}]),
                
                dict(label="Fifa18",
                     method="update",
                     args=[{"visible": [False, False, False, False,False,False,True, True, False, False,False,False]},
                           {"title": "Messi Vs Ronaldo: Fifa 18"}]),
                
                dict(label="Fifa19",
                     method="update",
                     args=[{"visible": [False,False, False, False,False,False,False, False,True, True,False,False]},
                           {"title": "Messi Vs Ronaldo: Fifa 19"}]),
                
                dict(label="Fifa20",
                     method="update",
                     args=[{"visible": [False,False, False, False,False,False,False, False, False, False,True,True]},
                           {"title": "Messi Vs Ronaldo: Fifa 20"}])
            ]),
        )
    ])




fig.update_layout(title={
        'text': "How do their stats compare?",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
    },
    showlegend=True
)

fig.show()
```
