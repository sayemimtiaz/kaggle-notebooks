# Indexing and Slicing with reindex()

- Reindexing is typically used with time series data when we have date-time index.
- One advantage of reindex() is that we can pass row and column labels to reindex() that are not in the data frame.

```python
summer = pd.read_csv('/kaggle/input/summer-olympics-dataset-18962012/summer.csv')
```

```python
summer
```

- We again have RangeIndex.

```python
summer.loc[[0, 5, 30000], ['Athlete', 'Medal']]
```

- Don't confuse it with the position based index. When using RangeIndex it is simply a coincidence that the row labels and the row position index are the same.

```python
#Error row 40000 does not exist
#summer.loc[[0, 5, 30000, 40000], ['Athlete', 'Medal']]
```

```python
summer.reindex(index = [0, 5, 30000, 40000], columns = ['Athlete', 'Medal'])
```

```python
summer.reindex(index = [0, 5, 30000, 40000], columns = ['Athlete', 'Medal','Age'])
```

- Reindex() gives us some flexibility but it does not work if we don't have unique column index or row index.

```python
summer = pd.read_csv('/kaggle/input/summer-olympics-dataset-18962012/summer.csv', index_col = 'Athlete')
```

```python
summer
```

- Now we have a non unique index.
- As long as you do not index a non unique axis reindex() works.

```python
summer.reindex(columns = ['Medal', 'Age'])
```

```python
# Error: ValueError: cannot reindex from a duplicate axis
#summer.reindex(index = ['PHELPS, Michael'], columns = ['Medal', 'Age'])
```

- It does not matter if we use a duplicate entry or a unique entry for index. As long as we have duplicates in the index we cannot reindex.

- NOTE: we always have to pass a list to the index and the columns parameter of reindex() even if we have only one entry.

## Putting it all together

```python
#BAD PRACTICE: Two indexing operations involved
#chained indexing
summer[['Year', 'Event', 'Medal']].loc['LEWIS, Carl']
```

```python
#BAD PRACTICE: Two indexing operations involved
# chained indexing can cause severe problems when we want to create slices or copies of dataframe
# or if we want to manipulate data in a data frame 
summer.loc['LEWIS, Carl'][['Year', 'Event','Medal']]
```

```python
#BEST PRACTICE
# Only one indexing operation involved no chained indexing.
summer.loc['LEWIS, Carl', ['Year', 'Event', 'Medal']]
```
