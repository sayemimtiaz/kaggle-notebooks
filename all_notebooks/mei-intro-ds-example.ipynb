## Exploring the data (1): Basic Statistics and Boxplots

### Basic Statistics
In this part of the activity you will output summary values from the table. Summary values 'summarise' the values for one field of the table, using data from all the records in the table. For example the mean or standard deviation.

The next block of code will output a list of the summary values available for the field 'Daily Mean Temperature'.

## Exploring the data (2): Visualisations
In this part of the activity you will explore a different type of visualisation for the data. The visualisation seen here is a *hexagonal bin plot* - these are like scatter diagrams but with the frequency in each hexagon represented by the shading. The are very useful for bigger data sets.

The next box of code creates a hexagonal bin plot for the Daily Mean Temperature and Daily Mean Pressure for 2015.

```python
# draw a hexagonal bin plot for life expectancy v GDP
heathrow_2015_data.plot.hexbin(x='Daily Mean Pressure', y='Daily Mean Temperature',gridsize=10, sharex=False)
plt.show()

# explore changing the size of the hexagons by changing the value in gridsize=10

# explore comparing other pairs of columns from the dataset
```
