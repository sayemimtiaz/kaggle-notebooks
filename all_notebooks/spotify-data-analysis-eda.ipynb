```python
plt.figure(figsize=(40,10))
sns.scatterplot(x=data['Artist'], y=data['Artist Followers'])
```

* Scatter plots' primary uses are to observe and show relationships between two numeric variables.
* This scatterplot is between 'Artist' and 'Artist Followers'

```python
plt.figure(figsize=(40,100))
# Bar chart showing song name and no of times it charted
sns.barplot(x=data['Number of Times Charted'], y=data["Song Name"])
# Add label for horizontal axis
plt.xlabel("")
# Add label for vertical axis
plt.title("Song Name and Number of Times Charted")

# Check your answer
```

* barplot() method. A barplot is basically used to aggregate the categorical data according to some methods and by default it's the mean.
* This is the bar plot between "Number of Times Charted" and "Song Name"

```python
plt.figure(figsize=(50,25))
# Line chart showing the number of times each genre has charted
sns.lineplot(x="Genre",y="Number of Times Charted",data=data)
plt.xlabel("Genre")
plt.title("Genre")
```

* This is the lineplot between "Genre" and "Number of Times Charted"
* Figure creates a new figure window using default property values. The resulting figure is the current figure. figure( Name,Value ) modifies properties of the figure using one or more name-value pair arguments.

```python
plt.figure(figsize=(50,80))
# Line chart showing the Streams to each song
sns.lineplot(x="Song Name",y="Week of Highest Charting",data=data)
plt.xlabel("Week of Highest Charting")
plt.title("Week of Highest Charting")
```

* This is the lineplot between "Song Name" and "Week of Highest Charting"
