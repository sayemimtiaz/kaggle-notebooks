```python
df = pd.read_csv("../input/modified-covid19-open-line-list/modified_COVID19_open_line_list.csv")
df.head().style.background_gradient(cmap='nipy_spectral')
```

```python
import plotly.express as px

# Grouping it by Respiratory symptoms  and Pneumonitis
plot_data = df.groupby(['respiratory symptoms', 'dizziness'], as_index=False).pneumonitis.sum()

fig = px.bar(plot_data, x='respiratory symptoms', y='pneumonitis', color='dizziness')
fig.show()
```

```python
plot_data = df.groupby(['age'], as_index=False).chills.sum()

fig = px.line(plot_data, x='age', y='chills')
fig.show()
```

```python
cnt_srs = df['dyspnea'].value_counts().head()
trace = go.Bar(
    y=cnt_srs.index[::-1],
    x=cnt_srs.values[::-1],
    orientation = 'h',
    marker=dict(
        color=cnt_srs.values[::-1],
        colorscale = 'Purples',
        reversescale = True
    ),
)

layout = dict(
    title='dyspnea',
    )
data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="dyspnea")
```

```python
Admission = df.groupby(["admission_hospital_to_confirmation"])["age"].sum().reset_index().sort_values("age",ascending=False).reset_index(drop=True)
Admission
```

```python
fig = go.Figure(data=[go.Bar(
            x=Admission['admission_hospital_to_confirmation'][0:10], y=Admission['age'][0:10],
            text=Admission['age'][0:10],
            textposition='auto',
            marker_color='black'

        )])
fig.update_layout(
    title='Hospital Admission',
    xaxis_title="Hospital Admission to Confirmation",
    yaxis_title="Age",
)
fig.show()
```

```python
fig = go.Figure(data=[go.Scatter(
    x=Admission['admission_hospital_to_confirmation'][0:10],
    y=Admission['age'][0:10],
    mode='markers',
    marker=dict(
        color=[145, 140, 135, 130, 125, 120,115,110,105,100],
        size=[100, 90, 70, 60, 60, 60,50,50,40,35],
        showscale=True
        )
)])
fig.update_layout(
    title='Hospital Admission',
    xaxis_title="Hospital Admission to confirmation",
    yaxis_title="Age",
)
fig.show()
```

```python
labels = ["admission_hospital_to_confirmation","age"]
values = Admission.loc[0, ["admission_hospital_to_confirmation","age"]]
df = px.data.tips()
fig = px.pie(Admission, values=values, names=labels, color_discrete_sequence=['royalblue','darkblue','lightcyan'])
fig.update_layout(
    title='Hospital admission to confirmation : '+str(Admission["age"][0]),
)
fig.show()
```
