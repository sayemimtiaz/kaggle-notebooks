```python
df = pd.read_csv('/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv')
df = df.sample(frac=1).reset_index(drop=True)
print(df.shape)
df.head()
```

```python
train_y = np.array(df['output'])
train_x = np.array(df.drop('output', axis=1))
```

## Let's create a neural network

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    #input layer
    tf.keras.layers.InputLayer(input_shape=(13)),
    #hiddens layers
    tf.keras.layers.Dense(32, activation='sigmoid'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(16, activation='sigmoid'),
    tf.keras.layers.Dropout(0.1),
    #output layer
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])

model.summary()
```

## We have to find the optimal epochs number 

```python
earlystopping = tf.keras.callbacks.EarlyStopping(monitor ="val_loss", mode ="min", patience = 400, restore_best_weights = True)

history = model.fit(train_x, train_y, epochs = 1500, validation_split = 0.3, callbacks = [earlystopping])
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
