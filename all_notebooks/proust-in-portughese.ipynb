```python
from csv import QUOTE_NONE
```

```python
def read_and_reformat(csv_path):
    df = pd.read_csv(csv_path,
                     sep='|',
                     encoding='iso-8859-1',
                     dtype=object,
                     header=None,
                     quoting=QUOTE_NONE,
                     names=['@paragraph@volume@chapter'])    
    df['@paragraph@volume@chapter'] = df['@paragraph@volume@chapter'].str.replace('#NAME\?', '')
    df['@paragraph@volume@chapter'] = df['@paragraph@volume@chapter'].str.strip(',')
    return df
```

```python
df = read_and_reformat('../input/marcel-proust-in-search-of-lost-time/proust_dataset_ptbr.csv')
df.head()
```

```python
def read_and_reformat(csv_path, encoding='iso-8859-1'):
    df1 = pd.read_csv(csv_path,
                     sep='|',
                     encoding=encoding,
                     dtype=object,
                     header=None,
                     quoting=QUOTE_NONE,
                     names=['@paragraph@volume@chapter'])    
    df1['@paragraph@volume@chapter'] = df1['@paragraph@volume@chapter'].str.replace('#NAME\?', '')
    df1['@paragraph@volume@chapter'] = df1['@paragraph@volume@chapter'].str.strip(',')
    return df1

df1 = read_and_reformat('../input/marcel-proust-in-search-of-lost-time/proust_dataset_ENG.csv', 'utf-8')
```

```python
##Code by Taha07  https://www.kaggle.com/taha07/data-scientists-jobs-analysis-visualization/notebook

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'black',
                      colormap='Set1',
                      height =2000,
                      width = 2000
                     ).generate(str(df1["@paragraph@volume@chapter"]))
plt.rcParams['figure.figsize'] = (12,12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Proust English Words")
plt.show()
```
