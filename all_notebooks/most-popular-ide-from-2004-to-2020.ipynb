```python
sns.set(style='white')
plt.figure(figsize=(35,15))
# create a color palette
palette = plt.get_cmap('prism')
 
# multiple line plot
num=0
for column in df.drop('Date', axis=1):
    num+=1
    plt.plot(df['Date'], df[column], marker='', color=palette(num), linewidth=3, alpha=0.9, label=column)

# Add legend
plt.legend(ncol=2)
 
# Add titles
plt.title("Most Popular IDE from 2004 to 2020", loc='left', fontsize=30, fontweight=0, color='k')
plt.xlabel("Time", fontsize=22)
plt.xticks(rotation=90)
plt.ylabel("Score", fontsize=22)
```
