```python
def cprint(string:str, end="\n"):
    """
    A little utility function for printing and stuff
    """
    _pprint(f"[black]{string}[/black]", end=end)

def set_palette(palette):
    """
    Function to set the color space
    """
    sns.set_palette(palette)
    cmap = plt.get_cmap(palette)
    colors = [cmap(i) for i in np.linspace(0, 1, 15)]
    return colors

def cout(string: str, color=Fore.RED, end="\n"):
    """
    Saves some work
    """
    print(color+string+Style.RESET_ALL, end=end)

def print_quantile_summary(column:str):
    """
    Prints Quantile Summary a bit nicely.
    """
    verbose = column.split(' ')[0].capitalize()
    cprint(f"[blue]Some Stats about {verbose} Scores:[/blue]", end="")

    print(f"Maximum Marks achieved in {verbose} were: ", end="")
    cout(f"{data[column].max()}", end="\n")

    print(f"Minimum Marks achieved in {verbose} were: ", end="")
    cout(f"{data[column].min()}", end="\n\n")

    print(f"25% Students scored less than ", end="")
    cout(f"{int(np.quantile(data[column], 0.25))}", end=" ")
    print(f"marks in {verbose}")

    print(f"50% Students scored less than ", end="")
    cout(f"{int(np.quantile(data[column], 0.50))}", end=" ")
    print(f"marks in {verbose}")

    print(f"75% Students scored less than ", end="")
    cout(f"{int(np.quantile(data[column], 0.75))}", end=" ")
    print(f"marks in {verbose}")

colors = set_palette("winter_r")
```

## Gender vs Race/Ethnicity Plots

Let's start by observing the trends of Gender column with respect to Race/Ethnicity column.

```python
unq_groups = data['race/ethnicity'].value_counts().index.tolist()

fe_list = []
ma_list = []
fe_std = []
ma_std = []
for gr in unq_groups:
    temp_condition = data['race/ethnicity'] == gr
    male_cond = data['gender'] == 'male'
    female_cond = data['gender'] == 'female'
    ma_list.append(data[temp_condition & male_cond].count().max())
    fe_list.append(data[temp_condition & female_cond].count().max())
    fe_std.append(data[temp_condition & female_cond]['gender'].index.values.mean())
    ma_std.append(data[temp_condition & male_cond]['gender'].index.values.mean())
    
width = 0.35
```

```python
fig, ax = plt.subplots(figsize=(8, 7))

ax.bar(unq_groups, ma_list, width, label='Male', color='blue')
ax.bar(unq_groups, fe_list, width, bottom=ma_list, label='Female', color='aqua')

ax.set_ylabel('Count')
ax.set_xlabel('Group')
ax.set_title('Count by Race/Ethnicity and Gender')
ax.legend()

plt.show()
```

# Hypothesis Testing ðŸ“‚

## Shapiro-Wilk Test

Shapiro-Wilk tests whether a data sample has a Gaussian distribution.

Assumptions:
* Observations in each sample are independent and identically distributed (iid).

Interpretation:
* H0: the sample has a Gaussian distribution.
* H1: the sample does not have a Gaussian distribution.

We will do this test on `Reading Score`, `Writing Score`, `Math Score` and `Average Score` and then store the results in a dataframe.
