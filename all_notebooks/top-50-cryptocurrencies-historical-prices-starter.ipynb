## Imports 

```python
import pandas as pd
import datetime as dt
import pandas_datareader.data as web
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

# Analysis of a single cryptocurrency

**The analysis is done for each day starting from January  2020 to July 2021 on Ethereum (ETH) cryptocurrency.**

```python
eth_data = pd.read_csv("../input/top-50-cryptocurrency-historical-prices/Ethereum.csv" , 
                       index_col =["SNo"] , 
                       parse_dates=['Date'])
start = "2020-01-01"
end = "2021-07-31"
mask = (eth_data['Date'] >= start) & (eth_data['Date'] <= end)
eth_data = eth_data.loc[mask]
eth_data.head()
```

```python
eth_data.describe()
```

## Area Chart for daily prices

```python
area = px.area(data_frame = eth_data ,
               y = "Price",
               x="Date", 
               title = 'Ethereum(ETH) Daily Price')

area.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

area.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')
area.update_layout(showlegend = True,
    title = {
        'text': 'Ethereum(ETH) Daily Price (2020-Present)',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

area.show()
```

## Candlestick Chart

```python
candlestick = go.Figure(data = [go.Candlestick(x =eth_data["Date"], 
                                               open = eth_data[('Open')], 
                                               high = eth_data[('High')], 
                                               low = eth_data[('Low')], 
                                               close = eth_data[('Price')])])
candlestick.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

candlestick.update_layout(
    title = {
        'text': 'Ethereum(ETH) Daily Price(2019-Present) Candelstick Chart',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

candlestick.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')
candlestick.show()
```

## OHLC Chart

```python
ohlc = go.Figure(data = [go.Ohlc(x = eth_data["Date"], 
                                               open = eth_data[('Open')], 
                                               high = eth_data[('High')], 
                                               low = eth_data[('Low')], 
                                               close = eth_data[('Price')])])

ohlc.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

ohlc.update_layout(
    title = {
        'text': 'Ethereum(ETH) Daily Price(2020-Present) OHLC Chart',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
ohlc.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')
ohlc.show()
```

## Daily Percentage Change

```python
line = px.line(data_frame=eth_data,
               x = "Date" ,
               y="Change %", 
               title = 'Ethereum(ETH) Daily Percentage Change(2019-Present)')

line.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

line.update_yaxes(title_text = 'Perctange Change', ticksuffix = '%')
line.update_layout(showlegend = True,
    title = {
        'text': 'Ethereum(ETH) Daily Percentage Change(2019-Present)',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

line.show()
```

# Analysis of all TOP 50 cryptocurrencies 

**The analysis is done for each day starting from January  2021 to July 2021 on all the 50 cryptocurrencies in this dataset.**

```python
data = pd.read_csv("../input/top-50-cryptocurrency-historical-prices/ALL_COMBINED.csv",parse_dates=['Date'])
start = "2021-01-01"
end = "2021-07-31"
mask = (data['Date'] >= start) & (data['Date'] <= end)
data = data.loc[mask]
data.head()
```

```python
data.describe()
```

## Area Plot Change of all 50 Cryptocurrencies 

```python
area = px.area(data_frame= data , 
               x = "Date" ,
               y= "Price", 
               line_group="Currency_Name" , 
               color = "Currency_Name" , 
               color_discrete_sequence=px.colors.qualitative.Alphabet_r,
               title = 'Area Plot for TOP 50 Cryptocurrencies (2021-Present)')

area.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

area.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')
area.update_layout(showlegend = True,
    title = {
        'text': 'Area Plot for TOP 50 Cryptocurrencies (Jan 2021-Present)',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

area.show()
```

## Daily Percentage Change of all 50 Cryptocurrencies

```python
area = px.area(data_frame = data,
                 y  = "Change %" , 
                 x = "Date" , 
                 line_group="Currency_Name", 
                 color = "Currency_Name",
                 color_discrete_sequence=px.colors.qualitative.Alphabet, 
                 title = 'Daily Percentage Change of all 50 Cryptocurrencies(JAN 2021- Present)')

area.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

area.update_yaxes(title_text = 'Percentage Change ', ticksuffix = '%')
area.update_layout(showlegend = True,
    title = {
        'text': 'Daily Percentage Change of all 50 Cryptocurrencies(JAN 2021- Present)',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

area.show()
```
