### Measures Of The Center:

A measure of central tendency (measure of center) is a value that attempts to describe a set of data by identifying the central position of the data set (as representative of a "typical" value in the set). We are familiar with measures of central tendency called the mean, median and mode.

Those Three Measure of the Center are as follows:

1. Mean: Measure of average of all the values in a sample is called Mean.
2. Median: Measure of the central value of the sample set is called Median.
3. Mode: The value most recurrent in the sample set is known as Mode.

## Measures of the Spread:

Measures of spread describe how similar or varied the set of observed values are for a particular variable (data item). Measures of spread include the range, quartiles and the interquartile range, variance and standard deviation. And they are as follows:

* **Range**: It is the given measure of how spread apart the values in a data set are.

$$range = x__max - x__min$$

* **Inter Quartile Range (IQR)**: It is the measure of variability, based on dividing a data set into quartiles.

$$Inter Quartile Range (IQR) = Q_3 - Q_1$$

* **Variance**: It describes how much a random variable differs from its expected value. It entails computing squares of deviations.
    * **Deviation** is the difference between each element from the mean.
    
    * **Population Variance** is the average of squared deviations.
    
        $$ S^2 = \frac{\Sigma(x_i - \bar{x})}{N} $$
        
    * **Sample Variance** is the average of squared differences from the mean.
    
        $$ S^2 = \frac{\Sigma(x_i - \bar{x})}{n-1} $$ 
        
* **Standard Deviation**: It is the measure of the dispersion of a set of data from its mean.

    $$ \sigma = \sqrt(\frac{\Sigma(x_i - \mu)}{N}) $$

In This Kernel we will cover Descriptive Statics topics using Python on Gap Minder Datasets. We will cover practical implementation of it using python.

**Data Sets Information** : This datasets provides data about the population, life expectancy and GDP in different countries of the world from 1952 to 2007. There is also a separate file for 2007.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
gapminder_filepath = "../input/gapminder-datacamp-2007/gapminder_full.csv"
gapminder_data = pd.read_csv(gapminder_filepath)
gapminder_data.head()
```

### Measure of Centre:

* Mean:
* Median:
* Mode:

```python
# Mean of data attribute Life expectancy:
mean = gapminder_data['life_exp'].mean()
print("Mean:", mean)

# Median of data attribute Life expectancy:
median = gapminder_data['life_exp'].median()
print("Median:", median)

# Median of data attribute Life expectancy:
mode = gapminder_data['life_exp'].mode()
print("Model:",mode)
```

### Measure of spread:

* Range
* Inter Quantile Range(IQR):
* Variance:
    * Deviation:
    * Population Variance:
    * Sample Variance:
 * Standard Deviation:

```python
def sam_variance(data):
    # Number of observations
    n = len(data)
    # Mean of the data
    mean = sum(data) / n -1
    # Square deviations
    deviations = [(x - mean) ** 2 for x in data]
    # Variance
    variance = sum(deviations) / n
    return variance
```

```python
from statistics import variance , stdev

# Range of given data attributes:
rang = gapminder_data['life_exp'].max() - gapminder_data['life_exp'].min()
print("Range:",rang)

# Interquartile Range(IQR) of given data attributes:
q3, q1 = np.percentile(gapminder_data['life_exp'], [75 ,25])
iqr = q3 - q1
print("IQR:",iqr)

# Population Variance of given data attributes:
var = variance(gapminder_data['life_exp'])
print("Variance:",var)

# Deviation.
dev = stdev(gapminder_data['life_exp'])
print("Standard Deviation:",dev)

# Sample Variance.
sam_var = sam_variance(gapminder_data['life_exp'])
print("Sample Variance", sam_var)
```
