```python
# Load Data
df = pd.read_csv('../input/nlp-tweet-sentiment-analysis/bitcointweets.csv', header=None)
df = df[[1,7]]
df.columns = ['tweet','label']
df.head()
```

```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, Div, Slider, CustomJS
from bokeh.layouts import Column
output_notebook() #create default state to generate the output
```

```python
# To interact with data in Bokeh, we have to use ColumnDataSource
cds = ColumnDataSource(df)

# Create simple UI consisting of slider representing the tweet index + div objects to show the tweet text & label
slider = Slider(title='Tweet Index', start=0, end=df.shape[0]-1, value=0)
div_text = Div(text="Tweet Text:", width=450, height=75)
div_label = Div(text="Tweet Label:", width=200, height=50)

# Define a callback function for the slider
slider_callback = CustomJS(args=dict(slider=slider, div_text=div_text, div_label=div_label, cds=cds), code='''
    var tweet_text = cds.data['tweet'][slider.value];
    var tweet_label = cds.data['label'][slider.value];
    div_text.text = 'Tweet Text:<br>' + tweet_text;
    div_label.text = 'Tweet Label:<br>' + tweet_label;
''') 

# Attach the callback to the 'value' property of slider
slider.js_on_change('value', slider_callback)

layout = Column(slider,div_text,div_label)
show(layout)
```
