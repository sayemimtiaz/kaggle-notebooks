#When is a football coach overtaken?

Last year, the board of a major European club was discussing the next coach and an important name was suggested. He was quickly dismissed: "It's out of date", were the words of an influential director. It is a common comment in football, and even some coaches have used it to define each other. They say that life goes fast, but not as fast as football. This is the point.

The way football leaves coaches behind is difficult to explain, but it is real. That is how someone like Giovanni Trapattoni, champion of the Champions League for Juventus (1985), can end up training Ireland; that José Mourinho ended up at Tottenham; and Carlo Ancelotti, at Everton. Suddenly, the coaches discover that they are not being sought out by the best clubs, because they are not winning the trophies of the same level, falling into a gradual fall.

#Rise, Peak and Decline

It is from these great titles that it is possible to calculate how long a coach can follow at the top with great jobs. According to a survey by the Independent, from England, the average is 12.3 years. It was arrived at from the history of 50 coaches, with different profiles. They fit the logic that careers tend to last about 30 years and usually go through three stages: rise, peak and decline.

This peak is the time between the first and the last great success of a coach. Great success in lifting is a national league title or reaching the Champions League final. In the case of team work, winning the Euro or reaching the decision of a World Cup.

This is not to say that those that are on top 10 will start to worry in the next two years. It is not a matter of fighting an “imaginary clock”, but of following the tangible changes of the game. Coaches themselves dictate this pace. The best are very intelligent, full of ideas that defy orthodox models.

Rinus Michels' Ajax took the game to a whole new level with “total football”. Opponents could not handle the pressure, which was a revolutionary idea.

This is crucial: a good idea is not a good idea forever. What worked so well for 10, 15 years may not work now.
