```python
df2=df1.drop('Date',axis=1).reset_index()
```

```python
lat_lon=[]
geolocator=Nominatim(user_agent="Kaggle")
for location in df2['Country']:
    location = geolocator.geocode(location)
    if location is None:
        lat_lon.append(np.nan)
    else:    
        geo=(location.latitude,location.longitude)
        lat_lon.append(geo)
```

```python
df2['geo_loc']=lat_lon
lat,lon=zip(*np.array(df2['geo_loc']))
df2['lat']=lat
df2['lon']=lon
df2.drop(['geo_loc'],axis=1,inplace=True)
```

```python
m = folium.Map(location=[54, 15], tiles='openstreetmap')
for id,row in df2.iterrows():
    folium.Marker(location=[row['lat'],row['lon']], popup=row['Confirmed']).add_to(m)

m
```

<small><a href='#0'>Go to top</a></small>  

<a id="12"></a><h1 style='background:#FFADB5;font-size: 25px; border:0; font-family: cursive;color:black;border-radius: 15px 45px'><center>12) Marker Cluster map for confirmed case in various countries.</center></h1>


```python
m = folium.Map(location=[54,15], tiles='cartodbpositron', zoom_start=0)

from folium.plugins import MarkerCluster
mc = MarkerCluster()
for idx, row in df2.iterrows():
    mc.add_child(folium.Marker([row['lat'], row['lon']],popup=row['Confirmed']))
m.add_child(mc)

m
```

```python
lat_lon=[]
geolocator=Nominatim(user_agent="app")
for location in df3['Name of State / UT']:
    location = geolocator.geocode(location)
    if location is None:
        lat_lon.append(np.nan)
    else:    
        geo=(location.latitude,location.longitude)
        lat_lon.append(geo)
        
df3['geo_loc']=lat_lon
lat,lon=zip(*np.array(df3['geo_loc']))
df3['lat']=lat
df3['lon']=lon
df3.drop(['geo_loc'],axis=1,inplace=True)
```
