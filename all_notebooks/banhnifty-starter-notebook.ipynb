```python
import numpy as np 
import pandas as pd 
from pathlib import Path
df = pd.DataFrame()
data_dir  = Path("/kaggle/input/bank-nifty-5-year-data")
for files in list(data_dir.glob("*.csv")):
    df_ = pd.read_csv(files.as_posix())
    df = df.append(df_)
    df["Date"] = pd.to_datetime(df["Date"])
```

```python
print(f"date range\nstart: {df.Date.min()} end: {df.Date.max()} ")
print("number of days: ", (df.Date.max() - df.Date.min()).days)
```
