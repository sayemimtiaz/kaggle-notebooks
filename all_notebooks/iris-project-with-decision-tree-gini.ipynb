```python
df = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
df.head()
# We printed our first 5 lines.


X = df.iloc[:, 0:4]
Y = df.iloc[:, 4]
# We specified our class column.


```

To get high accuracy from decision trees, we need to divide our data into two, prediction and learn. I divided it as 70 percent learning and 30 percent as prediction. I used the code below for this.


```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3)
```

Now I call the gini algorithm for decision tree study.

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=10, min_samples_leaf=8)
```


Let's explain our hyperparameters.
* max_depth = The depth of our tree. We indicate how many 'if' structures will be nested. 
* min_samples_leaf =It refers to the minimum number of instances required to split an internal node.

Let's run the algorithm

```python
clf_gini.fit(X_train, y_train)
```

Let's print out the algorithm's predictions now.

```python
y_pred = clf_gini.predict(X_test)
print(y_pred)
```

Our algorithm worked.Now we will enter random numbers ourselves for prediction. And the algorithm will predict which flower these values belong to.

```python
print(clf_gini.predict([[5,2,3,1]]))
# sepal_length = 5
# sepal_width=2
# petal_length=3	
# petal_width=1
# Which flower can have these values?
```

Let's measure the accuracy of our algorithm.

```python

print ("Accuracy is ", accuracy_score(y_test,y_pred)*100)
```
