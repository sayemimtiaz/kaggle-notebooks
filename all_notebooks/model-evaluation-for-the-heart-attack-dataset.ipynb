# **model evaluation for the heart attack dataset.**

About this dataset
Age : Age of the patient

Sex : Sex of the patient

exang: exercise induced angina (1 = yes; 0 = no)

ca: number of major vessels (0-3)

cp : Chest Pain type chest pain type

Value 1: typical angina
Value 2: atypical angina
Value 3: non-anginal pain
Value 4: asymptomatic
trtbps : resting blood pressure (in mm Hg)

chol : cholestoral in mg/dl fetched via BMI sensor

fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

rest_ecg : resting electrocardiographic results

Value 0: normal
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
thalach : maximum heart rate achieved

target : 0= less chance of heart attack 1= more chance of heart attack

```python
#split the data to feature and label

X = heart_data.iloc[:, 0: -1]
y = heart_data.iloc[:, -1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 0)
```

```python
#scalled the data
scaler = StandardScaler()

X_test_scaled = scaler.fit_transform(X_test)
X_train_scaled = scaler.fit_transform(X_train)

y_train_scaled = scaler.fit_transform(y_train)
y_test_scaled = scaler.fit_transform(y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 7)  
knn.fit(X_train_scaled, y_train)

knn_predicted = knn.predict(X_test_scaled)

#model fit check 
train_score = knn.score(X_train_scaled, y_train)
test_score = knn.score(X_test_scaled, y_test)

print('test fit score: {}'.format(knn.score(X_test_scaled, y_test)))
print('train fit score: {}'.format(knn.score(X_train_scaled, y_train)))
print('Confusion matrix: {}'.format(confusion_matrix(y_test, knn_predicted)))
print('Accuracy score: {}'.format(accuracy_score(y_test, knn_predicted)))
print('Y_test: {}'.format(y_test.head(25)))
print('predicted data: {}'.format(knn_predicted[:15]))
```

logistic regression



```python
lrc = LogisticRegression(C=10).fit(X_train_scaled, y_train)

pdct =lrc.predict(X_test_scaled)

print('model prediction: {}'.format(pdct))
print('model y test: {}'.format(y_test))
print('test fit score: {}'.format(lrc.score(X_test_scaled, y_test)))
print('train fit score: {}'.format(lrc.score(X_train_scaled, y_train)))
print('confussion metric: {}'.format(confusion_matrix(y_test, pdct)))
print('model prediction accuracy: {}'.format(accuracy_score(y_test, pdct)))
```

```python
#Suppot vector machine
svc = SVC()
svc.fit(X_train_scaled, y_train)
svc_predict = svc.predict(X_test_scaled)


print('model prediction: {}'.format(svc_predict))
print('model y test: {}'.format(y_test))
print('test fit score: {}'.format(svc.score(X_test_scaled, y_test)))
print('train fit score: {}'.format(svc.score(X_train_scaled, y_train)))
print('confussion metric: {}'.format(confusion_matrix(y_test, svc_predict)))
print('model prediction accuracy: {}'.format(accuracy_score(y_test, svc_predict)))
```

```python
dt = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
dt.fit(X_train_scaled, y_train)
dt_predict = dt.predict(X_test_scaled)

print('model prediction: {}'.format(dt_predict))
print('model y test: {}'.format(y_test))
print('test fit score: {}'.format(dt.score(X_test_scaled, y_test)))
print('train fit score: {}'.format(dt.score(X_train_scaled, y_train)))
print('confussion metric: {}'.format(confusion_matrix(y_test, dt_predict)))
print('model prediction accuracy: {}'.format(accuracy_score(y_test, dt_predict)))
```
