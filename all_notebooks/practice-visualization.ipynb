# Lineplot - [Compare Nasdaq and Kospi](https://www.kaggle.com/jeongbinpark/make-lineplot)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np


font = fm.FontProperties(fname='../input/myfont/210 Black.ttf')
nasdaq = pd.read_csv('../input/nasdaq-kospi-compare/nasdaq.csv')
kospi = pd.read_csv('../input/nasdaq-kospi-compare/kospi.csv')

def make_time(time):
  return time.replace('년', '-').replace('월','').replace(' ','')

nasdaq['날짜'] = nasdaq['날짜'].apply(make_time)
kospi['날짜'] = kospi['날짜'].apply(make_time)
nasdaq['날짜'] = pd.to_datetime(nasdaq['날짜'])
kospi['날짜'] = pd.to_datetime(kospi['날짜'])
def make_price(price):
  return price.replace(',', '')

kospi['종가'] = kospi['종가'].apply(make_price)
nasdaq['종가'] = nasdaq['종가'].apply(make_price)
kospi['종가'] = pd.to_numeric(kospi['종가'])
nasdaq['종가'] = pd.to_numeric(nasdaq['종가'])
kospi['종가'] = kospi['종가'] / kospi.loc[120,'종가']
nasdaq['종가'] = nasdaq['종가'] / nasdaq.loc[120,'종가']

fig = plt.figure(figsize=(22,12))
ax = plt.axes()
plt.plot(kospi['날짜'], kospi['종가'],lw=4)
plt.plot(nasdaq['날짜'], nasdaq['종가'],lw=4)
plt.scatter(x=kospi['날짜'][0],y = kospi['종가'][0],s=150)
plt.scatter(x=nasdaq['날짜'][0],y = nasdaq['종가'][0],s=150)
plt.text(s='475%',x=nasdaq['날짜'][0],y = nasdaq['종가'][0]-0.2,font=font, fontsize=22)
plt.text(s='147%',x=kospi['날짜'][0],y = kospi['종가'][0]-0.2,font=font, fontsize=22)
plt.grid(axis="y",alpha=0.5)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xticks(font=font, fontsize=18)
plt.yticks(font=font, fontsize=18)
plt.title("NASDAQ vs KOSPI ( Korean Stock Price Index )",font=font, fontsize=22)
plt.text(s='NASDAQ', x=18100, y=4,font = font, fontsize=18)
plt.text(s='KOSPI', x=18100, y=1.3,font = font, fontsize=18)
ax.set(facecolor = "whitesmoke")
plt.show()
```

# Barplot - [Just BarChart](https://www.kaggle.com/jeongbinpark/practice-barchart)

```python
fontsize=20
plt.rc('font', size=20)
N = 5
ind = np.arange(N)
width = 0.27

fig = plt.figure(figsize=(22,10))
ax = fig.add_subplot(111)

tflite = [26.6, 15.3, 12.2, 89.7, 79.3]
int8 = [7.09, 0, 3.44, 23.3, 21.2]
float16 = [13.4, 7.74, 6.13, 44.9, 39.7]

# yvals = [4, 9, 2]
rects1 = ax.bar(ind, tflite, width, color='grey')
# zvals = [1,2,3]
rects2 = ax.bar(ind+width, float16, width, color='silver')
# kvals = [11,12,13]
rects3 = ax.bar(ind+width*2, int8, width, color='lightgrey')

# ax.set_ylabel('Scores')
ax.set_ylabel('1',fontsize=18)
ax.set_xticks(ind+width)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

ax.set_xticklabels( ('1', '2', '3', '4', '5') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('1', '2', '3') ,fontsize=15,edgecolor ='white')


def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., h, '%d'%int(h),
                ha='center', va='bottom',fontsize=15)
# plt.gca().set_facecolor('lavender')
ax.grid(axis="y",alpha=0.5)
        
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```

# Barplot - MVTEC AD visualization 

```python
fontsize=20
x_label = ['bottle','cable' ,'capsule', 'carpet' ,'grid' ,'hazelnut' ,'leather' ,'metal_nut' ,'pill' ,'screw' ,'tile' ,'transistor' ,'wood' ,'zipper']
diffence = [0.32,0.16,0.25,-0.22,0.72,0.39,0.18,-0.09,0.29,0.42,0.17,0.37,0.35,0.13]
plt.rc('font', size=20)
fig = plt.figure(figsize=(22,10))
color_mode = ['grey','grey','grey','green','grey','grey','grey','green','grey','grey','grey','grey','grey','grey']
plt.bar(x_label, diffence, color=color_mode)
plt.ylabel('OCSVM - Reconstruction')
plt.xticks(rotation=45)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.grid(axis="y",alpha=0.5)
for i in range(len(x_label)):
    score_value = round(diffence[i],2)
    if score_value > 0:
        plt.text(s=f'{score_value}', x=i-0.26, y=diffence[i]+0.03,fontsize=20)
    else:
        plt.text(s=f'{score_value}', x=i-0.35, y=diffence[i]-0.05,fontsize=20)
        
plt.ylim(-0.35,0.8)
plt.show()
```

# Scatterplot - [The relationship between happiness and GDP](https://www.kaggle.com/jeongbinpark/how-to-make-meaningful-visualization)

### It's my favorite !

```python
data = pd.read_csv("../input/world-happiness-report-2021/world-happiness-report-2021.csv")
data = data[['Country name','Ladder score','Logged GDP per capita']]
data = data[:50]
x = np.array(data['Ladder score'])
for i in range(len(x)):
    fit = -0.0747281 *x*x*x + 1.05045076*x*x -3.16337971*x +7.10682006
bound_country = ['Kuwait','Guatemala','Jamaica','Costa Rica','Kosovo','Brazil','El Salvador','Uzbekistan','United Arab Emirates','Singapore','Ireland','Luxembourg']
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=30,color = 'dimgray')
plt.axis('off')

plt.plot(x,fit,lw=8,alpha=0.3)
for i in range(len(data)):
    if data.loc[i,'Country name'] in bound_country:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10,color='lightgrey')
        plt.scatter(x=data.loc[i,'Ladder score'],y=data.loc[i,'Logged GDP per capita'],color ='lightgrey',s=100)
    else:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)

plt.text(s='x: Life Satisfication',x=7.3,y=9.4,fontsize=20)
plt.text(s='y: GDP per Captia',x=7.3,y=9.3,fontsize=20)
        
plt.show()
```
