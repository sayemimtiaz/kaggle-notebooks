```python
def scatter(data, flag, legend, palette):
    for index, i in enumerate(list(data.index.values)):
        if i == 'RISING':
            end_index = index
            
    top = list(data.iloc[2 : end_index].index.values)
    top_count = data.iloc[2 : end_index].values.flatten()
    df = pd.DataFrame({flag : top, 'Interest' : top_count})
    
    plt.figure(figsize = (15, 5))
    sns.scatterplot(data = df, x = flag, y = 'Interest', hue = 'Interest', size = 'Interest', sizes = (200, 1000), legend = False, palette = palette)
    plt.xlabel(flag, fontsize = 20)
    plt.ylabel('', fontsize = 20)
    plt.xticks(rotation = 90, fontsize = 15, fontweight = 'bold')
    plt.yticks(fontsize = 15)
    plt.title(legend, fontsize = 25, fontweight = 'bold')
    plt.show()
```

<h1 id = 'sec13' style = 'background-color:black;color:white;font-family:COMIC SANS MS'> <center> ‚òÅ WORD CLOUDS | ‚úç UNIGRAMS | üìù BIGRAMS </center> </h1>

## **WORD CLOUDS | UNIGRAMS | BIGRAMS**

```python
def preprocess(df):
    tweets = []
    remove = ['bit', 'ly', 'co', 'http', 'com', 'rt', 'via']
    
    for tweet in df['tweet']:
        tweet = re.sub("[^a-zA-Z]", " ", tweet)
        tweet = re.sub(r"http\S+", "", tweet, flags = re.MULTILINE)
        tweet = tweet.lower()
        tweet = nltk.word_tokenize(tweet)
        tweet = [word for word in tweet if not word in set(stopwords.words("english"))]
        tweet = [word for word in tweet if not word in remove]
        lemmatizer = nltk.WordNetLemmatizer()
        tweet = [lemmatizer.lemmatize(word) for word in tweet]
        tweet = " ".join(tweet)
        tweets.append(tweet)
        
    df["processed_tweet"] = tweets
    return df
```

```python
def most_common_words(df, n = 11):
    count_vectorizer = CountVectorizer()
    tf = count_vectorizer.fit_transform(df)
    tf = tf.sum(axis=0) 
    tf = [(term, tf[0, i]) for term, i in count_vectorizer.vocabulary_.items()]
    tf = sorted(tf, key = lambda x: x[1], reverse=True)

    return tf[1 : n]
```

```python
def most_common_bigrams(df, n = 10):
    count_vectorizer = CountVectorizer(ngram_range=(2, 2))
    tf = count_vectorizer.fit_transform(df)
    tf = tf.sum(axis=0)
    tf = [(term, tf[0, i]) for term, i in count_vectorizer.vocabulary_.items()]
    tf = sorted(tf, key = lambda x: x[1], reverse=True)

    return tf[ : n]
```

```python
def pos_tagging(df):
    pos_tags = {'NNP' : 'Noun', 'VBG' : 'Verb', 'VBZ' : 'Verb', 'NNS' : 'Noun', 'VBP' : 'Verb', 'JJ' : 'Adjective', 'NN' : 'Noun', 'VB' : 'Verb', 'JJS' : 'Adjective', 'VBD' : 'Verb', 'RB' : 'Adverb'}
    df['POS'] = [pos_tags[x[1]] for x in nltk.pos_tag(df['Term'])]
    
    return df
```

```python
def generate_word_cloud(df, image_path, color):
    plt.subplots(figsize=(20,6))
    mask = np.array(Image.open(image_path))
    wc = WordCloud(stopwords=STOPWORDS, 
                   mask=mask, background_color="black", contour_color='black',
                   max_words=2000, max_font_size=2048,
                   width=mask.shape[1], height=mask.shape[0],
                   colormap=color)

    wc.generate(' '.join(df))
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

```python
def barplot(df, legend, hue, color):
    plt.figure(figsize = (15, 5))
    sns.barplot(data = df, x = 'Term', y = 'Frequency', hue = hue, palette = color, dodge = False)
    plt.xlabel('Term', fontsize = 20)
    plt.ylabel('Frequency', fontsize = 20)
    plt.xticks(rotation = 90, fontsize = 15, fontweight = 'bold')
    plt.yticks(fontsize = 15)
    plt.legend(loc = 'upper right', ncol = 5, prop={'size': 12})
    plt.title('Top 10 ' + legend + ' Kaggle Tweets', fontsize = 25, fontweight = 'bold')
    plt.show()
```

<h2 id = 'sec26' style = 'background-color:black;color:white;font-family:COMIC SANS MS'> <center> ü§© SENTIMENT ANALYSIS OF KAGGLE TWEETS OVER TIME ‚è≥ </center> </h2>

## **SENTIMENT ANALYSIS OF KAGGLE TWEETS OVER TIME**

```python
sia = SentimentIntensityAnalyzer()
categories = ['Positive', 'Negative', 'Neutral']
```

```python
years = [t_2010, t_2011, t_2012, t_2013, t_2014, t_2015, t_2016, t_2017, t_2018, t_2019, t_2020, t_2021]
colors = ['#0d0887', '#46039f', '#7201a8', '#9c179e', '#bd3786', '#d8576b', '#ed7953', '#fb9f3a', '#fdca26', '#f0f921', '#3cb043', '#3a5311']
for i in range(4):
    fig = go.Figure()
    for j in range(i * 3, (i * 3) + 3):
        neg = sum([sia.polarity_scores(x)['neg'] for x in years[j]]) / len(years[j])
        pos = sum([sia.polarity_scores(x)['neu'] for x in years[j]]) / len(years[j])
        neu = sum([sia.polarity_scores(x)['pos'] for x in years[j]]) / len(years[j])

        fig.add_trace(go.Scatterpolar(
              r = [pos, neg, neu],
              theta = categories,
              fill = 'toself',
              line =  dict(
              color = colors[j]),
              name = str(j + 2010)
        ))

    fig.update_layout(
      title = f'Sentiment Scores from {j - 2 + 2010} to {j + 2010}',
      polar=dict(
      bgcolor='grey',
      radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
    showlegend=True
    )

    fig.show()
```

<h2 id = 'sec27' style = 'background-color:black;color:white;font-family:COMIC SANS MS'> <center> üòä TOP 10 POSITIVE WORDS IN KAGGLE TWEETS OVER TIME ‚è≥ </center> </h2>

## **TOP 10 POSITIVE WORDS IN KAGGLE TWEETS OVER TIME**

```python
df_pos = pd.DataFrame()
annot_pos = pd.DataFrame()
df_neg = pd.DataFrame()
annot_neg = pd.DataFrame()
year_labels = [str(i) for i in range(2010, 2022)]

for index, i in enumerate(years):
    mcw = most_common_words(years[index], -1)
    words_pos = []
    words_neg = []
    pos = []
    neg = []

    for j in mcw:
        score = TextBlob(j[0]).sentiment.polarity
        if score > 0.0:
            pos.append(j[1])
            words_pos.append(j[0])
        elif score < 0.0:
            neg.append(j[1])
            words_neg.append(j[0])
            
    df_pos[year_labels[index]] = pos[ : 10]
    annot_pos[year_labels[index]] = words_pos[ : 10]
    df_neg[year_labels[index]] = neg[ : 10]
    annot_neg[year_labels[index]] = words_neg[ : 10]
    
columns = ['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th', '10th']
df_pos = df_pos.T
annot_pos = annot_pos.T
df_neg = df_neg.T
annot_neg = annot_neg.T
df_pos.columns = columns
annot_pos.columns = columns
df_neg.columns = columns
annot_neg.columns = columns
```

```python
def heatmap(df, annot, legend, color):
    plt.figure(figsize = (50, 50))
    sns.heatmap(df, annot = annot, fmt = '', annot_kws = {'size': 40}, cbar_kws={'label': 'Frequency'}, cmap = color)
    plt.title('Top 10 ' + legend + ' Words in Kaggle Tweets from 2010 to 2021', fontsize = 55, fontweight = 'bold')
    plt.xticks(fontsize = 40)
    plt.yticks(fontsize = 40)
    plt.xlabel('Top 10', fontsize = 50)
    plt.ylabel('Year', fontsize = 50)
    plt.show()
```

```python
heatmap(df_pos, annot_pos, 'Positive', 'flare')
```

<h2 id = 'sec29' style = 'background-color:black;color:white;font-family:COMIC SANS MS'> <center> üìå SUBJECTIVITY ANALYSIS OF KAGGLE TWEETS OVER TIME ‚è≥ </center> </h2>

## **SUBJECTIVITY ANALYSIS OF KAGGLE TWEETS OVER TIME**

```python
sub = []
obj = []
for i in years:
    score = sum([TextBlob(x).sentiment.subjectivity for x in i]) / len(i)
    if score > 0.5:
        sub.append(score)
        obj.append(1.0 - score)
    else:
        obj.append(score)
        sub.append(1.0 - score)
sub_obj = pd.DataFrame({'Year' : year_labels, 'Subjectivity' : sub, 'Objectivity' : obj})
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    y=sub_obj['Year'],
    x=sub_obj['Subjectivity'],
    name='Subjective',
    orientation='h',
    marker=dict(
        color='rgba(246, 78, 139, 0.6)',
        line=dict(color='rgba(246, 78, 139, 1.0)', width=3)
    )
))
fig.add_trace(go.Bar(
    y=sub_obj['Year'],
    x=sub_obj['Objectivity'],
    name='Objective',
    orientation='h',
    marker=dict(
        color='rgba(58, 71, 80, 0.6)',
        line=dict(color='rgba(58, 71, 80, 1.0)', width=3)
    )
))

fig.update_layout(barmode='stack', xaxis=dict(title='Score'), yaxis=dict(title='Year'),  title = 'Subjectivity Score')
fig.show()
```
