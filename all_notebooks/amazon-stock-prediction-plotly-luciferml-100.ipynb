<h1 style="text-align:center;font-weight: bold">About the Dataset</h1>
    <h3 style="text-align:left;font-weight: bold">
This dataset provides the history of daily prices of Amazon stock (AMZN). All the column descriptions are provided. Currency is USD.
    </h3>
<ul>
<li>Date : Date for the data</li>
<li>Open : the price at which a stock started trading when the opening bell rang</li>
<li>High : the highest price at which a stock traded during a period</li>
<li>Low :  the lowest price of the period</li>
<li>Close : the price of an individual stock when the stock exchange closed shop for the day.</li>
<li>Adj Close : is a more complex analysis that uses the closing price as a starting point, but it takes into account factors such as dividends, stock splits and new stock offerings to determine a value</li>
<li>Volume : he total number of shares that are actually traded (bought and sold) during the trading day or specified set period of time.</li>
</ul>

```python
features = dataset.drop(['Date', 'Close'], axis = 1)
labels= dataset['Close']
```

- Labels : Close

```python
params = {'fit_intercept': True, 'positive': True}
regressor = Regression(predictor = 'lin', params = params)
regressor.fit(features, labels)
result = regressor.result()
accuracy_scores[result['Regressor']] = result['Accuracy']
RMSE[result['Regressor']] = result['RMSE']
```
