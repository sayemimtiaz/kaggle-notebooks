```python
# Utility Functions

def count_df(df, feat):
    """
    Converts value counts to a neat dataframe
    
    This function creates a dataframe with 2 features
        > The first feature has each of the unique values
        in `feat`
        > The second feature has the corresponding count
        
    Parameters
    ----------
    df : pandas dataframe
        Dataframe
    feat : string
        Feature under consideration
        
    Returns
    -------
    df
        Dataframe
    """
    
    temp = pd.DataFrame(df[feat].value_counts().sort_index())
    temp.columns = ['count'] # `count` is like a keyword name for the feature here
    temp[feat] = temp.index
    
    return temp

def grouper(df, feat_list, uid, new_names=None):
    """
    Simply groups
    
    This function groups entries in`df` as per
    `feat_list` and counts as per `uid`
        
    Parameters
    ----------
    df : pandas dataframe
        Dataframe
    feat_list : list
        Features under consideration
    uid : string
        Identifier feature
    new_names : list
        List of new column names(opt)
        
    Returns
    -------
    df
        Dataframe
    """
    
    temp = df.groupby(feat_list, as_index=False).agg({uid: 'count'})
    if(new_names != None):
        temp.columns = new_names
    
    return temp


def count_plotter(df, feat, plt_title):
    """
    Plots count of unique values.
    
    This function helps plot the count of each value
    present in `feat` in dataframe `df`
    
    Parameters
    ----------
    df : pandas dataframe
        Dataframe
    feat : string
        Feature to plot for
        
    Returns
    -------
    Plot
        A line plot visualization
        
    Best Usage
    ----------
    1. For time series 
    """
    
    temp = count_df(df, feat)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=temp[feat], y=temp['count'],
                            line=dict(color='#f8766d')))
    #fig = px.line(temp, x=feat, y='count')
    
     # template enhancement
    fig.update_layout(
        template="plotly_dark",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 25},
            "yanchor": "top",
        },
        xaxis_title = feat,
        yaxis_title = "Count",
    )
    
    return fig.show()

def trim(x):
    """Trim whitespace from ends"""
    return x.strip()

def plot_bar_cplx(df, idf, feat, plt_title="You forgot the title", hor=False, how_many=5):
    """Plot a bar plot of values in a given feature
    Stands for `plot bar complex`
    """
    
    if(hor==False):
        df = df.groupby([feat], as_index=False).agg({idf: 'count'}).sort_values(by=idf, ascending=False).head(how_many)
        colors = ['#f8766d',] * df.shape[0]
        colors[0] = 'red'

        fig = go.Figure([go.Bar(x=df[feat], y=df[idf], marker_color=colors)])
    else:
        df = df.groupby([feat], as_index=False).agg({idf: 'count'}).sort_values(by=idf, ascending=True).tail(how_many)
        colors = ['#f8766d',] * df.shape[0]
        colors[(df.shape[0]-1)] = 'red'

        
    fig = go.Figure([go.Bar(y=df[feat], x=df[idf], orientation='h', marker_color=colors)])
    
    # template enhancement
    fig.update_layout(
        template="plotly_dark",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 25},
            "yanchor": "top",
        },
    )

    # show plot
    fig.show()
    
    
def plot_bar(df, idf, feat, plt_title="You forgot the title"):
    """Plot a bar plot of values in a given feature"""
    
    df = df.groupby([feat], as_index=False).agg({idf: 'count'}).sort_values(by=idf, ascending=False)
    colors = ['#f8766d',] * df.shape[0]
    colors[0] = colors[1] = colors[2] = 'red'
    
    fig = go.Figure([go.Bar(x=df[feat], y=df[idf], marker_color=colors)])
    
    # template enhancement
    fig.update_layout(
        template="plotly_dark",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 25},
            "yanchor": "top",
        },
        xaxis_title = "State",
        yaxis_title = "Count of people",
    )

    # show plot
    fig.show()
    
def spot_problems(df, feat, check_string):
    """
    Helps identify the `check_string` string inside
    the values in `df[feat]`
    -----
    
    For example, in the `fat_enc` feature in this notebook,
    problematic values in `Student's age` were {months, mon, 
    'days', 's', '-', '/', 'or'}
    """
    for f in df[feat]:
        try:
            if (check_string in f):
                print(f+'hello') # the hello helps look out for trailing spaces
        except:
            continue
```

<br><br>

Yes, this is an unconventional way to initiate a discussion on a policing issue. However, it is a known fact that a significant amount of deaths due to citizen-police encounters see an involvement of a firearm. Now, there are two ways to look at this - *Either the police use their firearms indiscriminately or everybody has easy access to firearms and therefore it becomes difficult for the cops to avoid theirs when encountering difficult situations*.  

**The former is an act against social responsibility, while the latter is a forced error. The former might be tackled with stringent policies and sanctions, but the latter will require a nationwide conscientious effort.**

```python
# Comparing gunshot related and unrelated deaths

temp = fat_enc.groupby(['Cause of death'], as_index=False).agg({'Unique ID': 'count'}).sort_values(by='Unique ID', ascending=False)
gunshot_related = temp[temp['Cause of death']=='Gunshot'].iloc[0]['Unique ID']
gunshot_unrelated = temp[temp['Cause of death']!='Gunshot'].iloc[0]['Unique ID']

# define variables
colors = ['#ff0000', '#f8766d']
plt_title = 'Fatal Encounters in the USA (2000-2020)'

# plot the figure
fig = go.Figure(data=[go.Pie(
    labels=['Involved a gun', 'Did not involve a gun'],
    values=[gunshot_related, gunshot_unrelated],
)])

#update figure
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                 marker=dict(colors=colors, line=dict(color="#000000", width=1)))

# template enhancement
fig.update_layout(
    template="plotly_dark",
    title={
        "text": plt_title,
        "font": {"family": "Rockwell", "size": 30},
        "yanchor": "top",
    }
)

# show plot
fig.show()
```

The figure above provides a high level understanding of the extent to which guns in the American society prove harmful to life. Of the over 25,000 deaths that have resulted from citizen-police interactions, 78% of these have seen an involvement of a gun. This is rather intuitive to understand because the dataset does account for only deaths due to interactions with the police(who do have access to firearms).  

However, [this report](https://www.pewresearch.org/fact-tank/2019/08/16/what-the-data-says-about-gun-deaths-in-the-u-s/) by the Pew Research Centre published in 2019 provides a very disturbing scenario of gun-related deaths. The report also suggests an increase in mass shooting incidents over the past decade or so([Source](https://www.pewresearch.org/fact-tank/2019/08/16/what-the-data-says-about-gun-deaths-in-the-u-s/ft_19-08-14_gundeaths_4/)).  

Therefore, it does seem as if gun-related violence is too easy to come by(based on the data provided). <div class="subsection">Are shootings affected by the racial status of people?</div>
<br><br>

In recent times, there have been several discussions on the media about police violence and the detrimental effects it has on people, especially the minorities. While most demonstrations against the forces are based on the grounds of racism(such as [#BlackLivesMatter](https://en.wikipedia.org/wiki/Black_Lives_Matter)), it is also important to analyze if the police shootings are driven by racism alone or if there are other factors.  

For example, it was noticed above how the majority of those shot by the officers were in posession of a firearm. In such situations, officers might have very little choice about what they should do. Such cases necessarily need not be catalysed by racism. However, killing an unarmed person even though there was little threat from the situation does indicate a possible bias.  

```python
def race_shooting_bias(swp, plt_title='You forgot the title'):
    """Plot race, shooting and bias plots for the given dataframe
    
    The `swp` parameter has to either be the shootings_wash_post.csv
    or a subset of it 
    """

    # convert abbreviations to clearer forms
    swp['race'] = swp['race'].replace({
        'A': 'Asian',
        'B': 'Black, non-Hispanic',
        'W': 'White, non-Hispanic',
        'N': 'Native American',
        'H': 'Hispanic',
        'O': 'Other',    
    })

    # actually dangerous proportion shot

    total_cnt = swp.groupby(['race'], as_index=False).agg({'id': 'count'})
    total_cnt.columns = ['race', 'total']

    armed = swp[(swp['armed']!='unarmed')]
    armed_cnt = armed.groupby(['race'], as_index=False).agg({'id': 'count'})
    armed_cnt.columns = ['race', 'armed_with_something']
    ar_prop = total_cnt.merge(right=armed_cnt)
    ar_prop['prop_armed'] = round((ar_prop['armed_with_something'] / ar_prop['total']) * 100, 2)

    g_armed = swp[(swp['armed']=='gun') | (swp['armed']=='knife')]
    g_armed_cnt = g_armed.groupby(['race'], as_index=False).agg({'id': 'count'})
    g_armed_cnt.columns = ['race', 'armed_with_gun']
    g_prop = ar_prop.merge(right=g_armed_cnt)
    g_prop['gun_armed'] = round((g_prop['armed_with_gun'] / g_prop['total']) * 100, 2)

    t_armed = swp[(swp['armed']=='toy weapon')]
    t_armed_cnt = t_armed.groupby(['race'], as_index=False).agg({'id': 'count'})
    t_armed_cnt.columns = ['race', 'armed_with_toy_weapon']
    t_prop = g_prop.merge(right=t_armed_cnt)
    t_prop['toy_weapon_armed'] = round((t_prop['armed_with_toy_weapon'] / t_prop['total']) * 100, 2)

    unarmed = swp[(swp['armed']=='unarmed')]
    unarmed_cnt = unarmed.groupby(['race'], as_index=False).agg({'id': 'count'})
    unarmed_cnt.columns = ['race', 'unarmed']
    prop = t_prop.merge(right=unarmed_cnt)
    prop['prop_unarmed'] = round((prop['unarmed'] / prop['total']) * 100, 2)

    # plot
    fig = go.Figure()
    fig.add_trace(go.Bar(x=prop['race'], y=prop['prop_unarmed'],
                            name='Proportion unarmed(no weapon at all)', marker_color="#1261a0"))
    fig.add_trace(go.Bar(x=prop['race'], y=prop['gun_armed'],
                            name='Proportion armed with a gun or knife', marker_color="#ff0000"))
    fig.add_trace(go.Bar(x=prop['race'], y=prop['toy_weapon_armed'],
                            name='Proportion carrying only a toy weapon', marker_color="#ffbf00"))

    # template enhancement
    fig.update_layout(
        template="plotly_dark",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 25},
            "yanchor": "top",
        },
        barmode='group',
        yaxis_title = "Percentage of people",
    )
    fig.show()
```

```python
race_shooting_bias(swp, plt_title='Race, shootings and bias in all states (2015-2020)')
```

ðŸ“ **NOTE**  
The above figure depicts only the most prevalent modes of being armed - Gun, knife, toy weapon or unarmed

ðŸ’¡ **INSIGHTS**

* The proportions of those armed with a gun or a knife at the time of their death at the hands of an on-duty officer are significantly high for all the races
    * It is the lowest for the Hispanic community at 66.7%, followed by the Asians at 67.39%
    * It is the highest for the Native Americans at 74.36%, followed by the White community at 72.41%
    * To put this more clearly, out of every 100 people shot by the cops from community {X}, {Y} people were armed at the time of the conflict, where
        * X is Asian, Y is 67
        * X is Hispanic, Y is 67
        * X is Black, Y is 70
        * X is White, Y is 72
        * X is Native American, Y is 74
* Inspite of the fact that shot white people tend to be armed more dangerously than shot black people(from above), the percentage of black people shot who were unarmed is almost twice the same metric for the whites.
    * Of 100 black people shot, there is a chance that 10 of them were unarmed at the time of the incident
    * Of 100 white people shot, there is a chance that 6 of them were unarmed at the time of the incident
* It is also to be noted that for every 100 white persons shot, there is a chance that 4 of them were only carrying a toy
    * The same metric for a black person is 2.74 i.e 3 in 100 black persons shot might have been carrying only a toy
    
This is a dichotomy. While a shot black person was less likely to be armed dangerously when compared to a white person, a white person carrying a toy weapon was more susceptible to be shot than a black person carrying one. **Though the first part of this statement indicates racial bias, the second part indicates misjudgement due to fear which is not necessarily racial.**

```python
# people_police_death ratios (for conflicts)

# people deaths
fes = grouper(fat_enc, ['Date (Year)', 'Location of death (state)'], 'Unique ID', ['year', 'state', 'people_deaths'])
fes = fes[(fes['year']>=2000) & (fes['year']<=2016)]
fes['state'] = fes['state'].apply(trim) # strip whitespace

# police deaths
pds = grouper(po_de_conflict, ['year', 'state'], 'person', ['year', 'state', 'police_deaths'])
pds = pds[(pds['year']>=2000) & (pds['year']<=2016)].reset_index(drop=True)
pds['state'] = pds['state'].apply(trim) # strip whitespace

# making the dataframe
people_police_deaths = pd.merge(
    fes,
    pds,
    on = ['year', 'state'],
)
people_police_deaths['ratio'] = people_police_deaths['people_deaths'] / people_police_deaths['police_deaths']
```

```python
states_deaths_police = people_police_deaths.groupby(['state'], as_index=False).\
    agg({'police_deaths': 'sum'}).\
    sort_values(by='police_deaths', ascending=False).\
    head(10)

states_deaths_people = people_police_deaths.groupby(['state'], as_index=False).\
    agg({'people_deaths': 'sum'}).\
    sort_values(by='people_deaths', ascending=False).\
    head(10)

colors = ['#ff0000',] * 10

fig = make_subplots(
    rows=1, cols=2, subplot_titles=("States with highest officer deaths", "States with highest non-officer deaths")
)

fig.add_trace(go.Bar(
    x=states_deaths_police['state'],
    y=states_deaths_police['police_deaths'],
    marker_color=colors,),
    row=1, col=1)
fig.add_trace(go.Bar(
    x=states_deaths_people['state'],
    y=states_deaths_people['people_deaths'],
    marker_color=colors,),
    row=1, col=2)

fig.update_layout(showlegend=False,
                  title={
                        "text": "Comparing officer and non-officer deaths(2000-2016)",
                        "font": {"family": "Rockwell", "size": 25},
                        "yanchor": "top",
                 },
                 template='plotly_dark')
# fig = px.bar(states_cop_deaths, x='state', y='police_deaths')
fig.show()
```

ðŸ’¡ **INSIGHTS**

*  Most police officers have been killed(in a fatal citizen-police encounter) in Texas, followed by New York and California
* Out of the 10 states with highest police officer deaths, 7 of these are also present in the list of 10 states with the highest number of non-officer deaths
