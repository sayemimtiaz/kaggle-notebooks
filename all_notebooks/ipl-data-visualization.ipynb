```python
delivery=pd.read_csv('../input/indian-premier-league-ipl-dataset/ipl_matches.csv')
delivery.head(3)
```

**Most runs scored in IPL from 2019-2021**

Below you can see the top 15 players with most runs.

```python
import matplotlib.pyplot as mlt
import seaborn as sns
mlt.subplots(figsize=(12,3))
max_runs=delivery.groupby(['striker'])['runs_off_bat'].sum()
ax=max_runs.sort_values(ascending=False)[:15].plot.bar(width=0.8,color=sns.color_palette('Blues_r',20))
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+50),fontsize=15)
mlt.show()
```

**Most successful batsmen against Bumrah in recent time**

Below you can see top 5 players who has scored the most against Bumrah in recent seasons

```python
mlt.subplots(figsize=(10,5))
bowler_name = ["JJ Bumrah"]
ct=delivery[delivery["bowler"].isin(bowler_name)]
cy=ct.groupby(['striker'])['runs_off_bat'].sum()
ax=cy.sort_values(ascending=False)[:5].plot.bar(width=0.8,color=sns.color_palette('Spectral',20))
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.10, p.get_height()),fontsize=15)
mlt.show()
```
