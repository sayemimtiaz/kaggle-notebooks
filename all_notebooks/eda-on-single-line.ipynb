# Introduction
In statistics, exploratory data analysis is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. EDA mainly focus on checking assumptions required for model fitting and hypothesis testing, and handling missing values and making transformations of variables as needed. There are lots of libraries available which made EDA easy and fast, but they require lots of code. Here, I am going to show how can we do EDA with single line of code. 



# 1. Panda profiling
Pandas profiling is an open source Python module with which we can quickly do an exploratory data analysis with just a few lines of code. Besides this it also generates interactive plot and reports in web format
 pandas profiling  save us all the work of visualizing and understanding the distribution of each variable. It generates a report with all the information easily available.

## Load some important libraries 

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
```

## Let's load our data
Here, I am going to perform analysis in very famous Titanic dataset

```python
df = pd.read_csv('../input/titanic-machine-learning-from-disaster/train.csv')
# let's see our data
df.head(4)
```

## let install panda_profiling first

```python
#pip install pandas-profiling
```

## Let's import panda_profiling 

```python
import pandas_profiling as pp
```

## Let's see the magic !!

```python
prof = pp.ProfileReport(df)
```

```python
prof
```

### The main advantage of using Pandas Profiling is that it can generate interactive reports in web format that can be presented to any person who doesn't know programming . So let's save our report in html file.

```python
prof.to_file(output_file = 'plot.html')
```

# 2.SweetViz
Sweetviz is an open source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with just two lines of code. Output is a fully self-contained HTML application.


```python
#pip install sweetviz
```

## Analyzing the dataframe
to analyze the dataframe we simply use analyze() function. Similarly, we use show_html() function to show the analysis

## Let's see the magic

```python
import sweetviz as sv

analyzed_report = sv.analyze(df)
analyzed_report.show_html('SWEETVIZ_REPORT.html') # to save and show the analyzed report
```

## To see the analyzed report

```python


from IPython.display import IFrame
IFrame(src = 'SWEETVIZ_REPORT.html',width=1000,height=600)

```

## We can compare train and test data using sweet viz. So let' try.....

```python
train_data = pd.read_csv("../input/titanic-machine-learning-from-disaster/train.csv")
test_data = pd.read_csv("../input/titanic-machine-learning-from-disaster/test.csv")

comparison_report = sv.compare(train_data,test_data)
comparison_report.show_html('Comparison.html')
```

## to visualize comparision report.....

```python
IFrame(src = 'Comparison.html',width=1000,height=500)
```
