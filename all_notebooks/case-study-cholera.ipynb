# Cholera

<hr>

**Cholera** is a disease caused by bacteria that produce a watery **diarrhea** that can rapidly lead to **dehydration**.

Cholera is an **acute diarrhoeal disease** that can kill within hours if left untreated.

A global strategy on cholera control with a target to reduce cholera deaths by 90% was launched in 2017.

Safe oral cholera vaccines should be used in conjunction with improvements in water and sanitation to control cholera outbreaks and for prevention in areas known to be high risk for cholera.

## Overview:
Cholera is an acute diarrhoeal infection caused by eating or drinking food or water that is contaminated with the bacterium Vibrio cholerae. Cholera remains a global threat to public health and is an indicator of inequity and lack of social development. Researchers have estimated that every year, there are 1.3 to 4.0 million cases of cholera, and 21 000 to 143 000 deaths worldwide due to the infection.
    
Cholera, caused by the bacteria Vibrio cholerae, is rare in the United States and other industrialized nations. However, globally, cholera cases have increased steadily since 2005 and the disease still occurs in many places including Africa, Southeast Asia, and Haiti. CDC responds to cholera outbreaks across the world using its Global Water, Sanitation and Hygiene (WASH) expertise.

Cholera can be life-threatening but it is easily prevented and treated. Travelers, public health and medical professionals and outbreak responders should be aware of areas with high rates of cholera, know how the disease spreads, and what to do to prevent it.

## Illness and Symptoms (as per **CDC**):
Approximately one in ten (5-10%) of infected persons will have severe cholera which in the early stages includes:
* profuse watery diarrhea, sometimes described as ‚Äúrice-water stools,‚Äù
* vomiting
* rapid heart rate
* loss of skin elasticity
* dry mucous membranes
* low blood pressure
* thirst
* muscle cramps
* restlessness or irritability

## Prevention:
   
* Most important prevention from **Cholera** is **Clean Water** and **Proper Sanitation**.

* **Vacciantion** of Oral Cholera is other most effected prevention, according to **WHO**.
    
* Basic Hygienic Behaviours like **Washing Hand** with **Soap** frequently and mostly before and after food.
    
* Increasing your Immune Stanima also best for fighting Cholera (**YOGA** and **Meditation** are great way to have it)
    
## Vaccines:
According to **WHO**, in year 2013 **OCV(Oral cholera vaccines)** creation took place and since than more than **50 Million** doses of these were given under different campaings.

# Some Preprocessing 
<hr>

Is Time for some **Data Preprocessing**

```python
# A Simple Column Names
chl_data = chl_data.rename(columns={"Number of reported cases of cholera":"Cases",
                                   "Number of reported deaths from cholera":"Deaths",
                                   "Cholera case fatality rate":"Fatality Rate"})
```

```python
# Count of Null Data
chl_data.isnull().sum()
```

```python
# Filling NULL with '0'
chl_data['Cases'] = chl_data['Cases'].fillna('0')
chl_data['Deaths'] = chl_data['Deaths'].fillna('0')
chl_data['Fatality Rate'] = chl_data['Fatality Rate'].fillna('0')

# Replacing 'Unknown' Values with '0'
chl_data['Cases'] = chl_data['Cases'].replace('Unknown', '0')
chl_data['Deaths'] = chl_data['Deaths'].replace('Unknown', '0')
chl_data['Fatality Rate'] = chl_data['Fatality Rate'].replace('Unknown', '0')

# Removing WhiteSpaces 
chl_data['Cases'] = chl_data['Cases'].str.replace(" ","")
chl_data['Deaths'] = chl_data['Deaths'].str.replace(" ","")
chl_data['Fatality Rate'] = chl_data['Fatality Rate'].str.replace(" ","")
```

```python
# Info about Dataframe
chl_data.info()
```

```python
# Converting 'Cases' & 'Deaths' Dtype from Object to int
chl_data['Cases'] = chl_data['Cases'].astype(int)
chl_data['Deaths'] = chl_data['Deaths'].astype(int)
```

```python
# Creating Dataframe w.r.t. Countries
a = ['Country']
b = ['Cases','Deaths']
country_wise = chl_data.groupby(a)[b].sum()

# Creating Dataframe w.r.t. Regions
a = ['WHO Region']
b = ['Cases','Deaths']
region_wise = chl_data.groupby(a)[b].sum()
```

```python
# Creating New columns(Fatality Rate, Recovery, and Recvery Rate) for **Country DataFrame**
country_wise["Fatality Rate"] = np.round(100*country_wise["Deaths"]/country_wise["Cases"],2)
country_wise["Recovery"] = country_wise["Cases"] - country_wise["Deaths"]
country_wise["Recovery Rate"] = np.round(100*(country_wise["Cases"] - country_wise["Deaths"])/country_wise["Cases"],2)
country_wise = country_wise[[ 'Cases', 'Recovery', 'Deaths', 'Recovery Rate', 'Fatality Rate']]

# Creating New columns(Fatality Rate, Recovery, and Recvery Rate) for **Region DataFrame**
region_wise["Fatality Rate"] = np.round(100*region_wise["Deaths"]/region_wise["Cases"],2)
region_wise["Recovery"] = region_wise["Cases"] - region_wise["Deaths"]
region_wise["Recovery Rate"] = np.round(100*(region_wise["Cases"] - region_wise["Deaths"])/region_wise["Cases"],2)
region_wise = region_wise[[ 'Cases', 'Recovery', 'Deaths', 'Recovery Rate', 'Fatality Rate']]

# Sorting both DataFrames w.r.t **Cases**
country_wise = country_wise.sort_values(by=['Cases'], ascending=False)
region_wise = region_wise.sort_values(by=['Cases'], ascending=False)
```

It's better to find out **Top 10** Countries in terms of Numbers.

# Time Graph
<hr>
**Time Graphs** are best to understand the growth or decline in a long period.

Here, I used **'Year'** column for that.

## Graphs for Top 10 Countries w.r.t. Cases
<hr>

```python
# Listing the Countries Names
country_list = country_wise.sort_values('Cases')["Country"].values[-10:]

# Ploting with Some Preprocessing
for country_name in country_list:
    df = chl_data[chl_data['Country'] == country_name]
    df = df.sort_values('Year')
    x = df['Cases']
    y = df['Deaths']
    x.reset_index(drop=True)
    case = 0
    death = 0
    cases = []
    deaths = []
    for i in range(len(x)):
        case = case + x.iloc[i]
        death = death + y.iloc[i]
        cases.append(case)
        deaths.append(death)
    df['Total Cases'] = cases
    df['Total Deaths'] = deaths
    max_total_cases_yr = df['Year'].max()
    max_total_cases = float(df[df['Year'] == max_total_cases_yr]['Total Cases'])
    max_total_deaths = float(df[df['Year'] == max_total_cases_yr]['Total Deaths'])
    if df['Cases'].sum() >= 0 and df.shape[0] > 5:
        fig, ax = plt.subplots()
        df.plot(x='Year', y='Total Cases', ax=ax)
        df.plot(x='Year', y='Total Deaths', ax=ax)
        plt.scatter(x= [max_total_cases_yr], y=[max_total_cases], s=70, clip_on=False, linewidth=0)
        plt.annotate(int(max_total_cases), xy=[max_total_cases_yr, max_total_cases], xytext=[7, -2], textcoords='offset points')
        plt.scatter(x= [max_total_cases_yr], y=[max_total_deaths], s=70, clip_on=False, linewidth=0)
        plt.annotate(int(max_total_deaths), xy=[max_total_cases_yr, max_total_deaths], xytext=[7, -2], textcoords='offset points')
        ax.set_title(country_name)
```

## Graphs for WHO Regions w.r.t. Cases
<hr>

```python
# Listing WHO Regions Names
region_wise = region_wise.reset_index()
region_list = region_wise.sort_values('Deaths')["WHO Region"]

df_region = chl_data.groupby(['WHO Region', 'Year']).sum()
df_region = df_region.reset_index()

for region_name in region_list:
    df = df_region[df_region['WHO Region'] == region_name]
    df = df.sort_values('Year')
    x = df['Cases']
    y = df['Deaths']
    x.reset_index(drop=True)
    case = 0
    death = 0
    cases = []
    deaths = []
    for i in range(len(x)):
        case = case + x.iloc[i]
        death = death + y.iloc[i]
        cases.append(case)
        deaths.append(death)
    df['Total Cases'] = cases
    df['Total Deaths'] = deaths
    max_total_cases_yr = df['Year'].max()
    max_total_cases = float(df[df['Year'] == max_total_cases_yr]['Total Cases'])
    max_total_deaths = float(df[df['Year'] == max_total_cases_yr]['Total Deaths'])
    if df['Cases'].sum() >= 0:
        fig, ax = plt.subplots()
        df.plot(x='Year', y='Total Cases', ax=ax)
        df.plot(x='Year', y='Total Deaths', ax=ax)
        plt.scatter(x= [max_total_cases_yr], y=[max_total_cases], s=70, clip_on=False, linewidth=0)
        plt.annotate(int(max_total_cases), xy=[max_total_cases_yr, max_total_cases], xytext=[7, -2], textcoords='offset points')
        plt.scatter(x= [max_total_cases_yr], y=[max_total_deaths], s=70, clip_on=False, linewidth=0)
        plt.annotate(int(max_total_deaths), xy=[max_total_cases_yr, max_total_deaths], xytext=[7, -2], textcoords='offset points')
        ax.set_title(region_name)
```
