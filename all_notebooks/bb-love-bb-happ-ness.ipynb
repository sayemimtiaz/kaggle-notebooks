```python
f,ax = plt.subplots(figsize=(12,12), dpi=100)
sns.heatmap(happy15.corr(), annot=True, linewidth=1, fmt=" .1F", linecolor="#E8DAEF", ax=ax)
plt.show()
```

<h3><b>First of all, the reason I created a correlation graph is to see the relationship between the features that I can compare more easily. As seen above, there are considerably interrelated issues such as 0.7 and 0.8. Since the aim of our topic is happiness, if we look at the features related to happiness, we can see that economy, family, health and freedom are in a right proportion with happiness.</b></h3>

```python
plt.style.use("seaborn")
plt.figure(figsize=(10,5), dpi=100)

x = happy15[["Economy (GDP per Capita)"]]
y = happy15[["Happiness Score"]]

plt.plot(x, label="Economy", color="#28B463", linestyle="-", marker=".", linewidth=1, markeredgecolor="#3498DB")
plt.plot(y, label="Happiness", color="#3498DB", linestyle="-", marker=".",linewidth=1)

plt.legend()
plt.show()
```

<h3><b>We need to graph the relationships we derive from the heatmap above. The first graphic I will draw is the relationship between Economy and Happiness on a line chart. As can be seen, despite the deviations in the economy, the line of economy and the line of happiness are almost parallel to each other. This means that one's own economy affects one's own happiness too much.</b></h3>

```python
plt.style.use("seaborn")
plt.figure(figsize=(8,4), dpi=150)
plt.scatter(happy15["Health (Life Expectancy)"], happy15["Happiness Score"], alpha= 0.5, color="r", marker="s")
plt.title("Health - Happiness Relations", size=20, color="#C0392B")
plt.xlabel("Health", size=15, color="#F39C12")
plt.ylabel("Happiness", size=15, color="#1ABC9C")

plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=100)
x = happy15[["Happiness Score"]]

plt.hist(x, orientation='horizontal', bins=15, color="#154360")
plt.xlabel("Happiness Scores", fontsize=15, color="#B3B6B7")
plt.title("Histogram of Happiness Scores", fontsize=20, color="#0E6655")

plt.show()
```
