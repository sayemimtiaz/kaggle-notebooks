```python
# Please give me a second to import all the libraries I need (or might need) üôÉ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Importing Completed")
```

Reading the csv file

```python
whr_2015 = pd.read_csv("../input/world-happiness/2015.csv", index_col = 'Happiness Rank')
print("Read")
```

Here's a sneak peek of the World Happiness Report data of 2015

```python
whr_2015
```

Is there any correlation between Happiness Score and Per Capita GDP?
I think that there should be a good correlation between them, let's find out...

```python
whr_2015['Economy (GDP per Capita)'].corr(whr_2015['Happiness Score'])
```

There's a good correlation between them!! Better to visualise this, including the Regional Distribution

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'Economy (GDP per Capita)', y = 'Happiness Score', hue = 'Region', data = whr_2015)
```

We can observe that the Sub-Saharan Countries lie in the bottom-left part of the graph.<br>
Is there any correlation between Happiness Score and Family?

```python
whr_2015['Family'].corr(whr_2015['Happiness Score'])
```

Here is also quite a bit of correlation. Plotting for better understanding...


```python
plt.figure(figsize = (20, 12))
sns.lmplot(x = 'Family', y = 'Happiness Score', hue = 'Region', data = whr_2015)
```

Checking the distribution of Happiness Score in different regions...

```python
plt.figure(figsize = (30, 8))
sns.swarmplot(x = 'Region', y = 'Happiness Score', data = whr_2015)

print('\nMean Happiness Score of different regions are as follows - ')
for i in pd.unique(whr_2015['Region']):
    gdp = whr_2015['Happiness Score'].where(whr_2015['Region'] == i).mean()
    print('{} - {}'.format(i, gdp))
```

Australia and New Zealand, North American, and Western Europian countries seem quite happy, and Sub-Saharan African countries seem quite unhappy ‚òπÔ∏è
<br>But why? Is that because of GDP? Let's find out that too...

```python
mean_gdp = whr_2015['Economy (GDP per Capita)'].mean()
mean_gdp
```

So mean GDP of the whole world was 0.8461372151898726, and mean GDP of Sub-Saharan African countries is - 


```python
mean_gdp_ssa = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Sub-Saharan Africa').mean()
mean_gdp_ssa
```

Well, thats... quite... low... let's see the distribution of GDP in Sub-Saharan Africa - 

```python
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Sub-Saharan Africa'), shade = True)
```

Now I am curious to see the distribution of other regions, plotting for other regions too üòÅ

```python
plt.figure(figsize = (20,15))
plt.subplot(3, 3, 1)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Western Europe'), shade = True, label = 'Western Europe')
plt.legend()
plt.subplot(3, 3, 2)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'North America'), shade = True, label = 'Western Europe')
plt.legend()
plt.subplot(3, 3, 3)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Australia and New Zealand'), shade = True, label = 'Australia and New Zealand')
plt.legend()
plt.subplot(3, 3, 4)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Middle East and Northern Africa'), shade = True, label = 'Middle East and Northern Africa')
plt.legend()
plt.subplot(3, 3, 5)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Latin America and Caribbean'), shade = True, label = 'Latin America and Caribbean')
plt.legend()
plt.subplot(3, 3, 6)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Southern Asia'), shade = True, label = 'Southern Asia')
plt.legend()
plt.subplot(3, 3, 7)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Central and Eastern Europe'), shade = True, label = 'Central and Eastern Europe')
plt.legend()
plt.subplot(3, 3, 8)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Eastern Asia'), shade = True, label = 'Eastern Asia')
plt.legend()
plt.subplot(3, 3, 9)
sns.kdeplot(data = whr_2015['Economy (GDP per Capita)'].where(whr_2015['Region'] == 'Southern Asia'), shade = True, label = 'Southern Asia')
plt.legend()
```
