<font color="sienna" size=+2>4.5 How much words do I make per comment</font>

```python
def remove_tag(string):
    text=re.sub('<.*?>','',string)
    return text

def remove_newline(string):
    text=re.sub('\n','',string)
    return text

def remove_url(string): 
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','',string)
    return text

def remove_number(text):
    line=re.sub(r'[0-9]+','',text)
    return line

def find_at(text):
    line=re.findall(r'(?<=@)\w+',text)
    return " ".join(line)

def remove_punct(text):
    line = re.sub(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*','',text)
    #string="".join(line)
    return line


my_discussion_list['comment']=my_discussion_list['Message'].apply(lambda x:remove_tag(str(x)))
my_discussion_list['comment']=my_discussion_list['comment'].apply(lambda x:remove_newline(x))
my_discussion_list['comment']=my_discussion_list['comment'].apply(lambda x:remove_url(x))
my_discussion_list['comment']=my_discussion_list['comment'].apply(lambda x:remove_number(x))
my_discussion_list['mentions']=my_discussion_list['comment'].apply(lambda x:find_at(x))
my_discussion_list['comment']=my_discussion_list['comment'].apply(lambda x:remove_punct(x))
my_discussion_list['text_length']=my_discussion_list['comment'].str.split().map(lambda x: len(x))
```

```python
if(len(my_discussion_list)>0):
    
    fig = go.Figure(go.Indicator(
        mode = "number",
        value = my_discussion_list['text_length'].mean(),
        title = {"text": "Average Length of Comments",'font': {'size':26}},
        number={'font':{'color': 'sienna','size':50}},
        domain = {'y': [0.50, 1], 'x': [0.50, 0.50]}))

    fig.add_trace(go.Histogram(x=my_discussion_list['text_length'],marker_color='sienna',name="Comment Length",
        opacity=1))


    fig.update_layout(title='Length of comments ',
                      xaxis_title="Length of Comments ",
                      yaxis_title="Number of Comments ",title_x=0.5)
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig.show()
else:
    print('You have zero Discussion')
```
