**Importing the Libraries**

```python
import numpy as np
import folium as fo
import pandas as pd
%matplotlib inline
```

***Creating the world map***

```python
map=fo.Map()
```

```python
map
```

Adding markers on the map

*Markers represent particular location on the map*

```python
x = fo.FeatureGroup(name='My Map')
```

Testing the marker on Chennai

```python
x.add_child(fo.Marker(location=[13.0827,80.2707],popup='Chennai',icon=fo.Icon(color='purple')))
```

```python
map.add_child(x)
```

**Data Processing**

```python
volcano=pd.read_csv('../input/volcano/Volcano.csv')
```

```python
lat_vo = list(volcano['Latitude'])
lon_vo = list(volcano['Longitude'])
name_vo = list(volcano['Name'])
```

```python
vol = fo.FeatureGroup(name='My Map')
```

```python
for lat,lon,name in zip(lat_vo,lon_vo,name_vo):
    vol.add_child(fo.Marker(location=[lat,lon] ,
                          popup=name , icon=fo.Icon(color='pink')))
```

**Marking Volcanoes with the color PINK** 

```python
map.add_child(vol)
```

**world.json for creating the polygon view**

```python
vol.add_child(fo.GeoJson(data=(open('../input/volcano/240.world.json','r',encoding='utf-8-sig').read())))
```

```python
map.add_child(vol)
```

```python
popu = pd.read_csv('../input/volcano/241.uscitiespop.csv')
```

```python
popu.head()
```

```python
lat_po = list(popu['lat'])
lon_po = list(popu['lon'])
name_po = list(popu['name'])
pop_po = list(popu['pop'])
```

```python
po = fo.FeatureGroup(name='My Map')
```

**MAking the markers colorful** â­

```python
def mar(popu):
    if(popu>35000):
        return 'red'
    elif(popu>10000 and popu<=35000):
        return 'blue'
    else:
        return 'green'
```

```python
for lat,lon,name,pop in zip(lat_po,lon_po,name_po,pop_po):
    po.add_child(fo.Marker(location=[lat,lon],popup=[pop,name] , 
                          icon=fo.Icon(color=mar(pop))))
```

```python
map.add_child(po)
```
