### Data Wrangling

In this project,  My cleaning and wrangling activities included:

1. Join two data frames:<br>
My aim is just getting vaccines' names from df_locations, based on that I joined two data frames

1. Deleting unwanted columns and rename column Iso:<br>
I am not interest in last_observation_date, source_name and website and also iso_code_y is same with iso_code_x, so I dropped iso_code_y and changed other one's name as 'iso_code'. There were two last row, it cause missing value in location so I dropped it.
  

```python
#merge two data frame
df = pd.merge(df,df_locations,on='location',how='left')
#drop unwanted columns
df.drop(['iso_code_y','last_observation_date','source_name','source_website'],axis=1,inplace=True)
df.rename(columns = {'iso_code_x':'iso_code'},inplace=True)
df.drop(df.tail(2).index, inplace = True) 
```

Understanding missing values

1. The iso-codes that are used for these countries : 'England','Northern Ireland','Scotland', 'Wales' are actually 'United Kingdom of Great Britain and Northern Ireland' and iso_code of them is 'GBR'[source](https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3). 'European Union' does not have any 3 letters iso_code, and 'Northern Cyprus' is actually dependent country of Turkey so it's code: 'TUR'.
1. The other missing values are related some countries begin to vaccinations later than other countries and the process is counting so I didnt do anything for them.
1. For playing with date base data, I splitted date, year, month and day. Also I added new columns for each of them.


```python
df[df.iso_code.isna()].location.unique()
```

```python
#Filling ISO-3 Codes
df['iso_code'] = df['iso_code'].fillna(pd.Series(np.where(df['location'] == 'England', 'GBR', 'NaN')))
df['iso_code'] = df['iso_code'].fillna(pd.Series(np.where(df['location'] == 'Northern Ireland', 'GBR', 'NaN')))
df['iso_code'] = df['iso_code'].fillna(pd.Series(np.where(df['location'] == 'Scotland', 'GBR', 'NaN')))
df['iso_code'] = df['iso_code'].fillna(pd.Series(np.where(df['location'] == 'Wales', 'GBR', 'NaN')))
df['iso_code'] = df['iso_code'].fillna(pd.Series(np.where(df['location'] == 'Northern Cyprus', 'TUR', 'NaN')))
```

```python
#Date Splitting
yseries = df['date'].map(lambda str: dt.datetime.strptime(str, '%Y-%m-%d').strftime('%Y'))
mseries = df['date'].map(lambda str: dt.datetime.strptime(str, '%Y-%m-%d').strftime('%m'))
dseries = df['date'].map(lambda str: dt.datetime.strptime(str, '%Y-%m-%d').strftime('%d'))

year=[]
month=[]
day=[]

for i in yseries.tolist():
    year.append(int(i))
for i in mseries.tolist():
    month.append(int(i))
for i in dseries.tolist():
    day.append(int(i))

df = pd.DataFrame({'location':df['location'],'iso_code':df['iso_code'],'date':df['date'],'year' : year, 'month' : month,'day':day,
                    'total_vaccinations':df['total_vaccinations'],'people_vaccinated':df['people_vaccinated'],
                    'people_fully_vaccinated':df['people_fully_vaccinated'],'daily_vaccinations_raw':df['daily_vaccinations_raw'],
                    'daily_vaccinations':df['daily_vaccinations'],'total_vaccinations_per_hundred':df['total_vaccinations_per_hundred'],
                    'people_vaccinated_per_hundred':df['people_vaccinated_per_hundred'],'people_fully_vaccinated_per_hundred':df['people_fully_vaccinated_per_hundred'],
                   'daily_vaccinations_per_million':df['daily_vaccinations_per_million'],'vaccines':df['vaccines']})


```
