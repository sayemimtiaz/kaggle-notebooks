```python
import numpy as np 
import pandas as pd 
from plotnine import *
from plotnine.data import mpg
from plotnine import options as op
from plotnine import watermark as wm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import geopandas as gpd
import random
from pywaffle import Waffle
import collections
```

```python
unicorn_data = pd.read_excel('/kaggle/input/list-of-unicorn-companies/uniorn_companies.xlsx')
unicorn_data = unicorn_data.dropna(axis=1,how='all')
print('Null Values in Columns:')
unicorn_data.isnull().sum()
```

Other than one row in Select Investors we dont have null values so we can dive into the data.

> **Top 3 Investment Groups:**<br>
> Top 3 invsetment groups that invested in most unicorns

```python
import collections
investor_list = list(unicorn_data['Select Investors'])
investors = []
for ele in investor_list:
    try:
        investors.append(ele.split(','))
    except:
        investors.append(ele)

investors_full = []
for sublist in investors:
    try:
        for item in sublist:
            if len(item) != 0:
                investors_full.append(item.strip())
    except:
        continue

occurences = {k: v for k, v in sorted(dict(collections.Counter(investors_full)).items(), key=lambda item: item[1], reverse=True)}
occurences_dict = {'company_name':['Sequoia Capital','Tencent Holdings', 'Accel'],'total_companies_invested':[60,25,23]}
occurences_df = pd.DataFrame.from_dict(occurences_dict)
```

```python
op.figure_size=(5,5)
ax = (ggplot(occurences_df)         
 + aes(x='company_name', y='total_companies_invested',fill='company_name')    
 + geom_col(size=15)
 + scale_fill_manual(values = ["#FFC133","#FD02A5","#F03812"])
 + labs(title = "Top 3 Unicorn Investment Groups")
 + labs(y = "Number of Companies", x = "Investement Group")
 + theme(
    panel_background = element_rect(fill = "black"),    
    plot_background = element_rect(fill = "black", color = "black"),
    legend_background = element_rect(fill = "black"),
    legend_key = element_blank(),
     
    panel_grid = element_line(size = 0.01),
    panel_grid_minor_y = element_blank(),
    panel_grid_major_y = element_blank(),
    
    legend_text = element_text(color = "white", size=13,face='bold'),
    axis_text_x = element_text(color = "white", size = 13,angle=345,face='bold'),
    axis_text_y = element_text(color = "white", size = 13, hjust = 1, margin={'b': 20, 't':10}),
    axis_title = element_text(color = "white", size = 14, hjust = 1,face = "bold"),
    plot_title = element_text(color = "white", face = "bold", size = 12, hjust = 4, margin={'b': 20, 't':10}),
    panel_spacing_x = 1
  )
)


fig = ax.draw()

accel = plt.imread('https://raw.githubusercontent.com/vineethbabu/unicorn_project_images/main/accel_1_70x55.png')
tencent = plt.imread('https://raw.githubusercontent.com/vineethbabu/unicorn_project_images/main/tencent_70x55.png')
sequio = plt.imread('https://raw.githubusercontent.com/vineethbabu/unicorn_project_images/main/sequio_70x55.png')
fig.figimage(accel, xo=100, yo=201, alpha=1, norm=None, cmap=None, vmin=None, vmax=None, origin=None, resize=False)
fig.figimage(tencent, xo=341, yo=212, alpha=1, norm=None, cmap=None, vmin=None, vmax=None, origin=None, resize=False)
fig.figimage(sequio, xo=222, yo=417, alpha=1, norm=None, cmap=None, vmin=None, vmax=None, origin=None, resize=False)
fig.show()
```

There is a raise in the number of unicorn companies in recent years. It is a very healthy sign to observe, lets now plot top 15 most valuable companies.   

```python
date_country_df = unicorn_data[['Country','Date Added']]
date_country_df = pd.DataFrame(date_country_df.groupby('Date Added')['Country'].apply(list))
date_country_df = date_country_df.reset_index()
date_country_df = pd.DataFrame(date_country_df.groupby(date_country_df['Date Added'].map(lambda x: x.year), as_index=False))

year_country_dict = {}
for i in range(date_country_df.shape[0]):
    year_country = list(date_country_df[1].iloc[i]['Country'])
    year_country_flat = [item for sublist in year_country for item in sublist]
    year_country_dict.update({date_country_df[0].iloc[i]:dict(collections.Counter(year_country_flat))})

select_countries = ['United States','India','United Kingdom','China','Germany']
ten_15_countries_unicorns = {'United States':[],'India':[],'United Kingdom':[],'China':[],'Germany':[]}
fifteen_21_countries_unicorns = {'United States':[],'India':[],'United Kingdom':[],'China':[],'Germany':[]}

for k,v in year_country_dict.items():
    if k>2010 and k<=2015:
        for key,value in v.items():
            if key=='United States':
                ten_15_countries_unicorns['United States'].append(value)
            elif key == 'India':
                ten_15_countries_unicorns['India'].append(value)
            elif key == 'United Kingdom':
                ten_15_countries_unicorns['United Kingdom'].append(value)
            elif key == 'China':
                ten_15_countries_unicorns['China'].append(value)
            elif key == 'Germany':
                ten_15_countries_unicorns['Germany'].append(value)
    if k>2015 and k<=2021:
        for key,value in v.items():
            if key=='United States':
                fifteen_21_countries_unicorns['United States'].append(value)
            elif key == 'India':
                fifteen_21_countries_unicorns['India'].append(value)
            elif key == 'United Kingdom':
                fifteen_21_countries_unicorns['United Kingdom'].append(value)
            elif key == 'China':
                fifteen_21_countries_unicorns['China'].append(value)
            elif key == 'Germany':
                fifteen_21_countries_unicorns['Germany'].append(value)     


for k, v in ten_15_countries_unicorns.items():
    if k == 'United States':
        ten_15_countries_unicorns['United States'] = [sum(ten_15_countries_unicorns['United States'])]
    if k == 'China':
        ten_15_countries_unicorns['China'] = [sum(ten_15_countries_unicorns['China'])]
    if k == 'India':
        ten_15_countries_unicorns['India'] = [sum(ten_15_countries_unicorns['India'])]
    if k == 'United Kingdom':
        ten_15_countries_unicorns['United Kingdom'] = [sum(ten_15_countries_unicorns['United Kingdom'])]
    if k == 'Germany':
        ten_15_countries_unicorns['Germany'] = [sum(ten_15_countries_unicorns['Germany'])]
        
for k, v in fifteen_21_countries_unicorns.items():
    if k == 'United States':
        fifteen_21_countries_unicorns['United States'] = [sum(fifteen_21_countries_unicorns['United States'])]
    if k == 'China':
        fifteen_21_countries_unicorns['China'] = [sum(fifteen_21_countries_unicorns['China'])]
    if k == 'India':
        fifteen_21_countries_unicorns['India'] = [sum(fifteen_21_countries_unicorns['India'])]
    if k == 'United Kingdom':
        fifteen_21_countries_unicorns['United Kingdom'] = [sum(fifteen_21_countries_unicorns['United Kingdom'])]
    if k == 'Germany':
        fifteen_21_countries_unicorns['Germany'] = [sum(fifteen_21_countries_unicorns['Germany'])]

ten_15_df = pd.DataFrame.from_dict(ten_15_countries_unicorns,orient='index')
fifteen_21_df = pd.DataFrame.from_dict(fifteen_21_countries_unicorns,orient='index')
five_year_plot = pd.DataFrame(columns = ['country','2010-2015','2015-2021'])
five_year_plot['2010-2015'] =  ten_15_df[0].values
five_year_plot['2015-2021'] =  fifteen_21_df[0].values + ten_15_df[0].values
five_year_plot['country'] = select_countries

five_year_plot['period_1'] = ['2010-2015']*5
five_year_plot['period_2'] = ['2015-2021']*5
five_year_plot = five_year_plot.set_index('country')
five_year_plot
```

> **Top 15 Companies Valuation vs Unicorn Age:**<br>
A bar plot with y axis containing valuation in billions and no of years since it became a unicorn

```python
date_creation_df = unicorn_data[['Date Added','Company','Valuation ($B) ', 'Category']]
date_creation_df.reset_index(level=0,inplace=True)
date_creation_df['Date Added'] = pd.to_datetime(date_creation_df['Date Added'], format='%m%d%y') 

now = pd.Timestamp('now')
date_creation_df['Date Added'] = date_creation_df['Date Added'].where(date_creation_df['Date Added'] < now, date_creation_df['Date Added'] -  np.timedelta64(100, 'Y')) 

date_creation_df['age'] = (now - date_creation_df['Date Added']).astype('<m8[Y]')
date_creation_df = date_creation_df.sort_values(by=['age'],ascending=False)
date_creation_df.reset_index()
date_creation_df = date_creation_df.drop(columns=['index'],axis=1)
date_creation_df.reset_index(level=0,inplace=True)
date_creation_df = date_creation_df.drop(columns=['index'],axis=1)
date_creation_df = date_creation_df.sort_values(by=['Valuation ($B) '],ascending=False).head(15)
date_creation_df.reset_index(level=0,inplace=True)

top_15_companies_age_valuation = pd.DataFrame(columns=['company','count','valuation_age'])
top_15_companies_age_valuation['company'] = ['']*date_creation_df.shape[0]*2
top_15_companies_age_valuation['count'] = ['']*date_creation_df.shape[0]*2
top_15_companies_age_valuation['valuation_age'] = ['']*date_creation_df.shape[0]*2
j = 0
for i in range(date_creation_df.shape[0]):
    top_15_companies_age_valuation['company'].iloc[j] = date_creation_df['Company'].iloc[i]
    top_15_companies_age_valuation['company'].iloc[j+1] = date_creation_df['Company'].iloc[i]
    top_15_companies_age_valuation['valuation_age'].iloc[j] = 'Valuation'
    top_15_companies_age_valuation['count'].iloc[j] = date_creation_df['Valuation ($B) '].iloc[i]
    top_15_companies_age_valuation['valuation_age'].iloc[j+1] = 'age'
    top_15_companies_age_valuation['count'].iloc[j+1] = date_creation_df['age'].iloc[i]
    j=j+2

op.figure_size=(8,5)
ax = (ggplot(top_15_companies_age_valuation)         
 + aes(x='company', y='count',fill='valuation_age')    
 + geom_col(size=20)
 + scale_fill_manual(values = ["#FE8826","#F7E1C0"])
 + labs(title = "Top 15 Companies Valuation and Age of Unicorn Status")
 + labs(y = "Valuation in $B; Age in Years", x = "Company")
 + theme(
    panel_background = element_rect(fill = "black"),    
    plot_background = element_rect(fill = "black", color = "black"),
    legend_background = element_rect(fill = "black"),
    legend_key = element_blank(),
   
    panel_grid = element_line(size = 0.3),
    panel_grid_minor_y = element_blank(),
    panel_grid_major_y = element_blank(),
    
    legend_text = element_text(color = "white",size = 10,face = "bold"),
    axis_text_x = element_text(color = "white", size = 13,angle=90,face = "bold"),
    axis_text_y = element_text(color = "white", size = 13, hjust = 1, margin={'b': 20, 't':10},face = "bold"),
    axis_title = element_text(color = "white", size = 14, hjust = 1,face = "bold"),
    plot_title = element_text(color = "white", face = "bold", size = 16, hjust = 4, margin={'b': 20, 't':10}),
    panel_spacing_x = 1
  )
)

fig = ax.draw()

fig.show()
```

We can see that the most valuable company **BYTEDANCE** became unicorn very recently but achieved very high valuation. Now I want to look into distribution in four select countries USA, UK, INDIA, CHINA.
