```python
df=pd.read_csv("../input/pesquisa-data-hackers-2019/datahackers-survey-2019-anonymous-responses.csv")
df.head()
```

```python
fig,axes = plt.subplots(1,1,figsize=(20,5))
sns.heatmap(df.isna(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
df = df.rename(columns={"('P1', 'age')":"age", "('P2', 'gender')": "gender", "('P3', 'living_in_brasil')": "resident", "('P34', 'kaggle_learn')": "kaggle", "('P20', 'linear regression')": "linear", "('P24', 'sql')": "sql", "('P24', 'imagens')": "images", "('P20', 'nlp')": "nlp", "('P20', 'gans')": "gans", "('P35', 'data_science_plataforms_preference')": "platforms", "('P27', 'google_data_studio')": "google"})
```

```python
px.histogram(df, x='resident', color='kaggle')
```

```python
fig = px.bar(df, y="kaggle", x="resident", color="age", orientation="h",
             color_continuous_scale='Bluered_r', hover_name="gender")

fig.show()
```

```python
fig = px.bar(df,
             y='kaggle',
             x='resident',
             orientation='h',
             color='age',
             title='Data Hackers Survey 2019',
             opacity=0.8,
             color_discrete_sequence=px.colors.diverging.Armyrose,
             template='plotly_dark'
            )
fig.update_xaxes(range=[0,35])
fig.show()
```

```python
corr=df[df.columns.sort_values()].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig = go.Figure(data=go.Heatmap(z=corr.mask(mask),
                                x=corr.columns.values,
                                y=corr.columns.values,
                                xgap=1, ygap=1,
                                colorscale="Rainbow",
                                colorbar_thickness=20,
                                colorbar_ticklen=3,
                                zmid=0),
                layout = go.Layout(title_text='Correlation Matrix', template='plotly_dark',
                height=900,
                xaxis_showgrid=False,
                yaxis_showgrid=False,
                yaxis_autorange='reversed'))
fig.show()
```

```python
fig = px.bar_polar(df, r="kaggle", theta="age", color="resident", template="plotly_dark",
            color_discrete_sequence= px.colors.sequential.Plasma_r)
fig.show()
```

```python
fig = px.parallel_categories(df, color="kaggle", color_continuous_scale=px.colors.sequential.Viridis)
fig.show()
```
