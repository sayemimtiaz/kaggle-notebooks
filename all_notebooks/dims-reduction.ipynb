```python
from sklearn.preprocessing import LabelEncoder
import umap
from cupy import asnumpy 
from sklearn.manifold import trustworthiness, TSNE 
from sklearn.preprocessing import LabelEncoder
import gc
import sklearn 
import cuml
from sklearn.metrics import calinski_harabasz_score, silhouette_score,davies_bouldin_score
from scipy.stats import spearmanr, mannwhitneyu
from scipy.spatial.distance import mahalanobis
lb_en = LabelEncoder()
enc_targets = lb_en.fit_transform(moa_features.loc[:,"multiclass_target"].values)
```

```python

%%time

class PCAPerformance:
  def __init__(self,features,labels,cat_cols,num_components):
    lbl_enc = LabelEncoder()
    features[cat_cols] = features[cat_cols].apply(lbl_enc.fit_transform) 
    self.features = features.values 
    self.labels = labels
    self.num_components = num_components
    
  def get_scores(self):
    trustworthiness_score = list()
    silhouette_coef = list()
    calinski_score = list()
    davies_score = list()
    
    for i in [11,12,13]:  
        np.random.seed(i)
        pca = cuml.decomposition.PCA(n_components=self.num_components,random_state=i)
        pca_transform = pca.fit_transform(self.features)
        trustworthiness_score.append(cuml.metrics.trustworthiness(asnumpy(self.features), asnumpy(pca_transform)))
        silhouette_coef.append(silhouette_score(self.features,self.labels))
        calinski_score.append(calinski_harabasz_score(self.features,self.labels))
        davies_score.append(davies_bouldin_score(self.features,self.labels))
        #spearmanr(self.features,self.labels)
        #mannwhitneyu(self.features,self.labels)
        #mahalanobis(self.features,self.features)
    out_dict =  {"trustworthiness":(np.mean(trustworthiness_score),np.std(trustworthiness_score)),
                 "silhouette_coef":(np.mean(silhouette_coef),np.std(silhouette_coef)),
                 "calinski_harabasz_score":(np.mean(calinski_score),np.std(calinski_score)),
                 "davies_bouldin_score":(np.mean(davies_score),np.std(davies_score))}
    return out_dict



me = PCAPerformance(moa_features.iloc[:,1:-1],enc_targets,moa_cat_cols,num_components=2)
me.get_scores() 
```
