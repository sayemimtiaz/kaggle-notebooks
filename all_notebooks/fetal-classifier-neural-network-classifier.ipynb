```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
import sklearn.model_selection as ms
import tensorflow as tf
from tensorflow import keras

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv(r'../input/fetal-health-classification/fetal_health.csv')
data.sample(5)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data_corr = data.corr()['fetal_health']
data_corr.sort_values()
```

```python
X = data.copy()
X.drop('fetal_health', axis=1, inplace=True)
y = data['fetal_health'].copy()
```

```python
X_train_full, X_test, y_train_full, y_test = ms.train_test_split(X, y, stratify=y, random_state=42)
X_train, X_validation, y_train, y_validation = ms.train_test_split(X_train_full, y_train_full, stratify= y_train_full, random_state=42)
```

```python
X_train.shape
```

# Building Neural Network Model

```python
model = keras.models.Sequential([
                    keras.layers.Input(shape=X_train.shape[1], name='input'),
                    keras.layers.Dense(250, activation='relu'),
                    keras.layers.Dense(125, activation='relu'),
                     keras.layers.Dense(4, activation='softmax', name='output')
                    ])

model.compile(
            loss='sparse_categorical_crossentropy',
            optimizer='sgd',
            metrics= ['accuracy']
            )

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_validation, y_validation))
```

# Performance

```python
metrics_df = pd.DataFrame(history.history)
metrics_df.sample()
```

```python
print(f'Avg Training Accuracy: {round(metrics_df.accuracy.mean() * 100, 1)}')
print(f'Avg Validation Accuracy: {round(metrics_df.val_accuracy.mean() * 100, 1)}')
```

```python
metrics_df.plot()
plt.ylim((0,1))
```

# Performance on Testing Data

```python
model.evaluate(X_test, y_test)
```
