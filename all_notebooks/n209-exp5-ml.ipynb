***3. Reading the data in***

```python
df = pd.read_csv("../input/spotify-dataset-19212020-160k-tracks/data.csv")

# take a look at the dataset
df.head()
```

***Selecting features for regression.***

```python
cdf = df[['acousticness','danceability','duration_ms','energy','explicit','instrumentalness','key','liveness','loudness','mode','popularity','speechiness','tempo','valence']]
cdf.head(14)
```

```python
plt.scatter(cdf.instrumentalness, cdf.speechiness,  color='red')
plt.xlabel("instrumentalness")
plt.ylabel("speechiness")
plt.show()
```

***Creating train and test dataset.***

```python
msk = np.random.rand(len(df)) < 0.7
train = cdf[msk]
test = cdf[~msk]
```

***Train data distribution.***

```python
plt.scatter(cdf.instrumentalness, cdf.speechiness,  color='red')
plt.xlabel("instrumentalness")
plt.ylabel("speechiness")
plt.show()
```

***4. Multiple Regression Model***

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['instrumentalness','duration_ms','energy']])
y = np.asanyarray(train[['tempo']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

***5. Prediction***

```python
y_hat= regr.predict(test[['instrumentalness','duration_ms','energy']])
x = np.asanyarray(test[['instrumentalness','duration_ms','energy']])
y = np.asanyarray(test[['tempo']])
print("Sum of squares: %.4f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.4f' % regr.score(x, y))
```
