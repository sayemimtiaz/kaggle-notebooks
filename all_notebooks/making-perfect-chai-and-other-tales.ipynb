# What makes a person click a video's thumbnail ‚ùì

As soon as any visitor enters the Youtube Homepage, the first thing they see is a screen with thumbnails. There are several thumbnails, most of which relate to the kind of content the particular visitor has *watched in the past*. The first hurdle a CTDS video must cross over is **getting a user X to click on a video's thumbnail**. But, what makes a user want to click on the thumbnail of a video?

From our experience on YouTube(if you don't have any, I recommend you try) we know that a visitor can see the following features of the thumbnail:
- The type of it
- The topic to be discussed, the ML hero and other info
- The duration of the episode

At this juncture, let's hypothesize that these 3 features that are visible to a visitor play an important role in whether the visitor chooses to click a thumbnail or not. We will see if they hold good. 


üîç **WHAT METRIC DEFINES SUCCESS HERE**

In order to evaluate the success rate of whether a visitor will click on a thumbnail or not, the click through rate(CTR) will be used. The CTR measures the ability of a video to prompt people to watch it after seeing it on their homepage, recommendation section or trending section. 

> CTR = Number of Clicks / Number of Impressions * 100

üìö **WHAT DO WE KNOW OR BACKGROUND KNOWLEDGE**

**A high CTR for a video indicates that the viewers saw something in its thumbnail that compelled them to click on it**

‚ùó **NOTE**

Every user does have a different purpose. So, it's not that simple to generalize all episodes. However, the idea here is not to compare videos, but to identify if the click through rate is determined by *what the visitor sees* on the thumbnail.

```python
# utility functions to visualize data


def across_epi_plot(df, feature, ind="release_date", title="You forgot the title"):
    """
    Plots an interactive line plot depicting the values of each episode
    across `feature` in the data
    -----
    
    df: The dataframe
    feature: The feature name from the relevant dataset
    ind: The independent feature across which the values were shown
         `release_date` was the default
    title: Title of the plot
    """

    # find the mean value
    mean_val = round(df[feature].mean(), 2)

    # set the plot title
    plt_title = title + " (Mean=" + str(mean_val) + ")"

    # plot the graph
    fig = px.line(df, x=ind, y=feature, hover_data=["episode_id", "episode_name"])
    fig.update_traces(mode="lines+markers", line_color="#e87d23")
    fig.update_layout(
        hoverlabel=dict(bgcolor="white", font_size=12, font_family="Rockwell")
    )
    fig.add_shape(
        # add a horizontal line
        type="line",
        x0="2019-07-19",
        y0=mean_val,
        x1="2020-06-20",
        y1=mean_val,
        line=dict(color="black", width=2, dash="dash"),
    )

    # template enhancement
    fig.update_layout(
        template="ggplot2",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 20},
            "xanchor": "center",
            "yanchor": "top",
        },
        xaxis=dict(range=["2019-07-19", "2020-06-20"]),
    )

    # show plot
    fig.show()


def across_epi_plot2(df, feature, ind="release_date", title="You forgot the title"):
    """
    Plots an interactive line plot depicting the values of each episode
    across `feature` in the data
    -----
    
    df: The dataframe
    feature: The feature name from the relevant dataset
    ind: The independent feature across which the values were shown
         `release_date` was the default
    title: Title of the plot
    """

    # find the mean value
    mean_val = round(df[feature].mean(), 2)

    # set the plot title
    plt_title = title + " (Mean=" + str(mean_val) + ")"

    # plot the graph
    fig = px.line(df, x=ind, y=feature)
    fig.update_traces(mode="lines+markers", line_color="#e87d23")
    fig.update_layout(
        hoverlabel=dict(bgcolor="white", font_size=12, font_family="Rockwell")
    )
    fig.add_shape(
        # add a horizontal line
        type="line",
        x0="2019-07-19",
        y0=mean_val,
        x1="2020-06-20",
        y1=mean_val,
        line=dict(color="black", width=2, dash="dash"),
    )

    # template enhancement
    fig.update_layout(
        template="ggplot2",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 20},
            "xanchor": "center",
            "yanchor": "top",
        },
        xaxis=dict(range=["2019-07-19", "2020-06-20"]),
    )

    # show plot
    fig.show()


def comp_bin_plot(
    df,
    f1,
    f2,
    l1,
    l2,
    title="You forgot the title",
    xtitle="You forgot",
    ytitle="You forgot",
):
    """
    Plots an interactive line plot depicting the values of each episode
    across features `f1` and `f2` in the data
    -----
    
    df: Dataframe
    f1: First feature
    f2: Second feature
    l1: Label 1
    l2: Label 2
    title: Title of the plot
    """

    # find the means
    m1 = round(df[f1].mean(), 2)
    m2 = round(df[f2].mean(), 2)

    # set the plot title
    plt_title = (
        title
        + " ("
        + l1
        + "_Mean="
        + str(m1)
        + ")"
        + " ("
        + l2
        + "_Mean="
        + str(m2)
        + ")"
    )

    # plot the graph
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df["episode_id"], y=df[f1], mode="lines", name=l1))
    fig.add_trace(go.Scatter(x=df["episode_id"], y=df[f2], mode="lines", name=l2))
    fig.update_traces(mode="lines+markers")
    fig.update_layout(
        hoverlabel=dict(bgcolor="white", font_size=12, font_family="Rockwell")
    )

    # template enhancement
    fig.update_layout(
        template="ggplot2",
        title={
            "text": plt_title,
            "font": {"family": "Rockwell", "size": 20},
            "xanchor": "center",
            "yanchor": "top",
        },
        xaxis_title=xtitle,
        yaxis_title=ytitle,
        xaxis=dict(range=["2019-07-19", "2020-06-20"]),
    )

    # show plot
    fig.show()


def plot_corr(df, feature, title="You forgot the title"):
    """
    Plot correlations of appropriate features against `feature`
    as a bar plot
    -----
    
    df: The dataframe
    feature: The feature whose correlation with other features
             is under consideration
    title: The title of the plot
    """

    # features to exclude
    ignore_list = [
        "youtube_thumbnail_type",
        "anchor_thumbnail_type",
        "anchor_plays",
        "spotify_starts",
        "spotify_streams",
        "spotify_listeners",
        "apple_listeners",
        "apple_listened_hours",
        "apple_avg_listen_duration",
    ]  # audio data is not considered here
    ignore_list.append(
        feature
    )  # self-correlation is not necessary; always going to be 1

    # prepare some correlation data
    sub_corr = pd.DataFrame(df.corr()[feature])
    sub_corr.columns = ["Correlation"]
    sub_corr["Impact"] = sub_corr["Correlation"]
    sub_corr = sub_corr.drop(ignore_list)

    # plot the graph
    fig = px.bar(
        sub_corr,
        x=sub_corr.index,
        y="Correlation",
        color="Correlation",
        color_continuous_scale="oranges",
    )

    # template enhancement
    fig.update_layout(
        template="plotly_dark",
        title={
            "text": title,
            "font": {"family": "Rockwell", "size": 20},
            "yanchor": "top",
        },
        xaxis_title="",
    )

    fig.show()
```

## Does clicking on the video mean that a person will watch the video?

This is a very important insight to have. If an episode has a high CTR and a low average percentage of video watched, that means the viewers didn't find the content compelling(or maybe they just had a whole different set of expectations).  

In order to check this, I have engineered a new feature *youtube_avg_watch_percentage* which is the average percentage of an episode watched by people.

> youtube_avg_watch_percentage = youtube_avg_watch_duration / episode_duration

```python
# feature engineering average watch percentage

epi['youtube_avg_watch_percentage'] = epi['youtube_avg_watch_duration'] / epi['episode_duration']
```

```python
# youtube_avg_watch_percentage across episodes

across_epi_plot(epi, 'youtube_avg_watch_percentage', title='Average watch percentage across episodes')
```

- The least of this metric was 1.8% that came in episode 46. This probably could have been because the topic looked very specific to a particular group of practioners interested in ML in Classical Japanese Literature.
    - However, the reason was quite different. On checking the youtube version of this episode, it became apparent to me that this episode on youtube was just an intro to the talk and the talk was available only as audio.

Before we start with this, let's aggregate all the transcripts together and create one main dataset with a few additional features that could help in analysis.

```python
# feature engineer the transcript features


def conv_to_sec(x):
    """ Time to seconds """

    t_list = x.split(":")
    if len(t_list) == 2:
        m = t_list[0]
        s = t_list[1]
        time = int(m) * 60 + int(s)
    else:
        h = t_list[0]
        m = t_list[1]
        s = t_list[2]
        time = int(h) * 60 * 60 + int(m) * 60 + int(s)
    return time


def get_durations(nums, size):
    """ Get durations i.e the time for which each speaker spoke continuously """

    diffs = []
    for i in range(size - 1):
        diffs.append(nums[i + 1] - nums[i])
    diffs.append(30)  # standard value for all end of the episode CFA by Sanyam
    return diffs


def transform_transcript(sub, episode_id):
    """ Transform the transcript of the given episode """

    # create the time second feature that converts the time into the unified qty. of seconds
    sub["Time_sec"] = sub["Time"].apply(conv_to_sec)

    # get durations
    sub["Duration"] = get_durations(sub["Time_sec"], sub.shape[0])

    # providing an identity to each transcript
    sub["Episode_ID"] = episode_id
    sub = sub[["Episode_ID", "Time", "Time_sec", "Duration", "Speaker", "Text"]]

    return sub


def combine_transcripts(sub_dir):
    """ Combine all the 75 transcripts of the ML Heroes Interviews together as one dataframe """

    episodes = []
    for i in range(1, 76):
        file = "E" + str(i) + ".csv"
        try:
            sub_epi = pd.read_csv(os.path.join(sub_dir, file))
            sub_epi = transform_transcript(sub_epi, ("E" + str(i)))
            episodes.append(sub_epi)
        except:
            continue
    return pd.concat(episodes, ignore_index=True)


# create the combined transcript dataset
sub_dir = "../input/chai-time-data-science/Cleaned Subtitles"
transcripts = combine_transcripts(sub_dir)

# display first 10 instances
transcripts.head(10)
```

üîç **ABOUT THE NEW DATASET - TRANSCRIPTS**
- Our main dataset has 7649 rows; the distribution of rows is non-uniform across episodes
- There is no transcript available for Episode 4. So, we have 74 episodes' transcripts in this dataframe
- And episode 69 was Sanyam's AMA, his gift to the community on his b'day ;) (May 27th). So, the transcripts here only have his dialogues(monologue!)
- I have introduced 3 main features as of now:
    - Episode_ID => To understand which episode the dialogue belongs to
    - Time_sec => Unify the time format (seconds)
    - Duration => Duration of each continuous speech by a speaker before the other speaker speaks (seconds)
- Also, I have added a 30 second duration at the end for the last call for action that the host makes at the end of each episode

With these new-found transcript features, we can introduce a new set of attributes - **Speaking speeds**. The number of words per each instance divided by the duration(seconds) to speak these by the speaker gives the speaking speed. The following dataset depicts the first 5 instances of this newly formed dataset.  

> You could ignore the warning generated beneath

```python
# generating some unusual features - Speech Speed Patterns


def host_speaking_speed(epi):
    """ Return the speaking speed of the host in words per minute """

    speeds = []

    df = transcripts[transcripts["Episode_ID"] == epi]
    df = df[df["Speaker"] == "Sanyam Bhutani"].reset_index()
    df = df[
        1 : (df.shape[0] - 1)
    ]  # the first and last dialogues are standard for all episodes

    for i in range(df.shape[0]):
        if df["Duration"].iloc[i] != 0:
            speeds.append(len(df["Text"].iloc[i].split()) / df["Duration"].iloc[i])
        else:
            continue
    return round(np.mean(speeds), 2)


def host_avg_speaking_speed():
    """ Returns a dataframe of Sanyam's average speaking speed across episodes """

    avg_speeds = []
    for epi in list(transcripts["Episode_ID"].unique()):
        avg_speeds.append(host_speaking_speed(epi))
    return pd.DataFrame(
        {
            "episode_id": list(transcripts["Episode_ID"].unique()),
            "host_avg_spk_speed": avg_speeds,
        }
    )


def guest_speaking_speed(epi):
    """ Return the speaking speed of the host in words per minute """

    speeds = []

    df = transcripts[transcripts["Episode_ID"] == epi]
    df = df[df["Speaker"] != "Sanyam Bhutani"].reset_index()

    # episode 69 was the host AMA session. So, there was no guest here => np.nan() will be added in place of guest speed here
    for i in range(df.shape[0]):
        if df["Duration"].iloc[i] != 0:
            speeds.append(len(df["Text"].iloc[i].split()) / df["Duration"].iloc[i])
        else:
            continue
    return round(np.mean(speeds), 2)


def guest_avg_speaking_speed():
    """ Returns a dataframe of Sanyam's average speaking speed across episodes """

    avg_speeds = []
    for epi in list(transcripts["Episode_ID"].unique()):
        avg_speeds.append(guest_speaking_speed(epi))
    return pd.DataFrame(
        {
            "episode_id": list(transcripts["Episode_ID"].unique()),
            "guest_avg_spk_speed": avg_speeds,
        }
    )


# create the dataset
spk_speed = pd.merge(
    host_avg_speaking_speed(), guest_avg_speaking_speed(), on="episode_id"
)

# feature engineer "both_avg"
spk_speed["both_avg"] = round(
    (spk_speed["host_avg_spk_speed"] + spk_speed["guest_avg_spk_speed"]) / 2, 2
)

# feature engineer "diff"
spk_speed["diff"] = round(
    (spk_speed["host_avg_spk_speed"] - spk_speed["guest_avg_spk_speed"]), 2
)

# display first 5 instances
spk_speed.head(5)
```

üîç **ABOUT THE NEW DATASET - SPK_SPEED**

- host_avg_spk_speed : The average number of words said by the host per second
- guest_avg_spk_speed : The average number of words said by the speaker per second
- both_avg : The average of bothe host's and guest's speak speeds
- diff : The difference between the host's and guest's speak speeds
    - If -ve, host speaks slower than guest
    - If +ve, host speaks faster than guest

```python
# This cell has been completely borrowed from Parul Pandey's starter kernel; hope that is okay!


def clean_text(text):
    """
    Preliminary cleaning operations
    """

    text = text.lower()
    # text = re.sub('\[.*?\]', '', text)
    text = re.sub("https?://\S+|www\.\S+", "", text)
    # text = re.sub('<.*?>+', '', text)
    # text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub("\n", "", text)
    text = re.sub("\w*\d\w*", "", text)
    return text


def text_preprocessing(text):
    """
    Cleaning and parsing the text.
    """

    tokenizer = nltk.tokenize.RegexpTokenizer(r"\w+")
    nopunc = clean_text(text)
    tokenized_text = tokenizer.tokenize(nopunc)
    remove_stopwords = [
        w for w in tokenized_text if w not in stopwords.words("english")
    ]
    combined_text = " ".join(remove_stopwords)
    return combined_text


def remove_stopwords(sen):
    """
    Remove stopwords
    """

    sen_new = " ".join([i for i in sen if i not in stopwords.words("english")])
    return sen_new
```

```python
# Make a dataset aggregating the content for each episode (only the guest)


def aggregate_trans(epi):
    """ Aggregate Transcripts for a whole episode """

    trans_epi = transcripts[transcripts["Episode_ID"] == epi]
    trans_epi = trans_epi[trans_epi["Speaker"] != "Sanyam Bhutani"]
    return " ".join(trans_epi["Text"])


# initialise empty data structures
trans_dict = {}
eid = []
content = []

# episode id
for ep in list(transcripts["Episode_ID"].unique()):
    eid.append(ep)
trans_dict["episode_id"] = eid

# aggregate content per episode
for ep in list(transcripts["Episode_ID"].unique()):
    content.append(aggregate_trans(ep))
trans_dict["content"] = content

# create the dataframe
epi_content = pd.DataFrame(trans_dict)

# cleaning up the transcripts (for summarization)
epi_content["cleaned"] = epi_content["content"].apply(clean_text)
epi_content["preprocessed"] = epi_content["cleaned"].apply(text_preprocessing)

# display first 10 instances
epi_content.head(10)
```

```python
# binning number of subscribers as well as number of likes


def make_sub_cat(x):
    """
    Bin into categories based on subscribers
    -----
    
    > Based on the 25th, 50th and 75th quantiles
    """

    if x <= 2:
        return "very low"
    elif (x > 2) and (x <= 6):
        return "low"
    elif (x > 6) and (x <= 16):
        return "medium"
    return "high"


def make_like_cat(x):
    """
    Bin into categories based on likes
    -----
    
    > Based on the 25th, 50th and 75th quantiles
    """

    if x <= 4:
        return "very low"
    elif (x > 4) and (x <= 13):
        return "low"
    elif (x > 13) and (x <= 24):
        return "medium"
    return "high"


epi["sub_cat"] = epi["youtube_subscribers"].apply(make_sub_cat)
epi["like_cat"] = epi["youtube_likes"].apply(make_like_cat)
```

```python
# calculate average sentence length


def get_avg_sent_length(text):
    """
    Returns the average sentence length for the given guest
    """

    sentences = sent_tokenize(text)
    avg_len = sum(len(sentence.split()) for sentence in sentences) / len(sentences)

    return avg_len


# feature engineer "avg_sent_length"
sub_data["avg_sent_length"] = sub_data["content"].apply(get_avg_sent_length)

# set the seaborn plotting style
sns.set_style("darkgrid")
sns.set(font_scale=1)

# set figure size
plt.figure(figsize=(10, 6))

# plot violin plot
sns.violinplot(
    y="avg_sent_length",
    data=sub_data,
    x="sub_cat",
    order=["very low", "low", "medium", "high"],
    palette="Reds",
)

# plot swarm plot
sns.swarmplot(
    y="avg_sent_length",
    data=sub_data,
    x="sub_cat",
    order=["very low", "low", "medium", "high"],
    color="black",
)

# re-assign axes labels
plt.xlabel("\nCategory of Episodes w.r.t Subscribers", fontsize=12)
plt.ylabel("Average Length of a Sentence by the Guest\n", fontsize=12)

# add title
plt.title(
    "Does the community prefer to subscribe when guests speak in shorter sentences?\n",
    fontsize=15,
)

# show plot
plt.show()
```

## Generating 20-line summaries for the 3 most loved episodes

If we have quick summaries for the CTDS episodes, we could probably understand the content that makes viewers subscribe to the channel and the content that does not. However, for the purpose of the analysis performed in this section, a 20-line summary was useful. It provides a good assortment of valuable content the guests in these 3 episodes had for the community. 


***NOTE:***  
This is where I learnt about something called [TextRank](https://web.eecs.umich.edu/~mihalcea/papers/mihalcea.emnlp04.pdf), an extractive summary generation algorithm built on top of Google's PageRank. To understand its crux in under 3 minutes, I would suggest [this video](https://www.speechtechmag.com/Articles/ReadArticle.aspx?ArticleID=127336). I am using the TextRank algorithm to pick up 20-line summaries from an episode. Not witholding any credit where it is due, I have used most of the code to do this from [here](https://www.analyticsvidhya.com/blog/2018/11/introduction-text-summarization-textrank-python/) and tweaked it at will, including modularizing it to suit my requirements.

```python
# Download GloVe embeddings
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove*.zip

# Extract word vectors from GloVe embeddings
word_embeddings = {}
f = open('glove.6B.100d.txt', encoding='utf-8')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    word_embeddings[word] = coefs
f.close()
```

```python
# Code to get summaries


def prep_sim_matrix(sentences, sentence_vectors):
    """
    Prepares a cosine similarity matrix
    """

    sim_mat = np.zeros([len(sentences), len(sentences)])
    for i in range(len(sentences)):
        for j in range(len(sentences)):
            if i != j:
                sim_mat[i][j] = cosine_similarity(
                    sentence_vectors[i].reshape(1, 100),
                    sentence_vectors[j].reshape(1, 100),
                )[0, 0]
    return sim_mat


def vectorize(sentences):
    """
    Convert text data into numbers/vectors
    """

    sentence_vectors = []  # initialise as empty
    for i in sentences:
        if len(i) != 0:
            v = sum([word_embeddings.get(w, np.zeros((100,))) for w in i.split()]) / (
                len(i.split()) + 0.001
            )
        else:
            v = np.zeros((100,))
        sentence_vectors.append(v)
    return sentence_vectors


def generate_sentences(episode):
    """
    Generate sentences for a given episode
    """

    temp = epi_content[epi_content["episode_id"] == episode]

    sentences = []
    for s in temp["cleaned"]:
        sentences.append(sent_tokenize(s))
    sentences = [y for x in sentences for y in x]

    return sentences


def get_summary(episode):
    """
    Get an N sentence (by default, N is 20) summary for a given episode
    """

    n_sent = 20  # tweak this paramater if you want more sentences in the summary

    # break the transcript to sentences
    sentences = generate_sentences(episode)
    
    # vectorize
    sent_vectors = vectorize(sentences)
    
    # cosine similarity matrix
    sim_mat = prep_sim_matrix(sentences, sent_vectors)

    # Summarization
    nx_graph = nx.from_numpy_array(sim_mat)
    scores = nx.pagerank(nx_graph)
    ranked_sentences = sorted(
        ((scores[i], s) for i, s in enumerate(sentences)), reverse=True
    )
    
    # display it
    print("SUMMARY FOR " + episode + "=>")
    for i in range(n_sent):
        print(ranked_sentences[i][1])
        print("xxxxxxxxxx")
    print("--------------------")
```

In order to understand the patterns of the host's interactions with the guests, I am creating a new dataset here(yes, again).

```python
# creating a new dataset to capture trends in Q and A


def count_questions(text):
    """
    Returns the number of question marks in the 'text'
    """

    return text.count("?")


def ques_ratio(df, n_ques):
    """
    Returns the ratio of number of questions / number of interactions
    by the host
    -----
    
    > An interaction is counted when the host talks
    > A question is counted everytime the host asks a question, irrespective
      of whether it is the same question asked in a different form
    > Can be greater than 1
    """

    return n_ques / df.shape[0]


def times_like_used(text):
    """
    Returns the number of times the host uses the word 'like'
    """

    return text.count("like")


def count_I(text):
    """
    Returns the number of times the host talks in
    first person
    -----
    > I think
    > I'll
    > I'm
    """

    i_sp = text.count("I ")
    i_ap = text.count("I'")
    return i_sp + i_ap


# code to make the dataframe
episodes = []
num_questions = []
interactions = []
ques_ratios = []
num_like_used = []
fp_used = []

for ep in list(transcripts["Episode_ID"].unique()):

    if ep == "E69":
        continue
    else:

        epi_t = transcripts[transcripts["Episode_ID"] == ep]
        epi_host = epi_t[epi_t["Speaker"] == "Sanyam Bhutani"]
        text = " ".join(epi_host["Text"])

        episodes.append(ep)
        num_questions.append(count_questions(text))
        interactions.append(epi_host.shape[0])
        ques_ratios.append(round(ques_ratio(epi_host, count_questions(text)), 2))
        num_like_used.append(times_like_used(text))
        fp_used.append(count_I(text))
        
# make the dataframe
host_pre = pd.DataFrame(
    {
        "episode_id": episodes,
        "num_questions_by_host": num_questions,
        "interactions": interactions,
        "ques_ratio": ques_ratios,
        "num_like_used": num_like_used,
        "first_person_usage": fp_used,
    }
)

# display first 10 rows
host_pre.head()
```
