### <b><mark style="background-color: #9B59B6"><font color="white">Connecting Culture and The Romantic Disease: TUBERCULOSIS</font></mark></b>

Interweaving the culture and science of tuberculosis

Author: Colin Martin - Published:February 27, 2014DOI:https://doi.org/10.1016/S0140-6736(14)60392-5

Tuberculosis has inspired poets, novelists, composers, and artists for centuries. Its hold on human creativity seems as tenacious as its persistent and devastating grip on global health.

TB was fashionable in certain creative circles during the early 19th century. The Romantic poet Lord Byron remarked to Thomas Moore, his eventual biographer: “I should like to die of a consumption…because the ladies would all say ‘Look at that poor Byron, how interesting he looks in dying!’”
