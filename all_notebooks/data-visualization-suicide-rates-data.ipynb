# Seaborn Tutorial

```python
# Sum suicides by countrys
data_country = suicide_data.groupby(["country"]).sum()["suicides_no"].reset_index().sort_values('suicides_no',ascending=False) 

# Visualization with Bar Plot
plt.figure(figsize=(15,10))
sns.barplot(x=data_country['country'], y=data_country['suicides_no'],palette = sns.color_palette("Paired"))
plt.xticks(rotation= 90)
plt.xlabel('Country')
plt.ylabel('Suicide')
plt.title('Suicide Rate Given Country')
```

```python
# Sum suicides by years
data_year = suicide_data.groupby(["year"]).sum()["suicides_no"].reset_index().sort_values("suicides_no",ascending=False)

plt.figure(figsize=(16,10))
ax= sns.barplot(x=data_year["year"], y=data_year["suicides_no"])
plt.xlabel('Year')
plt.ylabel('Suicides')
plt.title('Suicide Rate Given Year')
```

```python
# Visualization with Joint Plot
# kind : { ‚Äúscatter‚Äù | ‚Äúreg‚Äù | ‚Äúresid‚Äù | ‚Äúkde‚Äù | ‚Äúhex‚Äù }
g = sns.jointplot(data_year.year,data_year.suicides_no, kind="kde", color="#bfa9e0" ,size=7)
plt.savefig('graph.png')
```

```python
g = sns.jointplot("year", "suicides_no", data=data_year,size=6, ratio=3, color="#e8fb41")
```

```python
# Finding the year with the highest suicide
df = data_year.reset_index()
df = df.drop(["index"], axis=1)
max_suicide_year = df["year"][0]
max_suicide_year 
```

* As a result, we can see that the suicide rate of males  in 1999 was significantly higher than that of females, and this rate was mostly between the ages of 35 and 54.

# Plotly Tutorial

```python
# import graph objects as "go"
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
suicide_data.head()
```

```python
min_year = min(suicide_data["year"])
min_year
```

```python
# Avarage values by 1985
df = suicide_data[suicide_data["year"]==min_year].groupby("country").mean().sort_values("suicides_no", ascending=False).reset_index()
```

```python
# Visualization with Line Chart
# Creating trace1
trace1 = go.Scatter(
                    x = df.index,
                    y = df["gdp_per_capita ($)"],
                    mode = "lines",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= df.country)

data = [trace1]
layout = dict(title = 'GDP per capita in 1985',
              xaxis= dict(title= 'Country Rank',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

* The GDP per capita value of the countries with the highest suicide rate in 1985.

```python
# Creating trace2
trace2 = go.Scatter(
                    x = df.index,
                    y = df["suicides/100k pop"],
                    mode = "lines+markers",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.country)
data = [trace2]
layout = dict(title = 'Suicides/100k pop in 1985',
              xaxis= dict(title= 'Country Rank',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

* The suicides/100k pop value of the countries with the highest suicide rate in 1985.

```python
# Avarage suicide by country and age 
data_ca = suicide_data.groupby(["country","age"]).mean()["suicides_no"].reset_index()
data_ca
```

```python
# prepare data frames
df_age_5_14 = data_ca[data_ca.age == "5-14 years"].iloc[:100,:]
df_age_15_24 = data_ca[data_ca.age == "15-24 years"].iloc[:100,:]
df_age_25_34 = data_ca[data_ca.age == "25-34 years"].iloc[:100,:]
df_age_35_54 = data_ca[data_ca.age == "35-54 years"].iloc[:100,:]
df_age_55_74 = data_ca[data_ca.age == "55-74 years"].iloc[:100,:]
df_age_75 = data_ca[data_ca.age == "75+ years"].iloc[:100,:] 

# Visualization with Scatter Chart
# creating trace1
trace1 =go.Scatter(
                    x = df_age_5_14.index,
                    y = df_age_5_14.suicides_no,
                    mode = "markers",
                    name = "5-14 years",
                    marker = dict(color = 'rgb(188, 0, 6, 0.8)'),
                    text= df_age_5_14.country)
# creating trace2
trace2 =go.Scatter(
                    x = df_age_15_24.index,
                    y = df_age_15_24.suicides_no,
                    mode = "markers",
                    name = "15-24 years",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    text= df_age_15_24.country)
# creating trace3
trace3 =go.Scatter(
                    x = df_age_25_34.index,
                    y = df_age_25_34.suicides_no,
                    mode = "markers",
                    name = "25-34 years",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    text= df_age_15_24.country)
# creating trace4
trace4 =go.Scatter(
                    x = df_age_35_54.index,
                    y = df_age_35_54.suicides_no,
                    mode = "markers",
                    name = "35-54 years",
                    marker = dict(color = 'rgb(166, 255, 6, 0.8)'),
                    text= df_age_15_24.country)
# creating trace5
trace5 =go.Scatter(
                    x = df_age_55_74.index,
                    y = df_age_55_74.suicides_no,
                    mode = "markers",
                    name = "55-74 years",
                    marker = dict(color = 'rgb(249, 255, 6, 0.8)'),
                    text= df_age_15_24.country)
# creating trace6
trace6 =go.Scatter(
                    x = df_age_75.index,
                    y = df_age_75.suicides_no,
                    mode = "markers",
                    name = "75+ years",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= df_age_15_24.country)
data = [trace1, trace2,trace3,trace4,trace5,trace6]
layout = dict(title = 'Suicide Rate vs Countries by Age',
              xaxis= dict(title= 'Countries ',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Suicide Rate',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

* As a result, Japan, the Russian Federation, and the United States have greater differences in suicide rates by age compared to other countries.
* The highest suicide rate in most countries is between the ages of 35-54.

```python
df = suicide_data.groupby(["country","year"]).mean().sort_values("suicides_no", ascending=False).reset_index()
```

```python
# Visualization with Histogram 

# data preparation
data_2002 = df["suicides/100k pop"][df.year==2002][:70]
data_2013 = df["suicides/100k pop"][df.year==2013][:70]

trace1 = go.Histogram(
    x=data_2002,
    opacity=0.75,
    name = "2002",
    marker=dict(color='rgb(28, 18, 66, 0.3)'))
trace2 = go.Histogram(
    x=data_2013,
    opacity=0.75,
    name = "2013",
    marker=dict(color='rgb(218, 3, 0, 0.3)'))

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',
                   title=' Suicides ratio in 2002 and 2013',
                   xaxis=dict(title='Suicides Ratio'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# Visualization with Box Plot

# data preparation
data_1999 = df[df.year == 1999]

trace1 = go.Box(
    y=data_1999["gdp_per_capita ($)"],
    name = 'GDP per capita in 1999',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace1]
iplot(data)
```

```python
# Visualization with Scatter Matrix Plot

# import figure factory
import plotly.figure_factory as ff

# data preparation
dataframe = df[df.year == 1998]
data_1998 = dataframe.loc[:,["suicides_no","population", "gdp_per_capita ($)"]]
data_1998["index"] = np.arange(1,len(data_1998)+1)
# scatter matrix
fig = ff.create_scatterplotmatrix(data_1998, diag='box', index='index',colormap='Viridis',
                                  colormap_type='cat',
                                  height=700, width=700)
iplot(fig)
```

```python
# Visualization with Inset Plot

# first line plot
trace1 = go.Scatter(
    x=dataframe.index,
    y=dataframe["gdp_per_capita ($)"],
    name = "GDP per capita",
    marker = dict(color = 'rgb(9, 3, 130, 0.8)'),
)
# second line plot
trace2 = go.Scatter(
    x=dataframe.index,
    y=dataframe["suicides/100k pop"],
    xaxis='x2',
    yaxis='y2',
    name = "Suicides Ratio",
    marker = dict(color = 'rgb(167, 61, 130,0.8)'),
)
data = [trace1, trace2]
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
    title = 'GDP per capita and Suicides ratio in 1998'

)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# Visualization 3D Scatter Plot with Colorscaling

# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=dataframe.index,
    y=dataframe.suicides_no,
    z=dataframe["gdp_per_capita ($)"],
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(255,0,0)',        
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# Visualization with Multiple Subplots

trace1 = go.Scatter(
    x=dataframe.index,
    y=dataframe.suicides_no,
    name = "Suicides"
)
trace2 = go.Scatter(
    x=dataframe.index,
    y=dataframe.population,
    xaxis='x2',
    yaxis='y2',
    name = "Population"
)
trace3 = go.Scatter(
    x=dataframe.index,
    y=dataframe["gdp_per_capita ($)"],
    xaxis='x3',
    yaxis='y3',
    name = "GDP per capita"
)
trace4 = go.Scatter(
    x=dataframe.index,
    y=dataframe["suicides/100k pop"],
    xaxis='x4',
    yaxis='y4',
    name = "Suicides Ratio"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Suicides, population, GDP per capita and suicides Ratio in 1998'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

* As a result, we can see the relationship between population and suicide.
