```python
#Codes by ask9 https://www.kaggle.com/arbazkhan971/indians-vs-world/notebook

def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
    return color
```

```python
from pywaffle import Waffle
import random

role = df["Role"].value_counts()

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=role,
    colors = random_colors(len(role)),
    title={'label': 'Role Distribution', 'loc': 'left'},
    labels=["{}({})".format(a, b) for a, b in zip(role.index, role) ],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(df.groupby("Hero").get_group('Reinhardt')), 'framealpha': 0},
    font_size=30, 
    icons = 'child',
    figsize=(12, 5),  
    icon_legend=True
)
```
