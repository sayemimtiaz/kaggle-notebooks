```python
# Convert the date to age
from datetime import datetime, date

# the next function convert the date birth on age 

def age(born):
    born= datetime.strptime(born, "%d-%m-%y").date()  # "%d/%m/%Y"  %d-%m-%y
    today=date.today() # i'll use the today day because a I need the right format 
    return (today.year - born.year)-2 -((today.month,today.month) < (born.month,born.day)) # si el mes y dia de estudio es menor que el mes y dia de nacimiento se le resta 1, pues an no ha pasado su cumpleaños
```

```python
# create a new column with the age using de function age created 
data["Age"] = data["Date.of.Birth"].apply(age) 
# Now I'will erase the column day of birth 
data=data.drop(["Date.of.Birth"],axis=1) 
```

```python
# Eliminar outliers con estadistico z
"""
Z score is an important measurement or score 
that tells how many Standard deviation above or below a number is from the mean of the dataset

Any positive Z score means the no. of standard deviation above the mean and a negative score means no. 
of standard deviation below the mean
"""

# Z= X- M/S 
# La formula es la observacion menos la media sobre la desviación estandart 

# importamos el paquete 
from scipy import stats

# con esta linea de codigo aplicamos la función para encontrar el valor z de cada observación para nuestro dataframe
data['z_score']=stats.zscore(data['disbursed_amount']) # no la aplicaremos porque nos basaremos en el rango intercuartilico

# cualquier valor z > 3 o z < -3 se considera por lo general un outlier
data["z_score"] # vemos de manera simple cuantos datos tenemos como outlier 
```
