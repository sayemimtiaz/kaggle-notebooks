# Lending Club Case Study

### <font color=brown>Problem Statement</font>
A Consumer Finance company that specializes in lending various types of loans to customers in urban areas needs to make a decision if they should approve or deny the loan to the customer applying for the loan.

There are basically two types of risks associated with the bankâ€™s decision:
1. If the applicant is likely to repay the loan, then not approving the loan results in a loss of business to the company
2. If the applicant is not likely to repay the loan, i.e. he/she is likely to default, then approving the loan may lead to a financial loss for the company.

<b>The business objective is to identify patterns which indicate if a loan applicant is likely to default, which may be used for taking actions as given below:</b><br>
- Denying the loan
- Reducing the amount of loan
- Lending (to the risky applicants) at a higher interest rate etc.

When a person applies for a loan, there are two types of decisions that could be taken by the company:
1. <b>Loan accepted</b>: If the company approves the loan, there are 3 possible scenarios described below:
 1. Fully paid: Applicant has fully paid the loan (the principal and the interest rate)
 2. Current: Applicant is in the process of paying the instalments, i.e. the tenure of the loan is not yet completed. These candidates are not labelled as 'defaulted'.
 3. Charged-off: Applicant has not paid the instalments in due time for a long period of time, i.e. he/she has defaulted on the loan
2. <b>Loan rejected</b>: The company had rejected the loan (because the candidate does not meet their requirements etc.). Since the loan was rejected, there is no transactional history of those applicants with the company and so this data is not available with the company (and thus in this dataset).

<b>We are provided with a Comma Separated Values (CSV) and an excel files that contains the following data:</b><br>
- <i>loan.csv</i> - This file contains the customer details of the past loan applicants data and whether they defaulted or not
- <i>Data_Dictionary.xlsx</i> - This file describes the meaning of the variables provided in loan.csv file

### <font color=brown>Business Objectives</font>
Like most other lending companies, lending loans to â€˜riskyâ€™ applicants is the largest source of financial loss (called credit loss). The credit loss is the amount of money lost by the lender when the borrower refuses to pay or runs away with the money owed.
The goal is to identify the risky loan applicants using EDA, then such loans can be reduced thereby cutting down the amount of credit loss.

The company wants to understand the <b>driving factors (or driver variables)</b> behind loan default, i.e. the variables which are strong indicators of default so that this knowledge can be utilized for its portfolio and risk assessment.

#### Impute the data

```python
# The column 'emp_length' has some empty values
loan_cleaned_df['emp_length'].isnull().sum()
```

```python
# Replace these empty values with '0'
loan_cleaned_df['emp_length'].fillna('0',inplace=True)
```

```python
# Extracting the numeric value from emp_length and save it in a new column 'emp_experience_yrs' with data typecasted as 'int'
loan_cleaned_df['emp_experience_yrs']=loan_cleaned_df['emp_length'].str.extract('(\d+)').astype('float')
```

```python
# Deleting emp_length column as we will no more need this column
loan_cleaned_df = loan_cleaned_df.drop(['emp_length'], axis=1)
```

```python
# Let us chack the 'int_rate' column. We obseve that it is a string as it has '%' sign
loan_cleaned_df['int_rate']
```

```python
# Remove the '%' symbol from the column 'int_rate' and typecast it to float for future analysis
loan_cleaned_df['int_rate']=loan_cleaned_df['int_rate'].str.strip('%').astype('float')
```

```python
# Let us explore the loan_status column
loan_cleaned_df['loan_status'].value_counts()
```

```python
# Exclude the status 'Current' as we do not yet know the outcome of this if this would be defaulted or not
loan_cleaned_df=loan_cleaned_df[loan_cleaned_df['loan_status'].isin(['Fully Paid','Charged Off'])]
```

```python
# Create a column 'loan_defaulted' which will have value 1 if loan was defaulted else 0 if not defaulted
loan_cleaned_df['loan_defaulted']=loan_cleaned_df['loan_status'].apply(lambda x: 1 if x=='Charged Off' else 0 )
```

```python
# Let us explore the 'issue_d' column
loan_cleaned_df['issue_d']
```

```python
# Extract and format the year as '20xx' from 'issue_d' and save in new column'loan_funded_year' for future analysis
loan_cleaned_df['loan_funded_year'] = loan_cleaned_df['issue_d'].str.split('-').apply(lambda x:'20'+x[1]).astype('int')
```

```python
# Let us explore the column 'term'
loan_cleaned_df['term']
```

```python
# Remove the word 'months' from the column 'term',typecast it to 'int' and save in a new column 'term_in_months'
loan_cleaned_df['term_in_months']=loan_cleaned_df['term'].apply(lambda x: x.split()[0]).astype('int')
```

```python
# Delete the 'term' column as it is no more required
loan_cleaned_df = loan_cleaned_df.drop(['term'], axis=1)
```

```python
# Now we have a clean data set for our Analysis. let us check the final shape
loan_cleaned_df.shape
```

```python
# Let us now have a look at the final cleaned dataset
loan_cleaned_df
```
