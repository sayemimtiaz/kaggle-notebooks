Multiclass Classification, Decision Tree Algorithm

```python
    import pandas as pd
    from sklearn import tree
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import accuracy_score
    
    # Here i have used Decision Tree Algorithm 
    
    # Load the data set
    filename="//kaggle/input/iriscsv/Iris.csv"
    data_set = pd.read_csv(filename)
    
    data_set.head()

    # Visualize the data set using seaborn for specified columns
    # sb.violinplot(data=data_set, x='Species', y='SepalWidthCm')
    # plt.show()

    # Check whether data sets column contains any null value
    data_set.isnull().sum()

    # Check unique value on given column; unique = return the unique value
    data_set['Species'].unique()  # nunique = return the sum of unique value of column

    # Grouping the data set by passing column name data
    group_data = data_set.groupby(by='Species')
    # print(group_data.get_group('Iris-setosa'))

    # Drop the unwanted columns from dataset
    data_set.drop('Id', axis=1, inplace=True)

    # Divide the data set to get a input columns and target column
    input_column = data_set[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
                             'PetalWidthCm']]
    target_column = data_set['Species']

    # Categorical data - Get dummies for species column
    target_column_dummy = pd.get_dummies(target_column, prefix='Species')

    # Create object for tree
    object_tree = tree.DecisionTreeClassifier()

    # Splitting the model into train  and test set by passing input columns and target column
    x_train, x_test, y_train, y_test = train_test_split(input_column, target_column_dummy, test_size=0.2, random_state=0)

    # Train the model passing training values
    object_tree.fit(x_train, y_train)

    # Predict the output by passing test value
    predicted_x = object_tree.predict(x_test)

    # Predict the output for passing random values
    predicted_random_value = object_tree.predict([[5.0, 3.5, 1.3, 0.1]])
    # Answer is [[1 0 0]]

    # Test the score for our model by passing test set
    object_tree.score(x_test, y_test)

    # Another method to predict the accuracy score for our model
    accuracy_score(predicted_x, y_test)
```
