# Data Preprocessing

```python
# Grouping quality into high and low
df['quality'] = pd.cut(df['quality'], bins = (2,5,8), labels = ['low', 'high'])
df['quality'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
lab_enc = LabelEncoder()
```

```python
df['quality'] = lab_enc.fit_transform(df['quality'])
df.head(10)
```

# Feature Selection

```python
# Using train_test_split method for splitting dataset into 
# train and test data. 
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('quality', axis = 1) 
# All the variables are selected as independent variable except quality
Y = df['quality'] # Quality variable is target variable.
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.4, random_state = 52)
# Splitting dataset into 60% training set and 40% testing set with
# random state of 52.
```

```python
# Using standard scaling to get optimized result
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

# Machine Learning

```python
# Random forest , Neural Network and SVM are tested. First, SVM...
from sklearn.svm import SVC
```

```python
svm_clsf = SVC()
```

```python
svm_clsf.fit(X_train, Y_train)
svm_pred = svm_clsf.predict(X_test)
```

```python
# Using confusion matrix and classification report for 
# checking the performance of the model.
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(classification_report(svm_pred, Y_test))
print(confusion_matrix(svm_pred, Y_test))
```

Accuracy of SVM algorithm on this dataset is little bit low
And number of wrong labeled values are also high. 

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcs = RandomForestClassifier(n_estimators = 200)
```

```python
rfcs.fit(X_train, Y_train)
rfcs_pred = rfcs.predict(X_test)
```

```python
print(classification_report(rfcs_pred, Y_test))
print(confusion_matrix(rfcs_pred, Y_test))
```

Performance of random forest algorith is quite better than SVM
And number of wrong labeled values are less than SVM.

```python
# Neural Network
from sklearn.neural_network import MLPClassifier
```

```python
mlpc = MLPClassifier(hidden_layer_sizes = (11,11,11), max_iter = 500)
```

```python
mlpc.fit(X_train, Y_train)
mlpc_pred = mlpc.predict(X_test)
```

```python
print(classification_report(mlpc_pred, Y_test))
print(confusion_matrix(mlpc_pred, Y_test))
```

Neural network is the least accurate algorithm in this dataset
Because of less accuracy and more number of wrong labeled values.
