```python
data["PriceinGermany"] = data.PriceinGermany.str.strip('‚Ç¨')
data['PriceinGermany'] = data.PriceinGermany.str.replace(',','')
```

```python
data["PriceinGermany"] = data.PriceinGermany.astype(float)
data["PriceinGermany"]
```

```python
data["PriceinGermany"] = round(data["PriceinGermany"].fillna(data["PriceinGermany"].mean()))
```

```python
data["PriceinGermany"]
```

```python
data["PriceinUK"] = data.PriceinUK.str.strip('¬£')
data['PriceinUK'] = data.PriceinUK.str.replace(',','')
```

```python
data["PriceinUK"] = data.PriceinUK.astype(float)
data["PriceinUK"] = round(data["PriceinUK"].fillna(data["PriceinUK"].mean()))
```

Cleaning the rest of data


```python
data["Acceleration"] = data.Acceleration.str.strip(' sec')
data["Acceleration"] = data.Acceleration.astype(float)
data["TopSpeed"] = data.TopSpeed.str.strip(' km/h')
data["TopSpeed"] = data.TopSpeed.astype(float)
data["Range"] = data.Range.str.strip(' km')
data["Range"] = data.Range.astype(float)
data["Efficiency"] = data.Efficiency.str.strip(' Wh/km')
data["Efficiency"] = data.Efficiency.astype(float)
```

```python
data["FastChargeSpeed"] = data.FastChargeSpeed.str.strip("km/h")
```

```python
data["FastChargeSpeed"] = data["FastChargeSpeed"].replace("-","0")
```

```python
data["FastChargeSpeed"] = data.FastChargeSpeed.astype(float)
```

```python
data["FastChargeSpeed"] = round(data.FastChargeSpeed.replace(0,data["FastChargeSpeed"].mean()))
```

# Giving different parameters weightages to create an overall rating of the cars
I have given the weightages as per my needs that I would have wanted in my brand new electric car.

```python
a = 5 
# for efficiency
b = 4
# for FastChargeSpeed
c = 3
# for TopSpeed
d = 2
# for Acceleration
e = 1
# for Range
```

```python
rating = data["Efficiency"]*a +data["FastChargeSpeed"]*b + data["TopSpeed"]*c + data["Acceleration"]*d + data["Range"]*e
data["Rating"] = rating
 
```

```python
plt.figure(figsize = (12,7))
data_rat = data[["Name","Rating","PriceinGermany"]].sort_values(by = ["Rating"],ascending = False)
data_rat_20 = data_rat[:20]
sns.barplot(x = data_rat_20["Name"],y = data_rat_20["Rating"],palette = "viridis" )

plt.xticks(rotation = "vertical")
plt.xlabel("Car Name")
```

# Comparing the Ratings with the Prices, to find the  best choice for Electric Car

```python
plt.figure(figsize = (12,7))
sns.barplot(x = data_rat_20["Name"],y = data_rat_20["PriceinGermany"],palette = "viridis" )
plt.xticks(rotation = "vertical")
```

```python
rating_w_price = data["Rating"]**2/data["PriceinGermany"]
data["Rating_w_price"] = rating_w_price
```

```python
data_rp = data[["Name","Rating_w_price"]].sort_values(by = ["Rating_w_price"],ascending = False)
data_rp_20 = data_rp[:20]
plt.figure(figsize = (12,7))
sns.barplot(x = data_rp_20["Name"],y = data_rp_20["Rating_w_price"],palette = "cubehelix")
plt.xticks(rotation = "vertical")
```
