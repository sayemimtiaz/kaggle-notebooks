# Covid-19 Seems to Love Pd Fungus.

This notebook compares the known numbers of cases and deaths of Covid-19 to the known numbers of cases of white-nose syndrome in bats.

This comparison finds a noteworthy correlation between occurrences of Codid-19 and white-nose syndrome.

This finding implies a need to: (1) test the presence of WNS in regions where Covid-19 has occurred but WNS has not been checked, (2) test the presence of Covid-19 in regions where WNS has occurred, and (3) study whether and how WNS fungus spreads the Covid-19 virus.

### Covid-19

COronaVIrus Disease 2019 (COVID-19) was first reported in December 2019 in China's Wuhan city, and has now engulfed and locked down the entire world.

Caused aerially by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2), it leads to lungs congestion, fever, cough, muscle pain, sputum production, diarrhea, sore throat, loss of smell, and abdominal pain.

It was found that Wuhan's Covid-19 occurrences were primarily through bats. It is this finding that has prompted this notebook.

### White-Nose Syndrome (WNS)

In 2016, wildlife researchers in New York found some bats with an abnormal fuzzy white appearance on the muzzle, wings, and ears. This disease was named as White-Nose Syndrome (WNS).

Subsequent research has found that:
- WNS has killed more than 6 million bats in eastern states of USA during 2006-2019.
- WNS is present, and is spreading at an alarming rate, in bats in many regions of the world.
- WNS affects blood circulation, thermoregulation, water balance, and gas exchange in bats.
- WNS causes bats to arouse more often during hibernation, burn up fat reserves, and later starve to death.
- WNS is caused by Pseudogymnoascus destructans (Pd) fungus, growing on and in the skin of bats mainly during winter hibernation.

## Correlated Data

- Group Covid and WNS data in each bin of 2 latitude and 2 longitude.
- Join the Covid and WNS grouped data by bin.
- For each grouped record, assign severities in terms log of values.

```python
dfCg = dfC[['LatB','LngB','CovCases','CovDeaths']].groupby(['LatB','LngB']).sum().reset_index()
dfWg = dfW[['LatB','LngB','WnsYears']].groupby(['LatB','LngB']).max().reset_index()

dfCW = pd.merge(dfCg, dfWg, on=['LatB','LngB'], how='outer', indicator=False)
dfCW = dfCW.fillna(0)

dfCW['CovCas'] = dfCW.apply(lambda x: int(math.log(x['CovCases']+9,10)), axis=1)
dfCW['CovDth'] = dfCW.apply(lambda x: int(math.log(x['CovDeaths']+9,10)), axis=1)
dfCW['WnsYrs'] = dfCW.apply(lambda x: int(math.log(x['WnsYears']+1,2)), axis=1)

dfCW = dfCW.sort_values(by=['CovCases','CovDeaths','LatB'], ascending=False)

pd.options.display.float_format = '{:,.0f}'.format
dfCW.head()
```

### Correlational Maps

```python
zmap = folium.Map(location=[0, 0], zoom_start=2)

colorC = {0:'yellow', 1:'orange', 2:'coral', 3:'tomato', 4:'red'}

for la,lo,cca,cda,wya,cc,cd,wy in zip(dfCW['LatB'],dfCW['LngB'],dfCW['CovCases'],dfCW['CovDeaths'],dfCW['WnsYears'],dfCW['CovCas'],dfCW['CovDth'],dfCW['WnsYrs']):
    
    if cc > 0:
        folium.CircleMarker(
            [la, lo],
            radius = 3*cc+1,
            popup = str(cca) + '<br>' + str(cda) + '<br>' + str(wya) ,
            color = 'b',
            key_on = cc,
            threshold_scale = [0,1,2,3,4],
            fill_color = colorC[cc],
            fill = True,
            fill_opacity = 1
            ).add_to(zmap)
    
    if wy > 0:
        colorW = ['black','grean','green','blue','blue'][wy]
        folium.PolyLine([[la-1/2,lo-1/2],[la+1/2,lo+1/2]], color=colorW).add_to(zmap)
        folium.PolyLine([[la-1/2,lo+1/2],[la+1/2,lo-1/2]], color=colorW).add_to(zmap)
        
zmap
```

### Correlational Index

- Define Correlation index as "(4 - absolute of difference between CovCas and WnsYrs) if both are nonzero, and 0 otherwise".

```python
def corInd(c,w):
    if c>0 and w>0:
        return 4 - abs(c - w)
    else:
        return 0
```

```python
dfCW['CorrIndex'] = dfCW.apply(lambda x: corInd(x['CovCas'], x['WnsYrs']), axis=1)
dfCW['Bins'] = 1

dfCW.head()
```

```python
dfCor = dfCW[['CorrIndex','Bins']].groupby(['CorrIndex']).sum().reset_index()
dfCor['BinsPerc'] = 100*dfCor['Bins']/dfCW['Bins'].sum()

dfDict = pd.DataFrame(data=[[4,"Striking"],[3,"Noticeable"],[2,"Somewhat"],[1,"Faint"],[0,"None"]],
                        columns=['CorrIndex','Correlation'])

dfCor = pd.merge(dfCor, dfDict, on=['CorrIndex'], how='left', indicator=False)
dfCor = dfCor.sort_values(by=['CorrIndex'], ascending=False)

dfCor
```

### Conclusion and Recommendations

The correlation map and correlation index table presented here indicate that there exists a noteworthy correlation between Covid-19 occurrences and WNS occurrences.

This correlation implies two Covid-19 preventive measures. The regions where Covid-19 has occurred but WNS has not been checked should check occurrences of WNS as a preventive measure. The regions where WNS has occurred should check occurrences of Covid-19.

The correlation also implies that researchers should study whether WNS fungus is responsible for spreading the Covid-19 virus, and if yes, they should study whether the findings can lead to some cures for Covid-19.
