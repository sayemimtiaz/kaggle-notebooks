O Globo, (The Globe) is a Brazilian newspaper based in Rio de Janeiro. O Globo is the most prominent print publication in the Grupo Globo media conglomerate, which includes globo.com, which is the top 5 website in Brazil and top 134 worldwide.

Founded by journalist Irineu Marinho, owner of A Noite, it was originally intended as a morning daily to extend the newspaper interests of the company.
