# Pierre Gasly

The other young 24 year's old French Pilot. (Three Year of full F1 Competition)
Starting his first complete season in 2018, Pierre Gasly has made statement through his career.
In 2018, he didn't get has good results as Esteban Ocon starting his career. He finished 15th, with only 29 points. In deed, he had a low rate of Top Ten Qualifiers (28%), and his Top Ten presence was made from the same statement (23%).
But in this year, what is interesting to keep in mind, are his good transformation rates, leading him to pursue his career in 2019.

2019, Pierre Gasly started to perform.
He got his first podium ever, finishing second in the Brazilian Grand Prix.
Overall, he finished 7th with solid 95 points. He had a way better Top Ten Qualifiers Presence (62%) and an excellent presence in the Top Ten Race Results (67%).
This year was his best year in his career so far.

2020, new achievement for Pierre Gasly.
He got his first Victory ever in the Italian Grand Prix.
Overall, he finished 10th with good 75 points. He had a slightly worse compared to 2019 Top Ten Qualifiers Presence (64%) even if it's a lot of 10th, 9th, 8th place and a good enough presence in the Top Ten Race Results (58%).
This year was a good achievement year, and consistency remained there.
