# Introduction

The Nobel Prize is a set of annual international awards bestowed by Swedish and Norwegian institutions in recognition of academic, cultural, or scientific advances.

In 2019, 14 people were awarded with the Nobel Prize. Of these, only 1 was female (Esther Duflo, in the Economy category), representing 7% of the total in this year. And that makes it a pretty ordinary year, since the overall average is 0.47 women per year.

In the last 20 years, there have been more female Nobel laureates than in the first 90 years. However, comparing percentage values of the first decade (1900-1909) to the last decade (2010-2019), the inscrease was from 5 to 10%. It is progressing in the right direction, but maybe not fast enough!

Being interested in the subject and inspired by Iaroslava Mizai, I analyzed data of nobel prizes awarded to women from 1901 to 2019, and built this dashboard using **Tableau** to summarize them.

# Conclusions

The main conclusions of this analysis were:

- The first woman to win a Nobel Prize was Marie Curie, who was awarded in the Physics category in 1903. She is the only woman to have won multiple Nobel Prizes, since she was also awarded with the Nobel Prize in Chemistry in 1911. 

- Curie's daughter, Ir√®ne Joliot-Curie, won the Nobel Prize in Chemistry in 1935, making the two the only mother-daughter pair to have won Nobel Prizes.

- The most Nobel Prizes awarded to women in a single year was in 2009, when five women became laureates in four categories. However, of all 118 years analyzed in this dataset, 79 were marked for not having female representativity.

- The period between 2010 and 2019 was the decade in which most women got the award (13, representing 10.8% of the awards). On the order hand, no woman was awarded in the 50s. 

- Literature and Peace are the categories that rewards women the most (12.93% and 12.69% in total, respectively). However, Physics has the lowest representation of women: only 1,41%.

- On average, 0.47 women are awarded per year. That's right, LESS THAN ONE! Looking only at the last decade (2010-2019) this number has increased to 1.3 women per year. *Well, at least now we have a whole woman.*
