## 10. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan? *

> **We will use PIVOT TABLE to get the required dataframe.**

 
Actually pivot table comes very handy when you want to get the average values of certain column's data & it also
    provide neat and readable tabular expressions.

```python
df.pivot_table(index = "country", values = "hpw",columns = "salary", aggfunc = "mean")
```

```python
df[(df["country"]==' Japan')&(df.salary == " >50K")]["hpw"].mean()
```

>>>**Average hours per week in Japan who earns more than 50k is around 47.9**

```python
df[(df["country"]==' Japan')&(df.salary == ' <=50K')]["hpw"].mean()
```

>>>**Average hours per week in Japan who earns less than 50k is around 41**

# Data Visualisation:
*Now comes my favourite part of EDA ,so lets get our hand's dirtyðŸ˜‰*

# ( Age vs Hours-per-week)

```python
plt.scatter(df.hpw.head(1000),df.age.head(1000),marker = "^",s=50, color = "purple")
plt.xlabel("hours per week")
plt.ylabel("Age")
plt.gcf().set_size_inches(15,8)
plt.grid()
```

> **Observations:**

*Most people work at 40 hpw , followed by around 45,50 & 60 hpw.*

*People with age above 60 years , usually have less hpw.*

*Majority of people work in the range of 20-60 hpw.*




>>> **Men vs Women  (Age vs Hours-per-week)**

```python
plt.scatter(df[df.sex == " Male"].hpw.head(500),df[df.sex == " Male"].age.head(500),marker = "*",s=70,label="Male")
plt.scatter(df[df.sex == " Female"].hpw.head(500),df[df.sex == " Female"].age.head(500),marker = "o",s=20, label="Female")
plt.xlabel("hours per week")
plt.ylabel("Age")
plt.gcf().set_size_inches(15,8)
plt.legend()
```

>**Observations :**

*Less hpw(20-40) jobs are more occupied by women.*

*More hpw(50+) jobs are more occupied by men*


# (Salary vs  Sex vs hpw)

```python
sns.stripplot(x="sex",y = "hpw", data = df.head(5000), hue = "salary",size = 10, palette = "rocket")
plt.gcf().set_size_inches(15,8)
```

>**Observation :**

*Clearly Men gets better paid(50K +) as compared to Women , so do their hpw.*

# (Education vs Race vs Salary):

```python

sns.boxplot(x="salary",y= "education-num" ,hue ="race" ,data = df,width = 0.5,linewidth = 2.5,fliersize =0.4)
plt.gcf().set_size_inches(17,8)
```

>**Observation :**

*Those who have better education gets better paid*

*Asian-Pac-Islander, race clearly shines out in term of education*

# (Workclass vs Capital-gain):


```python
sns.catplot(x="workclass",y = "capital-gain", data = df, palette = "rocket")
plt.gcf().set_size_inches(15,8)
```
