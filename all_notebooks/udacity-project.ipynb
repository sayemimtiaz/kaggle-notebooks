### Data Cleaning (dropping unwanted columns ,reformate columns type and dealing with outliers)

***dropping N/As and unwanted***

```python
#creating varabiles for raw abd columns to not the The difference after dropping 
raw,col=raw_data.shape


raw_data.isnull().any()
```

```python
raw_data.columns
```

```python
#choosing the columns 
data=raw_data[['budget', 'genres','original_language',
       'popularity', 'production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime', 'title', 'vote_average',
       'vote_count']]

#dropping the N/A
df=data.dropna()
```

```python
#knowing how many col and rows we dropped 
new_raw,new_col=df.shape
print("we dropped:",raw-new_raw,'rows',',',col-new_col,'columns')
```

```python
df.head()
#we have in genres , production_countries , production_companies list of dicts ,,,, so we will try to turn it into str
```

**reformatting columns typs** : 

-we have in genres , production_countries and production_companies list of dicts 



-reformatting release_date to date type

```python
new_data=df.copy()

#function iterates over the list and return the first name in the first dictionary 
def name(col):
        names = [i['name'] for i in col]
        names = names[:1]
        return names
    
#function converts the list into a string 
def list_String(st):  
    string = " "  
    return (string.join(st))
    
#applying the functions:

new_data['genres'] =new_data['genres'].apply(literal_eval)
new_data['production_companies'] = new_data['production_companies'].apply(literal_eval)
new_data['production_countries']=new_data['production_countries'].apply(literal_eval)
#note:literal_eval is a function make the list readable 

new_data['genres'] = new_data['genres'].apply(name)
new_data['production_companies'] = new_data['production_companies'].apply(name)
new_data['production_countries']=new_data['production_countries'].apply(name)


new_data['genres'] = new_data['genres'].apply(list_String)
new_data['production_companies'] =new_data['production_companies'].apply(list_String)
new_data['production_countries']=new_data['production_countries'].apply(list_String)
```

```python
#reformatting the date type
new_data['release_date']=pd.to_datetime(new_data['release_date'])
```

### Question 2 ( what are the most popular genres ? )


```python
new_data.head(20).groupby(['genres']).count().plot.pie(y='budget',figsize=(10,10),autopct="%.2f")
plt.title('the largest genre')
plt.ylabel('');
```

***note***: the drama genre is the biggest in the 20 topp movie 

```python
tr =new_data.groupby(['genres','release_date']).count()[['budget']]
tr=tr.reset_index()
tr['count']=1
tr=tr.pivot_table(index='release_date',columns='genres',values='count',aggfunc='sum').resample('5Y').sum()

ttr=tr.T

ttr['sum']=ttr.sum(axis = 1, skipna = True) 

ttr.sort_values(by='sum',ascending=False,inplace=True)

ttr=ttr.head(6).drop('sum',axis=1)

ttr.T.plot(kind='barh',figsize=(10,18),grid=True)
plt.title('people teste changes over time');
```

##### ***note***: how people tests changes over time :

   #### -drama is only a head , only in the 90's era people starts to tend to Action and in the last 5 years .

### Question 3 (  what is the biggest country in movie industry ? )


```python
new_data.head(100).groupby(['production_countries']).count().plot.pie(y='budget',figsize=(10,10),autopct="%.2f")
plt.title('the counties movies producation');
plt.ylabel('number of movies');
```

```python
tr =new_data.groupby(['production_countries','release_date']).count()[['budget']]
tr=tr.reset_index()
tr['count']=1
tr=tr.pivot_table(index='release_date',columns='production_countries',values='count',aggfunc='sum').resample('10Y').sum()

ttr=tr.T

ttr['sum']=ttr.sum(axis = 1, skipna = True) 

ttr.sort_values(by='sum',ascending=False,inplace=True)

ttr=ttr.head(20).drop('sum',axis=1)

ttr.T.plot(kind='barh',figsize=(8,15),grid=True)
plt.title('people teste changes over time');
```

***note***: as we see the US has the biggest chunck in movie industry over time

### Question 4   what is the biggest company in movie industry ?
    

```python
new_data.head(20).groupby(['production_companies']).count().plot.pie(y='budget',figsize=(9,15),autopct="%.2f")
plt.title('the largest production_companies');
plt.ylabel('');
```

```python
tr =new_data.groupby(['production_companies','release_date']).count()[['budget']]
tr=tr.reset_index()
tr['count']=1
tr=tr.pivot_table(index='release_date',columns='production_companies',values='count',aggfunc='sum').resample('10Y').sum()

ttr=tr.T

ttr['sum']=ttr.sum(axis = 1, skipna = True) 

ttr.sort_values(by='sum',ascending=False,inplace=True)

ttr=ttr.head().drop('sum',axis=1)

ttr.T.plot(kind='barh',figsize=(15,10),grid=True)
plt.title('people teste changes over time');
```

### Question 5  comparing between revenue and budget .


```python
top_df=new_data.set_index('title')
top_df[['budget','revenue']].head(20).plot(kind='barh',stacked=True,figsize=(10,10),grid=True);
plt.title('the budget of the movie over the revenue of the 20 top movies');
plt.xlabel('money per hundred million ');
```

```python
new_data.index=new_data[['title']]
```

```python
new_data['profit']=new_data['revenue']-new_data['budget']
new_data.sort_values(by=['profit'],ascending=False)['profit'].head(20).plot(kind='barh',figsize=(10,10),grid=True);
plt.title('the most prfitable movies')
plt.xlabel('milion dillor')
```

***note***: as we see the dark knight has the largest scale ,whiche means he has the hightest revenue and budget as well

--------------------------------------------------------------------------------------------------------------

```python
new_data['profit']=new_data['revenue']/new_data['budget']
new_data.sort_values(by=['profit'],ascending=False)['profit'].head(20).plot(kind='barh',figsize=(10,10),grid=True);
plt.title('The film is the most profitable proportionately')
plt.xlabel('times profit')
```
