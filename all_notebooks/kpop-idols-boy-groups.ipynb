```python
fig = go.Figure();
fig.add_trace(go.Scatter(x = df['Members'].head(10),y = df['Orig. Memb.'],
                    mode='lines+markers',
                    name='Orig. Memb.'));

fig.update_traces(mode='lines+markers', marker_line_width=2, marker_size=10);

fig.update_layout(autosize=False, width=1000,height=700, legend_orientation="h");

fig.show();
```

```python
fig = px.treemap(df, path=['Members'], values='Orig. Memb.',
                  color='Orig. Memb.', hover_data=['Members'],
                  color_continuous_scale='rdylbu')
fig.show()
```
