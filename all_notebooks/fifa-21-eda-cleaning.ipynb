# **1. Importing Libraries & Data Exploration** <a class="anchor" id="1"></a>

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import pi #Used in Radar Plots
import seaborn as sns
sns.set_style(style='darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
fifa_21 = pd.read_csv('../input/fifa-21-complete-player-dataset/players_21.csv')
```

# **2. Data Manipulation & Cleaning** <a class="anchor" id="2"></a>

```python
useless_column = ['dob','sofifa_id','player_url','long_name','body_type','real_face','nation_position','loaned_from','nation_jersey_number']
```

```python
fifa_21 = fifa_21.drop(useless_column, axis = 1)
```

```python
fifa_21['BMI'] = fifa_21 ['weight_kg'] / (fifa_21['height_cm'] / 100) ** 2
```

```python
new_player_position = fifa_21['player_positions'].str.get_dummies(sep=',').add_prefix('Position')
```

```python
fifa_21 =  pd.concat([fifa_21,new_player_position],axis = 1)
fifa_21 =  fifa_21.drop('player_positions',axis=1)
```

```python
columns = ['ls','st','rs','lw','lf','cf','rf','rw','lam','cam','ram','lm','lcm','cm','rcm','rm','lwb','ldm', 'cdm','rdm','rwb','lb','lcb','cb','rcb','rb']
```

```python
for col in columns:
  fifa_21[col] = fifa_21[col].str.split('+',n=1,expand = True)[0]
        
```

```python
fifa_21[columns] = fifa_21[columns].fillna(0)
fifa_21[columns] = fifa_21[columns].astype(int)
```

```python
columns = ['dribbling','defending','physic','passing','shooting','pace']
```

```python
fifa_21[columns].isnull().sum()
```

```python
for col in columns:
  fifa_21[col] = fifa_21[col].fillna(fifa_21[col].median())
```

```python
fifa_21 = fifa_21.fillna(0)
```

```python
fifa_21.isnull().sum() #Verifying it Should be all zero
```

# **3. Let's Start with Analysis**<a class="anchor" id="3"></a>

```python
sns.relplot(x='overall',y='value_eur',hue='age',palette = 'viridis',size="BMI", sizes=(15, 200),aspect=2,data=fifa_21)
plt.title('Overall Rating vs Value in Euros',fontsize = 20)
plt.xlabel('Overall Rating')
plt.ylabel('Value in Euros')
plt.show()
```

```python
sns.relplot(x='potential',y='wage_eur',hue='age',palette = 'viridis',size="BMI", sizes=(15, 200),aspect=2,data=fifa_21)
plt.title('Potential Rating vs Wage in Euros',fontsize = 20)
plt.xlabel('Potential')
plt.ylabel('Wage in Euros')
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot('preferred_foot',data=fifa_21,palette='Blues')
plt.xlabel('Preferred Foot Players')
plt.ylabel('Count')
plt.title('Count of Preferred Foot')
Right,Left=fifa_21.preferred_foot.value_counts()
print('Left Preferred',Left)
print('Right Preferred',Right)
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot('international_reputation',data=fifa_21.head(100),palette='Blues')
plt.xlabel('International Reputation')
plt.ylabel('Count')
plt.title('Count of International Reputation of Top 100 Players')
plt.show()
```

```python
plt.figure(dpi=125)
sns.distplot(a=fifa_21['age'],kde=False,bins=20,color='red')
plt.axvline(x=np.mean(fifa_21['age']),c='green',ls='--',label='Mean Age of All Players')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(dpi=125)
sns.distplot(a=fifa_21['BMI'],kde=False,bins=20,color='green')
plt.axvline(x=np.mean(fifa_21['BMI']),c='black',ls='--',label='Mean BMI of All Players')
plt.legend()
plt.xlabel('BMI')
plt.ylabel('Count')
plt.title('Distribution of BMI')
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=125)
x=fifa_21['weight_kg']
y=fifa_21['pace']

sns.regplot(x,y,color='orange')
plt.title('Weight vs Pace',fontsize = 20)
plt.xlabel('Weight')
plt.ylabel('Pace')
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=125)
x=fifa_21['BMI']
y=fifa_21['pace']

sns.regplot(x,y,color='skyblue')
plt.title('BMI vs Pace',fontsize=20)
plt.xlabel('BMI')
plt.ylabel('Pace')
plt.show()
```

# **3.2 - The Biggest Rivalry - MESSI VS RONALDO**<a class="anchor" id="5"></a> 

```python
column_r=['short_name','overall','potential','pace','shooting','passing','dribbling','defending','physic']
rival = fifa_21.head(2)[column_r]
rival= rival.set_index('short_name')
rival.head()
```

```python
values = rival.iloc[0].tolist()
values += values [:1]
values
```

```python
values2 = rival.iloc[1].tolist()
values2 += values2[:1]
values2
```

```python
def RadarChart(player, data, player2, data2):
    Attributes = ["Overall","Potential","Pace","Shooting","Passing","Dribbling","Defending","Physical"]

    data += data [:1]
    data2 += data2 [:1]
    
    angles = [n / 8 * 2 * pi for n in range(8)]
    angles += angles [:1]
    
    angles2 = [n / 8 * 2 * pi for n in range(8)]
    angles2 += angles2 [:1]
    plt.figure(dpi=125)
    ax = plt.subplot(111, polar=True)
    ax = plt.subplot(111, polar=True)
    
    plt.xticks(angles[:-1],Attributes)

    ax.plot(angles,values)
    ax.fill(angles, values, 'teal', alpha=0.1)

    ax.plot(angles2,values2)
    ax.fill(angles2, values2, 'red', alpha=0.1)

    plt.figtext(0.1,0.98,player,color="steelblue")
    plt.figtext(0.1,0.94,"vs")
    plt.figtext(0.1,0.9,player2,color="darkorange")
    
    plt.show()
```

```python
RadarChart("Messi",[94.0, 94.0, 87.0, 92.0, 92.0, 96.0, 39.0, 66.0, 94.0],"Ronaldo",[93.0, 93.0, 90.0, 93.0, 82.0, 89.0, 35.0, 78.0, 93.0])
```

Out of 8 attributes - **Messi performed well in 5** and **Ronaldo in 3**
