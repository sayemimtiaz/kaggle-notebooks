Default argument example: 
def f(a, b=1):
   """ b = 1 is default argument"""  a is need to be filled
   
Flexible argument example: 
def f(*args):
 """ *args can be one or more"""

def f(** kwargs) 
 """ **kwargs is a dictionary"""


```python
def f(a,b=1,c=2):
    y=a+b+c
    return y
print (f(5)) #this case b and c is default a is flexible and it is 5.
print(f(5,4,3))
```

```python
def f (*args):
    for i in args:
        print (i)
f(1)
f(1,2,3,4)
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    for key,value in kwargs.items():
        print (key," ",value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

**ZIP example**

```python
# zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z=zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip) # unzip returns tuble
print(un_list1)
print(un_list2)
print(type(un_list2))
```

**LIST COMPREHENSİON**
    
    One of the most important topic of this kernel 
    We use list comprehension for data analysis often.
    less code more work
    
    How it works: Logically loops or conditions are on the end, first thing (result or condition) on the left the others is right. 
    

```python
num1 = [1,2,3]
num2=[i+1 for i in num1]
print (num2)
```

```python
# Conditionals on iterable
nm1 = [5,10,15]
nm2 =[i**2 if i==10 else i-5 if i<7 else i+5 for i in nm1]
print (nm2)
```
