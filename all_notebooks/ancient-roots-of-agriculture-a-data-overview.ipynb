Recent evidences suggests agriculture in rome, chile, chain, mexico, native american tribes, russia, india, yemen and few several other places. <strong>Natual Geographics Areas are old ruins of forgotten civilizations</strong>, where once innovations flourished, practised agriculture, have military, and other basic activities for successful administration. In current times, archeologists discovered as many as 35 such places, and recognized as NGAs, and These NGAs spread over almost all continents and evidences does support the pracise of agriculture at such spots. Eventualy these places become not-existant because of some natural calamity or war related things. Findings from these sites and carbon grading, suggests that agriculture at these places is nearly 10,000 years old.

Where does these NGAs located? as per recent findings,<strong> NGAs are located in North america,south america, europe, asia, and african continents.</strong> Though they are miles apart, few similarities in the practise of agriculture such as irrigation systems, agricultrue methods, fertilizers. 

Interestingly, at these <strong>NGAs few technologies and methods were invented and forgotten few times in a cycle</strong>, and this dataset has tried to cover such cyclic apperarences and disapperarences, throughout timeline from stone age to present period.

<p style = "font-size:30px; color: #007580 "><strong>Let's see where these Natutal Geographic Areas located around world.....</strong></p> 


```python
# https://stackoverflow.com/questions/25888396/how-to-get-latitude-longitude-with-python

def lat_log_finder(place):
    address = place
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'
    response = requests.get(url).json()
    lat = response[0]["lat"]
    log = response[0]["lon"]
    return [lat, log]



##  mapping NGA to actual locations with latitude and longitude values

NGA_list = df.NGA.unique().tolist()
 
NGA_corrected  =['Big Island Hawaii',
 'Cahokia',
 'Cambodia',
 'Central Java',
 'Chuuk Islands',
 'Cuzco',
 'Narmada river, india',
 'Finger Lakes, New york, usa',
 'Garo Hills',
 'Ghana',
 'Iceland',
 'Kachhi, pakistan',
 'Kansai',
 'Kalimantan',
 'Konya Plain',
 'Latium',
 'Lena River Valley',
 'ecuador,south america',
 'Middle Yellow River Valley',
 'Niger River',
 'North Colombia',
 'Orkhon Valley',
 'Port Moresby',
 'Paris Basin',
 'Sogdiana',
 'zhaotong',
 'ahvaz,iran',
 'Upper Egypt',
 'Valley of Oaxaca',
 'sana, yemen']

## mapping corrected NGA names  to get latitude and longitude locations
NGA_corrected_dir = { key:val for key,val in zip(NGA_list,NGA_corrected)}
df['locations']= df['NGA'].map(NGA_corrected_dir).apply(lambda x: lat_log_finder(x))                                      
df['latitude'] = df['locations'].apply(lambda x: x[0])
df['longitude'] = df['locations'].apply(lambda x: x[1])
df['latitude'] = df['latitude'].astype('float')
df['longitude'] = df['longitude'].astype('float')


## creating an point objects to make scatter  plot and locate  NGAs
crs = {'init':'epsg:4326'}
geometry = [Point(y,x) for x,y in zip(df['latitude'],df['longitude'])]
geo_df = gpd.GeoDataFrame(df.copy(),crs = crs, geometry = geometry)






#loading geodataframe
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world[world['name']!='Antarctica']

fig, ax  = plt.subplots(figsize = (20,15),dpi =100)

fig.patch.set_facecolor('#f6f5f5')
ax.set_facecolor('#f6f5f5')
world.dropna().plot( scheme='quantiles', 
                       k=10, legend = False, linewidth = 1,
                       ax = ax, color = colors[-2], alpha = 0.8)
geo_df.plot(color =colors[-1], marker = '^',ax = ax,markersize = 70)

loc_check = []
for loc,y,x in zip(df['NGA'], df['latitude'],df['longitude']):
    if loc in loc_check:
        continue
    if x < 20:
        ha = 'right'
        x_off = -4
        y_off = 1.5
    elif x>20:
        ha = 'left'
        x_off = 2
        y_off =0
    ax.annotate(loc, xy= (x,y),xytext= (x+x_off,y+y_off),
                    fontsize=11.5,fontfamily='serif',fontweight ='bold',ha=ha, va='bottom',color='black', alpha =0.8)
    loc_check.append(loc) 
    
for loc in ['left','right','top','bottom']:
    ax.spines[loc].set_visible(False)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)


### titles and tags



fig.text(0.12,0.77,'Ancient Roots of Agriculture: NGAs on Worldmap' ,**{'font':'serif', 'size':25,'weight':'bold',}, alpha = 0.9)

fig.text(0.12,0.725,'''This visualization shows actual locations of NGA (Natural Geographic Areas),
where lost civilizations which were found by archaeologists. Several Interesting 
things about agriculture are found at these locations.''',**{'font':'serif', 'size':14,}, alpha = 0.8)

fig.text(0.77,0.28,'© Made by bhuvanchennoju/Kaggle',{'font':'serif', 'size':8, 'weight':'bold',},alpha = 0.7)

fig.show()
```

With a glimpse on this timeline we got to know that we are handling data from mesothlic period to relatively currently period. Lets bin out time data into historical periods as per archeological timelines.


<ul style = "font-weight: bold; font-size:16px">Timelines are categorised as below
    <li> Mesolithic Period: -10000 to - 6000</li>
    <li> Neolithic Period : -6000 to -2200</li>
    <li> Bronze Age: -2200 to -800</li>
    <li> Iron Age: -800 to -52</li>

 <p style = "font-size:30px; color: #007580 "><strong> okay we have understood about historical timelines, 
 lets see active periods of NGAs with time period start and end in general.....</strong></p>

```python
####### binning date from and date to to respective time preriods

bins = [-10000,-6000,-2200,-800,-52,500,1500,1800,1990,2010]
time_periods = ['Mesolithic Period','Neolithic Period','Bronze Age',
                'Iron Age','Antiquity Ages',
                'Middle Ages','Early Modern Period',
                'Late Modern Period','Technology Period']

#df.drop(columns = ['locations'], inplace = True)
df['Start_Period'] = pd.cut(df['Date From'], bins = bins ,labels = time_periods)
df['End_Period'] = pd.cut(df['Date To'], bins = bins ,labels = time_periods)


value1 = (df[['NGA','Start_Period','End_Period']].groupby('NGA')['Start_Period']
                                             .value_counts().sort_index().unstack().fillna(0))

value2 = (df[['NGA','Start_Period','End_Period']].groupby('NGA')['End_Period']
                                             .value_counts().sort_index().unstack().fillna(0))


map_cols = [colors[1], colors[-2]]
colormap = matplotlib.colors.LinearSegmentedColormap.from_list("", map_cols)




#################### NGAs and Timeperiods
fig, ax = plt.subplots(1,2, figsize = (10,15),dpi = 100)
fig.patch.set_facecolor('#f5f6f6')
axes = ax.ravel()
for ax in axes:
    ax.set_facecolor('#f5f6f6')
    
sns.heatmap(value1,annot = True, ax = axes[0], vmin = 1,vmax=5,linewidth = 0.8, linecolor='#f5f6f6',
            square = True,cbar = False,cmap = colormap,annot_kws = {'font':'serif', 'size':8, 'weight':'bold',})
sns.heatmap(value2,annot = True, ax = axes[1], vmin = 1, vmax = 5,linewidth = 0.8, linecolor='#f5f6f6',
            square = True, cbar = False, cmap = colormap,annot_kws ={'font':'serif', 'size':8, 'weight':'bold',})



##### ticks and labels 
for ax in axes:
    ax.tick_params(axis='x', which='major', 
                    labelsize=9, labelbottom =True,
                    bottom=False, top = False, labeltop=True,rotation = 90)
    ax.set_ylabel('')
    ax.set_xticklabels(time_periods,{'font':'serif', 'size':9, 'weight':'bold',},alpha = 0.8)
    ax.set_yticklabels(NGA_list,{'font':'serif', 'size':9, 'weight':'bold',},alpha= 0.8)

axes[1].tick_params(axis='y', which='major', 
                    labelsize=9, labelright =True,left =False,right=False,labelleft=False,rotation = 0)

axes[0].set_xlabel('Start Period',{'font':'serif', 'size':9, 'weight':'bold',}, alpha = 0.8)
axes[1].set_xlabel('End Period',{'font':'serif', 'size':9, 'weight':'bold',},alpha=0.8)
axes[0].text(3,-4.5,'Start Period',{'font':'serif', 'size':9, 'weight':'bold',},alpha=0.8)
axes[1].text(3,-4.5,'End Period',{'font':'serif', 'size':9, 'weight':'bold',},alpha=0.8)

#### final writigs and titles


fig.text(-0.1,1.045,'Ancient Roots of Agriculture: Active Timeperiods of NGAs ' ,**{'font':'serif', 'size':20,'weight':'bold',}, alpha = 0.9)

fig.text(-0.1,1.02,'''This visualization shows when did an implementation was Started for a civilization at a given NGA,
and how many peroids does that existed before disapearence.''',**{'font':'serif', 'size':12,}, alpha = 0.8)

fig.text(0.85,0,'© Made by bhuvanchennoju/Kaggle',{'font':'serif', 'size':8, 'weight':'bold',},alpha = 0.7)

fig.show()
```

2. Agricultural Techniques: These are techniques involved in agriculture such as ploughing, sowing, irrigation, fertilizers, and harvesting.
3. Productivity and Crops: These are about which crops to grow, how much quantity to produce for a harvest.
