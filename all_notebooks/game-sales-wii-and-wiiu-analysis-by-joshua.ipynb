> # **3. Why Nintendo Wii Is A Massive Success For Nintendo**


As I say in the previous chapter, Wii seems a massive success with its 3 games being on the top 5 best selling games of all time. Lets compare its total games sales to other consoles.

First off, lets see total of global sales for Wii games

```python
print('Total Wii Games Sales (in million) : ',data_vg[data_vg.Platform == 'Wii'].Global_Sales.sum())
```

Total Nintendo Wii games sales in this are 909.810.000 copies, a really massive ammount. But lets compare it first to the other consoles.
Now, we should calculate the other Nintendo consoles's games sales first.

```python
data_vg.Platform.unique()
console_exwii = ['NES', 'GB', 'DS', 'X360', 'PS3', 'PS2', 'SNES', 'GBA','3DS', 'PS4', 'N64', 'PS', 'XB', 'PC', '2600', 'PSP', 'XOne', 'GC','WiiU', 'GEN', 'DC', 'PSV', 'SAT', 'SCD', 'WS', 'NG', 'TG16','3DO', 'GG', 'PCFX']
nintendo_consoles = ['Wii', 'NES', 'GB', 'DS', 'SNES', 'GBA', '3DS', 'N64', 'WiiU', 'GC']
```

```python
nintendo_gamesales = []
for i in range(len(nintendo_consoles)):
    a = data_vg[data_vg.Platform == nintendo_consoles[i]].Global_Sales.sum()
    nintendo_gamesales.append(a)
nintendo_dict = {'Nintendo Consoles' : nintendo_consoles,
                'Total Game Sales' : nintendo_gamesales}
data_vggs_nintendo = pd.DataFrame.from_dict(nintendo_dict, orient = 'index') 
data_vggs_nintendo = data_vggs_nintendo.transpose()
data_vggs_nintendo = data_vggs_nintendo.sort_values(['Total Game Sales'], ascending = False).reset_index(drop=True)
data_vggs_nintendo
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x = 'Nintendo Consoles', y = 'Total Game Sales', data = data_vggs_nintendo).set_title('Wii vs Other Nintendo Consoles', fontsize = 36)
```

With the graph shown above, it is so easy to see why Wii was and maybe still is the most successful console Nintendo has ever released. This console was only nearly matched by the most successful Nintendo handheld console, Nintendo DS. 

Next up, lets see how Nintendo Wii stands with other companies consoles.

```python
console_exnintendo = console_exwii
for i in nintendo_consoles:
    if i != 'Wii':
        console_exnintendo.remove(i)
    else:
        pass
console_exnintendo.append('Wii')
```

```python
console_gamesales = []
console_dict ={'Consoles' : console_exnintendo, 'Total Game Sales' : console_gamesales}
for i in range(len(console_exnintendo)):
    a = data_vg[data_vg.Platform == console_exnintendo[i]].Global_Sales.sum()
    console_gamesales.append(a)
data_vggs_ex_nintendo = pd.DataFrame(console_dict, columns = ['Consoles', 'Total Game Sales'])
data_vggs_ex_nintendo.sort_values(['Total Game Sales'], ascending = False).reset_index(drop = True)
```

Let's drop console with sales below 100 million

```python
data_vggs_ex_nintendo.info()
```

```python
data_vggs_ex_nintendo = data_vggs_ex_nintendo[data_vggs_ex_nintendo['Total Game Sales'] > 100]
data_vggs_ex_nintendo = data_vggs_ex_nintendo.sort_values(['Total Game Sales'], ascending = False).reset_index(drop = True)
data_vggs_ex_nintendo
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x = 'Consoles', y = 'Total Game Sales', data = data_vggs_ex_nintendo).set_title('Wii vs Other Consoles',fontsize = 36)
```

So, when compared to other consoles, Nintendo Wii still holds up. It places 4th overall, a really good place considering other Nintendo Consoles don't make it into this list. > # **4. Why Nintendo WiiU Is A Flop Compared to Nintendo Wii**


WiiU was released in 2012 and the console includes the main console with a new controller called WiiU gamepad. The gamepad is sort of the classic Wii gamepad with a touchscreen in the middle of it. In 2012, this kind of thing is really an innovative thing in console world. But innovative doesn't mean high sales, so lets see how WiiU does in term of its game sales.

![](https://d2pa5gi5n2e1an.cloudfront.net/webp/global/images/product/gameconsoles/Nintendo_Wii_U/Nintendo_Wii_U_L_1.jpg)

Lets see the WiiU total game sales

```python
print('Total Wii Games Sales (in million) : ',data_vg[data_vg.Platform == 'WiiU'].Global_Sales.sum())
```

Wow, a really decreased number compared to its predecessor. Now, to see it clearly lets just compare it to other nintendo console

```python
data_new_nintendo = data_vggs_nintendo.sort_values(['Total Game Sales'], ascending = True).reset_index(drop = True)
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x = 'Nintendo Consoles', y = 'Total Game Sales', data = data_new_nintendo).set_title('WiiU vs Other Nintendo Consoles', fontsize = 36)
```

![](https://media2.giphy.com/media/l3q2K5jinAlChoCLS/200w.gif)

Man, that games sales is really low. Well, it is true that this is one of the newer console and the age is still young compared to other consoles, but with the release of Nintendo Switch in early 2017, I think that WiiU games sales is not going to go head to head with Wii's sales.

I don't have this console personally, but I have seen their advertisements and review in many internet websites. From them, I believe that the console itself is good but everything else is a lackluster. There are some things that, at least for me, made WiiU a failure :

* **Really Confusing Controller Scheme :** This is the number one reason for me. Previous Wii Gamers went into WiiU with the mindset of WiiU is an upgraded Wii, but with the addition of the WiiU gamepad it makes it confusing. In a party games like Wii Party U or Mario Party 10, the players need to use normal Wii controllers (they might need to buy it themself !) with an additional WiiU gamepad and the usage is really confusing concept, well at least for me.

* **Nintendo's Market Target :** WiiU is targeted for children. This may not be a problem because Wii is also a family oriented console, but this I think create a wrong hardware choices. Because usually in a Nintendo games, family oriented games don't need high performance hardware capability because of their cartoony and simple graphics. This creates a new problem that I take on the next point

* **Underpowered Machine :** With the rise of AAA games at that time like Grand Theft Auto 5, Watch Dogs, and many more, WiiU is not a platform choice for the developer because the difficulty of porting and its underpowered machine compared to PS3 and XBOX 360. This makes WiiU placed in some sort of limbo in which it creates an exclusivity, but it is not a choice for AAA games developers. This doesn't mean mature games didn't have a place in WiiU, look at Bayonetta and Bayonetta 2, it's just that WiiU is not a choice for multi-platform AAA games.\

Those are the reasons for why I think WiiU flopped, if you want to know more there is a good article [here](https://www.lifewire.com/reasons-the-wii-u-is-a-failure-2498588) that lists why WiiU flops.
