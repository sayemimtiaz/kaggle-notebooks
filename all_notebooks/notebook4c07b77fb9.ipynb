```python
df['RAM'] = df['RAM'].fillna((df['RAM'].mean()))
```

```python
df['OnKameraMP'] = df['OnKameraMP'].fillna((df['OnKameraMP'].mean()))
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["Renk"]=le.fit_transform(df["Renk"])
df["Bluetooth"]=le.fit_transform(df["Bluetooth"])
df["CiftHat"]=le.fit_transform(df["CiftHat"])
df["4G"]=le.fit_transform(df["4G"])
df["3G"]=le.fit_transform(df["3G"])
df["Dokunmatik"]=le.fit_transform(df["Dokunmatik"])
df["WiFi"]=le.fit_transform(df["WiFi"])
df["FiyatAraligi"]=le.fit_transform(df["FiyatAraligi"])
```

```python
df.sample(10)
```

```python
X = df.drop("FiyatAraligi",axis=1)
y = df[['FiyatAraligi']]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=15/100, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
gaussian = GaussianNB()
model_gaussian = gaussian.fit(X_train, y_train)
y_predict_gnb = gaussian.predict(X_test)
```

```python
confusion_matrix(y_test, y_predict_gnb)
```

```python
accuracy_score(y_test, y_predict_gnb)
```

```python
print(classification_report(y_test, y_predict_gnb))
```

```python
dtc=DecisionTreeClassifier(criterion="entropy")
model_dtc = dtc.fit(X_train, y_train)
```

```python
y_predict_dtc = dtc.predict(X_test)
```

```python
confusion_matrix(y_test, y_predict_dtc)
```

```python
print(classification_report(y_test,y_predict_dtc))
```

```python
knn = KNeighborsClassifier()
model_knn= knn.fit(X_train, y_train)
y_predict_knn = knn.predict(X_test)
```

```python
confusion_matrix(y_test, y_predict_knn)
```

```python
print(classification_report(y_test,y_predict_knn))
```

```python
score_list = []
for each in range(2,50):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(X_train,y_train)
    score_list.append(knn2.score(X_test, y_test))

plt.plot(range(2,50),score_list)
plt.xlabel("En Yakın Komşu Sayilari")
plt.ylabel("Skor")
plt.show()
```
