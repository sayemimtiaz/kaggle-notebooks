Today we will discover how to use the knowledge of SQL with Pandas.
* SQL : SQL (pronounced "ess-que-el") stands for Structured Query Language. SQL is used to communicate with a database. According to ANSI (American National Standards Institute), it is the standard language for relational database management systems. SQL statements are used to perform tasks such as update data on a database, or retrieve data from a database. Some common relational database management systems that use SQL are: Oracle, Sybase, Microsoft SQL Server, Access, Ingres, etc. Although most database systems use SQL, most of them also have their own additional proprietary extensions that are usually only used on their system. However, the standard SQL commands such as "Select", "Insert", "Update", "Delete", "Create", and "Drop" can be used to accomplish almost everything that one needs to do with a database.
* Pandas : The Pandas Python library is an open-source library for data manipulation and analysis. Pandas is the most widely used data analysis library in Python; and rightly so, as with a single line of code, it can take data from a CSV, an Excel sheet or JSON and create the famous data object DataFrames.


DataFrames are essentially tabular data structures, i.e. data represented by columns and rows. This should be familiar to anyone working with Excel, SQL or SPSS and even R.

Now we will pass some SQL commands using pysqldf(â€˜SQL Commandâ€™):

```python
#SQLDF - Structured Query Language (SQL) on DataFrames (DF)
#A simple wrapper to run SQL (SQLite) queries on pandas.DataFrame objects (Python).
pysqldf = lambda q: sqldf(q, globals())
```

For example, we wanted to create a new DataFrame with Top 15 entries from our original DataFrame, df:

```python
df_10 = pysqldf("SELECT * FROM df LIMIT 15")
print(pysqldf('SELECT * FROM df_10'))
```

```python
df_10 = pysqldf("SELECT * FROM df1 LIMIT 15")
print(pysqldf('SELECT * FROM df_10'))
```

```python
print(pysqldf("""SELECT retweet_count, count(*) as 'TotalNo'
FROM df
GROUP BY retweet_count
HAVING COUNT(*) > 1000
ORDER BY 1
"""))
```

```python
print(pysqldf("""SELECT retweet_count, count(*) as 'TotalNo'
FROM df1
GROUP BY retweet_count
HAVING COUNT(*) > 1000
ORDER BY 1
"""))
```
