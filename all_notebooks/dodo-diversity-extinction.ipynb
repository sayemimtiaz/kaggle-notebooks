**<span style="color:#008080;">Tree of Life Project - Dataset Columns</span>**

Tree of Life Project - Species File - treeoflife_nodes.csv

node_id : Numeric identifier for the species in the tree

node_name : Name of the species or None if unknown

child_nodes : Number of child nodes

leaf_node : Whether or not the node is a leaf

tolorg_link : Whether or not there is a page on the website tolweb.org describing this species - http://tolweb.org/${node_name}/${node_id}

extinct : Whether the species is living or extinct
0 - living , 1 - extinct species

confidence : Confidence of placement in the tree structure.
0 - confident position , 1 - problematic position , 2 - unspecified position

phylesis : 0, 1 or 2 :
0 - monophyletic , 1 - uncertain monophyly , 2 - not monophyletic

```python
#Code by Baekhakjeon https://www.kaggle.com/baekhakjeon/explorer-tabular-playground-series-apr-2021

df_group_by_phylesis = df.groupby(by='phylesis').extinct.value_counts()
print(df_group_by_phylesis)
```

```python
#Code by Baekhakjeon https://www.kaggle.com/baekhakjeon/explorer-tabular-playground-series-apr-2021

phylesis_monophyletic = df_group_by_phylesis.loc[0, 1]/(df_group_by_phylesis.loc[0, 1]+df_group_by_phylesis.loc[0, 0])*100
phylesis_uncertain = df_group_by_phylesis.loc[1, 1]/(df_group_by_phylesis.loc[1, 1]+df_group_by_phylesis.loc[1, 0])*100
phylesis_not_monophyletic = df_group_by_phylesis.loc[2, 1]/(df_group_by_phylesis.loc[2, 1]+df_group_by_phylesis.loc[2, 0])*100
phylesis_by_class = [phylesis_monophyletic, phylesis_uncertain ,phylesis_not_monophyletic]
print('Phylesis Monophyletic rate is : {:.2f}%'.format(phylesis_monophyletic))
print('Phylesis Uncertain rate is : {:.2f}%'.format(phylesis_uncertain))
print('Not monophyletic phylesis rate is : {:.2f}%'.format(phylesis_not_monophyletic))
```

```python
#Code by Baekhakjeon https://www.kaggle.com/baekhakjeon/explorer-tabular-playground-series-apr-2021

plt.figure(figsize = (10, 5), dpi = 100)
plt.plot(['phylesis_monophyletic', 'phylesis_uncertain', 'phylesis_not_monophyletic'], phylesis_by_class, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=12)
for i, v in enumerate(phylesis_by_class):
    plt.text(i+0.12, v, "{:.2f}%".format(v), ha="center", fontsize=12)
plt.title("Phylesis by Class", fontsize=20)
plt.ylabel('Phylesis Rate(%)', fontsize=20)
plt.grid()
plt.show()
```

"The extinction of the dodo within less than a century of its discovery called attention to the previously unrecognised problem of human involvement in the disappearance of entire species. The dodo achieved widespread recognition from its role in the story of Alice's Adventures in Wonderland, and it has since become a fixture in popular culture, often as a symbol of extinction and obsolescence."

https://en.wikipedia.org/wiki/Dodo

```python
df.groupby(by=['extinct', 'phylesis']).confidence.value_counts()
```

temp.loc[0,0] First zero is for extinct (0=living and 1=extinct). Second zero is for phylesis (0,1,2)
temp.loc[1,0] One refers to extinct (1=extinct). Zero refers to phylesis (0=monophyletic) 

```python
#Code by Baekhakjeon https://www.kaggle.com/baekhakjeon/explorer-tabular-playground-series-apr-2021

temp = df.groupby(by=['extinct', 'phylesis']).confidence.count()
phylesis = ['phylesis_monophyletic', 'phylesis_uncertain', 'phylesis_not_monophyletic']
living_species = [temp.loc[0,0], temp.loc[0,1], temp.loc[0,2]]
extinct_species = [temp.loc[1,0], temp.loc[1,1], temp.loc[1,2]]

plt.figure(figsize = (10, 5), dpi = 100)

plt.plot(['monophyletic', 'uncertain', 'not_monophyl'], living_species, color='green', marker='*', linestyle='-', linewidth=2, markersize=12)
for i, v in enumerate(living_species):
    plt.text(i+0.12, v, "{:,d}".format(v), ha="center", fontsize=12)
plt.plot(['mono', 'monophyly', 'not mono'], extinct_species, color='pink', marker='o', linestyle='-', linewidth=2, markersize=12)
for i, v in enumerate(extinct_species):
    plt.text(i+0.12, v, "{:,d}".format(v), ha="center", fontsize=12)

plt.title("Phylesis", fontsize=20)
plt.ylabel('Number of Species', fontsize=20)
plt.xticks(fontsize=15)
format_change = lambda x: "{:,d}".format(x) # Labmda fuction to change displya yticks format.
yticks = [12000, 14000, 16000, 18000, 20000, 22000, 24000, 26000, 28000]
plt.yticks(yticks, [format_change(i) for i in yticks], fontsize=15)
plt.legend(['Living Species', 'Extinct Species'])
plt.grid()
plt.show()
#plt.xticks(rotation=45)It didn't work well
```
