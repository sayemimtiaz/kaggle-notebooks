# **<span style="color:#DC143C;">Diaphragm Pacing System (DPS) Background</span>**

Author: Shabir Bhimji

"The history of pacing the diaphragm is not new. The earliest record of phrenic stimulation for the treatment of asphyxia was reported in 1783. In the mid-1850s, French neurologists also proposed such an idea, but it was Hugo Wilhelm von Ziemssen who performed the first diaphragmatic pacing on a young female patient who had asphyxiated on charcoal vapor." 

"The true beginnings of diaphragmatic pacing started in the 1860s and 70s. The father of modern diaphragm pacing is Dr William W L Glenn from Yale University, who showed that the technique was not only practical but could be used clinically for the treatment for several medical disorders. With advances in technology, more refined and flexible electrodes were developed, and the thoracoscopic method of implantation became practical."

# **<span style="color:#DC143C;">Indications and Contraindications </span>**

"Diaphragm pacing is performed to provide ventilatory support in 2 main clinical scenarios:

"Central alveolar ventilation, or what is better known as sleep apnea.

"High spinal cord paralysis in which the drive for respiration is still present but the injury to the spinal cord prevents stimulation from the phrenic nerves".

**<span style="color:#DC143C;">DPS Contraindications</span>**

"Diaphragm pacing is contraindicated for patients in whom the phrenic nerve is not functional. Such patients include those with severe traumatic injury to the nerve, those with nerve tumors, and most of those with neuropathies. In addition, diaphragm pacing is contraindicated for patients with conditions in which the diaphragm itself is not functional."

https://emedicine.medscape.com/article/1970348-overview#a4

# **<span style="color:#DC143C;">Diaphragm Pacing May Be Harmful in ALS</span>**

Author: by Kristina Fiore

"Diaphragmatic pacing may be more harmful than helpful in amyotrophic lateral sclerosis (ALS), researchers reported."

"In a randomized controlled trial that was stopped early, patients treated with the NeuRx 4/4 Diaphragm Pacing System had more than a two-fold increased risk of death during the study than those who only had noninvasive ventilation."

"Researchers found that survival was shorter when diaphragmatic pacing was added to the mix.
A total of 78% of patients in the pacing group died, compared with 51% of those who had noninvasive ventilation alone." 

"Survival was significantly worse in those who had the diaphragm added to their treatment (11 months versus 22.5 months)."

https://www.medpagetoday.org/neurology/generalneurology/52940

"Diaphragmatic pacing should not be a routine treatment for patients with ALS in respiratory failure.
