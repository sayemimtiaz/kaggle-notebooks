<h1 style = "font-size:25px; font-family:Garamond ; font-weight : normal; background-color: #0a0a6e ; color : #f5f5fa; text-align: center; border-radius: 70px 70px;"> 1.About Dataset</h1>

<h1 style = "font-size:20px; font-family:Garamond ; font-weight : normal; color : #0a0a6e;"> This dataset only contains death numbers for both Palestine and Israel from 2000 - 2021.It has Age,Event Location Region,Event Location District,Date of Death,Gender,Type of Injuries and Reasons.</h1>

```python
#import all the libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
df=pd.read_csv('../input/../input/../input/palestine-body-count20002021/Palestine Body Count(2000-2021).csv')

df.head()
```

```python
dataTypeSeries = df.dtypes
#print(dataTypeSeries)
```

```python
df.fillna(value=0,inplace=True)
df["Age"]=df["Age"].apply(int)
#print(df.dtypes)
```

```python
df.rename(columns = {'Type of injury':'Type_of_injury','Event location - District':'Event_location_District','Event location - Region': 'Event_location_region','Name':'Total','Type of injury':'Type_of_injury'}, inplace=True)
df.columns


df["Date of death"]=pd.to_datetime(df["Date of death"])
df["Date of event"]=pd.to_datetime(df["Date of event"])

df["Death_year"]=df["Date of death"].dt.year
df["Death_month"]=df["Date of death"].dt.month
df["Death_day"]=df["Date of death"].dt.day

df["Event_year"]=df["Date of event"].dt.year
df["Event_month"]=df["Date of event"].dt.month
df["Event_day"]=df["Date of event"].dt.day
#df.head()
```

<h1 style = "font-size:25px; font-family:Garamond ; font-weight : normal; background-color: #0a0a6e ; color : #f5f5fa; text-align: center; border-radius: 70px 70px;">2. Palestine Vs Israel Body Count(2000-2021)</h1>

```python
#Palestine Vs Israel Body Count

pal7 = pd.DataFrame(df, columns = ['Death_year','Event_location_region','Total'])
pal7_c=pal7.groupby(['Death_year','Event_location_region']).count()
pal7_c = pal7_c.reset_index(inplace=False)
#print(pal7_c)

dfb7= pal7_c.pivot(index='Death_year',columns='Event_location_region',values='Total') 
dfb7.rename(columns = {'Gaza Strip':'Gaza_Strip','West Bank':'West_Bank'}, inplace=True)
dfb7 = dfb7.reset_index(inplace=False)
dfb7['Gaza_Strip'] = dfb7['Gaza_Strip'].replace(np.nan, 0)
sum_column = dfb7["Gaza_Strip"] + dfb7["West_Bank"]
dfb7["Palestine"] = sum_column
#print(dfb7)

#Chart-Palestine Vs Israel Body Count

font_color = '#030336'
fig, ax = plt.subplots(figsize = (15,7), dpi = 80)
ax.set_facecolor('#f7f9fc')
fig.patch.set_facecolor('#f7f9fc')
ax=sns.barplot(data=dfb7,x='Death_year',y=dfb7['Palestine'],color="#0acafa",label='Palestine')
ax=sns.barplot(data=dfb7,x='Death_year',y=-dfb7['Israel'],color="#080e4d",label='Israel')
ax.text(-1,2900, 'Palestine Vs Israel Body Count                                                                                    ', {'color':'#ebebf5','backgroundcolor':'#030336','font': 'serif','color':'#fcfcff', 'fontsize': 18, 'weight':'bold'},alpha = 0.9 )
ax.text(-1.2,2700, 'In Palestine highest death toll has been recorded in 3rd Gaza war in 2014 where as in Israel highest death toll has been', {'backgroundcolor':'#f7f9fc','font': 'serif','color':'#080e4d', 'fontsize': 12, 'weight':'bold'},alpha = 0.9 )
ax.text(-1.2,2550,' recorded in 2002.', {'backgroundcolor':'#f7f9fc','font': 'serif','color':'#080e4d', 'fontsize': 12, 'weight':'bold'},alpha = 0.9 )
plt.legend(title="Event Location Region")
plt.xticks(color=font_color)
plt.yticks(color=font_color)
ax.set_xlabel('Years',fontsize=15,fontweight="bold",color='#080e4d')
ax.set_ylabel('Total',fontsize=15,fontweight="bold",color='#080e4d')
ax.text(11, 2200, 
        '3rd Gaza War ⟶', 
         fontsize=10, 
         fontweight='bold', 
         fontfamily='serif',
         color='#080e4d')
```

<h1 style = "font-size:25px; font-family:Garamond ; font-weight : normal; background-color: #0a0a6e ; color : #f5f5fa; text-align: center; border-radius: 70px 70px;"> 5.Body Count by Palestine Districts</h1>

```python
#Body Count by Palestine District

pal2 = pd.DataFrame(df, columns = ['Event_location_District','Total'])
pal2_c=pal2.groupby(['Event_location_District']).count()
pal2_c = pal2_c.reset_index(inplace=False)
pal2_c.sort_values(by=['Total'], inplace=True)
#print(pal2_c)

#Chart-Body Count by Palestine District

data  = pal2_c['Total']
fig = plt.figure(figsize=(12,8), dpi = 80)
fig.patch.set_facecolor('#f7f9fc')
ax = plt.subplot(polar=True)
plt.axis('off')


labels = ['Gaza Strip ', 'Salfit',
         'Jericho', 'Tubas',
         'Qalqiliya', 'Gush Katif',
          'Al-Quds', 'East Jerusalem',
          'Bethlehem','Tulkarm',
          'Ramallah and al-Bira','Hebron',
          'Jenin','Nablus','Israel',
          'Deir al-Balah', 'Rafah',
          'Khan Yunis','North Gaza',
          'Gaza']

colors = [ '#324e85', '#324e85', '#324e85', '#324e85', '#324e85', '#324e85', '#324e85', '#324e85', '#324e85', '#324e85','#324e85','#324e85', '#324e85', '#324e85', '#060d1c', '#324e85', '#324e85','#324e85','#324e85','#324e85']
# Constants = parameters controling the plot layout:
upperLimit = 2000
lowerLimit = 800
labelPadding = 500

# Compute max and min in the dataset
max = data.max()

slope = (max - lowerLimit) / max
heights = slope * data.values + lowerLimit

# Compute the width of each bar. In total we have 2*Pi = 360°
width = 2*np.pi / len(data.index)

# Compute the angle each bar is centered on:
indexes = list(range(1, len(data)+1))
angles = [element * width for element in indexes]


# Draw bars
bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="#f6f5f5",
    color = colors
)

# Add labels
for bar, angle, height, label in zip(bars,angles, heights, labels):

    # Labels are rotated. Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle)

    # Flip some labels upside down
    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"

    # Finally add the labels
    ax.text(
        x=angle,
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor",**{'font':'serif', 'size':12, 'weight':'bold','color':'#030361'},alpha = 0.9) 
fig.text(0.25,1.1,'Body Count by Palestine Districts                                           ',{'backgroundcolor':'#030361','font': 'serif','color':'#f7f9fc', 'fontsize': 18, 'weight':'bold'},alpha = 0.9 )
fig.text(0.25,1.005,'Highest death has been recorded in the district Gaza.',{'backgroundcolor':'#f7f9fc','font': 'serif','color':'#030361', 'fontsize': 12, 'weight':'bold'},alpha = 0.9)

fig.show()
```

```python
import scipy
from scipy.stats import pearsonr

data1 = dfb7["Palestine"]
data2 = dfb7["Israel"]
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
  print('Probably independent')
else:
  print('Probably dependent')
print('p value is not significant so there is low correlation between Palestinians killed and Israelis killed.')
```
