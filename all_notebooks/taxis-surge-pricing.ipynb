```python
df = pd.read_csv("../input/taxi-pricing-with-mobility-analytics/sigma_cabs.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
from colorama import Fore, Style

def count(string: str, color=Fore.RED):
    """
    Utility function to string in colors
    """
    print(color+string+Style.RESET_ALL)
```

```python
def statistics(dataframe, column):
    count(f"The Average value in {column} is: {dataframe[column].mean():.2f}", Fore.CYAN)
    count(f"The Standard Deviation of {column} is: {dataframe[column].std():.2f}", Fore.CYAN)
    count(f"The Maximum value in {column} is: {dataframe[column].max()}", Fore.CYAN)
    count(f"The Minimum value in {column} is: {dataframe[column].min()}", Fore.CYAN)
    count(f"The 25th Quantile of {column} is: {dataframe[column].quantile(0.25)}", Fore.CYAN)
    count(f"The 50th Quantile of {column} is: {dataframe[column].quantile(0.50)}", Fore.CYAN)
    count(f"The 75th Quantile of {column} is: {dataframe[column].quantile(0.75)}", Fore.CYAN)
```

```python
statistics(df, "Surge_Pricing_Type")
```

```python
# Let's plot the age column too
plt.style.use("classic")
sns.distplot(df['Surge_Pricing_Type'], color='blue')
plt.title(f"Surge Pricing Type Distribution [\u03BC : {df['Surge_Pricing_Type'].mean():.2f} price | \u03C3 : {df['Surge_Pricing_Type'].std():.2f} price]")
plt.xlabel("Surge Pricing Type")
plt.ylabel("Count")
plt.show()
```

```python
#Code by Tanay Mehta https://www.kaggle.com/heyytanay/police-shooting-eda-interactive-map/notebook

plt.style.use("fivethirtyeight")
plt.figure(figsize=(16, 6))
sns.kdeplot(df.loc[df['Gender'] == 'Male', 'Surge_Pricing_Type'], label = 'Male',shade=True)
sns.kdeplot(df.loc[df['Gender'] == 'Female', 'Surge_Pricing_Type'], label = 'Female',shade=True)

# Labeling of plot
plt.xlabel('Surge Pricing Type')
plt.ylabel('Density')
plt.title('Distribution of Surge Pricing Type for Male and Female Individuals')
plt.show()
```

```python
# Pie Chart to see gender of individuals
targets = df['Gender'].value_counts().tolist()
values = list(dict(df['Gender'].value_counts()).keys())

fig = px.pie(
    values=targets, 
    names=values,
    title='Gender of Individuals in Taxi Surge Pricing',
    color_discrete_sequence=['blue', 'magenta']
)
fig.show()
```
