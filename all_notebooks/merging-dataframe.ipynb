**Penggabungan Data**

Sebutkan perbedaan antara full outer join, inner join, dan left join

1. Full Outer Join = menghasilkan gabungan index baris dan kolom dari setiap data yang digabung.
2. Inner Join ini akan menggabungkan data antara nilai key atau referensi yang beririsan.
3. Left Outer Join menghasilkan seluruh data yang kiri ditambah data yang kanan yang memiliki nilai key yang sama dengan data dari kiri.
4. Right Outer Join hanya kebalikan dari Left Outer Join.

**Penggabungan Data**

Pada kali ini gunakan data 'names1881.csv' dan 'names1981.csv', data ini berisi nama bayi yang lahir pada tahun 1881 atau 1981 di Amerika.

1. Baca data tersebut dengan memberikan nama column 'Name', 'Gender', 'Count'.
2. Tunjukan jumlah dimensi dari data 'names1881.csv' dan 'names1981.csv'
3. Gabungkan kedua data tersebut berdasarkan sumbu baris
4. Tunjukan jumlah dimensi dari data setelah di gabungkan.

```python
import pandas as pd
# read data
df1 = pd.read_csv('../input/names1881/names1881.csv', names=['Name', 'Gender', 'Count'])
df2 = pd.read_csv('../input/names1981/names1981.csv', names=['Name', 'Gender', 'Count'])

# periksa dimensi dari kedua data
print('Jumlah Dimensi data names1881.csv :', df1.shape)
print('Jumlah Dimensi data names1981.csv :', df2.shape)

# gabungkan data
data1 = [df1,df2]
data2 = pd.concat(data1)
print(data2)

# periksa dimensi data hasil gabungan
print('Jumlah Dimensi data gabungan :', data2.shape)
```

**Penggabungan Data**

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut :

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

# join data
data1 = pd.merge(df1, df2, on='id', how='outer')
data2 = pd.merge(df1, df2, on='id', how='inner')
data3 = pd.merge(df1, df2, on='id', how='left')
data4 = pd.merge(df1, df2, on='id', how='right')
print(data1)
print('='*20)
print(data2)
print('='*20)
print(data3)
print('='*20)
print(data4)
```

**Penggabungan Data berbeda label column Key**

Lakukan jenis penggabungan inner join untuk 2 data yang memiliki nama column key yang berbeda ini :

```python
data1 = {
    'key_data1' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'key_data2' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
data1 = {
    'key_data1' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'key_data2' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)
# Gabungkan kedua data tersebut
data = pd.merge(df1,df2, left_on='key_data1', right_on='key_data2', how='inner')
data
```
