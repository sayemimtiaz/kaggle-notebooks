To get a better understanding forest fires across the country, we need to understand the climatic changes experienced across different regions and in fact in every state of the country Brazil.
According to the Köppen system, Brazil hosts six major climatic subtypes:<br>
- **Desert** - is a climate in which there is an excess of evaporation over precipitation.
- **Equatorial** - climate is typically hot, very humid and wet. Can expect at least 60mm of rain fall every month of the year.
- **Tropical** -  the temperature remains relatively constant (hot) throughout the year. Sunlight is intense.
- **Semiarid** - receives precipitation below potential evapotranspiration (evaporation from Earth's land + Ocean + Plant transpiration), but not as low as a desert climate.
- **Oceanic** - Cool summers and cool but not cold winters, with a relatively narrow annual temperature range and few extremes of temperature. Lacks dry season.
- **Subtropical** - climate characterized by hot and humid summers, and cold to mild winters.

Following are the different types of climatic conditions that can experienced in different states or parts of the country Brazil in a day or across months of an year.

- **Desert** - is a climate in which there is an excess of evaporation over precipitation.
- **Equatorial** - climate is typically hot, very humid and wet. Can expect at least 60mm of rain fall every month of the year.
- **Tropical** -  the temperature remains relatively constant (hot) throughout the year. Sunlight is intense.
- **Semiarid** - receives precipitation below potential evapotranspiration (evaporation from Earth's land + Ocean + Plant transpiration), but not as low as a desert climate.
- **Oceanic** - Cool summers and cool but not cold winters, with a relatively narrow annual temperature range and few extremes of temperature. Lacks dry season.
- **Subtropical** - climate characterized by hot and humid summers, and cold to mild winters.

If we look at the **"Koppen's climate classification map for Brazil"** article an outcome of extensive research done by **Clayton Alcarde Alvares, Jose´ Luiz Stape, Paulo Cesar Sentelhas, Jose´ Leonardo de Moraes Gonc¸alves & Gerd Sparovek** (link : https://www.schweizerbart.de/papers/metz/detail/22/82078/Koppens_climate_classification_map_for_Brazil?af=search) published in Jan 2014 and per the Wikipedia (link : https://en.wikipedia.org/wiki/K%C3%B6ppen_climate_classification), 

Climates are divided into five climate groups (based on **seasonal precipitation** and **temperature patterns**):
- A (tropical): Af, Am, As and Aw
- B (dry): BSh, BSk, BWh and BWk
- C (temperate): Cfa, Cfb, Cfc, Cwa, Cwb, Cwc, Csa, Csb and Csc
- D (continental): Dfa, Dfb, Dfc, Dfd, Dwa, Dwb, Dwc, Dwd, Dsa, Dsb, Dsc, Dsd
- E (polar): ET and EF

A, B, C, D and E are the climate **main groups**.

2nd letter in these main groups **except E** represent seasonal precipitation.

3rd letter in these main groups **except A** represent temperature subgroup.

- **Aw** means "Tropical with dry winter". Rainfall expected is < 60 mm and annual rainfall < 25 mm. Dry season. Forest fire may break out.
- **As** means "Tropical with dry summer"; rainfall expected in a month < 60 mm and annual rainfall < 25 mm. Dry season. Forest fire may break out.
- **Cfa** means "Humid subtropical with hot summer" with temperature >= 22 and rainfall in the driest month > 40 mm. Dry summer with it may rain in the driest month in which case forest fire may break out.
- **Bsh** means "dry and semi-arid" annual rainfall will be very low. Dry season. Forest fire may break out.
- **Af** means "Tropical without dry season". Rainfall expected is >= 60 mm. Forest fire should not broke out in this scenario.
