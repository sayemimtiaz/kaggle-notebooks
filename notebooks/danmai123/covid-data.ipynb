**What is the cause of the staggering of global confirmed covid cases between May and June of 2020?**

The staggering of data here could have been caused by multiple reasons. The staggering of cases occurs near the end of May and into the beginning of June 2020. What is happening is that the cases are going up in a somewhat linear rate, then there is a steep drop down in the number of confirmed cases, that drop then goes up in a somewhat linear rate and this repeats several times. This could have happened due to the virus's seasonality, being more effective at spreading in certain seasons or climates. Scientists believe that colder weather and climates allow for a faster and more effective spreading of the virus. The US and India are two countries with high amounts of covid cases, so a fluctuation in their number of cases would make the biggest difference to the graph. From May to June (the time of the staggers in the graph), US is in the Spring season and India is in the Pre-Monsoon and Monsoon season. Both would have a fluctuation in temperature affecting the rate of covid cases, due to the seasons transitioning from a hot climate to a colder, rainy season, or from a cold, rainy season to a warmer season. These seasons have the biggest difference in daily temperature, where it could be 20 degrees Celsius one day, and 35 degrees the next day, affecting the rate of spread of the virus.



```python
covid_data = pd.read_csv("../input/novel-corona-virus-2019-dataset/covid_19_data.csv",parse_dates=['ObservationDate','Last Update'])
# This line of code basically assigns the data to the variable 'covid_data' and converts the columns, 'ObservationDate' and 'Last Update' into a date format.


cases_daily = covid_data.groupby(['ObservationDate'])['Country/Region'].nunique()
# This line of code groups the 'ObservationDate' and 'Coutnry/Reigon' data together, it then assign this data to the variable 'cases_daily'


#print(cases_daily.head())
# This line of code prints out the first 5 rows of data, it can help with testing


#The lines of code below show the data in the form of a graph


cases_daily.plot()
#sns.lineplot(data=cases)
#This line of code plots the data I made above into a line graph

plt.title('Countries With Observed Cases Daily')
plt.xlabel('Month Observed')
plt.ylabel('Countries')
# These 3 lines of code are used to give titles to the graph, x-axis and y-axis

plt.xticks(rotation = 45)
# This line of code rotates the labels for the x-axis by 45 degrees for ease of reading

plt.show() 
# This line of code hides a bunch of things that happens behind the scenes


covid_data['Country/Region'].nunique() 
# This line of code finds the amount of countries in the dataset and returns it as a string, it is a quick and efficient way to count the amount of things in a datset quickly
```
