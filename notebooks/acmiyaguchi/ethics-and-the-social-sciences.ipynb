```python
import semanticscholar as sch
import pandas as pd
from IPython.display import display

tagged = dict(
    principles=[
        ["10.15171/ijhpm.2016.55", 3],
        ["10.1186/1471-227X-11-16", 3],
        ["10.1186/s13073-014-0106-2", 2],
        ["10.1016/j.optm.2007.04.099", 2],
        ["PMID:17132331", 3],
        ["10.1016/j.socscimed.2015.11.017", 3],
        ["10.1093/phe/phu048", 2],
        ["10.1089/hs.2015.0077", 2],
        # WHO_CDS_EPR_GIP_2007_2c.pdf
        ["10.1186/1472-6939-7-5", 3],
    ],
    practice=[
        ["10.1186/s12889-015-2617-1", 3],
        ["10.1016/j.socscimed.2015.11.017", 2],
        ["10.1186/s12910-015-0025-9", 3],
    ],
    outreach=[["10.15171/ijhpm.2016.55", 3],],
    organization=[["10.1186/s12889-015-2617-1", 3],],
    quality=[["10.1016/j.trb.2014.08.004", 3], ["10.1186/s12910-015-0025-9", 3],],
    risk=[
        ["10.1093/ofid/ofx163.1031", 1],
        ["10.1016/j.jaad.2020.03.013", 1],
        ["10.1016/j.genhosppsych.2005.04.007", 2],
        ["10.1098/rstb.2004.1483", 3],
        ["10.1186/1472-6939-7-5", 3],
        ["10.1016/j.ijnss.2016.12.002", 2],
        ["10.1177/070674370404900612", 3],
    ],
    perception=[
        ["10.1093/jtm/taaa031", 2],
        ["10.1016/j.socscimed.2007.04.022", 3],
        ["10.1098/rsif.2014.1105", 3],
        ["10.1371/currents.dis.de9415573fbf90ee2c585cd0b2314547", 2],
        ["10.1089/hs.2016.0111", 1],
        ["10.1016/j.trb.2014.08.004", 3],
        ["10.1016/j.tele.2017.12.017", 3],
        ["10.3390/ijerph13080780", 3],
        ["10.1016/j.pubrev.2019.02.006", 2],
        ["10.1177/070674370404900612", 1],
        ["10.1016/S0140-6736(20)30379-2", 2],
    ],
)

def fetch_semantic_scholar(tagged: dict) -> pd.DataFrame:
    results = []
    for tag, paper_ids in tagged.items():
        for paper_id, score in paper_ids:
            # append the semantic scholar response with other metadata
            results.append({
                **dict(tag=tag, relevanceScore=score, queryId=paper_id),
                **sch.paper(paper_id)
            })
    df = pd.DataFrame(results)
    df["numCitations"] = df["citations"].apply(len)
    df["firstAuthor"] = df["authors"].apply(lambda x: x[0]["name"].split()[-1])
    return df


def filter_results(df, tag=None):
    cols = ["tag", "firstAuthor", "year", "title", "relevanceScore", "numCitations", "influentialCitationCount", "venue", "queryId"]
    
    if tag:
        df = df.query(f"tag=='{tag}'")
    return df[cols].sort_values(["relevanceScore", "numCitations"], ascending=False)
```

The WHO (2007) has produced guidelines around ethical considerations with respect to a pandemic influenza response. These are relevant in the COVID-19 pandemic, which draws similarities to other zoonotic diseases such as the 2003 SARS and 2009 H1N1 pandemics. The guidelines cover the following:

* Balancing rights, interests, and values
* Evidence base for public health measures
* Transparency, public engagement and social mobilization
* Information, education, and communication
* Resource constraints

A review of several CDC-sponsored meetings explores these considerations with regional and cultural perspectives. (Lor et al., 2016) The driving prompt is as follows:

Socioeconomic, geographical, and cultural differences between regions and countries affect the response to issues. Some key takeaways include the wide acceptance of transparency and inclusiveness for ethical decision-making, and a noted effectiveness of focusing on procedural ethics (shared values and fair procedures) to combat the common health threat of a global pandemic.

An ethical framework proposed by Kass (2005) for evaluating public health and avian influenza pandemic preparedness captures the multifaceted nature of balancing the complex nature of pandemic preparation and response. The practical nature of the framework strives to reduce burden while respecting autonomy of individuals and communities. She states that public health burdens fall under three major types:

* Risks to privacy and confidentiality
* Risks to liberty and self-determination
* Risks to justice

The risks outlined above are not isolated to public health. In particular, privacy and confidentiality are freedoms that could erode as the technology to perform surveillance at a massive scale evolves. There is some precedent for thinking about these issues. There is discussion on genomics around data sharing, management of incidental findings, and practices for obtaining informed consent (Geller 2014). The idea of preventative forensic screenings by Lombroso can provide strategies for balancing the risk-benefit of invasive surveillance (Munthe 2015). And finally, there are lessons to be learned from the cyber intelligence community around information sharing and organizational structures (Smith 2016).

Recent news highlights efforts to combat the spread of COVID-19, and may provide case studies for the ethical issues surrounding privacy in the context of technology.

There are decisions with more direct consquences on life. There has been considerable pressure to those working in the medical profession, and conflicting responsibilities to themselves, their families, their patients. This is discussed in (Ruderman 2006). Resource constraints force healthcare workers to make life and death decisions. An example of such decision is allocating a scarce supply of ventilators to a surge of new patients. Emergency care can be viewed from an ethical perspective (Aacharya 2011) which are summed in the following areas.

* Respect for Autonomy
* Nonmaleficence
    * Do no harm.
* Beneficence
    * One ought to prevent evil or harm.
    * One ought to remove evil or harm.
    * One ought to do or promote good.
* Justice
    * Allocation decisions must be made fairly.
    
The production of vaccines also requires a careful balance. While it may be tempting to point toward promising drugs as cure-alls, there is a process that must be followed. The ethics of an Ebola virus therapy trial is discussed in (Kombe 1982).

The current state of affairs also outlines the dilemma of many small businesses
that may suffer greatly during an economic downturn. As stringent physical
distancing measures start to relax, employers will need to plan for the
continuity of businesses. The perspective from optometry (Hom 2007) outlines

> Efforts to identify how the burden of responding to the outbreak and implementing public health measures affects the physical and psychological health of those providing care for Covid-19 patients and identify the immediate needs that must be addressed.

> Health care professionals (HCPs) and hospital systems that bore the brunt of the SARS outbreak continue to struggle with the aftermath of the crisis. Indeed, HCPs – both in clinical care and in public health – were severely tested by SARS. Unprecedented demands were placed on their skills and expertise, and their personal commitment to their profession was severely tried.

(Ruderman et al. 2006).

First, it is known that many medical workers will get some form of post traumatic stress disorder (PTSD). In a 
study of 102 medical professionals exposed to avian flu patients from Anhui Province in China, 20% exhibited symptoms 
of PTSD (Tang 2016). A study of the Toronto SARS outbreak estimated that a high degree of distress was expeienced 
by 29-35% of healtcare workers (Maunder 2004).

The impacts on hospital workers in the Toronto SARS outbreak is broken down as follows:

* Groups of healthcare workers
    * Those who were in contact with SARS patients
    * By professional discipline
    * Workers who are parents
* Mediating factors
    * Social isolation because of transmission potential
    * Stigma due to racism or public perception
    * Unusual degree of scrutiny from various sources
    * Job stress e.g. unfamiliar tasks
* Individual traits that contribute to stress (e.g. attachment insecurity)
* Cumulatative effect of multiple factors

In a study of a Taiwanese survey of stressors while caring for SARS patients (N=26), the following were the top stressors (Lee 2005).

1. Worry about being negligent and endangering coworkers (92.6%)
2. Frequent modification of infection control procedures (92.3%)
3. Uncertainty about when the epidemic will be under control (92.3%)

The top measures to reduce stress included:

1. Psychiatric services
2. Encouragement among coworkers
3. Hospital providing nutriments

Additionally, workers had their own coping strategies:

1. Taking protective measures
2. Actively aquiring more knowledge about SARS
3. Engaging in health-promoting behaviors

These three tables could be used to search for deeper exploration into the topic.

Existing personal protective equiptment should be donned effectively to maximize protection of workers and to ration supplies. Colored paints can be used to evaluate the effectivess of PPE (Eiras et al. 2017). Workers should expect mild skin irritation from n95 masks and googles, and to be careful when applying topical ointments. Full face respirators and face shields should also be considered (Kantor 2020).

# Drivers of fear, anxiety, and stigma

> Efforts to identify the underlying drivers of fear, anxiety and stigma that fuel misinformation and rumor, particularly through social media.


```python
display(filter_results(results, "perception"))
```

Misinformation can be driven by the psychosocial reactions to acute outbreaks.
Some drivers include fictional dramaticizations, mass media, and ethical
dilemmas (Pappas et al. 2009). Rational responses to fear can evolve into social
stigma, as seen in non-endemic populations such as Chinatown in New York (Depoux
et al. 2007). This stigma has also been observed in other outbreaks such as the
1993 US Hantavirus outbreak termed "Navajo disease". Fear and blame has been
associated with lack of information and perceived loss of control. Cultural
constructions of diease and merginalization within communities should be looked
into further.

A study about the swine flu describes panic during the early stages of the swine
flu pandemic, which turned out to be less impactful as the current pandemic
(Srinivasan 2009). It noted that levels of anxiety shift with the time-length
exposure of risk, and the most used source of information came from the internet
rather than the healthcare sector. It notes that the public can become apathetic
to panic.

Transportation research in regards to post-disaster evactuation and temporary
resettlement formulates logistical models with panic-induced psychological
penalties (Hu 2014). A similar modeling approach may be useful during a
transition to a post-quarantine/pre-vaccination economy. 

There are efforts of measuring the spread of information over the internet.
Virality of information via social media can be characterized by a diffusion
process, in the same way that physical viruses are modeled. This can be used to
measure a social response during a disease outbreak (Fast et al., 2015). Boolean
searches of Twitter posts have been used to increase situational awareness
during Superstorm Sandy (Smith et al., 2014). Public attention of the Ebola
outbreak in West Africa was measured in China and correlated between blogs and
media news (Liu et al. 2016). 

Social media can spread information quickly. An editorial notes that
"misinformation on Coronavirus may be the most contagious thing about it"
(Depoux 2020). An article by the Lancet highlights the response of verified key
information as a critical need during the pandemic (Lancet 2020). There seems to
be an established literature on rumor responses, such as in Taiwan via increased
media scrutiny and transparency (Hsu 2014). In the case of rumors of radiation
contaminated food in South Korea, only refutation was found to be effective in
reducing rumor beliefs among three strategies (Paek 2019). Additionally,
information seeking behavior has been studied in the context fo rumors around
the MERS outbreak (Lee 2018).
