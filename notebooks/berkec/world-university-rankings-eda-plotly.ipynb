# Teaching And Citation Score For top 100 University

```python
df = times_data.iloc[:100,:]

import plotly.graph_objs as go

trace1 = go.Scatter(
    x = df.world_rank,
    y = df.citations,
    mode = "lines+markers",
    name = "citations",
    marker = dict(color = 'rgba(153, 0, 76, 0.8)'),
    text = df.university_name)

trace2 = go.Scatter(
    x = df.world_rank,
    y = df.teaching,
    mode = "lines+markers",
    name = "teaching",
    marker = dict(color = 'rgba(0, 102, 102, 0.8)'),
    text = df.university_name)


data = [trace1,trace2]
layout = dict(title = "Citation and Teaching Score of Top 100 University",
             xaxis = dict(title="World Rank",ticklen = 5,zeroline=False))

fig = dict(data = data, layout = layout)
iplot(fig)
```

# Visualize with Scatter Chart

```python
df2013 = times_data[times_data.year == 2013].iloc[:100,:]
df2014 = times_data[times_data.year == 2014].iloc[:100,:]
df2015 = times_data[times_data.year == 2015].iloc[:100,:]    
df2016 = times_data[times_data.year == 2016].iloc[:100,:]

trace1 = go.Scatter(
    x = df2013.world_rank,
    y = df2013.citations,
    mode = "markers",
    name = "2013",
    marker = dict(color = 'rgba(205,102,34,0.8)'),
    text = df2013.university_name)
trace2 = go.Scatter(
    x = df2014.world_rank,
    y = df2014.citations,
    mode = "markers",
    name = "2014",
    marker = dict(color = 'rgba(26,47,185,0.8)'),
    text = df2014.university_name)
trace3 = go.Scatter(
    x = df2015.world_rank,
    y = df2015.citations,
    mode = "markers",
    name = "2015",
    marker = dict(color = 'rgba(174,26,185,0.8)'),
    text = df2015.university_name)
trace4 = go.Scatter(
    x = df2016.world_rank,
    y = df2016.citations,
    mode = "markers",
    name = "2016",
    marker = dict(color = 'rgba(242,115,19,0.8)'),
    text = df2016.university_name)
data = [trace1,trace2,trace3,trace3]

layout = dict(title = "Citation Versus World Rank of top 100 Universies with 2013,2014,2015 and 2016",
         xaxis = dict(title = "World Rank",ticklen = 5, zeroline = False),
         yaxis = dict(title = "Citation",ticklen = 5, zeroline = False))
fig = dict(data = data, layout = layout)
iplot(fig)
```

# Pie Chart

```python
df2016 = times_data[times_data.year == 2016].iloc[:10,:]
pie1 = df2016.num_students
pie1_list = [float(each.replace(',', '.')) for each in pie1]  # str(2,4) => str(2.4) = > float(2.4) = 2.4
labels = df2016.university_name
# figure
fig = {
  "data": [
    {
      "values": pie1_list,
      "labels": labels,
      "domain": {"x": [0, .5]},
      "name": "Number Of Students Rates",
      "hoverinfo":"label+percent+name",
      "hole": .3,
      "type": "pie"
    },],
  "layout": {
        "title":"Universities Number of Students rates",
        "annotations": [
            { "font": { "size": 20},
              "showarrow": False,
              "text": "Number of Students",
                "x": 0.20,
                "y": 1
            },
        ]
    }
}
iplot(fig)
```

# Histogram

```python
df2011 = times_data.student_staff_ratio[times_data.year == 2011]
df2012 = times_data.student_staff_ratio[times_data.year == 2012]

trace1 = go.Histogram(
    x = df2011,
    opacity = 0.75,
    name = '2011',
    marker = dict(color = 'rgba(185,16,4,0.6)'))

trace1 = go.Histogram(
    x = df2012,
    opacity = 0.8,
    name = '2012',
    marker = dict(color = 'rgba(4,40,185,0.6)'))

data = [trace1,trace2]
layout = go.Layout(barmode='overlay',
                  title = "Students-staff raito in 2011-2012",
                  xaxis = dict(title = 'student_staff_ratio'),
                  yaxis = dict(title = 'Count'))
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# WordCloud

```python
x2016 = times_data.country[times_data.year == 2016]
plt.subplots(figsize = (16,16))
wordcloud = WordCloud(
    background_color='black',
    width = 512,
    height = 384).generate(" ".join(x2016))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')
plt.show()
```
