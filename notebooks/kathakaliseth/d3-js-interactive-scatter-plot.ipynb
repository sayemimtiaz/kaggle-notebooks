```python
#Embedding the html string
html_string = """
<!DOCTYPE html>
<link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/4.7.0/css/font-awesome.min.css">
<meta charset="utf-8">
<style>
    .rectangle {
      fill: steelblue;
}
.rectangle:hover {
    fill: orange;
}
.axis {
  font: 10px sans-serif;
  fill: "b8b8b8";
}

.tooltip {
  background-color: black;
  border: #b8b8b8;
  border-radius: 5px;
  padding: 15px;
  min-width: 200px;
  text-align: left;
  color: white;
}

.tooltip2 {
  background-color: black;
  border: none;
  border-radius: 5px;
  padding: 15px;
  min-width: 400px;
  text-align: left;
  color: #663333;
}
.bar {
            fill: #5f89ad;
        }
        
        .axis {
            font-size: 13px;
            fill: #ffffff;
        }
        
        .axis path,
        .axis line {
            fill: none;
            display: none;
        }
        
        .label {
            font-size: 13px;
            color: #ffffff;
            fill: #ffffff;
        }
           .text2 {
            font-size: 13px;
             color: #ffffff;
 fill: #ffffff;
        }
       body {
  font-family: 'Open Sans', sans-serif;
}

div#layout {
  text-align: center;
}

div#container {
  width: 1000px;
  height: 600px;
  margin: auto;
  background-color: #2F4A6D;
}

svg {
  width: 100%;
  height: 100%;
}

.bar {
  fill: #80cbc4;
}

text2 {
  font-size: 12px;
  fill: #fff;
}

path {
  stroke: gray;
}

line {
  stroke: #b8b8b8;
}

line#limit {
  stroke: #FED966;
  stroke-width: 3;
  stroke-dasharray: 3 6;
}

.grid path {
  stroke-width: 0;
}

.grid .tick line {
  stroke: #b8b8b8;
  stroke-opacity: 0.3;
}

text.divergence {
  font-size: 14px;
  fill: #2F4A6D;
}

text.value {
  font-size: 14px;
}

text.title {
  font-size: 22px;
  font-weight: 600;
}
.xAxis line {
  stroke: #B8B8B8;
}

.yAxis line {
  stroke: #B8B8B8;
}
.node circle {
        stroke: #3a403d;
        stroke-width: .5px;
      }
.node:hover{
  stroke-width: 3px !important;
  opacity: 1 !important;
}


</style>
            <div class="chart" id="viz1" style="width:1000px;height:640px;background:#11111d";
            ;border:1px solid #999999;"></div>
 """

js_string="""
 require.config({
    paths: {
        d3: "https://d3js.org/d3.v4.min"
     }
 });
24
  require(["d3"], function(d3) {

   console.log(d3);
   var ds = [

{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":72,"Writing":74,"RW":73,"Score":72.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":90,"Writing":88,"RW":89,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":90,"Reading":95,"Writing":93,"RW":94,"Score":92.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":47,"Reading":57,"Writing":44,"RW":50.5,"Score":49.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":78,"Writing":75,"RW":76.5,"Score":76.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":83,"Writing":78,"RW":80.5,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":95,"Writing":92,"RW":93.5,"Score":91.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":40,"Reading":43,"Writing":39,"RW":41,"Score":40.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":64,"Reading":64,"Writing":67,"RW":65.5,"Score":65},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":38,"Reading":60,"Writing":50,"RW":55,"Score":49.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":54,"Writing":52,"RW":53,"Score":54.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":40,"Reading":52,"Writing":43,"RW":47.5,"Score":45},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":81,"Writing":73,"RW":77,"Score":73},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":78,"Reading":72,"Writing":70,"RW":71,"Score":73.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":50,"Reading":53,"Writing":58,"RW":55.5,"Score":53.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":75,"Writing":78,"RW":76.5,"Score":74},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":89,"Writing":86,"RW":87.5,"Score":87.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":18,"Reading":32,"Writing":28,"RW":30,"Score":26},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":46,"Reading":42,"Writing":46,"RW":44,"Score":44.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":54,"Reading":58,"Writing":61,"RW":59.5,"Score":57.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":69,"Writing":63,"RW":66,"Score":66},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":75,"Writing":70,"RW":72.5,"Score":70},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":44,"Reading":54,"Writing":53,"RW":53.5,"Score":50.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":73,"Writing":73,"RW":73,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":71,"Writing":80,"RW":75.5,"Score":75},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":74,"Writing":72,"RW":73,"Score":73},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":54,"Writing":55,"RW":54.5,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":69,"Writing":75,"RW":72,"Score":70.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":70,"Writing":65,"RW":67.5,"Score":68.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":70,"Writing":75,"RW":72.5,"Score":69},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":74,"Writing":74,"RW":74,"Score":72.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":65,"Writing":61,"RW":63,"Score":63},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":56,"Reading":72,"Writing":65,"RW":68.5,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":40,"Reading":42,"Writing":38,"RW":40,"Score":40},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":97,"Reading":87,"Writing":82,"RW":84.5,"Score":88.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":81,"Writing":79,"RW":80,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":81,"Writing":83,"RW":82,"Score":79.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":64,"Writing":59,"RW":61.5,"Score":57.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":75,"Reading":90,"Writing":88,"RW":89,"Score":84.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":57,"Reading":56,"Writing":57,"RW":56.5,"Score":56.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":61,"Writing":54,"RW":57.5,"Score":56.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":73,"Writing":68,"RW":70.5,"Score":66.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":58,"Writing":65,"RW":61.5,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":59,"Reading":65,"Writing":66,"RW":65.5,"Score":63.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":56,"Writing":54,"RW":55,"Score":53.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":54,"Writing":57,"RW":55.5,"Score":58.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":55,"Reading":65,"Writing":62,"RW":63.5,"Score":60.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":71,"Writing":76,"RW":73.5,"Score":71},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":57,"Reading":74,"Writing":76,"RW":75,"Score":69},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":84,"Writing":82,"RW":83,"Score":82.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":55,"Writing":48,"RW":51.5,"Score":52},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":77,"Reading":69,"Writing":68,"RW":68.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":44,"Writing":42,"RW":43,"Score":46.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":78,"Writing":75,"RW":76.5,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":71,"Reading":84,"Writing":87,"RW":85.5,"Score":80.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":33,"Reading":41,"Writing":43,"RW":42,"Score":39},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":85,"Writing":86,"RW":85.5,"Score":84.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":55,"Writing":49,"RW":52,"Score":52},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":58,"Reading":59,"Writing":58,"RW":58.5,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":0,"Reading":17,"Writing":10,"RW":13.5,"Score":9},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":79,"Reading":74,"Writing":72,"RW":73,"Score":75},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":39,"Reading":39,"Writing":34,"RW":36.5,"Score":37.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":61,"Writing":55,"RW":58,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":80,"Writing":71,"RW":75.5,"Score":73.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":58,"Writing":59,"RW":58.5,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":64,"Writing":61,"RW":62.5,"Score":64},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":45,"Reading":37,"Writing":37,"RW":37,"Score":39.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":60,"Reading":72,"Writing":74,"RW":73,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":58,"Writing":56,"RW":57,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":39,"Reading":64,"Writing":57,"RW":60.5,"Score":53.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":58,"Reading":63,"Writing":73,"RW":68,"Score":64.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":55,"Writing":63,"RW":59,"Score":60.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":41,"Reading":51,"Writing":48,"RW":49.5,"Score":46.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":57,"Writing":56,"RW":56.5,"Score":58},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":49,"Reading":49,"Writing":41,"RW":45,"Score":46.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":41,"Writing":38,"RW":39.5,"Score":41},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":30,"Reading":26,"Writing":22,"RW":24,"Score":26},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":80,"Reading":78,"Writing":81,"RW":79.5,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":61,"Reading":74,"Writing":72,"RW":73,"Score":69},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":68,"Writing":68,"RW":68,"Score":66},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":47,"Reading":49,"Writing":50,"RW":49.5,"Score":48.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":45,"Writing":45,"RW":45,"Score":46.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":50,"Reading":47,"Writing":54,"RW":50.5,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":64,"Writing":63,"RW":63.5,"Score":66.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":42,"Reading":39,"Writing":34,"RW":36.5,"Score":38.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":80,"Writing":82,"RW":81,"Score":78.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":76,"Reading":83,"Writing":88,"RW":85.5,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":71,"Writing":74,"RW":72.5,"Score":72},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":70,"Writing":67,"RW":68.5,"Score":65},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":86,"Writing":82,"RW":84,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":72,"Writing":74,"RW":73,"Score":70.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":27,"Reading":34,"Writing":36,"RW":35,"Score":32.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":79,"Writing":71,"RW":75,"Score":73.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":43,"Reading":45,"Writing":50,"RW":47.5,"Score":46},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":86,"Writing":92,"RW":89,"Score":85.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":78,"Reading":81,"Writing":82,"RW":81.5,"Score":80.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":66,"Writing":62,"RW":64,"Score":64.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":72,"Writing":70,"RW":71,"Score":68.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":67,"Writing":62,"RW":64.5,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":67,"Writing":62,"RW":64.5,"Score":64.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":67,"Writing":67,"RW":67,"Score":71},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":74,"Writing":74,"RW":74,"Score":72},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":91,"Writing":89,"RW":90,"Score":88.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":60,"Reading":44,"Writing":47,"RW":45.5,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":98,"Reading":86,"Writing":90,"RW":88,"Score":91.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":67,"Writing":72,"RW":69.5,"Score":65.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":100,"Writing":100,"RW":100,"Score":95.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":66,"Reading":63,"Writing":64,"RW":63.5,"Score":64.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":52,"Reading":76,"Writing":70,"RW":73,"Score":66},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":64,"Writing":72,"RW":68,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":77,"Reading":89,"Writing":98,"RW":93.5,"Score":88},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":55,"Writing":49,"RW":52,"Score":55.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":53,"Writing":47,"RW":50,"Score":51.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":58,"Writing":54,"RW":56,"Score":54.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":99,"Reading":100,"Writing":100,"RW":100,"Score":99.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":77,"Writing":74,"RW":75.5,"Score":78.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":75,"Reading":85,"Writing":82,"RW":83.5,"Score":80.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":78,"Reading":82,"Writing":79,"RW":80.5,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":63,"Writing":61,"RW":62,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":55,"Reading":69,"Writing":65,"RW":67,"Score":63},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":92,"Writing":89,"RW":90.5,"Score":86.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":91,"Reading":89,"Writing":92,"RW":90.5,"Score":90.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":93,"Writing":93,"RW":93,"Score":91.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":57,"Writing":56,"RW":56.5,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":83,"Reading":80,"Writing":73,"RW":76.5,"Score":78.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":95,"Writing":86,"RW":90.5,"Score":89.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":68,"Writing":67,"RW":67.5,"Score":69},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":77,"Writing":74,"RW":75.5,"Score":72},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":82,"Writing":74,"RW":78,"Score":79.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":49,"Writing":51,"RW":50,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":89,"Reading":84,"Writing":82,"RW":83,"Score":85},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":53,"Reading":37,"Writing":40,"RW":38.5,"Score":43.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":87,"Reading":74,"Writing":70,"RW":72,"Score":77},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":81,"Writing":84,"RW":82.5,"Score":80},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":79,"Writing":75,"RW":77,"Score":76},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":55,"Writing":48,"RW":51.5,"Score":53.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":51,"Reading":54,"Writing":41,"RW":47.5,"Score":48.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":55,"Writing":56,"RW":55.5,"Score":60.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":66,"Writing":67,"RW":66.5,"Score":64},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":61,"Writing":69,"RW":65,"Score":67},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":72,"Writing":71,"RW":71.5,"Score":73},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":59,"Reading":62,"Writing":64,"RW":63,"Score":61.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":42,"Reading":55,"Writing":54,"RW":54.5,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":43,"Writing":47,"RW":45,"Score":49},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":73,"Writing":78,"RW":75.5,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":22,"Reading":39,"Writing":33,"RW":36,"Score":31.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":84,"Writing":75,"RW":79.5,"Score":82.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":68,"Writing":66,"RW":67,"Score":69},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":75,"Writing":81,"RW":78,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":100,"Reading":100,"Writing":93,"RW":96.5,"Score":97.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":67,"Writing":69,"RW":68,"Score":66},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":67,"Writing":68,"RW":67.5,"Score":70.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":70,"Writing":66,"RW":68,"Score":65},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":49,"Writing":47,"RW":48,"Score":50},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":67,"Writing":61,"RW":64,"Score":63.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":70,"Reading":89,"Writing":88,"RW":88.5,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":66,"Reading":74,"Writing":78,"RW":76,"Score":72.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":60,"Reading":60,"Writing":60,"RW":60,"Score":60},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":61,"Reading":86,"Writing":87,"RW":86.5,"Score":78},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":66,"Reading":62,"Writing":64,"RW":63,"Score":64},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":82,"Reading":78,"Writing":74,"RW":76,"Score":78},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":75,"Reading":88,"Writing":85,"RW":86.5,"Score":82.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":53,"Writing":52,"RW":52.5,"Score":51.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":53,"Writing":49,"RW":51,"Score":51.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":92,"Writing":91,"RW":91.5,"Score":88},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":96,"Reading":100,"Writing":100,"RW":100,"Score":98.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":53,"Reading":51,"Writing":51,"RW":51,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":58,"Reading":76,"Writing":78,"RW":77,"Score":70.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":83,"Writing":78,"RW":80.5,"Score":76.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":75,"Writing":70,"RW":72.5,"Score":70.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":72,"Reading":73,"Writing":74,"RW":73.5,"Score":73},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":94,"Reading":88,"Writing":78,"RW":83,"Score":86.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":86,"Writing":81,"RW":83.5,"Score":82},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":67,"Writing":70,"RW":68.5,"Score":66.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":43,"Reading":51,"Writing":54,"RW":52.5,"Score":49.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":91,"Writing":87,"RW":89,"Score":86.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":46,"Reading":54,"Writing":58,"RW":56,"Score":52.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":77,"Writing":77,"RW":77,"Score":75},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":52,"Reading":70,"Writing":62,"RW":66,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":97,"Reading":100,"Writing":100,"RW":100,"Score":99},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":62,"Reading":68,"Writing":75,"RW":71.5,"Score":68.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":46,"Reading":64,"Writing":66,"RW":65,"Score":58.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":50,"Reading":50,"Writing":47,"RW":48.5,"Score":49},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":69,"Writing":70,"RW":69.5,"Score":68},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":45,"Reading":52,"Writing":49,"RW":50.5,"Score":48.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":67,"Writing":65,"RW":66,"Score":65.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":76,"Writing":65,"RW":70.5,"Score":73.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":66,"Writing":68,"RW":67,"Score":65.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":52,"Writing":45,"RW":48.5,"Score":48.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":88,"Writing":87,"RW":87.5,"Score":84},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":65,"Writing":69,"RW":67,"Score":66.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":83,"Writing":79,"RW":81,"Score":79.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":64,"Writing":66,"RW":65,"Score":64},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":62,"Writing":62,"RW":62,"Score":67},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":84,"Writing":85,"RW":84.5,"Score":79.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":55,"Writing":52,"RW":53.5,"Score":56},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":59,"Reading":69,"Writing":65,"RW":67,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":56,"Writing":51,"RW":53.5,"Score":54},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":45,"Reading":53,"Writing":55,"RW":54,"Score":51},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":78,"Reading":79,"Writing":76,"RW":77.5,"Score":77.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":84,"Writing":86,"RW":85,"Score":79},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":81,"Writing":77,"RW":79,"Score":74.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":77,"Writing":69,"RW":73,"Score":71.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":69,"Writing":68,"RW":68.5,"Score":64.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":41,"Writing":42,"RW":41.5,"Score":47.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":74,"Reading":71,"Writing":78,"RW":74.5,"Score":74.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":62,"Writing":62,"RW":62,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":80,"Writing":76,"RW":78,"Score":79},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":74,"Reading":81,"Writing":76,"RW":78.5,"Score":77},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":61,"Writing":66,"RW":63.5,"Score":61.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":80,"Reading":79,"Writing":79,"RW":79,"Score":79.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":35,"Reading":28,"Writing":27,"RW":27.5,"Score":30},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":42,"Reading":62,"Writing":60,"RW":61,"Score":54.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":60,"Reading":51,"Writing":56,"RW":53.5,"Score":55.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":91,"Writing":81,"RW":86,"Score":86.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":84,"Reading":83,"Writing":75,"RW":79,"Score":80.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":83,"Reading":86,"Writing":88,"RW":87,"Score":85.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":34,"Reading":42,"Writing":39,"RW":40.5,"Score":38.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":66,"Reading":77,"Writing":70,"RW":73.5,"Score":71},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":61,"Reading":56,"Writing":56,"RW":56,"Score":57.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":56,"Reading":68,"Writing":74,"RW":71,"Score":66},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":85,"Writing":73,"RW":79,"Score":81.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":65,"Writing":62,"RW":63.5,"Score":60.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":86,"Reading":80,"Writing":75,"RW":77.5,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":52,"Reading":66,"Writing":73,"RW":69.5,"Score":63.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":45,"Reading":56,"Writing":54,"RW":55,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":72,"Writing":71,"RW":71.5,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":50,"Writing":54,"RW":52,"Score":53.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":68,"Reading":72,"Writing":64,"RW":68,"Score":68},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":95,"Writing":94,"RW":94.5,"Score":92.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":64,"Writing":66,"RW":65,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":46,"Reading":43,"Writing":42,"RW":42.5,"Score":43.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":86,"Writing":83,"RW":84.5,"Score":78.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":92,"Reading":87,"Writing":78,"RW":82.5,"Score":85.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":83,"Reading":82,"Writing":84,"RW":83,"Score":83},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":75,"Writing":77,"RW":76,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":66,"Writing":67,"RW":66.5,"Score":65.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":64,"Reading":60,"Writing":74,"RW":67,"Score":66},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":52,"Writing":51,"RW":51.5,"Score":52.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":80,"Writing":80,"RW":80,"Score":81.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":73,"Reading":68,"Writing":66,"RW":67,"Score":69},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":83,"Writing":83,"RW":83,"Score":82},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":56,"Reading":52,"Writing":55,"RW":53.5,"Score":54.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":51,"Writing":43,"RW":47,"Score":51},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":74,"Writing":69,"RW":71.5,"Score":72.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":76,"Writing":71,"RW":73.5,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":89,"Reading":76,"Writing":74,"RW":75,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":58,"Reading":70,"Writing":68,"RW":69,"Score":65.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":64,"Writing":62,"RW":63,"Score":63.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":60,"Writing":53,"RW":56.5,"Score":60.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":47,"Reading":49,"Writing":49,"RW":49,"Score":48.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":71,"Reading":83,"Writing":83,"RW":83,"Score":79},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":60,"Reading":70,"Writing":70,"RW":70,"Score":66.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":80,"Writing":72,"RW":76,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":52,"Writing":52,"RW":52,"Score":52.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":73,"Writing":70,"RW":71.5,"Score":68.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":73,"Writing":68,"RW":70.5,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":78,"Reading":77,"Writing":77,"RW":77,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":75,"Writing":78,"RW":76.5,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":81,"Writing":81,"RW":81,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":64,"Reading":79,"Writing":77,"RW":78,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":79,"Writing":78,"RW":78.5,"Score":78.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":50,"Writing":51,"RW":50.5,"Score":48.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":99,"Reading":93,"Writing":90,"RW":91.5,"Score":94},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":73,"Writing":68,"RW":70.5,"Score":72.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":59,"Reading":42,"Writing":41,"RW":41.5,"Score":47.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":75,"Writing":81,"RW":78,"Score":73},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":72,"Writing":77,"RW":74.5,"Score":72.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":92,"Writing":95,"RW":93.5,"Score":91.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":71,"Reading":76,"Writing":70,"RW":73,"Score":72.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":63,"Writing":61,"RW":62,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":49,"Writing":42,"RW":45.5,"Score":49.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":47,"Reading":53,"Writing":58,"RW":55.5,"Score":52.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":70,"Writing":71,"RW":70.5,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":85,"Writing":76,"RW":80.5,"Score":83},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":83,"Reading":78,"Writing":73,"RW":75.5,"Score":78},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":92,"Writing":93,"RW":92.5,"Score":90},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":63,"Writing":75,"RW":69,"Score":65.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":86,"Writing":80,"RW":83,"Score":77},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":56,"Writing":57,"RW":56.5,"Score":62},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":52,"Writing":42,"RW":47,"Score":49},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":45,"Reading":48,"Writing":46,"RW":47,"Score":46.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":79,"Writing":84,"RW":81.5,"Score":78.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":70,"Reading":78,"Writing":78,"RW":78,"Score":75.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":37,"Reading":46,"Writing":46,"RW":46,"Score":43},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":82,"Writing":82,"RW":82,"Score":81.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":97,"Reading":82,"Writing":88,"RW":85,"Score":89},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":89,"Writing":82,"RW":85.5,"Score":79.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":88,"Reading":75,"Writing":76,"RW":75.5,"Score":79.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":76,"Writing":77,"RW":76.5,"Score":76.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":70,"Writing":68,"RW":69,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":86,"Reading":73,"Writing":70,"RW":71.5,"Score":76.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":60,"Writing":57,"RW":58.5,"Score":60},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":73,"Writing":75,"RW":74,"Score":71},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":78,"Reading":77,"Writing":80,"RW":78.5,"Score":78.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":67,"Reading":62,"Writing":60,"RW":61,"Score":63},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":46,"Reading":41,"Writing":43,"RW":42,"Score":43.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":74,"Writing":68,"RW":71,"Score":71},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":40,"Reading":46,"Writing":50,"RW":48,"Score":45.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":90,"Reading":87,"Writing":75,"RW":81,"Score":84},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":81,"Reading":78,"Writing":81,"RW":79.5,"Score":80},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":56,"Reading":54,"Writing":52,"RW":53,"Score":54},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":84,"Writing":81,"RW":82.5,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":76,"Writing":64,"RW":70,"Score":73.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":74,"Reading":75,"Writing":83,"RW":79,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":67,"Writing":69,"RW":68,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":99,"Reading":87,"Writing":81,"RW":84,"Score":89},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":52,"Writing":44,"RW":48,"Score":49},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":71,"Writing":67,"RW":69,"Score":63.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":57,"Writing":52,"RW":54.5,"Score":52.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":76,"Writing":80,"RW":78,"Score":76.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":60,"Writing":57,"RW":58.5,"Score":61},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":61,"Writing":68,"RW":64.5,"Score":65.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":68,"Reading":67,"Writing":69,"RW":68,"Score":68},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":64,"Writing":75,"RW":69.5,"Score":66},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":66,"Writing":65,"RW":65.5,"Score":67.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":82,"Writing":91,"RW":86.5,"Score":83.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":83,"Reading":72,"Writing":78,"RW":75,"Score":77.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":71,"Writing":69,"RW":70,"Score":67.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":56,"Reading":65,"Writing":63,"RW":64,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":79,"Writing":84,"RW":81.5,"Score":76.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":86,"Writing":79,"RW":82.5,"Score":80},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":81,"Writing":80,"RW":80.5,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":43,"Reading":53,"Writing":53,"RW":53,"Score":49.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":41,"Reading":46,"Writing":43,"RW":44.5,"Score":43.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":90,"Writing":94,"RW":92,"Score":88.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":61,"Writing":62,"RW":61.5,"Score":61.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":28,"Reading":23,"Writing":19,"RW":21,"Score":23.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":75,"Writing":77,"RW":76,"Score":78},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":41,"Reading":55,"Writing":51,"RW":53,"Score":49},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":60,"Writing":61,"RW":60.5,"Score":64},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":47,"Reading":37,"Writing":35,"RW":36,"Score":39.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":56,"Writing":53,"RW":54.5,"Score":57},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":90,"Reading":78,"Writing":81,"RW":79.5,"Score":83},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":83,"Reading":93,"Writing":95,"RW":94,"Score":90.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":68,"Writing":66,"RW":67,"Score":65},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":70,"Writing":69,"RW":69.5,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":49,"Reading":51,"Writing":43,"RW":47,"Score":47.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":24,"Reading":38,"Writing":27,"RW":32.5,"Score":29.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":35,"Reading":55,"Writing":60,"RW":57.5,"Score":50},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":61,"Writing":52,"RW":56.5,"Score":57},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":73,"Writing":63,"RW":68,"Score":65.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":76,"Writing":74,"RW":75,"Score":73},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":72,"Writing":67,"RW":69.5,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":73,"Writing":67,"RW":70,"Score":73},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":80,"Writing":75,"RW":77.5,"Score":75.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":61,"Writing":57,"RW":59,"Score":60},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":94,"Writing":95,"RW":94.5,"Score":88.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":75,"Reading":74,"Writing":66,"RW":70,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":74,"Writing":76,"RW":75,"Score":79},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":65,"Writing":69,"RW":67,"Score":62},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":57,"Writing":52,"RW":54.5,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":78,"Writing":80,"RW":79,"Score":73.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":46,"Reading":58,"Writing":57,"RW":57.5,"Score":53.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":71,"Writing":70,"RW":70.5,"Score":66.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":72,"Writing":70,"RW":71,"Score":67.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":61,"Writing":61,"RW":61,"Score":61.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":42,"Reading":66,"Writing":69,"RW":67.5,"Score":59},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":59,"Reading":62,"Writing":61,"RW":61.5,"Score":60.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":90,"Writing":89,"RW":89.5,"Score":86.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":62,"Writing":59,"RW":60.5,"Score":59.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":84,"Writing":78,"RW":81,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":58,"Writing":58,"RW":58,"Score":56},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":27,"Reading":34,"Writing":32,"RW":33,"Score":31},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":60,"Writing":58,"RW":59,"Score":59},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":49,"Reading":58,"Writing":60,"RW":59,"Score":55.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":58,"Writing":53,"RW":55.5,"Score":60},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":66,"Writing":61,"RW":63.5,"Score":62.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":64,"Writing":58,"RW":61,"Score":55.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":84,"Writing":85,"RW":84.5,"Score":80.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":77,"Writing":71,"RW":74,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":45,"Reading":73,"Writing":70,"RW":71.5,"Score":62.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":74,"Writing":72,"RW":73,"Score":73.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":97,"Writing":96,"RW":96.5,"Score":91.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":70,"Writing":73,"RW":71.5,"Score":67.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":46,"Reading":43,"Writing":41,"RW":42,"Score":43.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":90,"Writing":82,"RW":86,"Score":84},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":85,"Reading":95,"Writing":100,"RW":97.5,"Score":93.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":83,"Writing":77,"RW":80,"Score":77},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":64,"Writing":62,"RW":63,"Score":64},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":86,"Writing":83,"RW":84.5,"Score":83},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":100,"Writing":95,"RW":97.5,"Score":94},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":79,"Reading":81,"Writing":71,"RW":76,"Score":77},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":38,"Reading":49,"Writing":45,"RW":47,"Score":44},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":38,"Reading":43,"Writing":43,"RW":43,"Score":41.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":76,"Writing":75,"RW":75.5,"Score":72.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":73,"Writing":70,"RW":71.5,"Score":69},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":57,"Reading":78,"Writing":67,"RW":72.5,"Score":67.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":64,"Writing":64,"RW":64,"Score":63.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":70,"Writing":75,"RW":72.5,"Score":72.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":73,"Reading":67,"Writing":59,"RW":63,"Score":66.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":68,"Writing":77,"RW":72.5,"Score":74},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":67,"Writing":67,"RW":67,"Score":70},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":57,"Reading":54,"Writing":56,"RW":55,"Score":55.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":74,"Writing":77,"RW":75.5,"Score":72},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":45,"Writing":41,"RW":43,"Score":44.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":67,"Writing":63,"RW":65,"Score":60},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":89,"Writing":95,"RW":92,"Score":89.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":63,"Writing":57,"RW":60,"Score":64.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":60,"Reading":59,"Writing":54,"RW":56.5,"Score":57.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":54,"Writing":67,"RW":60.5,"Score":60},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":43,"Writing":43,"RW":43,"Score":46.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":65,"Writing":55,"RW":60,"Score":56.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":99,"Writing":100,"RW":99.5,"Score":95.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":59,"Writing":62,"RW":60.5,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":73,"Writing":68,"RW":70.5,"Score":68},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":65,"Writing":63,"RW":64,"Score":64.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":80,"Writing":77,"RW":78.5,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":52,"Reading":57,"Writing":56,"RW":56.5,"Score":55},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":84,"Writing":85,"RW":84.5,"Score":85.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":70,"Reading":71,"Writing":74,"RW":72.5,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":84,"Reading":83,"Writing":78,"RW":80.5,"Score":81.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":66,"Writing":60,"RW":63,"Score":65.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":67,"Writing":67,"RW":67,"Score":65.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":51,"Reading":72,"Writing":79,"RW":75.5,"Score":67.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":73,"Writing":69,"RW":71,"Score":75.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":74,"Writing":68,"RW":71,"Score":71},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":73,"Writing":67,"RW":70,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":59,"Writing":62,"RW":60.5,"Score":63},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":57,"Reading":56,"Writing":54,"RW":55,"Score":55.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":82,"Reading":93,"Writing":93,"RW":93,"Score":89.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":57,"Reading":58,"Writing":64,"RW":61,"Score":59.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":47,"Reading":58,"Writing":67,"RW":62.5,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":85,"Writing":80,"RW":82.5,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":41,"Reading":39,"Writing":34,"RW":36.5,"Score":38},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":67,"Writing":62,"RW":64.5,"Score":63.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":86,"Reading":83,"Writing":86,"RW":84.5,"Score":85},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":69,"Reading":71,"Writing":65,"RW":68,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":59,"Writing":53,"RW":56,"Score":59},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":68,"Reading":63,"Writing":54,"RW":58.5,"Score":61.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":66,"Writing":59,"RW":62.5,"Score":63},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":72,"Writing":70,"RW":71,"Score":67.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":56,"Writing":55,"RW":55.5,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":47,"Reading":59,"Writing":50,"RW":54.5,"Score":52},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":66,"Writing":66,"RW":66,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":50,"Reading":48,"Writing":53,"RW":50.5,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":68,"Writing":64,"RW":66,"Score":69},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":75,"Reading":66,"Writing":73,"RW":69.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":56,"Writing":51,"RW":53.5,"Score":59},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":89,"Reading":88,"Writing":82,"RW":85,"Score":86.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":81,"Writing":79,"RW":80,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":78,"Reading":81,"Writing":80,"RW":80.5,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":59,"Reading":73,"Writing":69,"RW":71,"Score":67},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":83,"Writing":76,"RW":79.5,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":79,"Reading":82,"Writing":73,"RW":77.5,"Score":78},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":74,"Writing":77,"RW":75.5,"Score":72.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":69,"Reading":66,"Writing":60,"RW":63,"Score":65},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":86,"Reading":81,"Writing":80,"RW":80.5,"Score":82.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":47,"Reading":46,"Writing":42,"RW":44,"Score":45},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":73,"Writing":72,"RW":72.5,"Score":75.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":64,"Reading":85,"Writing":85,"RW":85,"Score":78},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":100,"Reading":92,"Writing":97,"RW":94.5,"Score":96.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":77,"Writing":74,"RW":75.5,"Score":72},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":58,"Writing":49,"RW":53.5,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":61,"Writing":62,"RW":61.5,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":37,"Reading":56,"Writing":47,"RW":51.5,"Score":46.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":89,"Writing":89,"RW":89,"Score":85.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":54,"Writing":48,"RW":51,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":100,"Reading":100,"Writing":100,"RW":100,"Score":100},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":72,"Reading":65,"Writing":68,"RW":66.5,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":58,"Writing":55,"RW":56.5,"Score":55.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":54,"Reading":54,"Writing":45,"RW":49.5,"Score":51},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":70,"Writing":76,"RW":73,"Score":72.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":90,"Writing":91,"RW":90.5,"Score":86},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":58,"Writing":62,"RW":60,"Score":65},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":87,"Writing":91,"RW":89,"Score":87.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":26,"Reading":31,"Writing":38,"RW":34.5,"Score":31.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":72,"Reading":67,"Writing":65,"RW":66,"Score":68},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":77,"Reading":88,"Writing":85,"RW":86.5,"Score":83.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":91,"Reading":74,"Writing":76,"RW":75,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":83,"Reading":85,"Writing":90,"RW":87.5,"Score":86},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":69,"Writing":74,"RW":71.5,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":86,"Writing":84,"RW":85,"Score":79.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":67,"Writing":61,"RW":64,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":90,"Reading":90,"Writing":91,"RW":90.5,"Score":90.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":76,"Writing":83,"RW":79.5,"Score":76.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":62,"Writing":66,"RW":64,"Score":68},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":68,"Writing":72,"RW":70,"Score":73.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":55,"Reading":64,"Writing":70,"RW":67,"Score":63},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":71,"Writing":67,"RW":69,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":73,"Reading":71,"Writing":68,"RW":69.5,"Score":70.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":52,"Reading":59,"Writing":56,"RW":57.5,"Score":55.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":68,"Reading":68,"Writing":61,"RW":64.5,"Score":65.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":52,"Writing":46,"RW":49,"Score":52.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":49,"Reading":52,"Writing":54,"RW":53,"Score":51.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":74,"Writing":71,"RW":72.5,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":47,"Writing":56,"RW":51.5,"Score":54.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":60,"Reading":75,"Writing":74,"RW":74.5,"Score":69.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":64,"Reading":53,"Writing":57,"RW":55,"Score":58},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":79,"Reading":82,"Writing":82,"RW":82,"Score":81},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":65,"Reading":85,"Writing":76,"RW":80.5,"Score":75.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":64,"Writing":70,"RW":67,"Score":66},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":83,"Reading":83,"Writing":90,"RW":86.5,"Score":85.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":88,"Writing":90,"RW":89,"Score":86.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":64,"Writing":68,"RW":66,"Score":62},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":64,"Writing":66,"RW":65,"Score":66},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":48,"Writing":52,"RW":50,"Score":51.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":59,"Reading":78,"Writing":76,"RW":77,"Score":71},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":69,"Writing":68,"RW":68.5,"Score":67.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":71,"Writing":72,"RW":71.5,"Score":73},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":79,"Writing":82,"RW":80.5,"Score":78.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":94,"Reading":87,"Writing":92,"RW":89.5,"Score":91},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":61,"Writing":54,"RW":57.5,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":95,"Reading":89,"Writing":92,"RW":90.5,"Score":92},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":40,"Reading":59,"Writing":54,"RW":56.5,"Score":51},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":82,"Writing":80,"RW":81,"Score":81.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":70,"Writing":66,"RW":68,"Score":68},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":59,"Writing":54,"RW":56.5,"Score":56},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":78,"Writing":77,"RW":77.5,"Score":78},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":86,"Reading":92,"Writing":87,"RW":89.5,"Score":88.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":71,"Writing":73,"RW":72,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":50,"Writing":43,"RW":46.5,"Score":52.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":49,"Writing":52,"RW":50.5,"Score":54.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":54,"Reading":61,"Writing":62,"RW":61.5,"Score":59},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":77,"Reading":97,"Writing":94,"RW":95.5,"Score":89.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":87,"Writing":85,"RW":86,"Score":82.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":89,"Writing":84,"RW":86.5,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":66,"Reading":74,"Writing":73,"RW":73.5,"Score":71},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":66,"Reading":78,"Writing":78,"RW":78,"Score":74},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":78,"Writing":79,"RW":78.5,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":49,"Writing":52,"RW":50.5,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":91,"Reading":86,"Writing":84,"RW":85,"Score":87},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":58,"Writing":57,"RW":57.5,"Score":61.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":54,"Reading":59,"Writing":50,"RW":54.5,"Score":54.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":53,"Reading":52,"Writing":49,"RW":50.5,"Score":51.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":60,"Writing":59,"RW":59.5,"Score":62.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":56,"Reading":61,"Writing":60,"RW":60.5,"Score":59},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":36,"Reading":53,"Writing":43,"RW":48,"Score":44},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":29,"Reading":41,"Writing":47,"RW":44,"Score":39},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":74,"Writing":70,"RW":72,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":67,"Writing":73,"RW":70,"Score":69.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":47,"Reading":54,"Writing":53,"RW":53.5,"Score":51.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":61,"Writing":58,"RW":59.5,"Score":60.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":88,"Writing":94,"RW":91,"Score":87},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":73,"Reading":69,"Writing":68,"RW":68.5,"Score":70},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":66,"Reading":83,"Writing":83,"RW":83,"Score":77.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":51,"Reading":60,"Writing":58,"RW":59,"Score":56.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":66,"Writing":62,"RW":64,"Score":59.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":66,"Writing":71,"RW":68.5,"Score":74},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":97,"Reading":92,"Writing":86,"RW":89,"Score":91.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":69,"Writing":68,"RW":68.5,"Score":70.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":79,"Reading":82,"Writing":80,"RW":81,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":77,"Writing":79,"RW":78,"Score":79},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":95,"Writing":89,"RW":92,"Score":88.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":63,"Writing":66,"RW":64.5,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":78,"Reading":83,"Writing":80,"RW":81.5,"Score":80.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":92,"Reading":100,"Writing":97,"RW":98.5,"Score":96.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":72,"Reading":67,"Writing":64,"RW":65.5,"Score":67.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":67,"Writing":64,"RW":65.5,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":72,"Writing":69,"RW":70.5,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":79,"Reading":76,"Writing":65,"RW":70.5,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":87,"Reading":90,"Writing":88,"RW":89,"Score":88.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":40,"Reading":48,"Writing":50,"RW":49,"Score":46},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":62,"Writing":64,"RW":63,"Score":67.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":45,"Writing":40,"RW":42.5,"Score":46},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":32,"Reading":39,"Writing":33,"RW":36,"Score":34.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":55,"Reading":72,"Writing":79,"RW":75.5,"Score":68.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":67,"Writing":66,"RW":66.5,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":70,"Writing":70,"RW":70,"Score":64.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":66,"Writing":62,"RW":64,"Score":67},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":74,"Reading":75,"Writing":79,"RW":77,"Score":76},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":74,"Writing":74,"RW":74,"Score":70.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":96,"Reading":90,"Writing":92,"RW":91,"Score":92.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":63,"Reading":80,"Writing":80,"RW":80,"Score":74.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":51,"Writing":46,"RW":48.5,"Score":48.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":48,"Reading":43,"Writing":45,"RW":44,"Score":45.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":92,"Reading":100,"Writing":100,"RW":100,"Score":97.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":61,"Reading":71,"Writing":78,"RW":74.5,"Score":70},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":48,"Writing":47,"RW":47.5,"Score":52.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":68,"Reading":68,"Writing":67,"RW":67.5,"Score":67.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":75,"Writing":70,"RW":72.5,"Score":72},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":91,"Reading":96,"Writing":92,"RW":94,"Score":93},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":62,"Writing":56,"RW":59,"Score":57},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":50,"Reading":66,"Writing":64,"RW":65,"Score":60},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":81,"Writing":71,"RW":76,"Score":75.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":40,"Reading":55,"Writing":53,"RW":54,"Score":49.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":61,"Reading":51,"Writing":52,"RW":51.5,"Score":54.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":91,"Writing":89,"RW":90,"Score":87},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":48,"Reading":56,"Writing":58,"RW":57,"Score":54},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":61,"Writing":68,"RW":64.5,"Score":60.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":97,"Writing":96,"RW":96.5,"Score":91.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":79,"Writing":80,"RW":79.5,"Score":78.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":73,"Writing":78,"RW":75.5,"Score":71.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":73,"Reading":75,"Writing":80,"RW":77.5,"Score":76},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":77,"Writing":77,"RW":77,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":76,"Writing":76,"RW":76,"Score":72.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":55,"Reading":73,"Writing":73,"RW":73,"Score":67},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":63,"Writing":62,"RW":62.5,"Score":56.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":64,"Writing":65,"RW":64.5,"Score":61},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":48,"Reading":66,"Writing":65,"RW":65.5,"Score":59.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":57,"Writing":54,"RW":55.5,"Score":56.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":62,"Writing":50,"RW":56,"Score":61},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":68,"Writing":64,"RW":66,"Score":66.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":76,"Writing":73,"RW":74.5,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":92,"Reading":100,"Writing":99,"RW":99.5,"Score":97},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":56,"Reading":79,"Writing":72,"RW":75.5,"Score":69},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":30,"Reading":24,"Writing":15,"RW":19.5,"Score":23},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":54,"Writing":48,"RW":51,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":77,"Writing":73,"RW":75,"Score":73},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":82,"Writing":81,"RW":81.5,"Score":76},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":60,"Writing":63,"RW":61.5,"Score":59},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":29,"Reading":29,"Writing":30,"RW":29.5,"Score":29.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":78,"Writing":80,"RW":79,"Score":78},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":60,"Reading":57,"Writing":51,"RW":54,"Score":56},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":84,"Reading":89,"Writing":90,"RW":89.5,"Score":87.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":72,"Writing":62,"RW":67,"Score":69.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":84,"Writing":82,"RW":83,"Score":83.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":40,"Reading":58,"Writing":54,"RW":56,"Score":50.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":64,"Writing":62,"RW":63,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":63,"Writing":65,"RW":64,"Score":62},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":69,"Reading":60,"Writing":63,"RW":61.5,"Score":64},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":59,"Writing":66,"RW":62.5,"Score":61},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":94,"Reading":90,"Writing":91,"RW":90.5,"Score":91.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":77,"Writing":74,"RW":75.5,"Score":72},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":93,"Writing":93,"RW":93,"Score":89.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":60,"Reading":68,"Writing":72,"RW":70,"Score":66.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":37,"Reading":45,"Writing":38,"RW":41.5,"Score":40},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":78,"Writing":83,"RW":80.5,"Score":83},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":95,"Reading":81,"Writing":84,"RW":82.5,"Score":86.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":73,"Writing":68,"RW":70.5,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":35,"Reading":61,"Writing":54,"RW":57.5,"Score":50},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":63,"Writing":56,"RW":59.5,"Score":60.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":58,"Reading":51,"Writing":52,"RW":51.5,"Score":53.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":100,"Reading":96,"Writing":86,"RW":91,"Score":94},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":58,"Writing":62,"RW":60,"Score":60.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":100,"Reading":97,"Writing":99,"RW":98,"Score":98.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":69,"Reading":70,"Writing":63,"RW":66.5,"Score":67.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":48,"Writing":46,"RW":47,"Score":51.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":57,"Writing":46,"RW":51.5,"Score":50.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":44,"Reading":51,"Writing":55,"RW":53,"Score":50},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":64,"Writing":70,"RW":67,"Score":67},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":79,"Reading":60,"Writing":65,"RW":62.5,"Score":68},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":66,"Reading":74,"Writing":81,"RW":77.5,"Score":73.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":88,"Writing":85,"RW":86.5,"Score":82.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":84,"Writing":80,"RW":82,"Score":82.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":74,"Writing":64,"RW":69,"Score":69.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":80,"Writing":81,"RW":80.5,"Score":76},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":80,"Reading":92,"Writing":88,"RW":90,"Score":86.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":86,"Reading":76,"Writing":74,"RW":75,"Score":78.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":74,"Writing":73,"RW":73.5,"Score":74.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":41,"Reading":52,"Writing":51,"RW":51.5,"Score":48},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":88,"Writing":90,"RW":89,"Score":84},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":72,"Reading":81,"Writing":79,"RW":80,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":74,"Reading":79,"Writing":80,"RW":79.5,"Score":77.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":65,"Writing":60,"RW":62.5,"Score":65},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":81,"Writing":81,"RW":81,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":70,"Writing":65,"RW":67.5,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":62,"Writing":68,"RW":65,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":50,"Reading":53,"Writing":55,"RW":54,"Score":52.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":79,"Writing":81,"RW":80,"Score":76.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":51,"Reading":56,"Writing":53,"RW":54.5,"Score":53.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":80,"Writing":76,"RW":78,"Score":74.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":86,"Writing":98,"RW":92,"Score":89.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":70,"Writing":74,"RW":72,"Score":69.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":79,"Writing":79,"RW":79,"Score":77},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":67,"Writing":67,"RW":67,"Score":65.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":67,"Writing":64,"RW":65.5,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":66,"Writing":61,"RW":63.5,"Score":65.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":43,"Reading":60,"Writing":58,"RW":59,"Score":53.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":90,"Reading":87,"Writing":85,"RW":86,"Score":87.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":74,"Reading":77,"Writing":73,"RW":75,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":66,"Writing":63,"RW":64.5,"Score":67.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":71,"Writing":69,"RW":70,"Score":65},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":69,"Writing":67,"RW":68,"Score":67},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":63,"Writing":63,"RW":63,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":50,"Reading":60,"Writing":60,"RW":60,"Score":56.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":63,"Reading":73,"Writing":71,"RW":72,"Score":69},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":85,"Writing":87,"RW":86,"Score":83},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":74,"Writing":61,"RW":67.5,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":72,"Writing":77,"RW":74.5,"Score":76.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":66,"Reading":76,"Writing":68,"RW":72,"Score":70},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":52,"Reading":57,"Writing":50,"RW":53.5,"Score":53},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":78,"Writing":76,"RW":77,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":84,"Writing":84,"RW":84,"Score":77.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":77,"Writing":78,"RW":77.5,"Score":74.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":50,"Reading":64,"Writing":66,"RW":65,"Score":60},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":73,"Reading":78,"Writing":76,"RW":77,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":70,"Reading":82,"Writing":76,"RW":79,"Score":76},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":81,"Reading":75,"Writing":78,"RW":76.5,"Score":78},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":63,"Reading":61,"Writing":60,"RW":60.5,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":72,"Writing":74,"RW":73,"Score":71},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":60,"Reading":68,"Writing":60,"RW":64,"Score":62.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":55,"Writing":54,"RW":54.5,"Score":57},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":29,"Reading":40,"Writing":44,"RW":42,"Score":37.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":66,"Writing":68,"RW":67,"Score":65.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":94,"Reading":99,"Writing":100,"RW":99.5,"Score":97.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":75,"Writing":68,"RW":71.5,"Score":76},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":78,"Writing":73,"RW":75.5,"Score":76},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":58,"Writing":44,"RW":51,"Score":51.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":93,"Reading":90,"Writing":83,"RW":86.5,"Score":88.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":49,"Reading":53,"Writing":53,"RW":53,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":76,"Writing":78,"RW":77,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":66,"Reading":74,"Writing":81,"RW":77.5,"Score":73.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":77,"Writing":73,"RW":75,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":49,"Reading":63,"Writing":56,"RW":59.5,"Score":56},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":79,"Reading":89,"Writing":86,"RW":87.5,"Score":84.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":82,"Writing":90,"RW":86,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":72,"Writing":70,"RW":71,"Score":67},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":57,"Reading":78,"Writing":79,"RW":78.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":66,"Reading":66,"Writing":59,"RW":62.5,"Score":63.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":81,"Writing":82,"RW":81.5,"Score":80.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":67,"Writing":72,"RW":69.5,"Score":65.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":84,"Writing":87,"RW":85.5,"Score":86},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":64,"Writing":67,"RW":65.5,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":59,"Reading":63,"Writing":64,"RW":63.5,"Score":62},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":72,"Writing":65,"RW":68.5,"Score":66.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":46,"Reading":34,"Writing":36,"RW":35,"Score":38.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":59,"Writing":52,"RW":55.5,"Score":59},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":89,"Reading":87,"Writing":79,"RW":83,"Score":85},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":42,"Reading":61,"Writing":58,"RW":59.5,"Score":53.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":93,"Reading":84,"Writing":90,"RW":87,"Score":89},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":80,"Reading":85,"Writing":85,"RW":85,"Score":83.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":98,"Reading":100,"Writing":99,"RW":99.5,"Score":99},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":81,"Writing":84,"RW":82.5,"Score":82},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":60,"Reading":70,"Writing":74,"RW":72,"Score":68},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":76,"Reading":94,"Writing":87,"RW":90.5,"Score":85.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":73,"Reading":78,"Writing":72,"RW":75,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":96,"Reading":96,"Writing":99,"RW":97.5,"Score":97},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":76,"Reading":76,"Writing":74,"RW":75,"Score":75.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":91,"Reading":73,"Writing":80,"RW":76.5,"Score":81.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":72,"Writing":70,"RW":71,"Score":68},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":55,"Reading":59,"Writing":59,"RW":59,"Score":57.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":90,"Writing":88,"RW":89,"Score":84},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":50,"Reading":48,"Writing":42,"RW":45,"Score":46.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":47,"Reading":43,"Writing":41,"RW":42,"Score":43.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":74,"Writing":71,"RW":72.5,"Score":75.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":75,"Writing":77,"RW":76,"Score":72.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":51,"Writing":57,"RW":54,"Score":58.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":73,"Reading":92,"Writing":84,"RW":88,"Score":83},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":39,"Writing":37,"RW":38,"Score":43},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":68,"Reading":77,"Writing":80,"RW":78.5,"Score":75},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":46,"Writing":43,"RW":44.5,"Score":48},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":89,"Writing":94,"RW":91.5,"Score":90},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":55,"Reading":47,"Writing":44,"RW":45.5,"Score":48.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":58,"Writing":57,"RW":57.5,"Score":56},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":57,"Writing":59,"RW":58,"Score":61},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":92,"Reading":79,"Writing":84,"RW":81.5,"Score":85},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":53,"Reading":66,"Writing":73,"RW":69.5,"Score":64},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":71,"Writing":73,"RW":72,"Score":75},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":60,"Writing":55,"RW":57.5,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":73,"Writing":72,"RW":72.5,"Score":75},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":37,"Reading":57,"Writing":56,"RW":56.5,"Score":50},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":84,"Writing":82,"RW":83,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":59,"Reading":73,"Writing":72,"RW":72.5,"Score":68},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":55,"Writing":47,"RW":51,"Score":52.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":79,"Writing":74,"RW":76.5,"Score":75},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":75,"Writing":71,"RW":73,"Score":71.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":64,"Writing":68,"RW":66,"Score":67},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":60,"Writing":59,"RW":59.5,"Score":56.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":84,"Writing":86,"RW":85,"Score":85.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":69,"Writing":68,"RW":68.5,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":72,"Writing":65,"RW":68.5,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":79,"Reading":77,"Writing":75,"RW":76,"Score":77},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":90,"Writing":85,"RW":87.5,"Score":84},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":55,"Writing":53,"RW":54,"Score":55.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":95,"Writing":92,"RW":93.5,"Score":90.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":55,"Reading":58,"Writing":52,"RW":55,"Score":55},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":70,"Reading":68,"Writing":72,"RW":70,"Score":70},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":52,"Reading":59,"Writing":65,"RW":62,"Score":58.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":69,"Reading":77,"Writing":77,"RW":77,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":53,"Reading":72,"Writing":64,"RW":68,"Score":63},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":48,"Reading":58,"Writing":54,"RW":56,"Score":53.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":78,"Reading":81,"Writing":86,"RW":83.5,"Score":81.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":62,"Writing":63,"RW":62.5,"Score":62.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":60,"Reading":63,"Writing":59,"RW":61,"Score":60.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":72,"Writing":72,"RW":72,"Score":72.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":58,"Reading":75,"Writing":77,"RW":76,"Score":70},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":62,"Writing":60,"RW":61,"Score":66},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":71,"Writing":75,"RW":73,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":58,"Reading":60,"Writing":57,"RW":58.5,"Score":58.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":48,"Writing":49,"RW":48.5,"Score":49.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":73,"Writing":74,"RW":73.5,"Score":74},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":52,"Reading":67,"Writing":72,"RW":69.5,"Score":63.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":78,"Writing":79,"RW":78.5,"Score":73},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":65,"Writing":60,"RW":62.5,"Score":63.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":58,"Writing":55,"RW":56.5,"Score":54},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":72,"Writing":70,"RW":71,"Score":69.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":35,"Reading":44,"Writing":43,"RW":43.5,"Score":40.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":72,"Reading":79,"Writing":82,"RW":80.5,"Score":77.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":94,"Reading":85,"Writing":82,"RW":83.5,"Score":87},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":46,"Reading":56,"Writing":57,"RW":56.5,"Score":53},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":90,"Writing":84,"RW":87,"Score":83.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":76,"Reading":85,"Writing":82,"RW":83.5,"Score":81},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":52,"Reading":59,"Writing":62,"RW":60.5,"Score":57.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":91,"Reading":81,"Writing":79,"RW":80,"Score":83.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":32,"Reading":51,"Writing":44,"RW":47.5,"Score":42.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":72,"Reading":79,"Writing":77,"RW":78,"Score":76},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":19,"Reading":38,"Writing":32,"RW":35,"Score":29.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":68,"Reading":65,"Writing":61,"RW":63,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":52,"Reading":65,"Writing":61,"RW":63,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":48,"Reading":62,"Writing":60,"RW":61,"Score":56.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":60,"Reading":66,"Writing":70,"RW":68,"Score":65.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":66,"Reading":74,"Writing":69,"RW":71.5,"Score":69.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":89,"Reading":84,"Writing":77,"RW":80.5,"Score":83.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":42,"Reading":52,"Writing":51,"RW":51.5,"Score":48.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":57,"Reading":68,"Writing":73,"RW":70.5,"Score":66},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":70,"Reading":70,"Writing":70,"RW":70,"Score":70},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":70,"Reading":84,"Writing":81,"RW":82.5,"Score":78.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":60,"Writing":54,"RW":57,"Score":61},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":52,"Reading":55,"Writing":57,"RW":56,"Score":54.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":73,"Writing":68,"RW":70.5,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":76,"Reading":80,"Writing":73,"RW":76.5,"Score":76.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":94,"Writing":95,"RW":94.5,"Score":92},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":85,"Writing":87,"RW":86,"Score":84.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":76,"Writing":78,"RW":77,"Score":75.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":75,"Reading":81,"Writing":74,"RW":77.5,"Score":76.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":74,"Writing":75,"RW":74.5,"Score":71},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":41,"Reading":45,"Writing":40,"RW":42.5,"Score":42},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":90,"Reading":75,"Writing":69,"RW":72,"Score":78},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":54,"Writing":51,"RW":52.5,"Score":54.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":31,"Writing":36,"RW":33.5,"Score":39.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":45,"Reading":47,"Writing":49,"RW":48,"Score":47},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":54,"Reading":64,"Writing":67,"RW":65.5,"Score":61.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":87,"Reading":84,"Writing":76,"RW":80,"Score":82.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":80,"Writing":83,"RW":81.5,"Score":78.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":94,"Reading":86,"Writing":87,"RW":86.5,"Score":89},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":45,"Reading":59,"Writing":64,"RW":61.5,"Score":56},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":61,"Reading":70,"Writing":76,"RW":73,"Score":69},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":60,"Reading":72,"Writing":68,"RW":70,"Score":66.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":77,"Reading":91,"Writing":88,"RW":89.5,"Score":85.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":90,"Writing":92,"RW":91,"Score":89},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":78,"Reading":90,"Writing":93,"RW":91.5,"Score":87},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":49,"Reading":52,"Writing":51,"RW":51.5,"Score":50.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":71,"Reading":87,"Writing":82,"RW":84.5,"Score":80},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":58,"Writing":52,"RW":55,"Score":52.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":67,"Writing":58,"RW":62.5,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":56,"Reading":68,"Writing":70,"RW":69,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":69,"Writing":76,"RW":72.5,"Score":70},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":69,"Reading":86,"Writing":81,"RW":83.5,"Score":78.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":54,"Writing":53,"RW":53.5,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":61,"Reading":60,"Writing":57,"RW":58.5,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":86,"Writing":89,"RW":87.5,"Score":83},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":60,"Writing":58,"RW":59,"Score":60.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":82,"Writing":89,"RW":85.5,"Score":82.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":50,"Writing":45,"RW":47.5,"Score":51},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":60,"Reading":64,"Writing":74,"RW":69,"Score":66},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":64,"Writing":57,"RW":60.5,"Score":64.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":82,"Writing":79,"RW":80.5,"Score":78.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":58,"Reading":57,"Writing":53,"RW":55,"Score":56},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":66,"Reading":77,"Writing":73,"RW":75,"Score":72},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":39,"Reading":52,"Writing":46,"RW":49,"Score":45.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":58,"Writing":51,"RW":54.5,"Score":57.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":23,"Reading":44,"Writing":36,"RW":40,"Score":34.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":74,"Reading":77,"Writing":76,"RW":76.5,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":40,"Reading":65,"Writing":64,"RW":64.5,"Score":56.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":90,"Reading":85,"Writing":84,"RW":84.5,"Score":86.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":91,"Reading":85,"Writing":85,"RW":85,"Score":87},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":54,"Writing":50,"RW":52,"Score":56},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":72,"Writing":68,"RW":70,"Score":66.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":80,"Reading":75,"Writing":69,"RW":72,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":71,"Reading":67,"Writing":67,"RW":67,"Score":68.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":68,"Writing":63,"RW":65.5,"Score":64},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":87,"Reading":85,"Writing":93,"RW":89,"Score":88.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":67,"Writing":61,"RW":64,"Score":70},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":64,"Writing":55,"RW":59.5,"Score":60.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":97,"Reading":97,"Writing":96,"RW":96.5,"Score":96.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":75,"Reading":68,"Writing":65,"RW":66.5,"Score":69.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":65,"Reading":79,"Writing":81,"RW":80,"Score":75},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":52,"Reading":49,"Writing":46,"RW":47.5,"Score":49},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":87,"Reading":73,"Writing":72,"RW":72.5,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":53,"Reading":62,"Writing":53,"RW":57.5,"Score":56},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":81,"Reading":86,"Writing":87,"RW":86.5,"Score":84.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":39,"Reading":42,"Writing":38,"RW":40,"Score":39.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":71,"Writing":80,"RW":75.5,"Score":74},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":97,"Reading":93,"Writing":91,"RW":92,"Score":93.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":82,"Writing":88,"RW":85,"Score":84},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":59,"Reading":53,"Writing":52,"RW":52.5,"Score":54.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":61,"Reading":42,"Writing":41,"RW":41.5,"Score":48},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":78,"Reading":74,"Writing":72,"RW":73,"Score":74.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":51,"Writing":51,"RW":51,"Score":50.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":58,"Writing":47,"RW":52.5,"Score":54.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":70,"Reading":72,"Writing":76,"RW":74,"Score":72.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":82,"Reading":84,"Writing":78,"RW":81,"Score":81.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":90,"Reading":90,"Writing":82,"RW":86,"Score":87.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":43,"Reading":62,"Writing":61,"RW":61.5,"Score":55.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":80,"Reading":64,"Writing":66,"RW":65,"Score":70},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":82,"Writing":84,"RW":83,"Score":82.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":61,"Writing":54,"RW":57.5,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":72,"Writing":80,"RW":76,"Score":70.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":64,"Reading":76,"Writing":74,"RW":75,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":63,"Reading":64,"Writing":66,"RW":65,"Score":64.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":71,"Reading":70,"Writing":70,"RW":70,"Score":70.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":73,"Writing":71,"RW":72,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":46,"Writing":44,"RW":45,"Score":48.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":51,"Reading":51,"Writing":54,"RW":52.5,"Score":52},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":62,"Reading":76,"Writing":80,"RW":78,"Score":72.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":93,"Reading":100,"Writing":95,"RW":97.5,"Score":96},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":54,"Reading":72,"Writing":59,"RW":65.5,"Score":61.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":69,"Reading":65,"Writing":74,"RW":69.5,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":51,"Writing":48,"RW":49.5,"Score":47.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":86,"Reading":85,"Writing":91,"RW":88,"Score":87.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":92,"Writing":85,"RW":88.5,"Score":87.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":67,"Writing":73,"RW":70,"Score":63.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":74,"Writing":75,"RW":74.5,"Score":79},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":59,"Reading":62,"Writing":69,"RW":65.5,"Score":63.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":32,"Reading":34,"Writing":38,"RW":36,"Score":34.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":36,"Reading":29,"Writing":27,"RW":28,"Score":30.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":63,"Reading":78,"Writing":79,"RW":78.5,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":67,"Reading":54,"Writing":63,"RW":58.5,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":78,"Writing":82,"RW":80,"Score":75},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":85,"Reading":84,"Writing":89,"RW":86.5,"Score":86},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":73,"Reading":78,"Writing":74,"RW":76,"Score":75},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":34,"Reading":48,"Writing":41,"RW":44.5,"Score":41},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":93,"Reading":100,"Writing":100,"RW":100,"Score":97.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":67,"Reading":84,"Writing":84,"RW":84,"Score":78.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":88,"Reading":77,"Writing":77,"RW":77,"Score":80.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":48,"Writing":51,"RW":49.5,"Score":52},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":84,"Writing":91,"RW":87.5,"Score":84.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":67,"Reading":75,"Writing":72,"RW":73.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":70,"Reading":64,"Writing":70,"RW":67,"Score":68},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":50,"Reading":42,"Writing":48,"RW":45,"Score":46.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":69,"Reading":84,"Writing":82,"RW":83,"Score":78.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":52,"Reading":61,"Writing":66,"RW":63.5,"Score":59.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":47,"Reading":62,"Writing":66,"RW":64,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":46,"Reading":61,"Writing":55,"RW":58,"Score":54},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":70,"Writing":66,"RW":68,"Score":68},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"Completed","Math":100,"Reading":100,"Writing":100,"RW":100,"Score":100},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":44,"Reading":61,"Writing":52,"RW":56.5,"Score":52.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":57,"Reading":77,"Writing":80,"RW":78.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":91,"Reading":96,"Writing":91,"RW":93.5,"Score":92.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":69,"Reading":70,"Writing":67,"RW":68.5,"Score":68.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":35,"Reading":53,"Writing":46,"RW":49.5,"Score":44.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":66,"Writing":66,"RW":66,"Score":68},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":54,"Reading":65,"Writing":65,"RW":65,"Score":61.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":74,"Reading":70,"Writing":69,"RW":69.5,"Score":71},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":74,"Reading":64,"Writing":60,"RW":62,"Score":66},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":64,"Reading":56,"Writing":52,"RW":54,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":61,"Writing":71,"RW":66,"Score":65.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":46,"Reading":43,"Writing":44,"RW":43.5,"Score":44.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":56,"Writing":51,"RW":53.5,"Score":51.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":74,"Writing":70,"RW":72,"Score":70.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":57,"Writing":62,"RW":59.5,"Score":60.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":61,"Reading":71,"Writing":73,"RW":72,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":70,"Reading":75,"Writing":74,"RW":74.5,"Score":73},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"Completed","Math":98,"Reading":87,"Writing":90,"RW":88.5,"Score":91.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":70,"Reading":63,"Writing":58,"RW":60.5,"Score":63.6666666666667},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":67,"Reading":57,"Writing":53,"RW":55,"Score":59},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":57,"Reading":58,"Writing":57,"RW":57.5,"Score":57.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":85,"Reading":81,"Writing":85,"RW":83,"Score":83.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":77,"Reading":68,"Writing":69,"RW":68.5,"Score":71.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Master","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":72,"Reading":66,"Writing":72,"RW":69,"Score":70},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":78,"Reading":91,"Writing":96,"RW":93.5,"Score":88.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":66,"Writing":64,"RW":65,"Score":70.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":61,"Reading":62,"Writing":61,"RW":61.5,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":58,"Reading":68,"Writing":61,"RW":64.5,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":61,"Writing":58,"RW":59.5,"Score":57.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":82,"Reading":82,"Writing":80,"RW":81,"Score":81.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":58,"Writing":60,"RW":59,"Score":55.6666666666667},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":49,"Reading":50,"Writing":52,"RW":51,"Score":50.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":57,"Reading":75,"Writing":73,"RW":74,"Score":68.3333333333333},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":94,"Reading":73,"Writing":71,"RW":72,"Score":79.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":75,"Reading":77,"Writing":83,"RW":80,"Score":78.3333333333333},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":74,"Reading":74,"Writing":72,"RW":73,"Score":73.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":58,"Reading":52,"Writing":54,"RW":53,"Score":54.6666666666667},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":69,"Writing":69,"RW":69,"Score":66.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":57,"Writing":62,"RW":59.5,"Score":63.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":84,"Reading":87,"Writing":81,"RW":84,"Score":84},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"None","Math":92,"Reading":100,"Writing":100,"RW":100,"Score":97.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":45,"Reading":63,"Writing":59,"RW":61,"Score":55.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":81,"Writing":71,"RW":76,"Score":75.6666666666667},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":56,"Reading":58,"Writing":64,"RW":61,"Score":59.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":48,"Reading":54,"Writing":53,"RW":53.5,"Score":51.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":100,"Reading":100,"Writing":100,"RW":100,"Score":100},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":65,"Reading":76,"Writing":75,"RW":75.5,"Score":72},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":72,"Reading":57,"Writing":58,"RW":57.5,"Score":62.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":70,"Writing":72,"RW":71,"Score":68},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":66,"Reading":68,"Writing":64,"RW":66,"Score":66},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":63,"Writing":60,"RW":61.5,"Score":62},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":68,"Reading":76,"Writing":67,"RW":71.5,"Score":70.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":75,"Reading":84,"Writing":80,"RW":82,"Score":79.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Standard","PreparationCourse":"None","Math":89,"Reading":100,"Writing":100,"RW":100,"Score":96.3333333333333},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":78,"Reading":72,"Writing":69,"RW":70.5,"Score":73},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":53,"Reading":50,"Writing":60,"RW":55,"Score":54.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":49,"Reading":65,"Writing":61,"RW":63,"Score":58.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"None","Math":54,"Reading":63,"Writing":67,"RW":65,"Score":61.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":64,"Reading":82,"Writing":77,"RW":79.5,"Score":74.3333333333333},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":60,"Reading":62,"Writing":60,"RW":61,"Score":60.6666666666667},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":62,"Reading":65,"Writing":58,"RW":61.5,"Score":61.6666666666667},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":55,"Reading":41,"Writing":48,"RW":44.5,"Score":48},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":91,"Reading":95,"Writing":94,"RW":94.5,"Score":93.3333333333333},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":8,"Reading":24,"Writing":23,"RW":23.5,"Score":18.3333333333333},
{"Gender":"Male","Race":"GroupD","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":81,"Reading":78,"Writing":78,"RW":78,"Score":79},
{"Gender":"Male","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":79,"Reading":85,"Writing":86,"RW":85.5,"Score":83.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":78,"Reading":87,"Writing":91,"RW":89,"Score":85.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"None","Math":74,"Reading":75,"Writing":82,"RW":78.5,"Score":77},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":57,"Reading":51,"Writing":54,"RW":52.5,"Score":54},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"Associate","Lunch":"Standard","PreparationCourse":"None","Math":40,"Reading":59,"Writing":51,"RW":55,"Score":50},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":81,"Reading":75,"Writing":76,"RW":75.5,"Score":77.3333333333333},
{"Gender":"Female","Race":"GroupA","ParentalEducation":"SomeHighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":44,"Reading":45,"Writing":45,"RW":45,"Score":44.6666666666667},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":67,"Reading":86,"Writing":83,"RW":84.5,"Score":78.6666666666667},
{"Gender":"Male","Race":"GroupE","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":86,"Reading":81,"Writing":75,"RW":78,"Score":80.6666666666667},
{"Gender":"Female","Race":"GroupB","ParentalEducation":"SomeHighSchool","Lunch":"Standard","PreparationCourse":"Completed","Math":65,"Reading":82,"Writing":78,"RW":80,"Score":75},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Associate","Lunch":"Free_Reduced","PreparationCourse":"None","Math":55,"Reading":76,"Writing":76,"RW":76,"Score":69},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"Bachelor","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":72,"Writing":74,"RW":73,"Score":69.3333333333333},
{"Gender":"Male","Race":"GroupA","ParentalEducation":"HighSchool","Lunch":"Standard","PreparationCourse":"None","Math":63,"Reading":63,"Writing":62,"RW":62.5,"Score":62.6666666666667},
{"Gender":"Female","Race":"GroupE","ParentalEducation":"Master","Lunch":"Standard","PreparationCourse":"Completed","Math":88,"Reading":99,"Writing":95,"RW":97,"Score":94},
{"Gender":"Male","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"None","Math":62,"Reading":55,"Writing":55,"RW":55,"Score":57.3333333333333},
{"Gender":"Female","Race":"GroupC","ParentalEducation":"HighSchool","Lunch":"Free_Reduced","PreparationCourse":"Completed","Math":59,"Reading":71,"Writing":65,"RW":68,"Score":65},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Standard","PreparationCourse":"Completed","Math":68,"Reading":78,"Writing":77,"RW":77.5,"Score":74.3333333333333},
{"Gender":"Female","Race":"GroupD","ParentalEducation":"SomeCollege","Lunch":"Free_Reduced","PreparationCourse":"None","Math":77,"Reading":86,"Writing":86,"RW":86,"Score":83}

    ];

    function wrap(text, width) {
    text.each(function () {
        var text = d3.select(this),
            words = text.text().split(/\s+/).reverse(),
            word,
            line = [],
            lineNumber = 0,
            lineHeight = 1.5, // ems
            x = text.attr("x"),
            y = text.attr("y"),
            dy = 0, //parseFloat(text.attr("dy")),
            tspan = text.text(null)
                        .append("tspan")
                        .attr("x", x)
                        .attr("y", y)
                        .attr("dy", dy + "em");
        while (word = words.pop()) {
            line.push(word);
            tspan.text(line.join(" "));
            if (tspan.node().getComputedTextLength() > width) {
                line.pop();
                tspan.text(line.join(" "));
                line = [word];
                tspan = text.append("tspan")
                            .attr("x", x)
                            .attr("y", y)
                            .attr("dy", ++lineNumber * lineHeight + dy + "em")
                            .text(word);
            }
        }
    });
}

  // set the dimensions and margins of the graph
  var margin1 = {top: 50, right: 120, bottom: 130, left: 60},
      width1 = 1080 - margin1.left - margin1.right , //1404 - margin1.left - margin1.right,
      height1 = 680 - margin1.top - margin1.bottom;
  
  // append the svg object to the body of the page
  var svg1 = d3.select("#viz1")
    .append("svg")
  //  .attr("preserveAspectRatio", "xMinYMin meet")
   // Class to make it responsive.
  // .classed("svg-content-responsive", true)
      .attr("width", width1 + margin1.left + margin1.right)
      .attr("height", height1 + margin1.top + margin1.bottom)
    .append("g")
      .attr("transform",
            "translate(" + margin1.left + "," + margin1.top + ")");
  
  //Read the data
  var data1 = ds
  
  
    
  
    // ---------------------------//
    //       AXIS  AND SCALE      //
    // ---------------------------//
  
    // Add X axis
    var x1 = d3.scaleLinear()
      .domain([0, 100])
      .range([ 0, width1-240-140 ]);  //660
    svg1.append("g")
      .style("font-size",10)
      .style("font-weight",200)
       .style("stroke","#b8b8b8")
      .attr("transform", "translate(0," + (height1 + 40) + ")")
      .call(d3.axisBottom(x1).ticks(12).tickSizeOuter(0).tickSizeInner(0));
  
    // Add X axis label:
    svg1.append("text")
        .attr("text-anchor", "end")
        .attr("x", width1-400)
        .attr("y", height1+70 )        
        .attr("fill","#dcdcaa")
        .text("Math Score");
  
    // Add Y axis
    var y1 = d3.scaleLinear()
      .domain([0,100])
      .range([ (height1+40), 50]); //520
    svg1.append("g")
    .style("font-size",10)
      .style("font-weight",200)
      .style("stroke","#b8b8b8")
      .call(d3.axisLeft(y1).ticks(12).tickSizeOuter(0).tickSizeInner(0));
  
    // Add Y axis label:
    svg1.append("text")
        .attr("text-anchor", "end")
        .attr("x", -260)
        .attr("y", -30 )
        .attr("fill","#dcdcaa")
        .text("Average Reading & Writing Score")
        .attr("text-anchor", "start")
        .attr("transform", "rotate(270)");
  
    // Add a scale for bubble size
    var z1 = d3.scaleSqrt()
      .domain([0, 100])
//  .range([ 2, 28]);
   .range([ 4, 4]);

    var myColor2 = d3.scaleOrdinal()
          .domain([ 
"SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"
])
.range(["#ff0d57","#ff7fa7","pink","#7fc4fc","#1e88e5","blue"]);
   
var title = d3.scaleOrdinal()
          .domain(
          ["Gender","Parental Education", "Race / Ethnicity","Free Lunch","Preparatoty Course"]
)
.range([96,160,290,404,470]);

var myColorq = d3.scaleOrdinal()
          .domain([ "Female","Male"])

          .range(["#03aa99", "#aa4422"]);
   
var myColorr = d3.scaleOrdinal()
          .domain([ "Free_Reduced","Standard"])
          .range(["#03aa99", "#aa4422"]);

var myColors = d3.scaleOrdinal()
          .domain([ "Completed","None"])
.range(["#03aa99","#aa4422"]);

  
     var myColor3 = d3.scaleOrdinal()
            .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"])
            .range(["#3366cc", "#dc3912", "#109618", "orange", "#990099"])

  
    // ---------------------------//
    //      TOOLTIP               //
    // ---------------------------//
  
    // -1- Create a tooltip div that is hidden by default:
    var tooltip1 = d3.select('#viz1')
      .append("div")
        .style("opacity", 0)
        .attr("class", "tooltip")
        .style("background-color", "#111111")
        .style("stroke", "#111111")
        .style("border-radius", "5px")
        .style("padding", "10px")
        .style("color", "white")
        .style("font-size", "12px")
        .style("font-weight", 600)


  
    // -2- Create 3 functions to show / update (when mouse move but stay on same circle) / hide the tooltip
    var showTooltip1 = function(d) {
      tooltip1
        .transition()
        .duration(10)
      tooltip1
        .style("opacity", 0.8)
        .html( "Average Score: " +'<font color="#ff5588">'+format1(d.Score) + "  " +'</font>'+ "<br />" 
            + "Race/Ethnicity: " +'<font color="gold">'+d.Race +'</font>'+ "<br />"
            + "Gender: " +'<font color="gold">'+d.Gender +'</font>'+ "<br />"
            + "Parental Education: " +'<font color="gold">'+d.ParentalEducation +'</font>'+ "<br />"
            + "Lunch: " + '<font color="gold">'+d.Lunch + '</font>')            
        .style("left", (d3.mouse(this)[0]+30) + "px")
        .style("top", (d3.mouse(this)[1]+30) + "px")
    }
    var moveTooltip1 = function(d) {
      tooltip1
        .style("left", (d3.mouse(this)[0]+30) + "px")
        .style("top", (d3.mouse(this)[1]+30) + "px")
    }
    var hideTooltip1 = function(d) {
      tooltip1
        .transition()
        .duration(3000)
        .style("opacity", 0)
    }
  

  
    // ---------------------------//
    //       HIGHLIGHT GROUP      //
    // ---------------------------//
  
    // What to do when one group is hovered
    var highlight1 = function(d){
      // reduce opacity of all groups
      d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".counts").style("opacity", .05)
      d3.selectAll(".counts2").style("opacity", .05)
      d3.selectAll(".counts3").style("opacity", .05)
      d3.selectAll(".counts4").style("opacity", .05)

      // expect the one that is hovered
      d3.selectAll("."+d).style("opacity", 0.9).style("visibility","visible")  
      d3.selectAll(".counts"+d).style("opacity", 1)
      d3.selectAll(".counts2"+d).style("opacity", 1)
      d3.selectAll(".counts3"+d).style("opacity", 1)
      d3.selectAll(".counts4"+d).style("opacity", 1)

    }
  
    // And when it is not hovered anymore
    var noHighlight1 = function(d){
      d3.selectAll(".bubbles").style("opacity", 0.7)
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".bubbles2").style("opacity", 0)
      d3.selectAll(".bubblestp").style("opacity", 0) 
      d3.selectAll(".counts").style("opacity", 0.8)
      d3.selectAll(".counts2").style("opacity", 0.8)
      d3.selectAll(".counts3").style("opacity", 0.8)
      d3.selectAll(".counts4").style("opacity", 0.8)

    }

         // What to do when one group is hovered
         var highlight1p = function(d){
      // reduce opacity of all groups
      d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles2").style("opacity", 0) 
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".bubblestp").style("opacity", 0) 
      d3.selectAll(".countsp").style("opacity", .05) 
      d3.selectAll(".counts2p").style("opacity", .05)
      d3.selectAll(".counts3p").style("opacity", .05)
      d3.selectAll(".counts4p").style("opacity", .05)

      // expect the one that is hovered
      d3.selectAll("."+d).style("opacity", 0.9).style("visibility","visible")  
      d3.selectAll(".countsp"+d).style("opacity", 1) 
      d3.selectAll(".counts2p"+d).style("opacity", 1)
      d3.selectAll(".counts3p"+d).style("opacity", 1)
      d3.selectAll(".counts4p"+d).style("opacity", 1)
    }
  
    // And when it is not hovered anymore
    var noHighlight1p = function(d){
      d3.selectAll(".bubbles").style("opacity", 0.7)
      d3.selectAll(".bubblest").style("opacity", 0)
      d3.selectAll(".bubbles2").style("opacity", 0)
      d3.selectAll(".bubblestp").style("opacity", 0) 
      d3.selectAll(".countsp").style("opacity", 0.8) 
      d3.selectAll(".counts2p").style("opacity", 0.8)
      d3.selectAll(".counts3p").style("opacity", 0.8)
      d3.selectAll(".counts4p").style("opacity", 0.8)
    }
     // What to do when one group is hovered
     var highlight2 = function(d){
      // reduce opacity of all groups
      //d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles2").style("opacity", 0) 
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".bubblesq").style("opacity", 0) 
      d3.selectAll(".countsq").style("opacity", .05) 
      d3.selectAll(".counts2q").style("opacity", .05)
      d3.selectAll(".counts3q").style("opacity", .05)
      d3.selectAll(".counts4q").style("opacity", .05)


      // expect the one that is hovered
      d3.selectAll("."+d).style("opacity", 0.9).style("visibility","visible")  
      d3.selectAll(".countsq"+d).style("opacity", 1) 
      d3.selectAll(".counts2q"+d).style("opacity", 1)
      d3.selectAll(".counts3q"+d).style("opacity", 1)
      d3.selectAll(".counts4q"+d).style("opacity", 1)
    }
  
    // And when it is not hovered anymore
    var noHighlight2 = function(d){
      d3.selectAll(".bubbles").style("opacity", 0.7)
      d3.selectAll(".bubblest").style("opacity", 0)
      d3.selectAll(".bubbles2").style("opacity", 0)
      d3.selectAll(".bubblesq").style("opacity", 0) 
      d3.selectAll(".countsq").style("opacity", 0.8) 
      d3.selectAll(".counts2q").style("opacity", 0.8)
      d3.selectAll(".counts3q").style("opacity", 0.8)
      d3.selectAll(".counts4q").style("opacity", 0.8)

    }

     // What to do when one group is hovered
     var highlight3 = function(d){
      d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles2").style("opacity", 0) 
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".bubblesr").style("opacity", 0) 
      d3.selectAll(".countsr").style("opacity", .05) 
      d3.selectAll(".counts2r").style("opacity", .05)
      d3.selectAll(".counts3r").style("opacity", .05)
      d3.selectAll(".counts4r").style("opacity", .05)
      // expect the one that is hovered
      d3.selectAll("."+d).style("opacity", 0.9).style("visibility","visible")  
      d3.selectAll(".countsr"+d).style("opacity", 1) 
      d3.selectAll(".counts2r"+d).style("opacity", 1)
      d3.selectAll(".counts3r"+d).style("opacity", 1)
      d3.selectAll(".counts4r"+d).style("opacity", 1)

    }
  
    // And when it is not hovered anymore
    var noHighlight3 = function(d){
      d3.selectAll(".bubbles").style("opacity", 0.7)
      d3.selectAll(".bubblest").style("opacity", 0)
      d3.selectAll(".bubbles2").style("opacity", 0)
      d3.selectAll(".bubblesr").style("opacity", 0) 
      d3.selectAll(".countsr").style("opacity", 0.8) 
      d3.selectAll(".counts2r").style("opacity", 0.8)
      d3.selectAll(".counts3r").style("opacity", 0.8)
      d3.selectAll(".counts4r").style("opacity", 0.8)
    }
  
     // What to do when one group is hovered
     var highlight4 = function(d){
      // reduce opacity of all groups
      //d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles").style("opacity", .05)
      d3.selectAll(".bubbles2").style("opacity", 0) 
      d3.selectAll(".bubblest").style("opacity", 0) 
      d3.selectAll(".bubbless").style("opacity", 0) 
      d3.selectAll(".countss").style("opacity", .05) 
      d3.selectAll(".counts2s").style("opacity", .05)
      d3.selectAll(".counts3s").style("opacity", .05)
      d3.selectAll(".counts4s").style("opacity", .05)
      // expect the one that is hovered
      d3.selectAll("."+d).style("opacity", 0.9).style("visibility","visible")  
      d3.selectAll(".countss"+d).style("opacity", 1) 
      d3.selectAll(".counts2s"+d).style("opacity", 1)
      d3.selectAll(".counts3s"+d).style("opacity", 1)
      d3.selectAll(".counts4s"+d).style("opacity", 1)
    }
  
    // And when it is not hovered anymore
    var noHighlight4 = function(d){
      d3.selectAll(".bubbles").style("opacity", 0.7)
      d3.selectAll(".bubblest").style("opacity", 0)
      d3.selectAll(".bubbles2").style("opacity", 0)
      d3.selectAll(".bubbless").style("opacity", 0) 
      d3.selectAll(".countss").style("opacity", 0.8) 
      d3.selectAll(".counts2s").style("opacity", 0.8)
      d3.selectAll(".counts3s").style("opacity", 0.8)
      d3.selectAll(".counts4s").style("opacity", 0.8)
    }
  
   
   
    var formatNumber1 = d3.format(",.0f"),    // zero decimal places
    format1 = function(d) { return formatNumber1(d); }; 

    var formatNumber2 = d3.format(",.2f"),    // zero decimal places
    format2 = function(d) { return formatNumber2(d); }; 
    // ---------------------------//
    //       CIRCLES              //
    // ---------------------------//
              // Add dots
          // Add dots
          // Add dots
          svg1.append('g')
      .selectAll("dot")
      .data(data1)
      .enter()
      .append("circle")
        .attr("class", function(d) { return "bubbles2 " + d.ParentalEducation })
     //   .attr("class", function(d) { return "bubbles2 " + d.product_group })
        .attr("cx", function (d) { return x1(d.Math ); } )
        .attr("cy", function (d) { return y1(d.RW); } )
          .attr("r", function (d) { return z1(d.Score); } )
         .style("fill", function (d) { return myColor2(d.ParentalEducation); }  )
         .style("stroke", "#f0f0f0" )
         .style("stroke-width", 1 )
        .style("opacity",0.7)
        .style("stroke-opacity",0.7)
        .style("visibility","hidden")

            // Add dots
    svg1.append('g')
      .selectAll("dot")
      .data(data1)
      .enter()
      .append("circle")
        .attr("class", function(d) { return "bubblesq " + d.Gender })
        .attr("cx", function (d) { return x1(d.Math ); } )
        .attr("cy", function (d) { return y1(d.RW); } )
          .attr("r", function (d) { return z1(d.Score); } )
             .style("fill", function (d) { return myColorq(d.Gender); } )
         .style("stroke", "#f0f0f0" )
         .style("stroke-width", 1 )
        .style("opacity",0.7)
        .style("stroke-opacity",0.7)
        .style("visibility","hidden")
      // -3- Trigger the functions for hover
            // Add dots
            svg1.append('g')
      .selectAll("dot")
      .data(data1)
      .enter()
      .append("circle")
        .attr("class", function(d) { return "bubblesr " + d.Lunch })
        .attr("cx", function (d) { return x1(d.Math ); } )
        .attr("cy", function (d) { return y1(d.RW); } )
          .attr("r", function (d) { return z1(d.Score); } )
         .style("fill", function (d) { return myColorr(d.Lunch); } )
         .style("stroke", "#f0f0f0" )
         .style("stroke-width", 1 )
        .style("opacity",0.7)
        .style("stroke-opacity",0.7)
        .style("visibility","hidden")

        svg1.append('g')
      .selectAll("dot")
      .data(data1)
      .enter()
      .append("circle")
        .attr("class", function(d) { return "bubbless " + d.PreparationCourse })
        .attr("cx", function (d) { return x1(d.Math ); } )
        .attr("cy", function (d) { return y1(d.RW); } )
          .attr("r", function (d) { return z1(d.Score); } )
          .style("fill", function (d) { return myColors(d.PreparationCourse); } )
         .style("stroke", "#f0f0f0")
         .style("stroke-width", 1 )
        .style("opacity",0.7)
        .style("stroke-opacity",0.7)
        .style("visibility","hidden")


    // Add dots
    svg1.append('g')
      .selectAll("dot")
      .data(data1)
      .enter()
      .append("circle")
        .attr("class", function(d) { return "bubbles " + d.Race })
     //   .attr("class", function(d) { return "bubbles2 " + d.product_group })
        .attr("cx", function (d) { return x1(d.Math ); } )
        .attr("cy", function (d) { return y1(d.RW); } )
          .attr("r", function (d) { return z1(d.Score); } )
         .style("fill", function (d) { return myColor3(d.Race); } )
         .style("stroke", "#f0f0f0" )
         .style("stroke-width",1 )
        .style("opacity",0.7)
        .style("stroke-opacity",0.7)
      // -3- Trigger the functions for hover
      .on("mouseover", showTooltip1 )
      .on("mousemove", moveTooltip1 )
      .on("mouseleave", hideTooltip1 )


      // ---------------------------//
      //       LEGEND              //
      // ---------------------------//
  
      // Add legend: circles
  

  
      var yCircle1 = 20
      var xCircle1 = width1 -320
      // Add one dot in the legend for each name.
      var size1 = 20
      var allgroups1 = ["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"]

var avgcol = d3.scaleLinear()
      .domain([60, 75])
       .range([ "#ee3333","#33ee33" ]);  //660
 
   
var allgroups2 = ["Avg.","Math","Rd.","Wr."]   

var allgroups3 = ["GroupA","GroupB","GroupC","GroupD","GroupE"]

var allgroups4 = ["Gender","Parental Education", "Race / Ethnicity","Free Lunch", "Preparatoty Course"]

var allgroups5 = ["Female","Male"]

var allgroups6 = ["Free_Reduced","Standard"] 

var allgroups7 = ["Completed","None"] 

var label5 = d3.scaleOrdinal()
          .domain(["Female","Male"
])
         .range(["Female","Male"])




var icons = d3.scaleOrdinal()
    .domain(["Avg.","Math","Rd.","Wr."])
    .range(['\uf19d',
    '\uf1ec',                 
    '\uf02d',
    '\uf044',

]);

var label1 = d3.scaleOrdinal()
          .domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"
])
         .range(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])

var label3 = d3.scaleOrdinal()
          .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"
])
         .range(["GroupA","GroupB","GroupC","GroupD","GroupE"])

 var count1 = d3.scaleOrdinal()
          .domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])
         .range([65.1,
63.1,
68.5,
69.6,
71.9,
73.6])



var count2 = d3.scaleOrdinal()
          .domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])
         .range([63.5,
62.1,
67.1,
67.9,
69.4,
69.7

])

var count3 = d3.scaleOrdinal()
          .domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])
         .range([66.9,
64.7,
69.5,
70.9,
73.0,
75.4

])


var count4 = d3.scaleOrdinal()
.domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])
         .range([64.9,
62.4,
68.8,
69.9,
73.4,
75.7
])

var count1q = d3.scaleOrdinal()
          .domain(["Female","Male"])
         .range([69.6,
65.8
])

var count2q = d3.scaleOrdinal()
          .domain(["Female","Male"])
         .range([63.6,
68.7
])
var count3q = d3.scaleOrdinal()
.domain(["Female","Male"])
         .range([72.6,
65.5
])

var count4q = d3.scaleOrdinal()
.domain(["Female","Male"])
         .range([72.5,
63.3

])

var count1r = d3.scaleOrdinal()
          .domain(["Free_Reduced","Standard"])
         .range([62.2,
70.8

])

var count2r = d3.scaleOrdinal()
.domain(["Free_Reduced","Standard"])
         .range([58.9,
70.0
])
var count3r = d3.scaleOrdinal()
.domain(["Free_Reduced","Standard"])
         .range([64.7,
71.7
])

var count4r = d3.scaleOrdinal()
.domain(["Free_Reduced","Standard"])
         .range([63.0,
70.8
])

var count1s = d3.scaleOrdinal()
          .domain(["Completed","None"])
         .range([72.7,
65.0
])

var count2s = d3.scaleOrdinal()
.domain(["Completed","None"])
         .range([69.7,
64.1
])
var count3s = d3.scaleOrdinal()
.domain(["Completed","None"])
         .range([73.9,
66.5
])

var count4s = d3.scaleOrdinal()
.domain(["Completed","None"])
         .range([74.4,
64.5
])

var count1p = d3.scaleOrdinal()
          .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"])
         .range([63.0,
65.5,
67.1,
69.2,
72.8
])


var count2p = d3.scaleOrdinal()
          .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"])
         .range([61.6,
63.5,
64.5,
67.4,
73.8
])

var count3p = d3.scaleOrdinal()
          .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"])
         .range([64.7,
67.4,
69.1,
70.0,
73.0
])

var count4p = d3.scaleOrdinal()
              .domain(["GroupA","GroupB","GroupC","GroupD","GroupE"])
             .range([62.7,
65.6,
67.8,
70.1,
71.4
])

var text1 = d3.scaleOrdinal()
          .domain(["SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"])
         .range([
         "SomeHighSchool",
"HighSchool",
"SomeCollege",
"Associate",
"Bachelor", 
"Master"
        ])

        svg1.append('rect')
      .attr("x", xCircle1+30)
      .attr("y",60)
      .attr("width",310)
      .attr("height",480)
      .attr("fill","#cccccc")
       .attr("fill","#090909")
      .attr('stroke',"#b8b8b8")
      .attr("stroke-width",1)
      .attr("stroke-opacity",0.4)

 svg1.selectAll("myFeatureslabels1")
        .data(allgroups2)
        .enter()
        .append("text")
        .attr("x", function(d,i) { return xCircle1 + 180 +  i * (size1 + 20)})
        .attr("y",  function(d,i) { return 100-16}) // 100 is where the first dot appears. 25 is the distance between dots
           .style("fill", "#dcdcaa")           
          .text(function(d){ return(icons(d))})
         .attr('dominant-baseline', 'central')
         .attr('font-family', 'FontAwesome')
          .attr("font-weight",600)
          .attr("font-size","17px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")   

  svg1.selectAll("myFeatureslabels4")
        .data(allgroups4)
        .enter()
        .append("text")
        .attr("x", function(d,i) { return xCircle1 + 40 } )
        .attr("y",  function(d) { return title(d)}) // 100 is where the first dot appears. 25 is the distance between dots      
          .style("fill", "#dcdcaa") //"#dddddd")    
          .text(function(d){ return (d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")   
                // Add labels beside legend dots
  svg1.selectAll("myrect")
        .data(allgroups1)
        .enter()
        .append("circle")
        .attr("cx", function(d,i) { return xCircle1  +  1 * (size1 + 30)})
          .attr("cy",  function(d,i) { return 178 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("stroke", "#f0f0f0")
          .style("stroke-width", 1)
          .style("fill", function(d){ return myColor2(d)})
          .attr("r", 5)
          .on("mouseover", highlight1)
          .on("mouseleave", noHighlight1)
  
      // Add labels beside legend dots
      svg1.selectAll("mylabels")
        .data(allgroups1)
        .enter()
        .append("text")
        .attr("class", function(d) { return "mylabels " + d })
          .attr("x", function(d,i) { return xCircle1 + 20 +  1 * (size1 + 30)})
          .attr("y",  function(d,i) { return 180 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill",  "#999999")
          .text(function(d){ return label1(d)})
          .attr("font-weight",700)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
          .on("mouseover", highlight1)
          .on("mouseleave", noHighlight1)

      svg1.selectAll("myrect2")
        .data(allgroups3)
        .enter()
        .append("circle")
        .attr("cx", function(d,i) { return xCircle1  +  1 * (size1 + 30)})
          .attr("cy",  function(d,i) { return 306 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill", function(d){ return myColor3(d)})
          .style("stroke", "#f0f0f0")
          .style("stroke-width",1)
          .attr("r", 5)
           .on("mouseover", highlight1p)
           .on("mouseleave", noHighlight1p)
                // Add labels beside legend dots
      svg1.selectAll("mylabels2")
        .data(allgroups3)
        .enter()
        .append("text")
          .attr("class", function(d) { return "mylabels2 " + d })
          .attr("x", function(d,i) { return xCircle1 + 20 +  1 * (size1 + 30)})
          .attr("y",  function(d,i) { return 308 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill",  "#999999")
          .text(function(d){ return label3(d)})
          .attr("font-weight",700)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
           .on("mouseover", highlight1p)
           .on("mouseleave", noHighlight1p)


      svg1.selectAll("myCounts")
        .data(allgroups1)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts " + d })
         .attr("x", function(d,i) { return xCircle1 + 100 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 180 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count1(d)})
          .attr("font-weight",600)
          .attr("font-size","12px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

                          // Add labels beside legend dots
      svg1.selectAll("myCounts2")
        .data(allgroups1)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts2 " + d })
         .attr("x", function(d,i) { return xCircle1 + 140 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 180 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count2(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts3")
        .data(allgroups1)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts3 " + d })
         .attr("x", function(d,i) { return xCircle1 + 180 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 180 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count3(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts4")
        .data(allgroups1)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts4 " + d })
         .attr("x", function(d,i) { return xCircle1 + 220 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 180 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", .8)
          .text(function(d){ return count4(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")


      svg1.selectAll("myCountsp")
        .data(allgroups3)
        .enter()
        .append("text")
         .attr("class", function(d) { return "countsp " + d })
         .attr("x", function(d,i) { return xCircle1 + 100 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 308 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count1p(d)})
          .attr("font-weight",600)
          .attr("font-size","12px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

                          // Add labels beside legend dots
      svg1.selectAll("myCounts2p")
        .data(allgroups3)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts2p " + d })
         .attr("x", function(d,i) { return xCircle1 + 140 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 308 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count2p(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

    svg1.selectAll("myCounts3p")
        .data(allgroups3)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts3p " + d })
         .attr("x", function(d,i) { return xCircle1 + 180 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 308 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count3p(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts4p")
        .data(allgroups3)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts4p " + d })
         .attr("x", function(d,i) { return xCircle1 + 220 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 308 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count2(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count4p(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myrectq")
        .data(allgroups5)
        .enter()
        .append("circle")
        .attr("cx", function(d,i) { return xCircle1  +  1 * (size1 + 30)})
          .attr("cy",  function(d,i) { return 112 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("stroke", "#f0f0f0")
          .style("stroke-width", 1)
          .style("fill", function(d){ return myColorq(d)})
          .attr("r", 5)
          .on("mouseover", highlight2)
          .on("mouseleave", noHighlight2)
  
      // Add labels beside legend dots
      svg1.selectAll("mylabelsq")
        .data(allgroups5)
        .enter()
        .append("text")
        .attr("class", function(d) { return "mylabelsq " + d })
          .attr("x", function(d,i) { return xCircle1 + 20 +  1 * (size1 + 30)})
          .attr("y",  function(d,i) { return 114 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill",  "#999999")
          .text(function(d){ return d})
          .attr("font-weight",700)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
          .on("mouseover", highlight2)
          .on("mouseleave", noHighlight2)



      svg1.selectAll("myCountsq")
        .data(allgroups5)
        .enter()
        .append("text")
         .attr("class", function(d) { return "countsq " + d })
         .attr("x", function(d,i) { return xCircle1 + 100 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 114 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count1q(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count1q(d)})
          .attr("font-weight",600)
          .attr("font-size","12px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

                          // Add labels beside legend dots
      svg1.selectAll("myCounts2q")
        .data(allgroups5)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts2q " + d })
         .attr("x", function(d,i) { return xCircle1 + 140 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 114 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill", function(d){ return avgcol(count2q(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count2q(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts3q")
        .data(allgroups5)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts3q " + d })
         .attr("x", function(d,i) { return xCircle1 + 180 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 114 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count3q(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count3q(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts4q")
        .data(allgroups5)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts4q " + d })
         .attr("x", function(d,i) { return xCircle1 + 220 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 114 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count4q(d))}  )
          .style("opacity", .8)
          .text(function(d){ return count4q(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
  
          svg1.selectAll("myrectr")
        .data(allgroups6)
        .enter()
        .append("circle")
        .attr("cx", function(d,i) { return xCircle1  +  1 * (size1 + 30)})
          .attr("cy",  function(d,i) { return 422 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("stroke", "#f0f0f0")
          .style("stroke-width", 1)
          .style("fill", function(d){ return myColorr(d)})
          .attr("r", 5)
          .on("mouseover", highlight3)
          .on("mouseleave", noHighlight3)
  
      // Add labels beside legend dots
      svg1.selectAll("mylabelsr")
        .data(allgroups6)
        .enter()
        .append("text")
          .attr("class", function(d) { return "mylabelsr " + d })
          .attr("x", function(d,i) { return xCircle1 + 20 +  1 * (size1 + 30)})
          .attr("y",  function(d,i) { return 424 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill",  "#999999")
          .text(function(d){ return d})
          .attr("font-weight",700)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
          .on("mouseover", highlight3)
          .on("mouseleave", noHighlight3)



      svg1.selectAll("myCountsr")
        .data(allgroups6)
        .enter()
        .append("text")
         .attr("class", function(d) { return "countsr " + d })
         .attr("x", function(d,i) { return xCircle1 + 100 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 424 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count1r(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count1r(d)})
          .attr("font-weight",600)
          .attr("font-size","12px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

                          // Add labels beside legend dots
      svg1.selectAll("myCounts2r")
        .data(allgroups6)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts2r " + d })
         .attr("x", function(d,i) { return xCircle1 + 140 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 424 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill", function(d){ return avgcol(count2r(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count2r(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts3r")
        .data(allgroups6)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts3r " + d })
         .attr("x", function(d,i) { return xCircle1 + 180 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 424 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count3r(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count3r(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts4r")
        .data(allgroups6)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts4r " + d })
         .attr("x", function(d,i) { return xCircle1 + 220 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 424 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count4r(d))}  )
          .style("opacity", .8)
          .text(function(d){ return count4r(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myrects")
        .data(allgroups7)
        .enter()
        .append("circle")
        .attr("cx", function(d,i) { return xCircle1  +  1 * (size1 + 30)})
          .attr("cy",  function(d,i) { return 488 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("stroke", "#f0f0f0")
          .style("stroke-width", 1)
          .style("fill", function(d){ return myColors(d)})
          .attr("r", 5)
          .on("mouseover", highlight4)
          .on("mouseleave", noHighlight4)
  
      // Add labels beside legend dots
      svg1.selectAll("mylabelss")
        .data(allgroups7)
        .enter()
        .append("text")
          .attr("class", function(d) { return "mylabelss " + d })
          .attr("x", function(d,i) { return xCircle1 + 20 +  1 * (size1 + 30)})
          .attr("y",  function(d,i) { return 490 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill",  "#999999")
          .text(function(d){ return d})
          .attr("font-weight",700)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")
          .on("mouseover", highlight4)
          .on("mouseleave", noHighlight4)



      svg1.selectAll("myCountss")
        .data(allgroups7)
        .enter()
        .append("text")
         .attr("class", function(d) { return "countss " + d })
         .attr("x", function(d,i) { return xCircle1 + 100 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 490 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count1s(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count1s(d)})
          .attr("font-weight",600)
          .attr("font-size","12px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

                          // Add labels beside legend dots
      svg1.selectAll("myCounts2s")
        .data(allgroups7)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts2s " + d })
         .attr("x", function(d,i) { return xCircle1 + 140 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 490 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
          .style("fill", function(d){ return avgcol(count2s(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count2s(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts3s")
        .data(allgroups7)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts3s " + d })
         .attr("x", function(d,i) { return xCircle1 + 180 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 490 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count3s(d))}  )
          .style("opacity", 0.8)
          .text(function(d){ return count3s(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

          svg1.selectAll("myCounts4s")
        .data(allgroups7)
        .enter()
        .append("text")
         .attr("class", function(d) { return "counts4s " + d })
         .attr("x", function(d,i) { return xCircle1 + 220 +  1 * (size1 + 64)})
         .attr("y",  function(d,i) { return 490 + i * (size1*.56 + 5)}) // 100 is where the first dot appears. 25 is the distance between dots
         .style("fill", function(d){ return avgcol(count4s(d))}  )
          .style("opacity", .8)
          .text(function(d){ return count4s(d)})
          .attr("font-weight",600)
          .attr("font-size","11px")
          .attr("text-anchor", "left")
          .style("alignment-baseline", "middle")

svg1.append("text").attr("x",325).attr("y",-20).text("Student Performance in Exams").
     attr("fill","#f7d057").attr("font-size","20px").attr("font-weight",750); //#f7d057
  svg1.append("text").attr("x",220).attr("y",0).text("Hover by Gender / Parental Education / Race / Free Lunch / Preparatory Course" ).
       attr("fill","#dcdcaa").attr("font-size","14px"); //attr("font-style","italic");
      
  });        
"""
```
