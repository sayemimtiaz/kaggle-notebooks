**MATPLOTLIB**

Matplotlib graphics drawing package Python is one of the most important tools for preparing the program. Matplotlib, which contains a very strong package, can be tailored and output at a suitable height. Both two-dimensional and three-dimensional graphics can be produced.

Matplotlib graphics can be edited in a very subtle way, but here we will focus on their basic structure.

* Line plot represents the relationship between 2 variables.
* Scatter plot displays the value of 2 sets of data on 2 dimensions.
* Histogram is better when we need to see distribution of numerical data.
* Customization: Colors,labels,thickness of line, title, opacity, grid, figsize, ticks of axis and linestyle 











```python
#Line Plot

data.Age.plot(kind = 'line', color ='g',label='Age',linewidth=2,alpha=0.5,grid=True,linestyle=':')
data.Weight.plot(color='b',label='Weight',linewidth=1,alpha=0.5,grid=True,linestyle='-.')
plt.legend(loc='upper right')     # legend = puts label into plot

plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis') 
plt.title('Line Plot')            # title = title of plot
plt.show()


```

```python
# Scatter Plot
# x = age, y = weight
data.plot(kind='scatter', x='Age', y='Weight',alpha=0.5,color='yellow')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age-Weight Scatter Plot')

```

```python
# Histogram
# bins = numer of bar in figure
data.Weight.plot(kind ='hist',bins = 25,figsize =(12,12))
plt.show()
```

* built in scope: names in predefined built in scope module such as print, len

```python
# how to print
x = 2
def f():
    x = 3
    return x
print(x)        # x = 2 global scope
print(f())      # x = 3 local scope
```

```python
#default arguments
def f(x,y =1, z = 2):
    k =x + y + z
    return k
print(f(3))
# or change arguments
print(f(12,4,7))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,5,2,6,7)

# flexible arguments **kwargs --> dictionary
def f(**kwargs):
    """ print key and value dictonary """
    for key,value in kwargs.items():    #for loop
        print(key, " ", value)
f(country ='turkey', capital ='ankara', region ='ic anadolu' ,population =32135554)        
        
```

**ITERATORS**


* An iterator is an object that contains a countable number of values.

* An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.

* Technically, in Python, an iterator is an object which implements the iterator protocol, which consist of the methods __iter__() and __next__().

```python
# example of iterators
word = "software"
it = iter(word)
print(next(it))    # print next iteration
print(*it)         # print remaining iteration
```

```python
# zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
#unzip example (more)
un_zip = zip (*z_list)
un_list1,un_list2 =list(un_zip)   # unzip return tuble
print(un_list1)
print(un_list2)
print(type(un_list1))
```

**LIST COMPREHENSION**

* List comprehensions provide a concise way to create lists. 
Ex: num1 = [1,2,3] and we want to make it num2 = [2,3,4]. This can be done with for loop. However it is unnecessarily long. We can make it one line code that is list comprehension.


```python
# Example of list comprehension
num_list1 = [1,2,3]
num_list2 = [i + 1 for i in num_list1 ]
print(num_list2)
```



[i + 1 for i in num1 ]: list of comprehension
i +1: list comprehension syntax
i: iterator
num1: iterable object


```python
# Conditionals on iterable
num_list1 = [10,20,30]
num_list2 = [i**2 if i==20 else i-10 if i< 15 else i+5 for i in num_list1]
print(num_list2)
```
