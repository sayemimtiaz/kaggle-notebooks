```python
#Features and Labes
X=social_network_data.iloc[:,[2,3]]#Matrix, our features
y=social_network_data.iloc[:,4]#Vector, our target
```

# Data Preprocessing 

```python
#Scaling Our Features
#First let's split our data into test and train set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Now Let's Turn Features_Scaling
from sklearn.preprocessing import StandardScaler
#Instanciate our scaler
scaler=StandardScaler()
#Let's fit and transform our data on training set
X_train_std=scaler.fit_transform(X_train)
#Let's transsform our data on test set 
X_test_std=scaler.transform(X_test)
```

# Let's Model

```python
#We are going to use support vector machine:SVM
#We have classification situation so we are going tu use Support Vector Classifier:SVC
from sklearn.svm import SVC
#Instanciate Our Model
svc=SVC(random_state=0)#Here we are going to use linear svc but for non linear we specifiy the parameter kernel='rbf'
#Let's fit our model 
svc.fit(X_train_std,y_train)
#Let's Predict Now
y_pred=svc.predict(X_test_std)
```

# Evaluate Our Model

```python
#Evaluate our model by using confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
