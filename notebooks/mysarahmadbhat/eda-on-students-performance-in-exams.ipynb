<p style="font-family: Arials; font-size: 16px; font-style: normal; font-weight: normal; letter-spacing: 3px; color: #FF8C00; line-height:1.0">2.6 Bi-Variate Analysis:</p>

<p>'In this section we will do Bi-Variate Analysis' </p>

```python
plt.figure(figsize=(10,6))
sns.violinplot(x=df["gender"],y=df["math score"])
plt.title("DISTRIBUTION OF MATH SCORE WITH GENDER",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df["gender"],y=df["reading score"])
plt.title("DISTRIBUTION OF READING SCORE WITH GENDER",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x=df["gender"],y=df["writing score"])
plt.title("DISTRIBUTION OF WRITING SCORE WITH GENDER",fontsize=20)
plt.show()
```

```python
pd.crosstab(df["gender"],df["race/ethnicity"]).plot(kind="bar",figsize=(10,6))
plt.title("GENDER AND THEIR RACE",fontsize=20)
plt.show()
```

```python
sns.displot(data=df,x="percentage",hue="gender")
plt.title("PERCENTAGE DISTRIBUTION W.R.T GENDER",fontsize=20)
plt.show()
```

```python
fig = px.box(df, x='race/ethnicity', y='math score',labels=dict(
            Total="Passenger Class", Day_Avg="Age"))
fig.update_layout(title_text='Distribution of Math Score With Race/Ethnicity',title_x=0.5)
fig.show()
```

```python
fig = px.box(df, x='race/ethnicity', y='reading score',labels=dict(
            Total="Passenger Class", Day_Avg="Age"))
fig.update_layout(title_text='Distribution of READING Score With Race/Ethnicity',title_x=0.5)
fig.show()
```

```python
fig = px.box(df, x='race/ethnicity', y='writing score',labels=dict(
            Total="Passenger Class", Day_Avg="Age"))
fig.update_layout(title_text='Distribution of WRITING Score With Race/Ethnicity',title_x=0.5)
fig.show()
```

```python
sns.displot(data=df,x="percentage",hue="race/ethnicity")
plt.title("PERCENTAGE DISTRIBUTION W.R.T RACE/ETHNICITY",fontsize=20)
plt.show()
```

```python
fig = px.box(df, x='parental level of education', y='writing score',labels=dict(
            Total="Passenger Class", Day_Avg="Age"))
fig.update_layout(title_text='Distribution of WRITING Score With PARENTAL LEVEL OF EDUCATION',title_x=0.5)
fig.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x=df["parental level of education"],y=df["math score"])
plt.title("DISTRIBUTION OF MATH SCORE WITH PARENTAL LEVEL OF EDUCATION",fontsize=20)
plt.show()
```

```python
sns.displot(data=df,x="percentage",hue="parental level of education")
plt.title("PERCENTAGE DISTRIBUTION W.R.T PARENTAL LEVEL OF EDUCATION",fontsize=20)
plt.show()
```

```python
l=df.groupby(df["lunch"])["math score"].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=l["lunch"],y=l["math score"])
plt.title("AVERAGE MATH SCORE BASED ON LUNCH",fontsize=20)
plt.show()
```

```python
l=df.groupby(df["lunch"])["reading score"].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=l["lunch"],y=l["reading score"])
plt.title("AVERAGE READING SCORE BASED ON LUNCH",fontsize=20)
plt.show()
```

```python
l=df.groupby(df["lunch"])["writing score"].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=l["lunch"],y=l["writing score"])
plt.title("AVERAGE WRITING SCORE BASED ON LUNCH",fontsize=20)
plt.show()
```

```python
sns.displot(data=df,x="percentage",hue="lunch")
plt.title("PERCENTAGE DISTRIBUTION W.R.T LUNCH",fontsize=20)
plt.show()
```

```python
t=df.groupby(df["test preparation course"])['math score'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=t["test preparation course"],y=t["math score"])
plt.title("AVERAGE MATH SCORE WITH TEST PREPARATION COURSE",fontsize=20)
plt.show()
```

```python
t=df.groupby(df["test preparation course"])['reading score'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=t["test preparation course"],y=t["reading score"])
plt.title("AVERAGE READING SCORE WITH TEST PREPARATION COURSE",fontsize=20)
plt.show()
```

```python
t=df.groupby(df["test preparation course"])['writing score'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=t["test preparation course"],y=t["writing score"])
plt.title("AVERAGE WRITING SCORE WITH TEST PREPARATION COURSE",fontsize=20)
plt.show()
```

```python
sns.set_palette("bright")
sns.displot(data=df,x="percentage",hue="test preparation course")
plt.title("PERECENTAGE DISTRIBUTION W.R.T TEST PREPRATION SCORE",fontsize=20)
plt.show()
```
