# Interactive Text & Knowledge Base Search Tool: http://kdcovid.nl/

# Key Features

## -- Retrieval of Documents, Displaying Sentences Relevant to Query from Full-Text
## -- Disease & Gene Highlighting and Linking to UniProt & MedGen
## -- Knowledge-graphs showing disease-gene interactions from DisGeNET for each paper


## About
This tool retrieves papers by measuring similarity between queries and sentences in the full text of papers in
CORD19 corpus using a similarity metric derived from <a href="https://github.com/ncbi-nlp/BioSentVec"> BioSentVec </a>.
We also use the tool <a href="http://ibi.imim.es/befree/"> BeFree </a> for entity linking (Bravo et al, 2014 & 2015).
Genes are linked to <a href="https://www.uniprot.org/">UniProt</a> and diseases to <a href="https://www.ncbi.nlm.nih.gov/medgen">medgen</a>.
We display relationships between genes and diseases from <a href="https://www.disgenet.org/">DisGeNET</a> (v6) (Piñero et al, 2019).
Highlighting formatting is based on <a href="https://explosion.ai/demos/displacy">displaCy</a>.
This tool was developed by Manzil Zaheer (Google), Nicholas Monath (UMass Amherst), 
Shehzaad Dhuliawala (Microsoft), Taamannae Taabassum (University of Toronto),
Rajarshi Das (UMass Amherst), Bhuwan Dhingra (CMU), and Andrew McCallum (UMass Amherst).


## Retrieving relevant evidence for a query

Given a query, our tool retrieves papers and highlights relevant sentences within it. We do this by assigning a score to all sentences in the corpus. The score reflects the similarity of the sentence with respect to the query. We also assign score to each individual research paper as the maximum relevance score of a sentence in the paper. Our tool also displays context surrounding the high scoring sentence as it might contain relevant information as well. Next, we briefly describe how we score each sentence with respect to a given query.

## Scoring sentences using BioSentVec

The score of each sentence is computed as the inner product between the sentence and the query vector. We create vector representation of each sentence using a pre-trained [BioSentVec model](https://github.com/ncbi-nlp/BioSentVec). The BioSentVec model is a [sent2vec model](https://github.com/epfml/sent2vec) which has been trained on the pubmed corpus and outputs a sentence embedding by averaging the embeddings of the words in them. Although simple, sent2vec has shown to be effective for many downstream applications. We compute all sentences vectors and store them offline. Given a query, we first compute its distributed representation and then find the K-nearest sentence vectors (i.e. the top-K scoring sentences) from the corpus. As mentioned before, we score a document as the maximum score of any sentence within it. The figure below summarizes our system.

![KNN](http://kdcovid.nl/fig2.jpg)

## Generating Sentence Embeddings

The following code can be used to generate sentence embeddings:


````python
import sent2vec
from kdcovid.encode_sentences import encode
import os
import numpy as np
import torch

model_path = '../input/scriptmodel/script_model.bin'
model = sent2vec.Sent2vecModel()

if os.path.exists(model_path):
    model.load_model(model_path)
else:
    print("Not found")

sentence_vectors, sentence_metadata = encode(documents, model=model)
sentence_vectors = torch.from_numpy(np.vstack(sentence_vectors).astype(np.float32))
````

## Extracting Knowledge Graphs (KGs) from the evidence
The main aim for building this tool is to help bio-medical researchers find relevant information. However, it is unlikely that all the answers that they are looking for would be answered by our initial set of retrieved papers. Often times, the initial retrieved evidence provides a good starting point for retrieving more evidence (e.g. think about those instances where you click an anchor link in a wikipedia page to go into another wiki page). We hypothesize that the bio-medical entities (viruses, genes, drugs etc) which are mentioned in the retrieved research papers would be useful for the researchers. Therefore, we identify the entities present in the papers, and link them to knowledge bases from which the researchers can browse further information. We also display relationship edges between the entities (e.g. associations between genes and diseases). Below we briefly summarize each component

## Identifying entities in text
The first step is to identify the entities that are present in the retrieved papers. We use a simple approach of fuzzy string matching to match to a dictionary containing various entity names and their aliases. The dictionary is derived from the BeFree system. After finding the entities, we link the genes to the UniProt and diseases to the MedGen ontologies. We are also adding entity linking drugs to Drugbank soon.

## Finding relations between entities
There are several knowledge bases (KBs) that capture relation between entities. These KBs are either manually curated from the findings of several research papers over many years or have been automatically inferred from text using various relation extraction approaches. Finding and displaying these relations in our tool will allow researchers to get insights about findings that are not restricted to the retrieved paper. For our tool, we use the gene-disease association present in DisGenet. We are working on adding relations between genes, diseases and drugs as well. 

## Entity Linking Code

We use a modified version of the [BeFree](http://ibi.imim.es/befree/) package to do entity linking. For more details checkout our [code](https://github.com/nmonath/kdcovid)'s README. 

````python
from befree.src.ner import BeFree_NER_cord19
from kdcovid.parse_befree_output import parse_befree_output

identified_genes, identified_diseases = BeFree_NER_cord19.parse_cord19(documents)
combined_mentions = parse_befree_output(documents, identified_diseases, identified_genes)
````

# Pros / Cons

## Pros:

1. We present a simple and easy to use knowledge discovery tool to help bio-medical researchers that uses distributed representation of query and sentences to find relevant snippets from full text of research papers.
2. We also extract several biological entities such as genes and diseases, link them to existing large high quality (human curated) knowledge bases and show relevant gene-disease associations. Traversing the links (from the nodes) would help bio-medical researchers to browse and gather more information that is not just restricted to the findings of the retrieved papers
3. We also present an easy to use web-interface and knowledge base search tool for researchers

## Cons:

1. Our tool retrieves relevant papers for a specific query and highlights relevant spans of text. However, it does not summarize the results from all the retrieved papers nor does it provide aggregate level trends or directions which would be beneficial for a researcher using the tool.
2. Use of contextual embeddings: Contextual embeddings such as BERT, RoBERTa have proven to be more effective than models that use fixed vector representation of words. We are currently working on getting sentence representation using aforementioned models and hope to update our web-interface soon
3. Tagging chemical nodes: Even though we tagged genes and diseases, we haven’t yet tagged entities which represent chemical compounds. We plan to fix this soon as well.
4. Reasoning using evidence from text and knowledge graphs: This is a part of the larger research we plan to pursue. Using the retrieved evidence (both text and KG), one could “reason” and predict new associations between drugs and diseases or find out if a drug could be re-purposed to treat a new disease (drug-repurposing). We believe our tool gives a good starting point for doing such research.
5. Its a tool after all: We built this tool to aid bio-medical researchers to browse and find relevant information. We don’t intend to predict cure or treatment time. We would like our tool to help researchers come to an answer for several pressing questions that the world is facing in this crisis.
