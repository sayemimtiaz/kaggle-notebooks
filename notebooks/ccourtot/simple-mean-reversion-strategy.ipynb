```python
data = pd.read_csv('../input/bitstampUSD_1-min_data_2012-01-01_to_2017-05-31.csv')
```

```python
data = data.dropna()
```

```python
price = data['Volume_(Currency)']/data['Volume_(BTC)']
```

```python
rent = np.log(price/price.shift(1)).dropna()
```

```python
pnl = 0
pos = 0
for r in rent:
    pnl = pnl + pos*np.exp(r)
    pos = -r
    pnl = pnl + r
     
print(pnl)
```
