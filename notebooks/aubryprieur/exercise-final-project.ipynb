## Step 4: Visualize the data

Use the next code cell to create a figure that tells a story behind your dataset.  You can use any chart type (_line chart, bar chart, heatmap, etc_) of your choosing!

````python
# Create a plot

# Set the width andheight of the figure
plt.figure(figsize=(18,10))
plt.ylim(0, 1.75)
plt.xlim(0, 165)

# Add title
plt.title('World Hapiness Report 2019 : Indices by Overall rank')

# scatter plots
sns.regplot(x=my_data['Overall rank'],y=my_data['Healthy life expectancy'], scatter_kws={"color":'blue'}, line_kws={"color": "blue"}, label='Healthy life expectancy')
sns.regplot(x=my_data['Overall rank'],y=my_data['Social support'], label='Social support' )
sns.regplot(x=my_data['Overall rank'],y=my_data['Freedom to make life choices'], label='Freedom to make life choices' )
sns.regplot(x=my_data['Overall rank'],y=my_data['Generosity'], label='Generosity' )
sns.regplot(x=my_data['Overall rank'],y=my_data['Perceptions of corruption'], label='Perceptions of corruption' )
sns.regplot(x=my_data['Overall rank'],y=my_data['GDP per capita'], label='GDP per capita' )
plt.ylabel("indices")
plt.legend(loc="best")

# Check that a figure appears below
step_4.check()
````
