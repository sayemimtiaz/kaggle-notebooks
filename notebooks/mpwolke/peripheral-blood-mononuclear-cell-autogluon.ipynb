# **<span style="color:#DC143C;">Decreased Mitochondrial Function, Biogenesis, and Degradation in Peripheral Blood Mononuclear Cells from Amyotrophic Lateral Sclerosis Patients as a Potential Tool for Biomarker Research</span>**

Citation: Araujo, B.G., Souza e Silva, L.F., de Barros Torresi, J.L. et al. Decreased Mitochondrial Function, Biogenesis, and Degradation in Peripheral Blood Mononuclear Cells from Amyotrophic Lateral Sclerosis Patients as a Potential Tool for Biomarker Research. Mol Neurobiol 57, 5084–5102 (2020). https://doi.org/10.1007/s12035-020-02059-1

"In PBMCs from ALS patients, the authors observed an overall lower mitochondrial calcium uptake/retention, mitochondrial depolarization, and redox homeostasis unbalance. Moreover, ALS PBMCs showed a significant decrease in critical metabolic genes and a diminishment in mitochondrial biogenesis and content. These findings seem to occur in parallel to an increase in mitochondrial fission and autophagy-related gene expression, despite a reduction in mitochondrial degradation signaling. Accordingly, ALS cells present a decrease in energy-producing metabolic compounds, such as ATP production and pyruvate, corroborating the fact that mitochondrial deregulation can lead to an energetic deficiency in peripheral cells as an ultimate consequence of the disorder. Thus, mitochondrial function evaluation in PBMCs could be a valuable strategy to detect ALS, as well as to assess its progression and therapy efficacy."


**<span style="color:#DC143C;">Depolarized Mitochondria and Disturbances in Calcium Handling in ALS PBMCs</span>**

"Considering the importance of mitochondria function in keeping calcium handling, and because modifications in the transport of electron and the proton motive force can interfere with ATP production, we evaluated the cytosolic Ca2+ level in PBMCs from ALS and control subjects. As the authors can notice, there is a significant reduction in the Fluo-4 fluorescence in the ALS group, suggesting a decrease in cytosolic calcium in the patient’s cells. Curiously, in the presence of FCCP (5 μM), there is also a diminishment in the cytosolic calcium in ALS PBMCs."


**<span style="color:#DC143C;">Decreased Antioxidant Defense and Metabolic Gene Expression in ALS PBMCs</span>**

"Because oxidative stress is increased in peripheral cells from ALS subjects, they further investigated the expression of NFE2L2, a transcription factor related to the synthesis of antioxidant enzymes. As shown in, there is a massive reduction in NFE2L2 in ALS cells, indicating that ALS PBMCs have a decreased antioxidant defense."

**<span style="color:#DC143C;">Changes in Bioenergetics Metabolic Compounds in PMBCs from ALS Patients</span>**

"As the authors observed changes in mitochondrial function in ALS’ PBMCs, besides deviations in the expression of genes related to mitochondrial metabolism, biogenesis, and fission, they investigated whether the bioproducts related to energy metabolisms, such as ATP, ADP, pyruvate, and lactate, could vary. As observed, there is a significant decrease in ATP levels in PBMCs from ALS patients compared with control individuals; no modification is observed regarding ADP levels. As they can notice, the pyruvate level significantly dropped in ALS PBMCs, while the level of lactate is significantly elevated. This outcome suggests that pyruvate may be converted into lactate to produce energy since ATP synthesis, once oxidative phosphorylation might be jeopardized because of mitochondrial deregulation."

"Due to the increased ROS (Reactive oxygen species) level in ALS PBMCs, they decided to investigate NFE2L2 expression, which is implicated in cellular detoxifying systems. They observed a large diminishment in NFE2L2 expression (more than 50%) in ALS cells. The increased levels of ROS (Reactive oxygen species) could have been related to a failure of NFE2L2 in translocating to the nucleus, reducing the transcription of secondary factors of the antioxidant cascade. Interestingly, in previous work with PBMCs, the expression of NFE2L2 had no changes in relation to controls, but ALS patients were 10 years older than in their study, which could indicate a time-dependent NFE2L2 expression. In agreement with their outcome, the expression of NFE2L2 was declined in humans with ALS and in the SOD1-G93A model. Of importance, NFE2L2 overexpression was associated with more prolonged survival and neuroprotection in the SOD1-G93A mouse model."

"Their results indicate that PBMCs from ALS patients present a significant mitochondrial dysfunction that could be used as a powerful tool in investigating ALS disclosure and improvement. Importantly, the usage of PBMCs in biomarker research was already validated, and our findings can further contribute to future findings concerning new pharmacological strategies against ALS."

https://link.springer.com/article/10.1007/s12035-020-02059-1

**<span style="color:#DC143C;">Amyotrophic Lateral Sclerosis Multiprotein Biomarkers in Peripheral Blood Mononuclear Cells</span>**

Authors: Giovanni Nardo, Silvia Pozzi, Mauro Pignataro, Eliana Lauranzano, Giorgia Spano, Silvia Garbelli, Stefania Mantovani, Kalliopi Marinou, Laura Papetti, Marta Monteforte, Valter Torri,  Gianfranco Bazzoni, Christian Lunetta, Massimo Corbo, Gabriele Mora, Caterina Bendotti, Valentina Bonetto.
October 2011PLoS ONE 6(10):e25545 - DOI: 10.1371/journal.pone.0025545 - PubMed

" Specific molecular biomarkers are urgently needed to facilitate clinical studies and speed up the development of effective treatments. They used a two-dimensional difference in gel electrophoresis approach to identify in easily accessible clinical samples, peripheral blood mononuclear cells (PBMC), a panel of protein biomarkers that are closely associated with ALS. Validations and a longitudinal study were performed by immunoassays on a selected number of proteins. The same proteins were also measured in PBMC and spinal cord of a G93A SOD1 transgenic rat model." 

"The authors identified combinations of protein biomarkers that can distinguish, with high discriminatory power, ALS patients from healthy controls (98%), and from patients with neurological disorders that may resemble ALS (91%), between two levels of disease severity (90%), and a number of translational biomarkers, that link responses between human and animal model. They demonstrated that TDP-43, cyclophilin A and ERp57 associate with disease progression in a longitudinal study. Moreover, the protein profile changes detected in peripheral blood mononuclear cells of ALS patients are suggestive of possible intracellular pathogenic mechanisms such as endoplasmic reticulum stress, nitrative stress, disturbances in redox regulation and RNA processing. Their results indicate that PBMC multiprotein biomarkers could contribute to determine amyotrophic lateral sclerosis diagnosis, differential diagnosis, disease severity and progression, and may help to elucidate pathogenic mechanisms."

https://www.researchgate.net/publication/51716609_Amyotrophic_Lateral_Sclerosis_Multiprotein_Biomarkers_in_Peripheral_Blood_Mononuclear_Cells

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
#https://auto.gluon.ai/stable/tutorials/tabular_prediction/tabular-quickstart.html

#Code by Daiki Katsuragawa https://www.kaggle.com/daikikatsuragawa/tps-apr-2021-benchmark-using-autogluon/notebook

label = 'smpldn'
time_limit = 3600
predictor = TabularPredictor(label=label).fit(df, time_limit=time_limit, num_bag_folds=8)
```

#For tabular problems, fit() returns a Predictor object. For classification, you can easily output predicted class probabilities instead of predicted classes:
https://auto.gluon.ai/stable/tutorials/tabular_prediction/tabular-quickstart.html

```python
#https://auto.gluon.ai/stable/tutorials/tabular_prediction/tabular-quickstart.html

pred_probs = predictor.predict_proba(df)
pred_probs.head(5)
```

```python
results = predictor.fit_summary()
```

From this summary, we can see that AutoGluon trained many different types of models as well as an ensemble of the best-performing models. The summary also describes the actual models that were trained during fit and how well each model performed on the held-out validation data. We can view what properties AutoGluon automatically inferred about our prediction task:

```python
print("AutoGluon infers problem type is: ", predictor.problem_type)
print("AutoGluon identified the following types of features:")
print(predictor.feature_metadata)
```

```python
predictor.leaderboard(df, silent=True)
```
