To get an idea of the impact that the coronavirus can have on the country, just remember what health services were like even before the pandemic arrived.

"Until now, receiving medical services has been very difficult for the majority of the population. Many people need to travel from one part of the country to another for hours to the few existing hospitals to receive care. There are pregnant women who currently need to travel from a city to another to have a cesarean section "

There is no official data on the number of hospitals currently operating in Haiti or the number of health professionals, although, according to the local press, both have been decreasing in recent years.

A 2018 report by the St Luke Foundation and the Maryland Medical Center showed that year, across the country, there were only 90 beds for intensive care. The country has 10 million inhabitants.

Of these, only 45 had respirators, one of the most essential equipment for the most critical cases of coronavirus.https://translate.google.com.br/translate?hl=en&sl=pt&u=https://www.bbc.com/portuguese/internacional-52055474&prev=search&pto=aue

```python
#Codes from Paul Mooney (visualization of most common words)

%matplotlib inline
def plotWordFrequency(input):
    f = open(Haiti,'r')
    words = [x for y in [l.split() for l in f.readlines()] for x in y]
    data = sorted([(w, words.count(w)) for w in set(words)], key = lambda x:x[1], reverse=True)[:40] 
    most_words = [x[0] for x in data]
    times_used = [int(x[1]) for x in data]
    plt.figure(figsize=(20,10))
    plt.bar(x=sorted(most_words), height=times_used, color = 'red', edgecolor = 'black',  width=.5)
    plt.xticks(rotation=45, fontsize=18)
    plt.yticks(rotation=0, fontsize=18)
    plt.xlabel('Most Common Words:', fontsize=18)
    plt.ylabel('Number of Occurences:', fontsize=18)
    plt.title('Most Commonly Used Words: %s' % (Haiti), fontsize=24)
    plt.show()
```
