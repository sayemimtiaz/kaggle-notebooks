```python
df['data'] = pd.to_datetime(df['data']).dt.normalize()
daily = df.sort_values(['data','stato','denominazione_regione'])
latest = df[df.data == daily.data.max()]
latest.head()
```

```python
df_groupby_regione = latest.groupby("denominazione_regione")[['totale_positivi', 'deceduti', 'dimessi_guariti','terapia_intensiva','totale_positivi','totale_ospedalizzati', 'totale_casi', 'isolamento_domiciliare', 'tamponi']].sum().reset_index()
dgr = df_groupby_regione 
dgr.head()
```

```python
import plotly.express as px
fig = px.bar(dgr[['denominazione_regione', 'terapia_intensiva']].sort_values('terapia_intensiva', ascending=False), 
             y="terapia_intensiva", x="denominazione_regione", color='denominazione_regione', 
             log_y=True, template='ggplot2', title='ICU patients by Regions')
fig.show()
```

```python
dgrs_el = dgr.sort_values(by=['deceduti'],ascending = False)
dgrs_el.head()
```

```python
from wordcloud import WordCloud
def nonan(x):
    if type(x) == str:
        return x.replace("\n", "")
    else:
        return ""

text = ' '.join([nonan(abstract) for abstract in df["denominazione_regione"]])
wordcloud = WordCloud(max_font_size=None,colormap='RdGy', background_color='green', collocations=False,
                      width=1200, height=1000).generate(text)
fig = px.imshow(wordcloud)
fig.update_layout(title_text='Regioni Italiane')
```
