## **Contribution of Gender to Purchase**

```python
df1= df.groupby(['Gender'])['Purchase'].agg('sum').reset_index(name="total_Purchase")
df2= df.groupby(['Gender'])['Purchase'].agg('mean').reset_index(name="total_Purchase")
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,10)) #ax1,ax2 refer to your two pies
# 1,2 denotes 1 row, 2 columns
ax1.pie(df1['total_Purchase'],labels=df1['Gender'],colors =["pink","blue"], explode= (0,0.05),autopct = '%1.1f%%') #plot first pie
ax1.set_title('Percentage of total Purchase across Gender',fontsize = 12)
ax1.legend( ['female', 'male']) #you would need to keep the sequence same

ax2.pie(df2['total_Purchase'],labels=df2['Gender'],colors =["pink","blue"], explode= (0,0.05),autopct = '%1.1f%%') #plot first pie
ax2.set_title('Percentage of mean Purchase across Gender',fontsize = 12)
ax2.legend( ['female', 'male'])

fig.show()
```

***Contribution of male gender towards purchase  is almost 3 times as female gender.
Also mean purchase value for male is higher than male.***

## **Contributions of various Age-groups to purchase**

```python
df3= df.groupby(['Age'])['Purchase'].agg('sum').reset_index(name="total_Purchase")
labels = df3.Age
men_means = df3.total_Purchase
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.fM' % (x * 1e-6)


formatter = FuncFormatter(millions)


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Total Purchase',color = '#019f66')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.set_ylabel('Total Purchase')
ax.set_title('Total Purchase across different Age group')
ax.set_xticks(x)
ax.yaxis.set_major_formatter(formatter)

ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)

fig.set_size_inches(10.5, 7.5, forward=True)
fig.tight_layout()

plt.show()
```

***From the above graph we can conclude that age group 26-35 contribute the most towards purchase and the age group below 17 contribute minimum.***

## **Contributions of various occupations to purchase**

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly import tools


df_2= df.groupby(['Occupation'])['Purchase'].agg('sum').reset_index(name="Total Purchase")
fig = px.bar(df_2, x="Occupation", y="Total Purchase")
fig.update_layout( annotations=[
        dict(
            x=xpos,
            y=ypos,
            xref='x',
            yref='y',
            text=str(ypos),
            


            ax=0,
            ay=-40
    ) for xpos, ypos in zip(df_2['Occupation'], df_2['Total Purchase'])
    ],
    font_size = 7,
    title={
        'text': "Total Purchase across Occupation",
        'y':0.97,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
        xaxis_type="category")


fig.show()
```

***From the above graph we can conclude that occupation of category 4 purchases the maximum and that of category 8 purchases the minimum.***

## **Contributions of different Cities to purchase**

```python
df_3= df.groupby(['City_Category'])['Purchase'].agg('sum').reset_index(name="total_Purchase")
labels = df_3.City_Category
men_means = df_3.total_Purchase
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.fM' % (x * 1e-6)


formatter = FuncFormatter(millions)


x = np.arange(len(labels))  # the label locations
width = 0.6 # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Total Purchase',color='#d25f2e')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.set_xlabel('City')
ax.set_ylabel('Total Purchase')
ax.set_title('Total Purchase across different Cities')
ax.set_xticks(x)
ax.yaxis.set_major_formatter(formatter)

ax.set_xticklabels(labels)



def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)

fig.set_size_inches(8.5, 6.5, forward=True)
fig.tight_layout()

plt.show()
```

***From the above graph we can conclude that contribution of City category B in purchases is maximum and that of category A is minimum.***

## **Percentage share of total Purchase across Marital_Status**

```python
import plotly.graph_objects as go
df6= df.groupby(['Marital_Status'])['Purchase'].agg('sum').reset_index(name="total_Purchase")
fig = go.Figure(data=[go.Pie(labels=df6['Marital_Status'], values=df6['total_Purchase'], hole=.5, textinfo='label + percent',    textposition='inside',)])
fig.update_layout(
    title={
        'text': "Percentage of total Purchase across Marital_Status",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

        

fig.show()
```

***From the above Donut-chart we can conclude that unmarried people has purchased more than married people.***

## **Contributions of different products to purchase**
