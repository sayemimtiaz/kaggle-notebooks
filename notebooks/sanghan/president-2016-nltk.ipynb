```python
import pandas as pd
from textblob import TextBlob
import sqlite3

conn = sqlite3.connect('../input/database.sqlite')

e = pd.read_sql_query("""
   Select ExtractedBodyText
   FROM Emails
   WHERE ExtractedBodyText like '%President%'
   LIMIT 1000
""", conn)
```

```python
blobs = [TextBlob(i) for i in e.ExtractedBodyText.values]

sen, pol = sum([i.sentiment.subjectivity for i in blobs])/1000, sum([i.sentiment.polarity for i in blobs])/1000
print("Average Subjecvity", sen,sep=":")
print("Average Polarity", pol,sep=":")
```
