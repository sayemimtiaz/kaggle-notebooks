## Problem statement

We are given coid-19 data for india, aggregated at state level, and asked to find some useful insights from this.
<br>
### Feature Information

1. **State/UTs** - Names of Indian States and Union Territories.
2. **Total Cases** - Total number of confirmed cases
3. **Active** - Total number of active cases
4. **Discharged** - Total number of discharged cases
5. **Deaths** - Total number of deaths
6. **Active Ratio (%)** - Ratio of number of active cases to total cases
7. **Discharge Ratio (%)** - Ratio of number of discharged cases to total cases
8. **Death Ratio (%)** - Ratio of number of deaths to total cases

```python
data.columns
```

```python
# TOTAL CASES 
total_cases = data[['State/UTs','Total Cases']].sort_values(by=['Total Cases'],ascending=False).reset_index(drop=True)
```

```python
plt.figure(figsize=(8,10))
ax =sns.barplot(x='Total Cases',y='State/UTs',data=total_cases)
ax.bar_label(ax.containers[0],padding=10,fmt='%g');
plt.show()
```
