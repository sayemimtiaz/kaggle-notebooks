```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country='Italy'


df1=pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header= 0)
df1=df1[df1['country']==selected_country]
df1['date']= pd.to_datetime(df1['date']).dt.strftime('%Y/%m/%d')

df2=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2=df2[df2['country']==selected_country]
df2['date']= pd.to_datetime(df2['date']).dt.strftime('%Y/%m/%d')

df=df2.loc[:,['daily_vaccinations','date']]
df3=pd.merge(df1,df, how='left',on='date')

df1.index=pd.to_datetime(df1['date'])
df2.index=pd.to_datetime(df2['date'])
print(np.unique(df2['country']))
```

```python
fig,ax=plt.subplots(figsize=(10,5))

ax.plot(df1.index,df1['daily_new_cases'],label='Cases')
ax1=ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r',label='Vacciantions')


plt.show()
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Dailly new vaccinations')

layout_obj=go.Layout(title='COVID vs. Vaccines', xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Dailly new deaths',side='right',overlaying='y'))
fig=go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
```
