## Hypothesis testing: Do women fear certain phenomena significantly more than men? 

```python
phobia_datas.reset_index(drop=True, inplace=True)
demographic_datas.reset_index(drop=True, inplace=True)
data = pd.concat([phobia_datas, demographic_datas], axis = 1)
```

```python
data.isnull().sum()
```

```python
# dealing nan values
# gender is not important
phobia_list = list(phobia_datas.columns.values)

for i in phobia_list:
    data[i].fillna(0.0,inplace = True)
```

```python
data.isnull().sum()
```

```python
#  phoiba ratings by genders


#phobia list 

phobia_list = list(phobia_datas.columns.values)

phobia_ratio = []

for each in phobia_list:
    x  = data[each][data.Gender == "female"]
    ratio1 = sum(x)/len(x)
    limited_float = round(ratio1, 2)
    phobia_ratio.append(limited_float)

dic1  = {"phobies": phobia_list, "average scores by females": phobia_ratio}
female_phobia_rate  =pd.DataFrame(dic1)
female_phobia_rate.head()
```

```python
phobia_ratio2 = []

for each in phobia_list:
    y  = data[each][data.Gender == "male"]
    ratio2 = sum(y)/len(y)
    limited_float2 = round(ratio2, 2)
    phobia_ratio2.append(limited_float2)
dic2  = {"phobies": phobia_list, "average scores by males": phobia_ratio2}
male_phobia_rate  =pd.DataFrame(dic2)
male_phobia_rate.head()
```

```python
f,ax1 = plt.subplots(figsize = (20,15))
sns.pointplot(x = "phobies", y = "average scores by females", data = female_phobia_rate, color = "orange", alpha = 0.6, label= "female")
sns.pointplot(x = "phobies", y = "average scores by males", data = male_phobia_rate, color = "blue", alpha = 0.6, label = "male")
ax1.legend(loc = "upper right", frameon = True )
ax1.set(xlabel = "phobies", ylabel = "average scores ", title = "average phobie scores given gender")
plt.show()
```

```python
# distinction between rates
```

```python

counter = 0 
distinction = []
for each in phobia_ratio:
    
    val  = each - phobia_ratio2[counter]
    limited_float3 = round(val,2)
    counter = counter + 1
    distinction.append(limited_float3)
```

```python

dis_dic  = {"phobies":phobia_list, "distinction": distinction }
dis_data = pd.DataFrame(dis_dic)
dis_data.head()
```

```python
### barplot
plt.figure(figsize = (20,8))
ax = sns.barplot(data = dis_data, x = "phobies", y="distinction", alpha = 0.6)

ax.set_title("Phobie Rating Gap Between Genders")
ax.text(7,1,"Bigger = Female's phobie score is bigger than Male's")
plt.show()

```

## Hypothesis testing: Do the left handed people have different interests than right handed?

```python
"""
- needed data's:
    hobbie_and_interests_datas
    demographic_datas   
    
- dealing null values (just handle interests)
"""
```

```python
hobbie_and_interests_datas.columns
```

```python
hobbie_and_interests_datas.isnull().sum()
```

```python
# dealing null values
hobbie_and_interests_datas = hobbie_and_interests_datas.fillna(0.0)
hobbie_and_interests_datas.isnull().sum()
```

```python
hob_dem = pd.concat([hobbie_and_interests_datas,demographic_datas["Left - right handed"]],axis= 1)
```

```python
hob_dem.head()
```

```python
hob_dem["Left - right handed"].value_counts()
```

```python

interests = list(hobbie_and_interests_datas.columns)
```

```python
right = []
left = []
for each in interests:
    a = hob_dem[each][hob_dem["Left - right handed"] == "right handed"]
    b = hob_dem[each][hob_dem["Left - right handed"] == "left handed"]
    
    right_ratio = sum(a)/len(a)
    left_ratio = sum(b)/len(b)
    right_ratio = round(right_ratio,2)
    left_ratio = round(left_ratio,2)
    right.append(right_ratio)
    left.append(left_ratio)
    
```

```python
hand_dic = {"hobbie": interests, "right handed": right, "left handed": left}
data_hand = pd.DataFrame(hand_dic)
```

```python
f, ax  = plt.subplots(figsize = (25,10))
sns.pointplot(data = data_hand , x = "hobbie", y = "right handed", color = "orange", alpha = 0.6, label = "right handed")
sns.pointplot(data = data_hand , x = "hobbie", y = "left handed", color = "green", alpha = 0.6, label = "left handed")
ax.set(xlabel ="hobbies", ylabel= "used hand", title = "İnterests By Given Used Hands")
plt.xticks(rotation = 90)
plt.text(0.3,4.5,"* Right Handed", color = "orange", alpha = 0.6,fontsize = 20)
plt.text(0.3,4.3,"* Left Handed", color = "green", alpha = 0.6,fontsize = 20)
plt.show()
```

```python
counter2 = 0
listt = []
for each in right:
    var1  = each - left[counter2]
    counter2 += 1
    var1 = round(var1,2)
    listt.append(var1)

dicc =  {"hobbies":interests, "distinction": listt}
data_dicc = pd.DataFrame(dicc)
data_dicc.head()

```

```python
plt.figure(figsize = (19,15))
sns.barplot(data = data_dicc , x = "hobbies", y = "distinction")
plt.xticks(rotation = 90 )
plt.text(0.5,0.3,"Right Handed", color = "red", alpha = 0.4, fontsize = 15)
plt.text(0.5,-0.35,"Left Handed",color = "red", alpha = 0.4, fontsize = 15)
plt.title("The Gap Between The Areas of Interest Selected According to The Hand Used ")
plt.show()
```
