# Load the data

```python
df = pd.read_csv("/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
df.head()
```

# Data visualization

```python
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax =  plt.subplots(2, 4, figsize=(18, 10))

# stroke
sns.countplot(df["stroke"], ax=ax[0,0])
# gender
sns.countplot(df["gender"], ax=ax[0,1])
# hypertension
sns.countplot(df["hypertension"], ax=ax[0,2])
# Residence_type
sns.countplot(df["Residence_type"], ax=ax[0,3])
# heart_disease
sns.countplot(df["heart_disease"], ax=ax[1,0])
# ever_married
sns.countplot(df["ever_married"], ax=ax[1,1])
# work_type
sns.countplot(df["work_type"], ax=ax[1,2])
ax[1,2].tick_params('x', labelrotation=45)
# smoking_status
sns.countplot(df["smoking_status"], ax=ax[1,3])
ax[1,3].tick_params('x', labelrotation=45)


fig.show()
plt.xticks(rotation=45)
```
