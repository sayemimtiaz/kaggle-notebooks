# **Goal:**

We all know Dask is good at reading the big dataset comparing Pandas. However, we would like to do a small bench marking with simple %timeit function to figure out how much it is taking on both libraries.

```python
import gc
import dask.dataframe as ddf
import os
```

```python
FILEPATH = '/kaggle/input/fifa19/data.csv'
```

```python
def get_filesize(filepath):
    
    file_bytes = os.stat(filepath).st_size / 1024
    
    file_kb = file_bytes / (1024)
    
    file_mb = file_bytes / (1024 * 1024)
    
    return file_kb
```

```python
get_filesize(FILEPATH)
```

```python
def get_data_pandas(filepath = FILEPATH):
    
    df2 = pd.read_csv(filepath)
    
    return df2
```

## Results

```python
results = pd.DataFrame(columns = ['Rows', 'Cols', 'Time Taken'])
```

```python
results.loc['Fifa Dataset - Dask']   = (18207, 89, '7.39 µs')
results.loc['Fifa Dataset - Pandas'] = (18207, 89, '9.06 µs')
```

```python
results
```

**Results:**

* For 18207 rows * 89 columns, we could save 1.30 µs which is quite good. 
* If we use big dataset, it would help us to save much time.
* We have to specify the dataset manually which might be a little annoying. However, that's the only one way we could get higher performance
