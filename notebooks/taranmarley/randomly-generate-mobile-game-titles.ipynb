# Random Name Generator

Looking at this dataset I thought it would be a bit of fun to turn it into a random name generator for mobile games. 

Mobile game titles are a particularly good fit for random generation because they often focus on simple nouns and verbs. Due to this I think the results came out quite well.

```python
df = pd.read_csv("/kaggle/input/top-play-store-games/android-games.csv")
df.info()
```

So here I use regex to remove any characters that aren't in the english alphabet. This help gets rid of numbers, colons and non-english words. 

I also get rid of words that are three characters or less as these are unlikely to be helpful in generating good names, this removes words like "the" and "and".

```python
import re
i = 0

game_title_words = []
regex = re.compile('[^a-zA-Z]')
for word_string in df['title'].unique():
    for word in word_string.split(' '):
        sub_word = regex.sub('', word)
        if sub_word == word:
            if sub_word != '':
                if len(sub_word) > 3 and sub_word not in game_title_words and len(sub_word) <= 10:
                    game_title_words.append(word)
print(game_title_words[:10])
```

The first ten words look fine above so time to make the generator.
For this I've set it to write a title from 2 to 4 words long. 

I also capitalize words that are in lower case while leaving Uppercase words alone. I felt this looks better.

```python
import random
random.seed(3)
titles = []
for i in range(10):
    title_length = random.randrange(2,5,1)
    title = ''
    for j in range(0,title_length):
        titleword = ''
        if j > 0:
            title += ' '
        titleword += random.choice(game_title_words)
        if titleword[0].islower():
            titleword = titleword.capitalize()        
        title += titleword
    titles.append(title)
print(titles)
```

Personally I think this works pretty well. Not every Title generated will be great but I think it produces an believable mobile game title the vast majority of the time. 

You can change the seed above to see new titles. Some can be pretty amusing.
