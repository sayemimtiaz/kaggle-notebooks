<h3><font style="color:#e52165">SimpleImputer Ä°le doldurma</font></h3>

SimpleImputer ile eksik verileri doldurmak daha kolaydÄ±r. Fakat teoride yukarÄ±da gÃ¶sterdiÄŸimiz yÃ¶ntemler ile aynÄ±dÄ±r.

```python
# Import iÅŸlemi yapmalÄ±yÄ±z;

from sklearn.impute import SimpleImputer
```

```python
# missing_values: neleri doldurmak istediÄŸimizi belirtiriz, np.nan eksik verileri doldurur
# strategy: hangi yÃ¶ntemi kullanmak istediÄŸimizi belirleriz, "mean" ortalamadÄ±r.

imputer = SimpleImputer(missing_values = np.nan,strategy="mean")
```

```python
# YukarÄ±daki iÅŸlemde, yapmak istediÄŸimiz iÅŸlemlerin detaylarÄ±nÄ± verdik.
# Åimdi ise bunu uygulamak istediÄŸimiz deÄŸiÅŸken(ler)i belirtiyoruz.

imputer.fit_transform(df[["Runtime"]])
```

> GÃ¶rmÃ¼ÅŸ olduÄŸunuz gibi, bir tane imputer nesnesi oluÅŸturduÄŸumuz zaman, bunu istediÄŸimiz deÄŸiÅŸkenlere fit_transform edebiliyoruz.

```python
!pip install ycimpute  
# Gerekli KÃ¼tÃ¼phaneyi import ediyoruz.

from ycimpute.imputer import knnimput
```

```python
# Bu iÅŸlem iÃ§in numeric datalar gerekli bu yÃ¼zden veri setimizden sadece numerik datalarÄ± Ã§ekiyoruz

df_numeric = df._get_numeric_data()
```

```python
# Ä°leride iÅŸimize yaracaÄŸÄ± iÃ§in kolon isimlerini bir deÄŸiÅŸeken kaydettik
# df'yi array'e Ã§evirdik cÃ¼nkÃ¼ algoritmada kullanacaÄŸÄ±z

var_names = list(df_numeric)
n_df = np.array(df_numeric)
```

```python
# k parametresi komÅŸuluk sayÄ±sÄ±dÄ±r. (KNN AlgoritmasÄ± ile alakalÄ±dÄ±r. AraÅŸtÄ±rabilirsiniz.)
# Bu algorit
# YukarÄ±da, algoritma iÃ§in array'e Ã§evirdiÄŸimiz "n_df" deÄŸiÅŸkenini kullanÄ±yoruz.

dff = knnimput.KNN(k=4).complete(n_df)
```
