# Data Visualization, Analysis, Insights

I found it intriguing that this database had a feature for age generations. Although I do find this information useful holistically - this information would be more insightful if used to compare generations between different countries. 

```python
generation_rate = df['suicides_no'].groupby(df.generation).sum().sort_values(ascending= False)
generation_rate
```

```python
import matplotlib.pyplot as plt 
from numpy import median
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 9), subplot_kw=dict(aspect="equal"))

wedges, texts = ax.pie(generation_rate, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(generation_rate.index[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Suicide Numbers Amongst Generations")

plt.show()
```

There is nothing in this data that states how many reported suicides there were based off the total population. "suicides/100k population" is close, but not exactly what I want (some countries do not have some years reported). I figured that since this dataset gives reported suicides by year, by demographic, I could analyze the countries separately by their year. So, I created another feature on a separate dataframe: "suicide/population" as well as "gdpcapita_by_year".

```python
country_rate = df['population'].groupby(df['country']).sum()
country_rate = country_rate.to_frame()
country_rate['avg_population'] = country_rate['population'] / df['population'].groupby(df['country']).count()
country_rate.round()
country_rate['suicide_no_avg'] = df['suicides_no'].groupby(df['country']).mean()
country_rate['avg_population'] = country_rate['avg_population'].round()
country_rate['suicide/population'] =  country_rate['suicide_no_avg'] / country_rate['avg_population']
country_rate['gdp_per_capita'] = df['gdp_per_capita'].groupby(df['country']).mean()
country_rate
```

```python
country_rate_time = df['suicides_no'].groupby(df['country_year']).sum()
country_rate_time = country_rate_time.to_frame()
country_rate_time['total_pop_by_year'] = df['population'].groupby(df['country_year']).sum()
cols = list(country_rate_time.columns)
country_rate_time['suicide/population'] = country_rate_time[cols[0]] / country_rate_time[cols[1]]
country_rate_time['gdpcapita_by_year'] = df['gdp_per_capita'].groupby(df['country_year']).first()
country_rate_time.head(25)
```

```python
cols = list(country_rate_time.columns)
country_rate_time.sort_values(by=[cols[2]], ascending=False)
```

However, an important factor to consider when analyzing the suicide rates of countries is to understand how many suicides have been recorded out of total population.
```python
def country_pct_change(country_df):
    country_df = country_df.groupby(country_df['year']).sum()['suicides_no']
    return country_df.pct_change().to_frame()

def country_totals(country_df):
    country_df = country_df.groupby(country_df['year']).sum()['suicides_no']
    return country_df.to_frame()

def country_per_capita(country_df):
    country_df = country_df.groupby(country_df['year']).first()['gdp_per_capita']
    return country_df.to_frame()
```

```python
from sklearn.preprocessing import LabelEncoder

s = (df.dtypes == 'object')
object_cols = list(s[s].index)

label_encoder = LabelEncoder()
for col in object_cols:
    df[col] = label_encoder.fit_transform(df[col])
df.info()
```

```python
cr = country_rate
cr.drop(['population', 'avg_population', 'suicide_no_avg'], axis=1)
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs

kmodel = KMeans(n_clusters=2)
yhat = kmodel.fit_predict(cr)
yhat
```

```python
cr['cluster'] = yhat
df0 = cr[cr.cluster==0]
df1 = cr[cr.cluster==1]

plt.scatter(df0['suicide/population'], df0.gdp_per_capita, color = 'red')
plt.scatter(df1['suicide/population'], df1.gdp_per_capita, color = 'blue')
```
