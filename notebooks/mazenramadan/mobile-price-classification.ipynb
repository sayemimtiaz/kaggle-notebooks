```python
x = train.drop('price_range', axis=1)
```

```python
x.head()
```

```python
y = train['price_range']
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
test_scaled = scaler.fit_transform(test)
```

```python
x_scaled
```

```python
test_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
```

```python
model_knn.fit(x_train, y_train)
```

```python
y_pred = model_knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
model_knn_acc = accuracy_score(y_test, y_pred)
```

```python
model_knn_acc
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model_dt.fit(x_train_scaled, y_train)
```

```python
y_pred = model_dt.predict(x_test_scaled)
```

```python
y_pred
```

```python
model_dt_acc = accuracy_score(y_test, y_pred)
```

```python
model_dt_acc
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
```

```python
model_lr.fit(x_train_scaled, y_train)
```

```python
y_pred = model_lr.predict(x_test_scaled)
```

```python
y_pred
```

```python
model_lr_acc = accuracy_score(y_test, y_pred)
```

```python
model_lr_acc
```

```python
#As u can see Logistic Regression scored the best accuracy and it will be the final model
```
