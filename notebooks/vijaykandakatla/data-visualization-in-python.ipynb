As we know data is generally of 2 types,Quantitative and Qualitative. 
**Quantitaive Data**-Quantitative data is information about quantities; that is, information that can be measured and written down with numbers. This type of data is measured using numbers and values, which makes it a more suitable candidate for data analysis. These are furthur broken down to 2 types namely discrete(eg, how many toys,how many teams) and continuous data(eg. age,height,weight).discrete data can furthur be nominal,ordinal, Binary kind.
**Qualitative Data** -This is the information about Qualities. these are categorically describes the topic, instead of measuring with a number.eg.sex,yes or no problems,rating.this is further divided to nominal,ordinal or binary nature.

Columns:---- Description ----                                                             Data Type

* exang -----exercise induced angina (1 = yes; 0 = no)-----                             Qualitative_binary data
* oldpeak----ST depression induced by exercise relative to rest  ------                 Quantitave_continuous_data 
* slope ----the slope of the peak exercise ST segment     ------                        Quantitave_discrete_data
* ca  ------number of major vessels (0-3) colored by flourosopy -------                 Quantative_nominal_data
* thal -----3 = normal; 6 = fixed defect; 7 = reversable defect ------                  Qualitative_ordinal_data
* target----- 1 or 0-------                                                             Qualitative_binary_data

*8.JOINTPLOT*
* This is drawn between two features.


```python
sns.set(style="white")
sns.jointplot(x="chol",y="age",kind="kde",color="g",data=df)


sns.set(style="white")
sns.jointplot(x="thalach",y="cp",kind="hex",color="g",data=df)

sns.set(style="white")
sns.jointplot(x="trestbps",y="ca",kind="hex",color="g",data=df)


sns.set(style="white")
sns.jointplot(x="oldpeak",y="thal",kind="kde",color="g",data=df)


sns.set(style="white")
sns.jointplot(x="ca",y="thal",kind="kde",color="g",data=df)


sns.set(style="white")
sns.jointplot(x="ca",y="cp",kind="kde",color="g",data=df)

```
