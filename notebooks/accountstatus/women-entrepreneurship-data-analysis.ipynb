# Importing the packages

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import plotly
plotly.offline.init_notebook_mode (connected = True)
```

# Importing Data 

```python
data=pd.read_csv('../input/women-entrepreneurship-and-labor-force/Dataset3.csv',sep=';')
```

```python
data.head()
```

# Having a look at the Women Entrepreneurship Index All Over The World

```python
fig = go.Figure(data=go.Choropleth(
    locations=data['Country'], # Spatial coordinates
    z = data['Women Entrepreneurship Index'].astype(float), # Data to be color-coded
    locationmode = 'country names', # set of locations match entries in `locations`
    colorscale = 'electric',
    colorbar_title = "Millions USD",
    
))

fig.update_layout(
    title_text = '2015 Women Entrepreneurship Index',
    geo_scope='world', # limite map scope to world

)

fig.show()
```

#### Well from this we can clear see the difference between the different countries on the basis of the women entrepreneurship index . We can clearly see that the highest index could be found out in Australia and the least in India but does that tell all ??

# Making a column for Ratio and having a look at it

```python
data['Ratio']=[data['Women Entrepreneurship Index'].values[i]/data['Entrepreneurship Index'].values[i] for i in range(len(data))]
```

## Adding ratio column to the data 

```python
data.head()
```

```python
fig = go.Figure(data=go.Choropleth(
    locations=data['Country'], # Spatial coordinates
    z = data['Ratio'].astype(float), # Data to be color-coded
    locationmode = 'country names', # set of locations match entries in `locations`
    colorscale = 'electric',
    colorbar_title = "Millions USD",
    
))

fig.update_layout(
    title_text = '2015 Women Entrepreneurship Ratio',
    geo_scope='world', # limite map scope to world

)

fig.show()
```

* From this we can clearly see that india is not the worst with ratio cause it has 1 ratio which means it treats men and women as equal and give both men and women the same amount of the entrepreneurship opportunities. 
* Having a look at Australia we can clearly see it provides nearly same entrepreneurship opportunites to both men and women .
* I also found out countries with more than 1 ratio like mexico which means country shows more attention towards women of the country and provide them with more entrpreneurship opportunites than other countries 
* Another finding was about countries like Turkey where we can see the index is about 0.7 which shows lower amount of ratio than any other country .

# Inflation Of Different countries and it's relation to Women Entrpreneurship Index

```python
data=data.sort_values('Women Entrepreneurship Index',ascending=False)
```

```python
fig=px.scatter(data,x='Women Entrepreneurship Index',y='Inflation rate',template='plotly_dark',trendline="ols",trendline_color_override="cyan")
fig.show()
```

#### From this plot we can clearly see that when the Women Entrepreneurship Index of a country increases we can see that the inflation rate of the country decreases which shows the stablilty in the economy of the country .

#### So we can clearly see that if women of a country are provided good amount of opportunities it's sure that country is moving on the path of the economical prosperity and stability .

# Effect of Female Labour force on the Inflation

```python
fig=px.scatter(data,x='Female Labor Force Participation Rate',y='Inflation rate',template='plotly_dark',trendline="ols",trendline_color_override="cyan")
fig.show()
```

#### Well we can see that the female labour force has but a small impact on the Inflation rate .

# Does development of a country effect the Women Entrepreneurship Index of a country

```python
group=data.groupby('Level of development').mean()
```

```python
px.bar(group,x=group.index,y='Women Entrepreneurship Index')
```

#### Well we can clearly see that the developed countries has higher amount of the Women Entrepreneurship index than the developing countries ....
