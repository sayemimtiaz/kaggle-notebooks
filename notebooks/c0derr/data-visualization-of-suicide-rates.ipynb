# WORD CLOUD
Word Cloud is a data visualization technique used for represending text data in which the size of each word indicates its frequency or importance.

```python
!pip install wordcloud
```

```python
# Python program to generate WordCloud 
  
# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
  

comment_words=' '
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in data.country: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words  + words+ ' '
  
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
import pandas as pd
data=pd.read_csv("/kaggle/input/suicide-rates-overview-1985-to-2016/master.csv")

#data.columns
#data.info() 

import numpy as np
import pandas as pd 
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go


dataframe=data[(data['gdp_per_capita ($)']>20000) & (data['population']<100000)]

# first plot
trace1 = go.Scatter (x = dataframe.country,
                     y = dataframe.suicides_no,
                     name = 'Suicide Score',
                     marker = dict(color = 'rgba(10,140,2,0.8)'),
                    )
# second line
trace2 = go.Scatter (x = dataframe.country,
                     y = dataframe.population,
                     xaxis = 'x2',
                     yaxis = 'y2',
                     name = 'Population',
                     marker = dict(color = 'rgba(10,60,160,0.8)'))

data = [trace1, trace2]
layout = go.Layout (xaxis2 = dict(domain = [0.6, 0.95], anchor = 'y2'),
                    yaxis2 = dict(domain = [0.6, 0.95], anchor = 'x2'),
                    title = 'income and teaching vs world rank of universities')

fig = go.Figure(data=data, layout=layout)
iplot(fig)
plt.savefig('inset-plot.png')
```
