#### Version 3: test on random features - see that previous results unreliable, propose the solution 

We take the first dataset and the smallest considered number of features - 48.
And generate completely random features of the same size. And look for the RocAuc we obtain from several trials.

Just the third trial gives us RocAuc 0.68 and nineth trial gives 0.7 which is better than modeling 0.66.
Thus the modeling is unreliable.

Actually we can understand what happening without simulation.
The problem is that when we make GridSearch with big number of possible params - everytime we get a random RocAuc with std about 0.05 (that std have been estimated https://www.kaggle.com/alexandervc/als-transciptomics-no-markers-beware-overfits ) 
thus taking the best would correspond just taking the best value from  mean = 0.5 , std = 0.05 random variable,
which of course can be easyly go quite high when number of params configurations is high. 

**The possible solution** - choose params with one  split(s) to folds, but then estimate the quality of the model
retraining it on other split fold or even averaging several splits to folds.  (Or even more sophisticated schemes like described here: https://habr.com/ru/company/ods/blog/336168/ )

# Testing modeling scheme on random data - shows that previous modeling scheme is UNRELIABLE - the reason is in GridSearch.

We take the first dataset and the smallest considered number of features - 48.
And generate completely random features of the same size. And look for the RocAuc we obtain from several trials.

Just the third trial gives us RocAuc 0.68 and nineth trial gives 0.7 which is better than modeling 0.66.
Thus the modeling is unreliable.

Actually we can understand what happening without simulation.
The problem is that when we make GridSearch with big number of possible params - everytime we get a random RocAuc with std about 0.05 (that std have been estimated https://www.kaggle.com/alexandervc/als-transciptomics-no-markers-beware-overfits ) 
thus taking the best would correspond just taking the best value from  mean = 0.5 , std = 0.05 random variable,
which of course can be easyly go quite high when number of params configurations is high. 

**The possible solution** - choose params with one  split(s) to folds, but then estimate the quality of the model
retraining it on other split fold or even averaging several splits to folds.  (Or even more sophisticated schemes like described here: https://habr.com/ru/company/ods/blog/336168/ )
