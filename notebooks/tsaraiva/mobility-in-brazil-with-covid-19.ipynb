>The **COVID-19** is an infectious decease recently discovered, and the majority of affected people present mild or moderate symptoms. Despite that, some people could face severe consequences. Nowadays, the COVID-19 is severely impacting the world due to its high rates of infections, which collapses the health systems of affected countries, resulting in several deaths.<br>
The transmission of COVID-19 occurs through contact with the secretions of a sick person or infected surfaces. As long as an efficient treatment or a vaccine is not available, the only way to avoid the transmission and the consequent collapse of health systems, with more deaths, is through **social isolation**, despite its economic implications. Several scientific publications and entities defend social isolation as a way to control the spread of COVID-19. This is defended for example by the **World Health Organization** <a href="https://www.who.int/emergencies/diseases/novel-coronavirus-2019">[2]</a>, the **UK government** <a href="https://www.gov.uk/government/publications/covid-19-stay-at-home-guidance/stay-at-home-guidance-for-households-with-possible-coronavirus-covid-19-infection"> [3]</a>, and the **Center for Disease Control and Prevention of the United States** <a href="https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/social-distancing.html">[4]</a>.<br>
In **Brazil**, the first recorded case of infection was on Feb 25, 2020, and now (May 2020) the country has the most cases and deaths per million in **Latin America**, even with probably substantial underestimates <a href="https://www.thelancet.com/journals/lancet/article/PIIS0140-6736(20)31095-3/fulltext">[1]</a>. Unfortunately, some people in Brazil disbelief the need for social isolation <a href="https://www1.folha.uol.com.br/internacional/en/scienceandhealth/2020/04/of-15-countries-brazil-is-2nd-with-greatest-disbelief-in-social-isolation-says-study.shtml">[5]</a>, as long as the COVID-19 deaths rate increases each day.<br>
The intention here is to analize the mobility of the population during the COVID-19 evolution in Brazil, using public available data. It is used open data about the COVID-19 cases from the BR government <a href="https://covid.saude.gov.br">[6]</a>, and mobility data from Apple <a href="https://www.apple.com/covid19/mobility">[7]</a>, Google <a href="https://www.google.com/covid19/mobility/">[8]</a>, besides the map of a company named Inloco <a href="https://mapabrasileirodacovid.inloco.com.br/pt/?hsCtaTracking=68943485-8e65-4d6f-8ac0-af7c3ce710a2%7C45448575-c1a6-42c8-86d9-c68a42fa3fcc">[9]</a>. In this analysis we used the numbers of deaths, instead of those of infections, as they are more accurate, considering the probably high underreporting in Brazil <a href="https://www.ft.com/content/7b46b8c3-b395-45ba-af68-50280cfeed47?shareType=nongift">[10]</a>.<br>

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

### Datasets

```python
br_states_population = pd.read_csv('/kaggle/input/brazilianstates/states.csv')
dadosms_macro = pd.read_csv('/kaggle/input/corona-virus-brazil/brazil_covid19_macro.csv', parse_dates=['date'], index_col=['date'])
dadosms_cities = pd.read_csv('/kaggle/input/corona-virus-brazil/brazil_covid19.csv', parse_dates=['date'], index_col=['date'])
dadosgoogle = pd.read_csv('/kaggle/input/covid19-mobility-datasets-google-and-apple/Global_Mobility_Report.csv', parse_dates=['date'], index_col=['date'], low_memory=False)
dadosapple = pd.read_csv('/kaggle/input/covid19-mobility-datasets-google-and-apple/applemobilitytrends.csv').query('region == "Brazil"').T[13:].rename(columns={12: 'driving', 13:'transit', 14: 'walking'})
```

### Functions

```python
#Function to calc number of deaths per day
def calcdiario2(data):
    obdiario = []
    for i in range(len(data)):
        if i == 0:
            obdiario.append(data[i])
        else:
            obdiario.append(data[i] - data[i-1])
    obdiario = pd.DataFrame({'date':data.index,'obdiario':obdiario}).set_index('date')
    return obdiario

#Function to create dataset of deaths and mobility
def create_df_mob_ob2(datact, obdiarioct, state):
    obdiario_ct = pd.DataFrame({'datact': datact.values, 'deaths': obdiarioct['obdiario']})
    obdiario_ct = obdiario_ct.set_index('datact')
    obdiario_ct.index.rename('date', inplace=True)
    google_res = dadosgooglebr[dadosgooglebr['region'] == state]['residential']
    final_res_ob_ct = obdiario_ct
    final_res_ob_ct['google_res'] = google_res
    return final_res_ob_ct

#Function to calculate accumulated deaths
def acummdeaths(state):
    dadosms_st = dadosms_cities[dadosms_cities['state'] == state]
    obst = dadosms_st.groupby(by='date').sum()['deaths']
    datast = obst.index
    return obst, datast
```

### Pre-processing

```python
# Extract of population of states in mi 
br_states_population = br_states_population.assign(Population=(br_states_population.Population / 1000000).astype('float64'))
br_states_population = br_states_population.set_index('State')

# Extract of covid-19 cases in Brasil
databrasil = dadosms_macro.index
obbrasil = dadosms_macro['deaths']

#Number of deaths per day in Brazil
obdiariobrasil = calcdiario2(obbrasil)
ma7_br = obdiariobrasil.rolling(window=7).mean()

#Calc the deaths per milion in Brazil
obpmbr = round(obbrasil/br_states_population['Population'].sum(),1)
obdiariopmbr = round(obdiariobrasil/br_states_population['Population'].sum(),1)
ma7_pmibr = obdiariopmbr.rolling(window=7).mean()

# Accumulated deaths in states
cities_grouped= dadosms_cities.groupby(by=['date', 'state']).sum()['deaths']
cities_grouped = pd.DataFrame(cities_grouped).reset_index()
cities_grouped = cities_grouped.sort_values(by='deaths', ascending=False)
cities_grouped2 = cities_grouped.groupby(by='state').max()['deaths'].sort_values(ascending=False)
cities_grouped2 = pd.DataFrame(cities_grouped2).reset_index()

# Accumulated deaths/mi in states
cities_grouped2_n = cities_grouped2.set_index('state')
cities_dea_mi = cities_grouped2_n['deaths'] / br_states_population['Population']
cities_dea_mi = cities_dea_mi.reset_index().rename(columns={'index':'state', 0: 'deaths'})
cities_dea_mi.loc[:,'deaths'] = round(cities_dea_mi['deaths'],1)
cities_dea_mi = cities_dea_mi.sort_values(by='deaths', ascending=False)

#Deaths in RJ
obrj, datarj = acummdeaths('Rio de Janeiro')
obdiariorj = calcdiario2(obrj)
ma7_obrj = obdiariorj.rolling(window=7).mean()

#Deaths in SP
obsp, datasp = acummdeaths('São Paulo')
obdiariosp = calcdiario2(obsp)
ma7_obsp = obdiariosp.rolling(window=7).mean()

#Deaths in AM
obam, dataam = acummdeaths('Amazonas')
obdiarioam = calcdiario2(obam)
ma7_obam =obdiarioam.rolling(window=7).mean()

#Mobilty data from Google
dadosgoogle.rename(columns={'country_region_code': 'crc', 'country_region': 'country', 'sub_region_1':'region', 'sub_region_2':'sr2',
       'retail_and_recreation_percent_change_from_baseline': 'retail', 'grocery_and_pharmacy_percent_change_from_baseline':'grocery_and_pharmacy',
       'parks_percent_change_from_baseline': 'parks', 'transit_stations_percent_change_from_baseline': 'transit_stations',
       'workplaces_percent_change_from_baseline': 'workplaces', 'residential_percent_change_from_baseline': 'residential'}, inplace=True)
dadosgooglebr = dadosgoogle[dadosgoogle['country'] == 'Brazil']
dadosgooglebr_geral = dadosgooglebr[dadosgooglebr['region'].isnull()]
datasgoogle_geral = dadosgooglebr_geral.index

dadosapple_br = dadosapple
dadosapple_br_driving = dadosapple['driving']
dadosapple_br_transit = dadosapple_br['transit']
dadosapple_br_walking = dadosapple_br['walking']

#Unifying mobility data from Apple e Google
mobility_frame = { 'br_google_retail': dadosgooglebr_geral['retail'], 'br_google_grocery_and_pharmacy': dadosgooglebr_geral['grocery_and_pharmacy'],
         'br_google_parks': dadosgooglebr_geral['parks'], 'br_google_transit_stations': dadosgooglebr_geral['transit_stations'],
         'br_google_workplaces': dadosgooglebr_geral['workplaces'], 'br_google_residential': dadosgooglebr_geral['residential'], 
         'br_apple_driving': dadosapple_br_driving,'br_apple_transit': dadosapple_br_transit,
         'br_apple_walking': dadosapple_br_walking}

mobility_frame_df = pd.DataFrame(mobility_frame)
mobility_frame_df = mobility_frame_df.reset_index().melt('index', var_name='mobility', value_name='vals').rename(columns={'index': 'date'})

#Final datasets mobility for Apple and Google
mobility_google = mobility_frame_df[mobility_frame_df['mobility'].str.contains('google')]
mobility_google = mobility_google.assign(vals=(mobility_google.vals.astype('float64')))
mobility_apple = mobility_frame_df[mobility_frame_df['mobility'].str.contains('apple')]
mobility_apple = mobility_apple.assign(vals=(mobility_apple.vals.astype('float64')))

#Create DF with daily deaths and mobility for BR
obdiario_br = pd.DataFrame({'databrasil': databrasil, 'obdiariobrasil': obdiariobrasil['obdiario'].values})
obdiario_br = obdiario_br.set_index('databrasil')
obdiario_br.index.rename('date', inplace=True)
google_res_br = mobility_frame_df[mobility_frame_df['mobility'] == 'br_google_residential'][['date', 'vals']]
google_res_br = google_res_br.set_index('date')

final_res_ob_br = obdiario_br
final_res_ob_br['google_res_br'] = google_res_br

#Create DF with daily deaths and mobility for states
final_res_ob_rj = create_df_mob_ob2(datarj, obdiariorj, 'State of Rio de Janeiro')
final_res_ob_sp = create_df_mob_ob2(datasp, obdiariosp, 'State of São Paulo')
final_res_ob_am = create_df_mob_ob2(dataam, obdiarioam, 'State of Amazonas')

#Dataframe with Google residential mobility and deaths per million in states
res_states_mean = dadosgooglebr.groupby(by='region').mean()['residential'].sort_values(ascending=False)
res_states_mean = round(res_states_mean,2)
res_states_mean = res_states_mean.reset_index()
res_states_mean['region'] = res_states_mean['region'].replace({'Federal District': 'Distrito Federal', 'State of Rio de Janeiro': 'Rio de Janeiro', 'State of Piauí': 'Piauí',
        'State of Ceará': 'Ceará', 'State of São Paulo': 'São Paulo', 'State of Rio Grande do Sul': 'Rio Grande do Sul','State of Pernambuco': 'Pernambuco', 'State of Bahia': 'Bahia',
        'State of Rio Grande do Norte': 'Rio Grande do Norte', 'State of Santa Catarina': 'Santa Catarina', 'State of Espírito Santo':'Espírito Santo', 'State of Goiás': 'Goiás',
        'State of Amapá': 'Amapá', 'State of Paraíba':'Paraíba', 'State of Sergipe':'Sergipe', 'State of Minas Gerais':'Minas Gerais', 'State of Paraná':'Paraná', 'State of Amazonas':'Amazonas',
       'State of Pará':'Pará', 'State of Alagoas':'Alagoas', 'State of Maranhão':'Maranhão', 'State of Tocantins':'Tocantins', 'State of Roraima':'Roraima', 'State of Acre':'Acre',
        'State of Mato Grosso':'Mato Grosso', 'State of Rondônia':'Rondônia', 'State of Mato Grosso do Sul':'Mato Grosso do Sul'})
res_states_mean.rename(columns={'region':'state'}, inplace=True)
res_states_mean.set_index('state', inplace=True)
cities_dea_mi2 = cities_dea_mi.set_index('state')
res_states_mean['deaths'] = cities_dea_mi2['deaths']
res_states_mean.reset_index(inplace=True)
res_states_mean.rename(columns={'state': 'State', 'residential':'Mean of residential mobility', 'deaths': 'Deaths per million'}, inplace=True)

#To plot the graphs
%matplotlib inline
plt.style.use('ggplot')
```

# Total deaths in the country

>First, before the mobility issues, it is presented the context of COVID-19 through the number of deaths, which, due to the sub notification problem, is more accurate that the number of infections. In the next two graphs, it is possible to observe the ascending curve of accumulated deaths over time, and the number of records notifieds per day in absolute values in the first graph and per million in the second.<br><br>
Since the first death in Brazil on March 17, 2020, the mean of new deaths reported is lower on Sundays and Mondays. This problem occurs because the daily numbers communicated by the government are not about the last 24h, [they could be related to weeks ago](https://www.terra.com.br/vida-e-estilo/saude/atraso-em-notificacoes-faz-numeros-da-covid-19-cair-aos-domingos-e-segundas,4fe30cd864373e5da9bd2c951327723f8wy0p6zq.html). This is more one challenge to make predictions based on these numbers. Thus, it is presented also in each graph a line with the 7-day simple moving average [(SMA)](https://www.sciencedirect.com/topics/engineering/moving-average) of the number of notifications per day. Each point in the SMA curve corresponds to the average of the last 7 days.

```python
fig, ax1 = plt.subplots(figsize=(21,10))

plt.title("Deaths accumulated and per day in BR", fontsize=40)

ax1.set_ylabel('Accum. deaths', color='tab:blue', fontsize=35)
ax1.tick_params(axis='y', labelcolor='tab:blue', labelsize=35, length=10, width=10)
ax1.plot(databrasil, obbrasil, lw=8, color='tab:blue', label='Accum. deaths -> last: {} in {}'.format(obbrasil.iloc[-1], max(databrasil.date)))
ax1.legend(loc=1, bbox_to_anchor=(0.68,1), fontsize=30)
ax1.set_xlabel('Date', color='black', fontsize=30)
ax1.tick_params(axis='x', labelcolor='black', labelsize=30, length=10, width=10, rotation=75)

ax2 = ax1.twinx()
ax2.set_ylabel('Deaths p/ day', color='tab:red', fontsize=35)
ax2.tick_params(axis='y', labelcolor='tab:red', labelsize=35, length=10, width=10)
ax2.bar(obdiariobrasil.index, obdiariobrasil['obdiario'], color='tab:red', label='Deaths p/ day -> last: {} in {}'.format(obdiariobrasil['obdiario'].iloc[-1], max(obdiariobrasil.index.date)))

ax3 = ax2
ax3.plot(ma7_br.index, ma7_br['obdiario'], lw=8, color='purple', label='7-day Moving Average  -> last: {} in {}'.format(round(max(ma7_br[ma7_br['obdiario'].notna()].values)[0],2), max(ma7_br.index.date)))
ax3.legend(loc=1, bbox_to_anchor=(0.79,0.9), fontsize=30)

x= np.arange(np.datetime64(min(databrasil)), np.datetime64(max(databrasil) + timedelta(days=1)), timedelta(days=5)).astype(datetime)
k = []
for i in x:
    k.append(i.date())
x = k
ax2.set_xticks(x)
ax2.set_xticklabels(x)

fig.tight_layout()
plt.show();
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title("Deaths accumulated and per day p/ mi in BR", fontsize=40)

ax1.set_ylabel('Accum. deaths/mi', color='tab:blue', fontsize=35)
ax1.tick_params(axis='y', labelcolor='tab:blue', labelsize=35, length=10, width=10)
ax1.plot(databrasil, obpmbr.values, lw=8, color='tab:blue', label='Accum. deaths/mi p/ day -> last: {} in {}'.format(obpmbr.values[len(obpmbr.values)-1], max(databrasil.date)))
ax1.legend(loc=1, bbox_to_anchor=(0.78,1), fontsize=30)
ax1.set_xlabel('Date', color='black', fontsize=30)
ax1.tick_params(axis='x', labelcolor='black', labelsize=30, length=10, width=10, rotation=75)

ax2 = ax1.twinx()
ax2.set_ylabel('Deaths/mi p/ day', color='tab:red', fontsize=35)
ax2.tick_params(axis='y', labelcolor='tab:red', labelsize=35, length=10, width=10)
ax2.bar(obdiariopmbr.index, obdiariopmbr['obdiario'], color='tab:red', label='Deaths/mi p/ day -> last: {} in {}'.format(obdiariopmbr.values[len(obdiariopmbr.values)-1][0], max(obdiariopmbr.index.date)))
ax2.legend(loc=1, bbox_to_anchor=(0.66,0.9), fontsize=30)

ax3 = ax2
ax3.plot(ma7_pmibr.index, ma7_pmibr['obdiario'], lw=8, color='purple', label='7-day Moving Average  -> last: {} in {}'.format(round(max(ma7_pmibr[ma7_pmibr['obdiario'].notna()].values)[0],2), max(ma7_pmibr.index.date)))
ax3.legend(loc=1, bbox_to_anchor=(0.76,0.9), fontsize=30)

x= np.arange(np.datetime64(min(databrasil)), np.datetime64(max(databrasil) + timedelta(days=1)), timedelta(days=5)).astype(datetime)
k = []
for i in x:
    k.append(i.date())
x = k
ax2.set_xticks(x)
ax2.set_xticklabels(x)

fig.tight_layout()
plt.show();
```

# Deaths in States of Brazil

>The next graphs show a macro view of the death numbers for the states of Brazil. The first graph contains the population of each state. The second figure shows the evolution of deaths in accumulated absolute numbers. The third graph shows the total of deaths in absolute numbers in each state, and the fourth graph presents he values per million, considerind the population of each state. <br><br>
It is possible to observe that considering absolute values, São Paulo has the higher number of deaths, followed by Rio de Janeiro, and Ceará. # Mobility in Brazil

>To analize the mobility in Brazil during COVID-19 pandemic it is used here Google and Apple mobility reports, and as a reference the map of the company named Inloco, available publicly. First a brief explanation of each mobility source used:

>>Google
>>>The mobility reports from Google uses percentage change of visits and permanence time to show tendences of displacement of people over time in different category of locations such as retail and leisure, markets and pharmacies, parks, public transport stations, workplaces and residential areas. The reference value used in its dataset is the median of the corresponding day of the week, over the five-week period from January 3 to February 6, 2020. According to Google, the most recent data represents approximately 2 3 days ago, which is the time needed to produce the data sets.<br>
**Source: [Google mobility - COVID19](https://www.google.com/covid19/mobility/)**

>>Apple
>>>The data from Apple shows a relative volume of direction requests by type of transport (diving, walking or public transportation) made to Apple maps compared to a reference value in January 13, 2020.<br>
**Source: [Apple mobility - COVID19](https://www.apple.com/covid19/mobility)**

>>Inloco
>>>Inloco is a company that created an index that aims to show the percentage of the Brazilian population that is respecting the distancing recommendation. They inform in its website that use data from telephony, GPS and Wi-Fi  networks. Despite the data used by Inloco does not represent the entire Brazilian population, they supposedly have a large database. There are some news about its map, like in [en](https://labs.ebanx.com/en/notes/in-loco-social-isolation-brazil-data/), and [pt-br](https://tecnoblog.net/332496/in-loco-divulga-mapa-com-indice-de-isolamento-social-no-brasil/).<br>
**Source: [Inloco map - COVID19](https://www.inloco.com.br/covid-19)**



>For this section, it was considered data at the country level. As can be seen in the correlation heatmap in the first figure of this section, there is a high negative correlation between the values of residential mobility from the Google dataset and the other values. This negative correlation makes sense, and it is expected since the other values refer to tendencies of mobility to several different places or displacements using different ways. On the other hand, residential mobility refers to the tendency of going to the or staying at home.<br><br>
In the second graph are the mobility indicators of Apple and Google over time. It is possible to observe that when the residential rate increases, the other values decrease. There is an increment in the residential indicator of around 20% in the analyzed time frame. The Inloco Map shows an increment of social isolation rate in the same time interval. The last graph shows that the residential mobility indicator starts to increase with the first registers of deaths by COVID-19. <br><br>
**It is necessary to note that any conclusion using this data must consider that the registers of death are provided by the authorities delayed, and there is a period since the infection until the deaths. Thus, it is expected that it takes some days until social isolation results in effects to reduce the number of deaths records.**
