The Global mapping of the percentage of the population that are fully vaccinated

The global mapping of each country's vaccination rate reveals the uneven distribution of the vaccines for the COVID-19 across the regions. For example, as of May 19, the only country that finished vaccinating the 10% of the population is Morocco. Why is Morocco doing so well? This may be attributed to the vaccine distribution from China, and also the fact that the country's repoting and data-management system of the population's vaccination rate is working better than other African countries. 

I furthur investigated the way in which Morocco has done this. As I search, I came aross many news articles, saying Morocco is to be emulated by other countries. Morocco was very fast in responding to the covid-19 at an early stage, and the king immediately secured the speacial fund for the fight against the covid, just like on war. The vaccinationn strategy was that whoever have the elderly and those suffering from chronic illnesses, irrespective of their citizenship. The country's relative economic affluence and the scientific vaccination strategy together co-contributed to the Morocco's success in vaccination, within the African continent.

[2]https://newafricanmagazine.com/25754/

```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,40.0))
fig.show()
```

The global mapping of the recent earthquake

As we can see, Alaska has been experiencing a series of small-scale earthquales for the past three to four days. According to a Volcano Discory article, the United States Geological Survey says that these earthquakes are too small perhaps to be noticed. (source:https://www.volcanodiscovery.com/earthquakes/quake-info/6257345/mag3quake-May-25-2021-23-km-SSW-of-White-Mountain-Alaska.html)

```python
df = pd.read_csv('../input/202105recent-earthquake/query.csv',header=0)
df.index = pd.to_datetime(df['time'])

df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,3.0))
fig.show()
```
