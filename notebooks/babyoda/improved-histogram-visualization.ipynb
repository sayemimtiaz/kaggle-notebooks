```python
#Definition of the Necesarry Modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from termcolor import colored
import random
import plotly.figure_factory as ff
import warnings            
warnings.filterwarnings("ignore") 
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go

#Importing the Datasets
Datax1 = pd.read_csv("../input/internet-access-oecd/DP_LIVE_25112020170111306.csv")
Datax1.columns = Datax1.columns.str.title()

Datax2 = pd.read_csv("../input/access-to-computers-from-home-oecd/DP_LIVE_04122020114826086.csv")
Datax2.columns = Datax2.columns.str.title()

#Reshaping the Datasets
Datax1.drop(["Subject","Indicator","Measure","Frequency","Flag Codes"], axis="columns", inplace=True)
Datax1.rename(columns={"Value": "Internet_Access"}, inplace = True)

Datax2.drop(["Subject","Indicator","Measure","Frequency","Flag Codes"], axis="columns", inplace=True)
Datax2.rename(columns={"Value": "Computer_Access"}, inplace = True)

#Combining the Datasets
Datax = pd.merge(Datax1, Datax2, on=["Location","Time"])
Datax.head(20)
```

```python
variables = ["Internet_Access","Computer_Access"]

for x in variables:
    
    #Convertion of Final Dataset to Necessary Format    
    CountryGroups = Datax.groupby("Location")
    Data1 = pd.concat([pd.Series(v[x].tolist(), name=k) for k, v in CountryGroups], axis=1)    
    
        #Managing Group Names as a List
    ln = len(Data1.columns)
    
    #Limiting the Number of Countries
    if ln >= 10:
        lnm = range(10)
    else:    
        lnm = range(ln)
        
    Group = []   
    for i in lnm:
        col= Data1.columns[i]
        Group.append(col)
    
    Data = pd.melt(Data1.reset_index(), id_vars=["index"],value_vars=Group)
    Data.columns = ["Index", "Country", x]
    
    #Setting the Dataframe Name
    Data1.name = str(x)
    
    #Histograms
    color_provider3 = []
    for i in Group:

        #Creating Random Colors for Each Box Plot
        r = random.randrange(0,255)
        b = random.randrange(0,255)
        g = random.randrange(0,255)
        color_provider = (r, g, b)
        color_provider2 = "rgb"+str(color_provider)
        color_provider3.append(color_provider2)

    hist_data = []
    for i in Data1.columns[:10]:
        x = Data1[i].dropna()
        hist_data.append(x)

    # Create distplot with curve_type set to 'normal'
    fig = ff.create_distplot(hist_data, Group, curve_type="normal", colors=color_provider3)

    # Add title
    fig.update_layout(title_text="Distplot with Normal Distribution for "+str(Data1.name))
    fig.show()

```
