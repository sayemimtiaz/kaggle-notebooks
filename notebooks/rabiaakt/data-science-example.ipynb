**Nested Function**
* There is a LEGB rule that is search local scope, enclosing function, global and built in scopes, respectively.

```python
def nested():
    
    def nest():
        x=5
        y=7
        z=x+y
        return z
    return nest()**2
nested()
```

**ARGUMENTS**

    Default argument example:
    def f(a, b=1):

      """ b = 1 is default argument"""

    Flexible argument example:
    def f(*args):

     """ *args can be one or more"""


    def f(** kwargs)

     """ **kwargs is a dictionary"""




```python
def arg(*args):
    for i in args:
        print(i)
arg(1,2,4,5,6,6,5,650)

def argk(**kwargs):
    for key in kwargs.items():
        print(key)
        
argk(country = 'turkey', capital = 'ankara', population = 426865685)
```
