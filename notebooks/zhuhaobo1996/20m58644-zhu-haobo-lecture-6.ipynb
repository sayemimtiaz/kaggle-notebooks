### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image
import os
South_Korea_dataset = 'South Korea'
South_Korea_Vaccination = 'South Korea'
```

### Import data

```python
covid_global_df = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
vaccination_df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
covid_global_df.index = pd.to_datetime(covid_global_df['date'])
vaccination_df.index = pd.to_datetime(vaccination_df['date'])
print(covid_global_df.columns)
print(vaccination_df.columns)
#print(np.unique(covid_global_df['country']))
#print(np.unique(vaccination_df['country']))

data_south_korea = covid_global_df[covid_global_df['country']== South_Korea_dataset]
vaccination_south_korea = vaccination_df[vaccination_df['country']== South_Korea_dataset]
```

### Analyses and Results

#### 1. Basic graph

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(data_south_korea.index, data_south_korea['daily_new_cases'],'r',label='daily cases')
ax.legend(loc='upper left')
ax_ = ax.twinx()
ax_.plot(vaccination_south_korea.index, vaccination_south_korea['daily_vaccinations'],'y',label='daily_vaccinations')
ax_.legend(loc='lower right')
plt.show()
```

First of all, I draw the basic graph, mark the two polylines with double vertical coordinates, and added legends in the upper left corner and lower right corner.


#### 2.Improved Visualization

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=data_south_korea.index,y=data_south_korea['daily_new_cases'],yaxis='y1',name='daily confirmed cases')
daily_vaccinations = go.Scatter(x=vaccination_south_korea.index,y=vaccination_south_korea['daily_vaccinations'],yaxis='y2',name='daily vaccinations')
layout_obj = go.Layout(title='Confirmed cases & Vaccinations',xaxis=dict(title='Date'),yaxis=dict(title='Daily confrimed cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
figure = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(figure)
```

Then I improve the display of the image, import the Plotly package for more accurate image marking, and  mark the horizontal and vertical axes of data and Legend, and display relevant data when the mouse is moved.


#### 3.Colored tables

```python
style_object = data_south_korea.style.background_gradient(cmap='jet').highlight_max('daily_new_cases').set_caption('Daily cases')
display(style_object)
```

```python
rank_covid_global_df = covid_global_df.groupby('country').max()['cumulative_total_cases'].sort_values(ascending= False)
rank_vaccination_df = vaccination_df.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending= False)
print(rank_covid_global_df.to_string())
print(rank_vaccination_df.to_string())
```

##### The prevention method of government
