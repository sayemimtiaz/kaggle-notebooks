# Classification/ Prediction


### Preprocessing data for classficiation models


#### Label encoding

```python
object_cols=['gender','workex','specialisation','status']
```

```python
df1
```

```python
df2 = df1.copy()
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder

 
object_cols=['gender','workex','specialisation','status','ssc_b','hsc_b']

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    df2[col] = label_encoder.fit_transform(df2[col])
df2.head()
display(df2.info(),df2)
```

```python
# Using get dummies for 3 class categorical variables (hsc_s and degree_t)
for column in ['hsc_s', 'degree_t']:
    dummies = pd.get_dummies(df2[column])
    df2[dummies.columns] = dummies
```

```python
df2
```

```python
# Let's plot correlation matrix to find out less correlated variable to drop them

cor=df2.corr()
plt.figure(figsize=(14,6))
sns.heatmap(cor,annot=True,cmap=plt.cm.Blues)
```

### Dropping few columns with extreme correlation score 

```python
df2.drop(['Others', 'Arts','degree_t','hsc_s','ssc_b','hsc_b', 'salary'], axis=1, inplace=True)
```

```python
corr=df2.corr()
plt.figure(figsize=(14,6))
sns.heatmap(cor,annot=True,cmap=plt.cm.Blues)
```

## Creating Training and Test dataÂ¶


```python
# target vector
y = df2['status']
```

```python
y
```

```python
# dropping as it is taken as target variable
df2.drop('status', axis = 1, inplace = True)
```

```python
df2.columns
```

```python
df2.head()
```

```python
# scaling the data
sc = StandardScaler()
# predictors
X = sc.fit_transform(df2)

# Let us now split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=42, shuffle=True)


print("X-Train:",X_train.shape)
print("X-Test:",X_test.shape)
print("Y-Train:",y_train.shape)
print("Y-Test:",y_test.shape)
```

## Logistic Regression

```python
#creating our model instance
model1 = LogisticRegression()
#fitting the model
model1.fit(X_train, y_train)
```

```python
# predicting the target vectors
y_pred=model1.predict(X_test)
```

```python
y_pred
```

### confusion matrix

```python
conf_mat = pd.DataFrame(confusion_matrix(y_test, y_pred))
fig = plt.figure(figsize=(6,5))
sns.heatmap(conf_mat, annot=True, annot_kws={"size": 16}, fmt='g',cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.xlabel("Predicted values")
plt.ylabel("True values")
plt.show()
```

```python
#Classification Report
print(classification_report(y_test, y_pred))
```

## Accuracy

```python
accuracy_score(y_pred, y_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)

print(f"The accuracy on test set using Logistic Regression is: {np.round(accuracy, 3)*100.0}%")
```

## Roc Curve

```python
# plotting the ROC curve
auc_roc = roc_auc_score(y_test, model1.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, model1.predict_proba(X_test)[:,1])

plt.plot(fpr, tpr, color='blue', lw=2, 
         label='Average ROC curve (area = {0:0.3f})'.format(auc_roc))
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--', 
         label= 'Average ROC curve (area = 0.500)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
auc(fpr, tpr)
```

```python
# Auc Area Under The Curve 
auc_score = auc(fpr, tpr)
print(f"Our auc_score came out to be {round(auc_score, 3)}.")
```

# I got a test_score of 89.6000001% and an 0.952 AUC score where it is proved that my model is 95.2% correctly predicted which is great.
