```python
def extract_crew(text):
    x=[]
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            x.append(i['name'])
    y=x[:]
    return y

        
```

```python
def keyword_col(text):
    x = []
    for i in ast.literal_eval(text):
        x.append(i['name'])
    y=x[:]
    return y
```

```python
def remove_spaces(L):
    x=[]
    for i in L:
        x.append(i.replace(" ",""))
    y = x[:]
    return y
```

```python
final['genres']= final['genres'].apply(remove_spaces)
final['cast']= final['cast'].apply(remove_spaces)
final['crew']= final['crew'].apply(remove_spaces)
final['keywords']= final['keywords'].apply(remove_spaces)
final['overview']= final['overview'].apply(remove_spaces)
```

```python
final['metadata']= final['genres']+final['keywords']+final['overview'] + final['cast'] + final['crew']
```

```python
def list_to_string(L):
    str=" "
    return (str.join(L))
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(stop_words='english', max_features=5000 )
```

```python
X = cv.fit_transform(final['metadata']).toarray()
```

```python
X.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(X)
```

```python
def recommend(movie):
    # recommend 5 closest recommendations
    movie_index = final[final['title_x'] == movie].index[0]
    similarity_scores = similarity_matrix[movie_index]
    values = sorted(list(enumerate(similarity_scores)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in values:
        print(final.iloc[i[0]].title_x)
        
    
```
