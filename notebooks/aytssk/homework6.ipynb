```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1 = 'Malta'
selected_country_2 = 'Malta'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_country_1]

df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_2]

df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed_cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='COVID vs. Vaccines', xaxis=dict(title='Date'), yaxis=dict(title='Daily Cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

the global ranking of Malta in terms of the total number of cases 

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)
#print(df1.to_string())
df1.to_csv('rank_cases.csv')
rc = pd.read_csv('./rank_cases.csv')
print(rc[rc['country'] == 'Malta'])
```

the global ranking of Malta in terms of the total vaccinations per number of people

```python
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=True)
#print(df2.to_string())
df2.to_csv('rank_people_fully_vaccinated.csv')
rpfv = pd.read_csv('./rank_people_fully_vaccinated.csv')
print(rpfv[rpfv['country'] == 'Malta'])
```

***Summary of the news***  
"Malta has the lowest number of COVID-19 cases and mortality rate in the EU."  
　This fact was revealed in the latest weekly data released by the European Centre for Disease Prevention and Control (ECDC). According to the ECDC data, Malta has a mortality rate of 1.94, which is the lowest in Europe. Denmark has the second lowest mortality rate at 2.23.  
　Furthermore, there have been no COVID-related deaths in Malta in the past two weeks.This is one of the longest periods of time since the pandemic caused by COVID-19 in early 2020 has started.  
    When Malta began recording a number of new cases on March 10 ( just before restrictions were ordered ), the island's two weeks cases rate was 828.47, which is the highest number in Europe at the time.  
　In addition, Malta has seen very few new cases. As a result, the hospitalization rate has also plummeted. In addition to this, Malta has the highest vaccination rate in Europe. About 67% of the people who need to be vaccinated have received at least one dose of the vaccine.  
　The ECDC data also revealed the situation among the elderly. Last week, only two seniors tested positive for the COVID-19, one is between the ages of 65 and 79 and the other is over the age of 80. One of them is a first person who get vaccinated against COVID-19 when the vaccination program was launched in December.
With the infection rate dropping, Malta has steadily begun to deregulate its lockdown.  Restaurants, bars, and gyms are going to be reopened in the future.

***explanation for the patterns***  
  The number of new cases has been low for the past few weeks.
  It was the second longest period of time between the start of COVID-19 case statistics on February 15, 2020, to appearance of the first case on March 7, 2020.  
　This means that vaccination and tighter regulations have paid off.
As you can see from the graph, vaccination in Malta started on January 18, 2021. At this time, there were still no strong regulations in Malta. However, on March 10, 2021, Malta experienced a record number of cases.  
  This led the government to tighten regulations and increase the number of vaccinations, resulting in a dramatic decrease in the number of new cases. In just two months, Malta have been able to control the COVID-19.  
　Malta is a small country, but I think we should learn from Malta. By thoroughly regulating and efficiently spreading the vaccination, Japan will be able to control the explosive outbreak.
