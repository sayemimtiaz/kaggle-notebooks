# Data Prepration To Feed into Classifier

```python
from sklearn.model_selection import train_test_split
```

```python
#Getting All the columns
col=heart.columns
```

```python
# Data points on which classifier will be trained
X=heart[col[:-2]]
X
```

```python
#Target Values
Y=heart['output'].values
```

```python
# Splitting into Test and Train data. Test data is used for evaluation
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Random Forest Classification -- Baseline

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train,y_train)
```

# Evluating The Model

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
ypred=clf.predict(X_test)
acc=accuracy_score(y_test, ypred) # 86 % Not Bad. But Can We improve?
prc=precision_score(y_test, ypred)
recall=recall_score(y_test, ypred)
f1=f1_score(y_test, ypred)
roc_auc=roc_auc_score(y_test, ypred)
print("Accuracy  ",acc)
print("Precision  ",prc)
print("Recall  ",recall)
print("F1  ",f1)
print("ROC_AUC  ",roc_auc)
```

# Feature Engineering

#### L1 Based Feature Selection

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
```

```python
lsvc = LinearSVC(C=0.01, penalty="l1", dual=False).fit(X_train, y_train) # Usually applied on whole dataset!!!!!
```

```python
model = SelectFromModel(lsvc, prefit=True)
```

```python
X_train_new=model.transform(X_train)
X_test_new=model.transform(X_test)   # Transforming the dataset based on Selected feature
```

```python
print("Old Shape =",X_train.shape," New Shape=",X_train_new.shape)
print("Yippe we reduced the Dimesions From ",X_train.shape[1]," to",X_train_new.shape[1])
```

```python
clf2=RandomForestClassifier(max_depth=2, random_state=0)
clf2.fit(X_train_new,y_train)
yprednew=clf2.predict(X_test_new)
acc=accuracy_score(y_test, yprednew) # 86 % Not Bad. But Can We improve?
prc=precision_score(y_test, yprednew)
recall=recall_score(y_test, yprednew)
f1=f1_score(y_test, yprednew)
roc_auc=roc_auc_score(y_test, yprednew)
print("Accuracy  ",acc)
print("Precision  ",prc)
print("Recall  ",recall)
print("F1  ",f1)
print("ROC_AUC  ",roc_auc)
```

Not So Much here, (maybe in Nest dataset ðŸ˜‰), so definitely use it.

The model complexity which was earlier e^13 has reduced to e^7 as Model complexity is exponentially proportional to the dimesion of the data used to train it. # Hyper Parameter Tuning

```python
import optuna
```

```python
import optuna
from sklearn.model_selection import train_test_split

def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 2, 255),
        "max_depth": trial.suggest_int("max_depth", 2, 30),
        "min_weight_fraction_leaf": trial.suggest_float("min_weight_fraction_leaf", 0.1, 0.5),
        "max_leaf_nodes": trial.suggest_int("max_leaf_nodes", 2, 20),
        "random_state": trial.suggest_int("random_state", 0, 100),
    }
    clf=RandomForestClassifier(**params)
    
    model = clf.fit(X_train,y_train)
    preds = clf.predict(X_test)
    acc = accuracy_score(y_test, preds)
    return acc

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
```

```python

print("Number of finished trials: ", len(study.trials))
print("Best trial: ", study.best_trial.params)
```

# XGBOOST

```python
import xgboost as xgb
```

```python
def objective(trial):
    param = {
        'eval_metric' : 'auc',
        'booster' : 'gbtree',
        #'tree_method' : 'gpu_hist' , 
        'use_label_encoder' : False , 
        'lambda' : trial.suggest_loguniform('lambda' , 1e-5 , 1.0),
        'alpha' : trial.suggest_loguniform('alpha' , 1e-5 , 1.0),
        'colsample_bytree' : trial.suggest_uniform('colsample_bytree' , 0 , 1.0),
        'subsample' : trial.suggest_uniform('subsample' , 0 , 1.0),
        'learning_rate' : trial.suggest_uniform('learning_rate' , 0 , 0.02),
        'n_estimators' : trial.suggest_int('n_estimators' , 1 , 9999),
        'max_depth' : trial.suggest_int('max_depth' , 1 , 20),
        'random_state' : trial.suggest_categorical('random_state' , [0,42,2021]),
        'min_child_weight' : trial.suggest_int('min_child_weight' , 1 , 300),
        'gamma' : trial.suggest_loguniform('gamma' , 1e-5 , 1.0)
    }
    model = xgb.XGBClassifier(**param)  
    
    model.fit(X_train,y_train,eval_set=[(X_test,y_test)],early_stopping_rounds=100,verbose=False)
    
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    return acc

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
```

```python
print('number of the finished trials:' , len(study.trials))
print('the parametors of best trial:' , study.best_trial.params)
print('best value:' , study.best_value)
```
